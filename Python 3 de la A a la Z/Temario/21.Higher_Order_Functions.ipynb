En Python, las funciones se tratan como ciudadanos de primera clase, lo que le permite realizar las siguientes operaciones en las funciones:

* Una funci贸n puede tomar una o m谩s funciones como par谩metros

En esta secci贸n, cubriremos:

## **Funci贸n como par谩metro**

````python
def sum_numbers(nums):  # funci贸n normal
    return sum(nums)    # una funci贸n triste que abusa de la funci贸n de suma incorporada: <

def higher_order_function(f, lst):  # funcionar como un par谩metro
    summation = f(lst)
    return summation
result = higher_order_function(sum_numbers, [1, 2, 3, 4, 5])
print(result)       # 15
````

## **Funci贸n como valor de retorno**

````python
def square(x):          # a square function
    return x ** 2

def cube(x):            # a cube function
    return x ** 3

def absolute(x):        # an absolute value function
    if x >= 0:
        return x
    else:
        return -(x)

def higher_order_function(type): # a higher order function returning a function
    if type == 'square':
        return square
    elif type == 'cube':
        return cube
    elif type == 'absolute':
        return absolute

result = higher_order_function('square')
print(result(3))       # 9
result = higher_order_function('cube')
print(result(3))       # 27
result = higher_order_function('absolute')
print(result(-3))      # 3
````

Puede ver en el ejemplo anterior que la funci贸n de orden superior est谩 devolviendo diferentes funciones dependiendo del par谩metro pasado

## **Python Closures**

Python permite que una funci贸n anidada acceda al 谩mbito externo de la funci贸n envolvente. Esto se conoce como Cierre. Echemos un vistazo a c贸mo funcionan los cierres en Python. En Python, el cierre se crea anidando una funci贸n dentro de otra funci贸n encapsuladora y luego devolviendo la funci贸n interna. Vea el ejemplo a continuaci贸n.

````python
def add_ten():
    ten = 10
    def add(num):
        return num + ten
    return add

closure_result = add_ten()
print(closure_result(5))  # 15
print(closure_result(10))  # 20
````

## **Decoradores en Python**

Un decorador es un patr贸n de dise帽o en Python que permite a un usuario agregar una nueva funcionalidad a un objeto existente sin modificar su estructura. Los decoradores generalmente se llaman antes de la definici贸n de una funci贸n que desea decorar.

### **Crear decoradores**

Para crear una funci贸n decoradora, necesitamos una funci贸n externa con una funci贸n contenedora interna.

````python
# Normal function
def greeting():
    return 'Welcome to Python'
def uppercase_decorator(function):
    def wrapper():
        func = function()
        make_uppercase = func.upper()
        return make_uppercase
    return wrapper
g = uppercase_decorator(greeting)
print(g())          # WELCOME TO PYTHON
````

````python
## Implementemos el ejemplo anterior con un decorador.

'''Esta funci贸n decoradora es una funci贸n de orden superior
que toma una funci贸n como par谩metro'''
def uppercase_decorator(function):
    def wrapper():
        func = function()
        make_uppercase = func.upper()
        return make_uppercase
    return wrapper
@uppercase_decorator
def greeting():
    return 'Welcome to Python'
print(greeting())   # WELCOME TO PYTHON
````

## **Aplicar m煤ltiples decoradores a una sola funci贸n**

````python

'''Estas funciones de decorador son funciones de orden superior.
que toman funciones como par谩metros'''

# First Decorator
def uppercase_decorator(function):
    def wrapper():
        func = function()
        make_uppercase = func.upper()
        return make_uppercase
    return wrapper

# Second decorator
def split_string_decorator(function):
    def wrapper():
        func = function()
        splitted_string = func.split()
        return splitted_string

    return wrapper

@split_string_decorator
@uppercase_decorator     # el orden con los decoradores es importante en este caso - la funci贸n .upper() no funciona con listas
def greeting():
    return 'Welcome to Python'
print(greeting())   # WELCOME TO PYTHON
````

## **Aceptar par谩metros en funciones de decorador**

La mayor铆a de las veces necesitamos que nuestras funciones tomen par谩metros, por lo que es posible que debamos definir un decorador que acepte par谩metros.

````python
def decorator_with_parameters(function):
    def wrapper_accepting_parameters(para1, para2, para3):
        function(para1, para2, para3)
        print("I live in {}".format(para3))
    return wrapper_accepting_parameters

@decorator_with_parameters
def print_full_name(first_name, last_name, country):
    print("I am {} {}. I love to teach.".format(
        first_name, last_name, country))

print_full_name("Asabeneh", "Yetayeh",'Finland')
````

## **Funciones integradas de orden superior**

Algunas de las funciones integradas de orden superior que cubrimos en esta parte son map() , filter y reduce . La funci贸n lambda se puede pasar como un par谩metro y el mejor caso de uso de las funciones lambda es en funciones como mapa, filtro y reducci贸n.

### **Python - Funci贸n de mapa**

La funci贸n map() es una funci贸n integrada que toma una funci贸n y es iterable como par谩metros.

```
    # syntax
    map(function, iterable)
```

````python
numbers = [1, 2, 3, 4, 5] # iterable
def square(x):
    return x ** 2
numbers_squared = map(square, numbers)
print(list(numbers_squared))    # [1, 4, 9, 16, 25]
# Lets apply it with a lambda function
numbers_squared = map(lambda x : x ** 2, numbers)
print(list(numbers_squared))    # [1, 4, 9, 16, 25]
````

````python
numbers_str = ['1', '2', '3', '4', '5']  # iterable
numbers_int = map(int, numbers_str)
print(list(numbers_int))    # [1, 2, 3, 4, 5]
````

````python
names = ['Asabeneh', 'Lidiya', 'Ermias', 'Abraham']  # iterable

def change_to_upper(name):
    return name.upper()

names_upper_cased = map(change_to_upper, names)
print(list(names_upper_cased))    # ['ASABENEH', 'LIDIYA', 'ERMIAS', 'ABRAHAM']

# Let us apply it with a lambda function
names_upper_cased = map(lambda name: name.upper(), names)
print(list(names_upper_cased))    # ['ASABENEH', 'LIDIYA', 'ERMIAS', 'ABRAHAM']
````

Lo que realmente hace el mapa es iterar sobre una lista. Por ejemplo, cambia los nombres a may煤sculas y devuelve una nueva lista.

### **Python - Funci贸n de filtro**

La funci贸n filter() llama a la funci贸n especificada que devuelve un valor booleano para cada elemento de la iterable (lista) especificada. Filtra los elementos que cumplen los criterios de filtrado.

```
    # syntax
    filter(function, iterable)
```

````python
# Lets filter only even nubers
numbers = [1, 2, 3, 4, 5]  # iterable

def is_even(num):
    if num % 2 == 0:
        return True
    return False

even_numbers = filter(is_even, numbers)
print(list(even_numbers))       # [2, 4]
````

````python
numbers = [1, 2, 3, 4, 5]  # iterable

def is_odd(num):
    if num % 2 != 0:
        return True
    return False

odd_numbers = filter(is_odd, numbers)
print(list(odd_numbers))       # [1, 3, 5]
````

````python
# Filter long name
names = ['Asabeneh', 'Lidiya', 'Ermias', 'Abraham']  # iterable
def is_name_long(name):
    if len(name) > 7:
        return True
    return False

long_names = filter(is_name_long, names)
print(list(long_names))         # ['Asabeneh']
````

### **Python - Funci贸n de reducci贸n**

La funci贸n reduce() est谩 definida en el m贸dulo functools y debemos importarla desde este m贸dulo. Al igual que el mapa y el filtro, toma dos par谩metros, una funci贸n y un iterable. Sin embargo, no devuelve otro iterable, sino un solo valor.

````python
from functools import reduce

numbers_str = ['1', '2', '3', '4', '5']  # iterable
def add_two_nums(x, y):
    return int(x) + int(y)

total = reduce(add_two_nums, numbers_str)
print(total)    # 15
````
