# **15. Visualización de Datos usando Matplotlib de Python**

## **Introducción a la Visualización de Datos**

La visualización de datos es una herramienta fundamental en la comunicación efectiva de información compleja. La capacidad de transformar datos en información visualmente atractiva y fácil de entender es esencial para la toma de decisiones informadas en cualquier campo, desde la ciencia y la tecnología hasta los negocios y la política.

La visualización de datos implica la creación de gráficos, diagramas, tablas y otros medios visuales para representar información. Esto puede incluir datos cuantitativos (como números y estadísticas) y cualitativos (como categorías y descripciones).

Al visualizar datos, es importante considerar el objetivo de la visualización y el público al que va dirigido. Los gráficos deben ser claros, concisos y fáciles de entender, y deben resaltar las tendencias y patrones importantes en los datos.

Hay muchos tipos diferentes de gráficos y visualizaciones de datos disponibles, y cada uno es adecuado para diferentes tipos de datos y propósitos. Algunos ejemplos incluyen histogramas, gráficos de barras, gráficos de líneas, gráficos de dispersión y diagramas de caja y bigotes.

En resumen, la visualización de datos es una herramienta poderosa para la comunicación efectiva de información compleja. La capacidad de transformar datos en información visualmente atractiva y fácil de entender es esencial para la toma de decisiones informadas en cualquier campo.

## **Anatomía de la Figura de Matplotlib**

La biblioteca Matplotlib es una herramienta muy popular y útil en Python para la visualización de datos. La mayoría de las visualizaciones en Matplotlib se construyen a partir de una figura y uno o más ejes. La figura es el contenedor principal para todos los elementos de la visualización, mientras que los ejes son los objetos que se utilizan para crear los gráficos y agregar etiquetas y anotaciones.

A continuación, se detalla la anatomía de una figura de Matplotlib:

* Título de la figura: se utiliza para describir brevemente la visualización y proporcionar información general sobre el conjunto de datos.

* Ejemplo de gráfico: un ejemplo de lo que se visualizará en la figura.

* Ejes: los ejes se dividen en dos tipos: los ejes X (horizontal) y los ejes Y (vertical). Son los objetos en los que se crean los gráficos.

* Etiquetas de los ejes: etiquetas para los ejes X e Y.

* Leyenda: se utiliza para describir brevemente los diferentes elementos en la visualización.

* Ticks: se utilizan para marcar los puntos a lo largo del eje.

* Spines: son las líneas que conectan los ticks y forman el marco de la visualización.

* Grid: se utiliza para ayudar a leer los valores de los ejes.

* Anotaciones: texto adicional que se agrega a la visualización para proporcionar más información.

En resumen, una figura de Matplotlib consta de varios elementos, incluidos los ejes, etiquetas de los ejes, leyendas, ticks, spines, grid y anotaciones. Cada uno de estos elementos juega un papel importante en la visualización de datos efectiva.

## **Creación de Figuras Simples**

```python
import matplotlib.pyplot as plt
import numpy as np

# Configurar el entorno
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Agregar el gráfico de líneas con opciones personalizadas
ax.plot(x, y, color='blue', linestyle='--', linewidth=2, label='Seno(x)')

# Agregar etiquetas de los ejes con opciones personalizadas
ax.set_xlabel('Eje X', fontsize=12)
ax.set_ylabel('Eje Y', fontsize=12)

# Agregar título a la figura con opciones personalizadas
ax.set_title('Gráfico de la Función Seno', fontsize=14)

# Agregar una cuadrícula al gráfico
ax.grid(True, linestyle='--', alpha=0.6)

# Agregar una leyenda
ax.legend(loc='upper right', fontsize=10)

# Cambiar el tamaño de los ejes
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)

# Personalizar las marcas de los ejes
ax.tick_params(axis='both', labelsize=10)

# Mostrar la figura
plt.show()
```

## **Generación de Datos Aleatorios y Contexto de Ejercicios**

La generación de datos aleatorios es una técnica muy útil en la visualización de datos, ya que nos permite crear conjuntos de datos ficticios para experimentar con diferentes técnicas y gráficos. En Python, la biblioteca NumPy proporciona varias funciones para generar datos aleatorios, como `random.rand()`, `random.randn()`, `random.randint()`, entre otras.

En el contexto de ejercicios de visualización de datos, se puede utilizar la generación de datos aleatorios para crear conjuntos de datos ficticios con diferentes distribuciones y patrones, para luego visualizarlos y experimentar con diferentes gráficos y técnicas de visualización. Por ejemplo, se pueden generar datos aleatorios con una distribución normal, uniforme o exponencial, y luego visualizarlos con histogramas, gráficos de barras o gráficos de líneas.

También se pueden utilizar conjuntos de datos reales de diferentes fuentes, como bases de datos públicas o archivos CSV, y luego preprocesarlos y transformarlos para adaptarlos a las necesidades de visualización. En cualquier caso, la generación de datos aleatorios y el uso de conjuntos de datos reales son herramientas importantes en el aprendizaje y práctica de la visualización de datos.

## **Personalización de Figuras**

La personalización de figuras es una parte importante de la visualización de datos, ya que nos permite adaptar la apariencia y el estilo de los gráficos para resaltar la información más relevante y transmitir el mensaje de forma clara y efectiva. En Matplotlib, existen muchas opciones y parámetros para personalizar las figuras, como colores, estilos de línea, tipos de marcadores, tamaños de fuente, títulos y etiquetas de ejes, entre otros.

A continuación, se muestran algunos ejemplos de cómo personalizar una figura de Matplotlib:

```python
import matplotlib.pyplot as plt
import numpy as np

# Configurar el entorno
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Agregar los gráficos de líneas con diferentes estilos y colores
ax.plot(x, y1, color='blue', linestyle='-', linewidth=2, label='Seno')
ax.plot(x, y2, color='red', linestyle='--', linewidth=2, label='Coseno')

# Agregar título y etiquetas de los ejes
ax.set_title('Gráfico de Funciones Seno y Coseno', fontsize=14)
ax.set_xlabel('Eje X', fontsize=12)
ax.set_ylabel('Eje Y', fontsize=12)

# Agregar leyenda y personalizar su estilo
ax.legend(loc='upper right', fontsize=12, frameon=False)

# Ajustar los límites de los ejes
ax.set_xlim([0, 10])
ax.set_ylim([-1.5, 1.5])

# Personalizar el estilo de los ticks y las etiquetas
ax.tick_params(axis='both', which='major', labelsize=10, width=1.5, length=6)
ax.tick_params(axis='both', which='minor', labelsize=8, width=1, length=4)

# Agregar cuadrícula
ax.grid(True, linestyle='--', alpha=0.6)

# Personalizar el fondo del gráfico
ax.set_facecolor('lightgray')

# Mostrar la figura
plt.show()
```

En este ejemplo, se agregan dos gráficos de líneas con diferentes estilos de línea y colores. Además, se agregan etiquetas a los ejes y un título a la figura. Se agrega una leyenda para identificar los gráficos y se ajustan los límites de los ejes para resaltar los valores más importantes. También se personaliza el estilo de los ticks y las etiquetas de los ejes para mejorar la legibilidad del gráfico.

Estos son solo algunos ejemplos de cómo personalizar una figura en Matplotlib. Con la práctica y el conocimiento de las diferentes opciones y parámetros disponibles, se pueden crear figuras de visualización de datos atractivas y efectivas para comunicar la información de manera clara y precisa.

## **Gráficas de Barras Verticales**

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
x = ['A', 'B', 'C', 'D', 'E']
y = [20, 35, 30, 25, 15]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Agregar los datos como barras verticales con personalizaciones
ax.bar(x, y, color='royalblue', alpha=0.7, edgecolor='black', linewidth=1.2, label='Frecuencia')

# Personalizar el título y etiquetas de los ejes
ax.set_title('Gráfico de Barras Verticales', fontsize=14)
ax.set_xlabel('Categorías', fontsize=12)
ax.set_ylabel('Frecuencia', fontsize=12)

# Personalizar el estilo de las etiquetas en el eje x con FixedLocator y FixedFormatter
ax.set_xticks(np.arange(len(x)))
ax.set_xticklabels(x, fontsize=10, rotation=0)

# Personalizar el estilo de los ticks y las etiquetas del eje y
ax.tick_params(axis='y', which='major', labelsize=10, width=1.2, length=6)

# Agregar valores en las barras
for i, v in enumerate(y):
    ax.text(i, v + 1, str(v), ha='center', va='bottom', fontsize=10)

# Agregar una leyenda
ax.legend(loc='upper right', fontsize=10)

# Añadir una cuadrícula
ax.yaxis.grid(True, linestyle='--', alpha=0.6)

# Personalizar el fondo del gráfico
ax.set_facecolor('lightgray')

# Mostrar la figura
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
x = ['A', 'B', 'C', 'D', 'E']
y1 = [20, 35, 30, 25, 15]
y2 = [15, 25, 20, 30, 20]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Ajustar la posición de las barras
bar_width = 0.35
x_pos = np.arange(len(x))

# Agregar los datos como barras verticales con personalizaciones
ax.bar(x_pos - bar_width/2, y1, bar_width, color='royalblue', alpha=0.7, label='Datos 1', edgecolor='black', linewidth=1.2)
ax.bar(x_pos + bar_width/2, y2, bar_width, color='orangered', alpha=0.7, label='Datos 2', edgecolor='black', linewidth=1.2)

# Personalizar el título y etiquetas de los ejes
ax.set_title('Gráfico de Barras Verticales Agrupadas', fontsize=14)
ax.set_xlabel('Categorías', fontsize=12)
ax.set_ylabel('Frecuencia', fontsize=12)

# Ajustar la posición de las etiquetas del eje x
ax.set_xticks(x_pos)
ax.set_xticklabels(x, fontsize=10)

# Personalizar el estilo de los ticks y las etiquetas del eje y
ax.tick_params(axis='y', which='major', labelsize=10, width=1.2, length=6)

# Agregar una leyenda
ax.legend(fontsize=10)

# Añadir valores en las barras
for i, v in enumerate(y1):
    ax.text(i - bar_width/2, v + 0.5, str(v), ha='center', va='bottom', fontsize=10)
for i, v in enumerate(y2):
    ax.text(i + bar_width/2, v + 0.5, str(v), ha='center', va='bottom', fontsize=10)

# Añadir una cuadrícula
ax.yaxis.grid(True, linestyle='--', alpha=0.6)

# Personalizar el fondo del gráfico
ax.set_facecolor('lightgray')

# Mostrar la figura
plt.show()
```

## **Gráficas de Barras Apiladas**

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D', 'E']
valores1 = [10, 24, 36, 40, 15]
valores2 = [5, 12, 20, 28, 8]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Ajustar el ancho de las barras
bar_width = 0.6

# Colores personalizados para las barras
colores1 = 'royalblue'
colores2 = 'orange'

# Crear la gráfica de barras apiladas con personalizaciones
ax.bar(categorias, valores1, bar_width, color=colores1, edgecolor='black', linewidth=1.2, alpha=0.7, label='Valores 1')
ax.bar(categorias, valores2, bar_width, color=colores2, edgecolor='black', linewidth=1.2, alpha=0.7, label='Valores 2', bottom=valores1)

# Personalizar el título y etiquetas de los ejes
ax.set_title('Gráfico de Barras Apiladas', fontsize=14)
ax.set_xlabel('Categorías', fontsize=12)
ax.set_ylabel('Valores', fontsize=12)

# Personalizar el estilo de los ticks y las etiquetas del eje x
ax.tick_params(axis='x', which='major', labelsize=10, width=1.2, length=6)

# Añadir valores en las barras
for i, v1, v2 in zip(range(len(categorias)), valores1, valores2):
    ax.text(i, v1 + v2, f'{v1 + v2}', ha='center', va='bottom', fontsize=10, color='black')

# Añadir una cuadrícula
ax.yaxis.grid(True, linestyle='--', alpha=0.6)

# Personalizar el fondo del gráfico
ax.set_facecolor('lightgray')

# Ajustar el límite del eje y
ax.set_ylim([0, 100])

# Agregar leyenda
ax.legend(loc='upper right')

# Mostrar la figura
plt.show()
```

## **Gráficas de Dispersión**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos aleatorios
x = np.random.rand(50)
y = np.random.rand(50)

# Crear la gráfica de dispersión con personalizaciones
plt.scatter(x, y, color='blue', s=50, marker='o', alpha=0.7, label='Datos')

# Personalizar el título y etiquetas de los ejes
plt.title('Gráfica de Dispersión', fontsize=14)
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)

# Personalizar el estilo de los ticks y las etiquetas de los ejes
plt.tick_params(axis='both', which='major', labelsize=10, width=1.2, length=6)

# Agregar una cuadrícula
plt.grid(True, linestyle='--', alpha=0.6)

# Añadir una línea de regresión
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), color='red', linestyle='--', linewidth=2, label='Línea de Regresión')

# Personalizar el fondo del gráfico
plt.gca().set_facecolor('lightgray')

# Agregar leyenda
plt.legend(loc='upper right')

# Mostrar la gráfica
plt.show()
```

## **Gráficas de Pie**

```python
import matplotlib.pyplot as plt

# Definir los datos
categorias = ['A', 'B', 'C', 'D', 'E']
frecuencias = [20, 10, 30, 15, 25]

# Personalizar colores
colores = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink']

# Separación de la rebanada "C" para resaltarla
separacion = (0, 0, 0.1, 0, 0)

# Crear la gráfica de pastel con personalizaciones
plt.pie(frecuencias, labels=categorias, colors=colores, explode=separacion, autopct='%1.1f%%',
        shadow=True, startangle=140)

# Configurar el título
plt.title('Gráfica de Pastel Personalizada')

# Mostrar la gráfica
plt.show()
```

## **Combinar Gráficas**

```python
import matplotlib.pyplot as plt
import numpy as np

# Definir los datos
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([10, 15, 13, 17, 20])
y2 = np.array([5, 7, 8, 12, 11])

# Crear la figura con dos subfiguras
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Configurar la primera subfigura como una gráfica de barras
ax1.bar(x, y1, color='skyblue', edgecolor='gray')
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Y1')
ax1.set_title('Gráfica de Barras')

# Configurar la segunda subfigura como una gráfica de línea
ax2.plot(x, y2, marker='o', linestyle='-', color='orange', markersize=8, markerfacecolor='red')
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Y2')
ax2.set_title('Gráfica de Línea')

# Ajustar los espacios entre las subfiguras
plt.subplots_adjust(wspace=0.3)

# Mostrar la figura
plt.show()
```
