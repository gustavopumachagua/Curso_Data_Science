## **Anatom铆a de la Figura de Matplotlib**

La mayor铆a de las visualizaciones en Matplotlib se construyen a partir de una figura y uno o m谩s ejes. En resumen, una figura de Matplotlib consta de varios elementos, incluidos los ejes, etiquetas de los ejes, leyendas, ticks, spines, grid y anotaciones. Cada uno de estos elementos juega un papel importante en la visualizaci贸n de datos efectiva.

## **Creaci贸n de Figuras Simples**

```python
import matplotlib.pyplot as plt
import numpy as np

# Configurar el entorno
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Agregar el gr谩fico de l铆neas con opciones personalizadas
ax.plot(x, y, color='blue', linestyle='--', linewidth=2, label='Seno(x)')

# Agregar etiquetas de los ejes con opciones personalizadas
ax.set_xlabel('Eje X', fontsize=12)
ax.set_ylabel('Eje Y', fontsize=12)

# Agregar t铆tulo a la figura con opciones personalizadas
ax.set_title('Gr谩fico de la Funci贸n Seno', fontsize=14)

# Agregar una cuadr铆cula al gr谩fico
ax.grid(True, linestyle='--', alpha=0.6)

# Agregar una leyenda
ax.legend(loc='upper right', fontsize=10)

# Cambiar el tama帽o de los ejes
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)

# Personalizar las marcas de los ejes
ax.tick_params(axis='both', labelsize=10)

# Mostrar la figura
plt.show()
```

## **Personalizaci贸n de Figuras**

La personalizaci贸n de figuras es una parte importante de la visualizaci贸n de datos, ya que nos permite adaptar la apariencia y el estilo de los gr谩ficos para resaltar la informaci贸n m谩s relevante y transmitir el mensaje de forma clara y efectiva. En Matplotlib, existen muchas opciones y par谩metros para personalizar las figuras, como colores, estilos de l铆nea, tipos de marcadores, tama帽os de fuente, t铆tulos y etiquetas de ejes, entre otros.

A continuaci贸n, se muestran algunos ejemplos de c贸mo personalizar una figura de Matplotlib:

```python
import matplotlib.pyplot as plt
import numpy as np

# Configurar el entorno
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Agregar los gr谩ficos de l铆neas con diferentes estilos y colores
ax.plot(x, y1, color='blue', linestyle='-', linewidth=2, label='Seno')
ax.plot(x, y2, color='red', linestyle='--', linewidth=2, label='Coseno')

# Agregar t铆tulo y etiquetas de los ejes
ax.set_title('Gr谩fico de Funciones Seno y Coseno', fontsize=14)
ax.set_xlabel('Eje X', fontsize=12)
ax.set_ylabel('Eje Y', fontsize=12)

# Agregar leyenda y personalizar su estilo
ax.legend(loc='upper right', fontsize=12, frameon=False)

# Ajustar los l铆mites de los ejes
ax.set_xlim([0, 10])
ax.set_ylim([-1.5, 1.5])

# Personalizar el estilo de los ticks y las etiquetas
ax.tick_params(axis='both', which='major', labelsize=10, width=1.5, length=6)
ax.tick_params(axis='both', which='minor', labelsize=8, width=1, length=4)

# Agregar cuadr铆cula
ax.grid(True, linestyle='--', alpha=0.6)

# Personalizar el fondo del gr谩fico
ax.set_facecolor('lightgray')

# Mostrar la figura
plt.show()
```

En este ejemplo, se agregan dos gr谩ficos de l铆neas con diferentes estilos de l铆nea y colores. Adem谩s, se agregan etiquetas a los ejes y un t铆tulo a la figura. Se agrega una leyenda para identificar los gr谩ficos y se ajustan los l铆mites de los ejes para resaltar los valores m谩s importantes. Tambi茅n se personaliza el estilo de los ticks y las etiquetas de los ejes para mejorar la legibilidad del gr谩fico.

Estos son solo algunos ejemplos de c贸mo personalizar una figura en Matplotlib. Con la pr谩ctica y el conocimiento de las diferentes opciones y par谩metros disponibles, se pueden crear figuras de visualizaci贸n de datos atractivas y efectivas para comunicar la informaci贸n de manera clara y precisa.

## **Gr谩ficas de Barras Verticales**

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
x = ['A', 'B', 'C', 'D', 'E']
y = [20, 35, 30, 25, 15]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Agregar los datos como barras verticales con personalizaciones
ax.bar(x, y, color='royalblue', alpha=0.7, edgecolor='black', linewidth=1.2, label='Frecuencia')

# Personalizar el t铆tulo y etiquetas de los ejes
ax.set_title('Gr谩fico de Barras Verticales', fontsize=14)
ax.set_xlabel('Categor铆as', fontsize=12)
ax.set_ylabel('Frecuencia', fontsize=12)

# Personalizar el estilo de las etiquetas en el eje x con FixedLocator y FixedFormatter
ax.set_xticks(np.arange(len(x)))
ax.set_xticklabels(x, fontsize=10, rotation=0)

# Personalizar el estilo de los ticks y las etiquetas del eje y
ax.tick_params(axis='y', which='major', labelsize=10, width=1.2, length=6)

# Agregar valores en las barras
for i, v in enumerate(y):
    ax.text(i, v + 1, str(v), ha='center', va='bottom', fontsize=10)

# Agregar una leyenda
ax.legend(loc='upper right', fontsize=10)

# A帽adir una cuadr铆cula
ax.yaxis.grid(True, linestyle='--', alpha=0.6)

# Personalizar el fondo del gr谩fico
ax.set_facecolor('lightgray')

# Mostrar la figura
plt.show()
```

## **Gr谩ficas de Dispersi贸n**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos aleatorios
x = np.random.rand(50)
y = np.random.rand(50)

# Crear la gr谩fica de dispersi贸n con personalizaciones
plt.scatter(x, y, color='blue', s=50, marker='o', alpha=0.7, label='Datos')

# Personalizar el t铆tulo y etiquetas de los ejes
plt.title('Gr谩fica de Dispersi贸n', fontsize=14)
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)

# Personalizar el estilo de los ticks y las etiquetas de los ejes
plt.tick_params(axis='both', which='major', labelsize=10, width=1.2, length=6)

# Agregar una cuadr铆cula
plt.grid(True, linestyle='--', alpha=0.6)

# A帽adir una l铆nea de regresi贸n
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), color='red', linestyle='--', linewidth=2, label='L铆nea de Regresi贸n')

# Personalizar el fondo del gr谩fico
plt.gca().set_facecolor('lightgray')

# Agregar leyenda
plt.legend(loc='upper right')

# Mostrar la gr谩fica
plt.show()
```

## **Combinar Gr谩ficas**

```python
import matplotlib.pyplot as plt
import numpy as np

# Definir los datos
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([10, 15, 13, 17, 20])
y2 = np.array([5, 7, 8, 12, 11])

# Crear la figura con dos subfiguras
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Configurar la primera subfigura como una gr谩fica de barras
ax1.bar(x, y1, color='skyblue', edgecolor='gray')
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Y1')
ax1.set_title('Gr谩fica de Barras')

# Configurar la segunda subfigura como una gr谩fica de l铆nea
ax2.plot(x, y2, marker='o', linestyle='-', color='orange', markersize=8, markerfacecolor='red')
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Y2')
ax2.set_title('Gr谩fica de L铆nea')

# Ajustar los espacios entre las subfiguras
plt.subplots_adjust(wspace=0.3)

# Mostrar la figura
plt.show()
```
