## **Personalizaci贸n de Figuras**

La personalizaci贸n de figuras es una parte importante de la visualizaci贸n de datos, ya que nos permite adaptar la apariencia y el estilo de los gr谩ficos para resaltar la informaci贸n m谩s relevante y transmitir el mensaje de forma clara y efectiva. A continuaci贸n, se muestran algunos ejemplos de c贸mo personalizar una figura de Matplotlib:

```python
import matplotlib.pyplot as plt
import numpy as np

# Configurar el entorno
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Agregar los gr谩ficos de l铆neas con diferentes estilos
ax.plot(x, y1, color='blue', linestyle='-', linewidth=2, label='Seno')
ax.plot(x, y2, color='red', linestyle='--', linewidth=2, label='Coseno')

# Agregar t铆tulo y etiquetas de los ejes
ax.set_title('Gr谩fico de L铆neas')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')

# Agregar leyenda y personalizar su estilo
ax.legend(loc='lower left', frameon=False, fontsize=12)

# Ajustar los l铆mites de los ejes
ax.set_xlim([0, 10])
ax.set_ylim([-1.5, 1.5])

# Personalizar el estilo de los ticks y las etiquetas
ax.tick_params(axis='both', which='major', labelsize=12, color='black', width=2, length=6)

# Mostrar la figura
plt.show()
```

En este ejemplo, se agregan dos gr谩ficos de l铆neas con diferentes estilos de l铆nea y colores. Adem谩s, se agregan etiquetas a los ejes y un t铆tulo a la figura. Tambi茅n se personaliza el estilo de los ticks y las etiquetas de los ejes para mejorar la legibilidad del gr谩fico.

Estos son solo algunos ejemplos de c贸mo personalizar una figura en Matplotlib. Con la pr谩ctica y el conocimiento de las diferentes opciones y par谩metros disponibles, se pueden crear figuras de visualizaci贸n de datos atractivas y efectivas para comunicar la informaci贸n de manera clara y precisa.

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
x = ['A', 'B', 'C', 'D', 'E']
y1 = [20, 35, 30, 25, 15]
y2 = [15, 25, 20, 30, 20]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Ajustar la posici贸n de las barras
bar_width = 0.35
x_pos = np.arange(len(x))

# Agregar los datos como barras verticales
ax.bar(x_pos - bar_width/2, y1, bar_width, color='blue', alpha=0.6, label='Datos 1')
ax.bar(x_pos + bar_width/2, y2, bar_width, color='orange', alpha=0.6, label='Datos 2')

# Agregar t铆tulo y etiquetas de los ejes
ax.set_title('Gr谩fico de Barras Verticales Agrupadas')
ax.set_xlabel('Categor铆as')
ax.set_ylabel('Frecuencia')

# Ajustar la posici贸n de las etiquetas del eje x
ax.set_xticks(x_pos)
ax.set_xticklabels(x)

# Agregar leyenda
ax.legend()

# Mostrar la figura
plt.show()
```

## **Combinar Gr谩ficas**

```python
import matplotlib.pyplot as plt
import numpy as np

# Definir los datos
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([10, 15, 13, 17, 20])
y2 = np.array([5, 7, 8, 12, 11])

# Crear la figura con dos subfiguras
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Configurar la primera subfigura como una gr谩fica de barras
ax1.bar(x, y1)
ax1.set_xlabel('X')
ax1.set_ylabel('Y1')
ax1.set_title('Gr谩fica de Barras')

# Configurar la segunda subfigura como una gr谩fica de l铆nea
ax2.plot(x, y2)
ax2.set_xlabel('X')
ax2.set_ylabel('Y2')
ax2.set_title('Gr谩fica de L铆nea')

# Ajustar los espacios entre las subfiguras
plt.subplots_adjust(wspace=0.3)

# Mostrar la figura
plt.show()
```
