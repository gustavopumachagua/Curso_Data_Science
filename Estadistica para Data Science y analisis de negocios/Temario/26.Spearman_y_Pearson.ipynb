# **26. Diferencias entre la Correlación de Spearman y Pearson**

## **1. Correlación de Spearman**

La correlación de Spearman es una medida de la relación entre dos variables que se basa en los rangos de los datos en lugar de en los valores absolutos. Se utiliza cuando los datos no siguen una distribución normal o cuando hay valores extremos. El coeficiente de correlación de Spearman varía entre -1 y 1, donde un valor de 1 indica una correlación positiva perfecta, un valor de -1 indica una correlación negativa perfecta y un valor de 0 indica que no hay correlación.

Aquí te muestro un ejemplo de cómo calcular el coeficiente de correlación de ```Spearman``` en Python utilizando la función spearmanr de la librería ```scipy.stats```. También te muestro cómo visualizar los datos utilizando una gráfica de dispersión.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# Definimos los datos de ejemplo
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 5, 1])

# Calculamos el coeficiente de correlación de Spearman
coeficiente, p_valor = spearmanr(x, y)

# Creamos una gráfica de dispersión
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f"Coeficiente de correlación de Spearman: {coeficiente:.2f}")

# Mostramos la gráfica
plt.show()
````

En este ejemplo, primero importamos las librerías necesarias. Luego, definimos dos arrays de ejemplo ```x``` e ```y```.

Después, utilizamos la función ```spearmanr``` de la librería ```scipy.stats``` para calcular el coeficiente de correlación de Spearman entre x e y. La función devuelve dos valores: el coeficiente de correlación y el p-valor asociado.

Finalmente, creamos una gráfica de dispersión utilizando la librería ```matplotlib``` y la función ```scatter```. Añadimos el coeficiente de correlación en el título de la gráfica utilizando la f-string notation.

## **2. Relación monotónica positiva**

Una relación monotónica positiva es aquella en la que a medida que aumenta el valor de una variable, también aumenta el valor de la otra variable. Es decir, cuando se grafica la relación, los puntos tienden a agruparse en una línea recta con una pendiente positiva.

A continuación, te muestro un ejemplo de cómo crear una relación monotónica positiva en Python utilizando la librería ```numpy``` y la librería ```matplotlib```.

````python
import numpy as np
import matplotlib.pyplot as plt

# Creamos un array de valores x y calculamos valores de y con una relación monotónica positiva
x = np.arange(0, 10, 0.5)
y = 2*x + np.random.normal(scale=1, size=len(x))

# Graficamos los puntos
plt.scatter(x, y)

# Agregamos una línea recta que representa la relación monotónica positiva
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--")

# Etiquetamos los ejes y agregamos un título
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ejemplo de una relación monotónica positiva")

# Mostramos la gráfica
plt.show()
````

En este ejemplo, primero importamos las librerías necesarias. Luego, creamos un array de valores x que van desde 0 a 10 con un intervalo de 0.5.

Después, creamos un array de valores y que se calculan a partir de x con una relación monotónica positiva de $y = 2*x$. Además, añadimos ruido aleatorio a los valores de y utilizando la función ```np.random.normal```.

A continuación, graficamos los puntos utilizando la función ```scatter``` de ```matplotlib```.

Para crear la línea recta que representa la relación monotónica positiva, utilizamos la función ```np.polyfit``` de numpy para ajustar una línea recta a los datos de x e y, y luego utilizamos la función ```np.poly1d``` para generar una función polinómica a partir de los coeficientes del ajuste. Finalmente, graficamos la línea recta utilizando la función ```plot``` de ```matplotlib```.

Por último, añadimos etiquetas a los ejes y un título a la gráfica utilizando las funciones ```xlabel```, ```ylabel``` y ```title```, respectivamente.

## **3. Relación monotónica negativa**

Una relación monotónica negativa es aquella en la que a medida que aumenta el valor de una variable, disminuye el valor de la otra variable. Es decir, cuando se grafica la relación, los puntos tienden a agruparse en una línea recta con una pendiente negativa.

A continuación, te muestro un ejemplo de cómo crear una relación monotónica negativa en Python utilizando la librería numpy y la librería matplotlib.

````python
import numpy as np
import matplotlib.pyplot as plt

# Creamos un array de valores x y calculamos valores de y con una relación monotónica negativa
x = np.arange(0, 10, 0.5)
y = -2*x + np.random.normal(scale=1, size=len(x))

# Graficamos los puntos
plt.scatter(x, y)

# Agregamos una línea recta que representa la relación monotónica negativa
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--")

# Etiquetamos los ejes y agregamos un título
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ejemplo de una relación monotónica negativa")

# Mostramos la gráfica
plt.show()
````

En este ejemplo, primero importamos las librerías necesarias. Luego, creamos un array de valores x que van desde 0 a 10 con un intervalo de 0.5.

Después, creamos un array de valores y que se calculan a partir de x con una relación monotónica negativa de $y = -2*x$. Además, añadimos ruido aleatorio a los valores de y utilizando la función ```np.random.normal```.

A continuación, graficamos los puntos utilizando la función scatter de matplotlib.

Para crear la línea recta que representa la relación monotónica negativa, utilizamos la función ```np.polyfit``` de numpy para ajustar una línea recta a los datos de ```x``` e ```y```, y luego utilizamos la función ```np.poly1d``` para generar una función polinómica a partir de los coeficientes del ajuste. Finalmente, graficamos la línea recta utilizando la función plot de matplotlib.

Por último, añadimos etiquetas a los ejes y un título a la gráfica utilizando las funciones ```xlabel```, ```ylabel``` y ```title```, respectivamente.

Una relación monotónica inexistente es aquella en la que no existe una relación clara entre los valores de las dos variables. Es decir, cuando se grafica la relación, los puntos están dispersos sin una tendencia clara a agruparse en una línea recta.

A continuación, te muestro un ejemplo de cómo crear una relación monotónica inexistente en Python utilizando la librería numpy y la librería matplotlib.

## **5. Spearman (monotónico) vs Pearson (lineal)**

Tanto la correlación de Spearman como la de Pearson miden la relación entre dos variables, pero la correlación de Spearman mide la relación monotónica mientras que la de Pearson mide la relación lineal. Es decir, la correlación de Spearman es más adecuada para medir relaciones entre variables no lineales o variables ordinales, mientras que la correlación de Pearson es más adecuada para medir relaciones entre variables continuas lineales.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, pearsonr

# Creamos dos arrays de valores x e y con valores aleatorios
x = np.random.normal(size=50)
y = np.random.normal(size=50)

# Creamos una variable z que es igual a x^2
z = x**2

# Graficamos los puntos
plt.scatter(x, y)

# Calculamos la correlación de Spearman
rho, pval = spearmanr(x, y)
print(f"Coeficiente de correlación de Spearman: {rho:.2f}")

# Calculamos la correlación de Pearson
r, pval = pearsonr(x, y)
print(f"Coeficiente de correlación de Pearson: {r:.2f}")

# Graficamos la línea de la relación monotónica
x_ord = np.sort(x)
y_ord = y[np.argsort(x)]
plt.plot(x_ord, y_ord, color="r")

# Etiquetamos los ejes y agregamos un título
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ejemplo de una relación no lineal entre X e Y")

# Mostramos la gráfica
plt.show()
````

## **6. Fórmula matemática**

La fórmula matemática para la correlación de Spearman es la siguiente en código

$r_s = 1 - (6 * S) / (n * (n**2 - 1))$


Donde:
- ```r_s``` es la correlación de Spearman
- ```S``` es la suma de los cuadrados de las diferencias de los rangos
- ```n``` es el número de observaciones

La fórmula matemática para la correlación de Pearson es la siguiente en código

$r = (sum((x - x_bar) * (y - y_bar) for x, y in zip(data_x, data_y))) / ((n - 1) * s_x * s_y)$


Donde:
- `r` es la correlación de Pearson
- $data_x$ y $data_y$ son las dos variables
- `n` es el número de observaciones
- $x_bar$ y $y_bar$ son las medias de las dos variables
- $s_x$ y $s_y$ son las desviaciones estándar de las dos variables
