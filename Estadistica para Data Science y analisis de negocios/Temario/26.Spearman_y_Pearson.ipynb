La correlaci贸n de Spearman es una medida de la relaci贸n entre dos variables que se basa en los rangos de los datos en lugar de en los valores absolutos.
````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# Definimos los datos de ejemplo
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 5, 1])

# Calculamos el coeficiente de correlaci贸n de Spearman
coeficiente, p_valor = spearmanr(x, y)

# Creamos una gr谩fica de dispersi贸n
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')

# Personalizamos el t铆tulo con el coeficiente y el valor p
plt.title(f"Scatter Plot - Spearman's R: {coeficiente:.2f}, p-value: {p_valor:.2f}")

# Mostramos la gr谩fica
plt.show()
````

## **5. Spearman (monot贸nico) vs Pearson (lineal)**

Tanto la correlaci贸n de Spearman como la de Pearson miden la relaci贸n entre dos variables, pero la correlaci贸n de Spearman mide la relaci贸n monot贸nica mientras que la de Pearson mide la relaci贸n lineal. Es decir, la correlaci贸n de Spearman es m谩s adecuada para medir relaciones entre variables no lineales o variables ordinales, mientras que la correlaci贸n de Pearson es m谩s adecuada para medir relaciones entre variables continuas lineales.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, pearsonr

# Creamos dos arrays de valores x e y con valores aleatorios
x = np.random.normal(size=50)
y = np.random.normal(size=50)

# Creamos una variable z que es igual a x^2
z = x**2

# Graficamos los puntos con colores y tama帽os de marcadores personalizados
plt.scatter(x, y, label='Datos', color='c', marker='o', s=60)

# Graficamos la l铆nea de la relaci贸n monot贸nica en un color diferente
x_ord = np.sort(x)
y_ord = y[np.argsort(x)]
plt.plot(x_ord, y_ord, color='r', label='L铆nea de relaci贸n')

# Etiquetamos los ejes y agregamos un t铆tulo
plt.xlabel("Variable Independiente (X)")
plt.ylabel("Variable Dependiente (Y)")
plt.title("Relaci贸n No Lineal entre X e Y")

# Agregamos una leyenda
plt.legend()

# Calculamos la correlaci贸n de Spearman
rho, pval = spearmanr(x, y)
print(f"Coeficiente de correlaci贸n de Spearman: {rho:.2f}")

# Calculamos la correlaci贸n de Pearson
r, pval = pearsonr(x, y)
print(f"Coeficiente de correlaci贸n de Pearson: {r:.2f}")

# Mostramos la gr谩fica
plt.grid(True)  # Agregamos una cuadr铆cula de fondo
plt.show()
````

## **6. F贸rmula matem谩tica**

La correlaci贸n de Spearman se utiliza para medir la relaci贸n monot贸nica (no necesariamente lineal) entre dos variables. La f贸rmula matem谩tica para la correlaci贸n de Spearman se calcula de la siguiente manera:

Supongamos que tenemos dos conjuntos de datos $(X, Y)$, y los rangos correspondientes de esos conjuntos se denotan como $(R_X, R_Y)$.

La f贸rmula para la correlaci贸n de Spearman $(\rho)$ es:

$\rho = 1 - \frac{6 \sum d_i^2}{n(n^2-1)}$

donde:

- $(d_i)$ es la diferencia entre los rangos de la variable $(X)$ y la variable Y para la i-茅sima observaci贸n.
- $(n)$ es el n煤mero total de observaciones.

Esta f贸rmula implica sumar las diferencias al cuadrado y luego utilizar esos valores en la f贸rmula para calcular la correlaci贸n de Spearman.

Es importante tener en cuenta que la correlaci贸n de Spearman produce valores en el rango $[-1, 1]$, donde:

- $(\rho = 1)$: Hay una correlaci贸n perfecta y positiva entre las variables.
- $(\rho = -1)$: Hay una correlaci贸n perfecta y negativa entre las variables.
- $(\rho = 0)$: No hay correlaci贸n.

Esta medida de correlaci贸n es 煤til cuando las variables no siguen una relaci贸n lineal y se busca evaluar la relaci贸n basada en el orden de los valores en lugar de sus magnitudes exactas.

La correlaci贸n de Pearson se utiliza para medir la fuerza y direcci贸n de una relaci贸n lineal entre dos variables continuas. La f贸rmula matem谩tica para la correlaci贸n de Pearson $(r)$ entre dos conjuntos de datos $(X, Y)$ con $(n)$ observaciones es la siguiente:

$r = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum_{i=1}^{n} (X_i - \bar{X})^2 \sum_{i=1}^{n} (Y_i - \bar{Y})^2}}$

Donde:

- $(X_i)$ y $(Y_i)$ son los valores individuales de las dos variables.
- $(\bar{X})$ y $(\bar{Y})$ son las medias de las variables $X$ e $Y$, respectivamente.

Esta f贸rmula implica restar las medias de los valores individuales, multiplicar esas diferencias y sumar esos productos. El resultado se divide por el producto de las desviaciones est谩ndar de ambas variables.

El coeficiente de correlaci贸n de Pearson, $(r)$, var铆a en el rango de -1 a 1, donde:

- $(r = 1)$: Hay una correlaci贸n perfecta y positiva entre las variables.
- $(r = -1)$: Hay una correlaci贸n perfecta y negativa entre las variables.
- $(r = 0)$: No hay correlaci贸n lineal.

Es importante tener en cuenta que la correlaci贸n de Pearson asume que las variables est谩n distribuidas de manera normal y que la relaci贸n entre ellas es lineal. Tambi茅n es sensible a los valores at铆picos.
