# **23. Acelera el procesamiento de datos con Funciones Universales (UFuncs) de Numpy**

## **Introducci贸n a las funciones universales**

Las Funciones Universales (UFuncs) de Numpy son funciones que se aplican a cada elemento de un arreglo de Numpy de forma r谩pida y eficiente. Estas funciones incluyen operaciones matem谩ticas b谩sicas, funciones trigonom茅tricas, funciones exponenciales y logar铆tmicas, entre otras.

```python
import numpy as np

# Crear un arreglo de Numpy unidimensional
arr = np.array([1, 2, 3, 4, 5])

# Aplicar la funci贸n cuadrado a cada elemento del arreglo
square_arr = np.square(arr)
print(square_arr)

# Aplicar la funci贸n exponencial a cada elemento del arreglo
exp_arr = np.exp(arr)
print(exp_arr)

# Aplicar la funci贸n seno a cada elemento del arreglo
sin_arr = np.sin(arr)
print(sin_arr)
```

## **Funciones universales b谩sicas: add, subtract, divide, multiple, power y absolute**

```python
import numpy as np

# Crear dos arreglos de Numpy unidimensionales
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 7, 8, 9, 10])

# Sumar los elementos de los dos arreglos
sum_arr = np.add(arr1, arr2)
print(sum_arr)

# Restar los elementos de los dos arreglos
sub_arr = np.subtract(arr2, arr1)
print(sub_arr)

# Dividir los elementos de los dos arreglos
div_arr = np.divide(arr1, arr2)
print(div_arr)

# Multiplicar los elementos de los dos arreglos
mul_arr = np.multiply(arr1, arr2)
print(mul_arr)

# Elevar los elementos de arr1 a los elementos de arr2
pow_arr = np.power(arr1, arr2)
print(pow_arr)

# Obtener el valor absoluto de los elementos de arr1
abs_arr = np.absolute(arr1)
print(abs_arr)
```

## **Ejemplo de funci贸n universal para el c谩lculo del 铆ndice de la masa corporal**

```python
import numpy as np

# Crear un arreglo de Numpy bidimensional con datos de peso (en kg) y altura (en m)
datos = np.array([[70, 1.80],
                  [80, 1.75],
                  [65, 1.68]])

# Obtener los pesos y alturas por separado
pesos = datos[:, 0]
alturas = datos[:, 1]

# Calcular el 铆ndice de masa corporal (IMC)
imc = np.divide(pesos, np.power(alturas, 2))
print(imc)
```
