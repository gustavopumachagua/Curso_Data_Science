# **29. Distribución Normal (o Gaussiana) y la Regla Empírica 68-95-99.7**

![distribucion normal](../imagenes%20Estadisticas/distribucion_normal.jpg "gaussiano")

## **1. Introducción a la distribución normal**

La distribución normal, también conocida como distribución de Gauss, es una distribución de probabilidad continua que es ampliamente utilizada en estadística debido a que muchos fenómenos naturales y sociales se distribuyen de manera aproximadamente normal. La distribución normal se caracteriza por su media (μ) y su desviación estándar (σ), y tiene una forma de campana simétrica alrededor de su media.

En Python, se puede generar una distribución normal utilizando la librería NumPy y la función ```numpy.random.normal()```. A continuación se muestra un ejemplo de cómo generar una distribución normal con media 0 y desviación estándar 1 y graficar su densidad de probabilidad utilizando la librería Matplotlib:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de una distribución normal con media 0 y desviación estándar 1
mu, sigma = 0, 1
datos = np.random.normal(mu, sigma, 1000)

# Graficar la densidad de probabilidad de la distribución normal
count, bins, ignored = plt.hist(datos, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
````

Este código generará una distribución normal con media 0 y desviación estándar 1 utilizando la función ```numpy.random.normal()```, y luego graficará su densidad de probabilidad utilizando la función ```plt.hist()``` y la función ```plt.plot()```. La densidad de probabilidad se grafica utilizando la fórmula matemática de la distribución normal, que es ```1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) )```. El resultado será una gráfica de campana simétrica alrededor de la media 0.

## **2. Ejemplos de variables distribuidas normalmente**

* Ejemplo 1:

Generar datos aleatorios distribuidos normalmente y visualizar su distribución
Para generar datos aleatorios distribuidos normalmente se puede utilizar la función ```numpy.random.normal()```. A continuación se presenta un ejemplo de cómo generar 1000 datos aleatorios con una media de 50 y una desviación estándar de 10:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios distribuidos normalmente
mu, sigma = 50, 10 # media y desviación estándar
datos = np.random.normal(mu, sigma, 1000)

# Visualizar la distribución de los datos
plt.hist(datos, bins=50, density=True, alpha=0.5, color='blue')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Distribución normal')
plt.show()
````

* Ejemplo 2:

 Generar datos aleatorios distribuidos normalmente y visualizar su distribución utilizando seaborn
Otra forma de visualizar la distribución normal de los datos es utilizando la librería seaborn. A continuación se presenta un ejemplo de cómo generar 1000 datos aleatorios con una media de 0 y una desviación estándar de 1 y visualizar su distribución utilizando un gráfico de densidad:

````python
import numpy as np
import seaborn as sns

# Generar datos aleatorios distribuidos normalmente
mu, sigma = 0, 1 # media y desviación estándar
datos = np.random.normal(mu, sigma, 1000)

# Visualizar la distribución de los datos utilizando seaborn
sns.kdeplot(datos, shade=True)
````

* Ejemplo 3:

 Generar datos aleatorios con varias distribuciones normales y visualizar su distribución
También es posible generar datos aleatorios con varias distribuciones normales utilizando la función ```numpy.random.normal()``` y luego visualizar su distribución utilizando la librería seaborn. A continuación se presenta un ejemplo de cómo generar 1000 datos aleatorios con tres distribuciones normales diferentes y visualizar su distribución utilizando un gráfico de densidad:

````python
import numpy as np
import seaborn as sns

# Generar datos aleatorios distribuidos normalmente con tres distribuciones diferentes
datos1 = np.random.normal(0, 1, 500)
datos2 = np.random.normal(5, 2, 300)
datos3 = np.random.normal(10, 3, 200)
datos = np.concatenate((datos1, datos2, datos3))

# Visualizar la distribución de los datos utilizando seaborn
sns.kdeplot(datos, shade=True)
````

## **3. Explicación Distribución Normal**

La distribución normal, también conocida como distribución gaussiana, es una distribución de probabilidad continua que se utiliza en muchos campos de la estadística y las ciencias naturales. Esta distribución es simétrica alrededor de su media y su forma es una campana.

La distribución normal se define por dos parámetros: su media ($\mu$) y su desviación estándar ($\sigma$). La media es el punto central de la distribución y la desviación estándar mide la dispersión de los datos alrededor de la media.

En Python, podemos generar una distribución normal utilizando la librería ```numpy``` y la función ```numpy.random.normal()```. En el siguiente ejemplo, generaremos una muestra aleatoria de datos normalmente distribuidos con una media de 0 y una desviación estándar de 1:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra aleatoria de datos normalmente distribuidos
mu, sigma = 0, 1 # media y desviación estándar
datos = np.random.normal(mu, sigma, 1000)

# Graficar el histograma de la muestra
count, bins, ignored = plt.hist(datos, 30, density=True)

# Graficar la curva de la distribución normal
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')

# Mostrar la gráfica
plt.show()
````

En este ejemplo, la muestra de datos se genera utilizando la función ```numpy.random.normal()```. Luego, se grafica un histograma de la muestra y se superpone una curva de la distribución normal utilizando la fórmula matemática de la distribución normal. Finalmente, se muestra la gráfica resultante.

El resultado será una gráfica con la forma típica de una distribución normal: una curva en forma de campana, centrada en la media y con una dispersión determinada por la desviación estándar.

## **4. Regla Empírica 68-95-99.7**

La regla empírica 68-95-99.7, también conocida como la regla de tres sigmas, es una regla empírica que se aplica a una distribución normal. La regla establece que el 68% de los datos se encuentran dentro de una desviación estándar de la media, el 95% de los datos se encuentran dentro de dos desviaciones estándar de la media, y el 99.7% de los datos se encuentran dentro de tres desviaciones estándar de la media.

A continuación se muestra un ejemplo de cómo se puede aplicar la regla empírica en Python utilizando la librería ```numpy``` para generar datos aleatorios distribuidos normalmente y la librería ```matplotlib``` para graficar la distribución y las áreas bajo la curva correspondientes a cada intervalo de la regla empírica:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios distribuidos normalmente con media de 0 y desviación estándar de 1
datos = np.random.normal(0, 1, 100000)

# Calcular la media y la desviación estándar de los datos
media = np.mean(datos)
desv_std = np.std(datos)

# Calcular los límites de los intervalos de la regla empírica
lim_inf_1 = media - desv_std
lim_sup_1 = media + desv_std

lim_inf_2 = media - 2 * desv_std
lim_sup_2 = media + 2 * desv_std

lim_inf_3 = media - 3 * desv_std
lim_sup_3 = media + 3 * desv_std

# Graficar la distribución de los datos y los intervalos de la regla empírica
plt.hist(datos, bins=50, density=True, alpha=0.6)
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 100)
plt.plot(x, 1/(desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2/(2 * desv_std**2)), color='black')
plt.axvline(lim_inf_1, color='red', linestyle='--')
plt.axvline(lim_sup_1, color='red', linestyle='--')
plt.axvline(lim_inf_2, color='green', linestyle='--')
plt.axvline(lim_sup_2, color='green', linestyle='--')
plt.axvline(lim_inf_3, color='purple', linestyle='--')
plt.axvline(lim_sup_3, color='purple', linestyle='--')

# Agregar leyenda y título al gráfico
plt.legend(['Distribución normal', '1 desv. estándar', '2 desv. estándar', '3 desv. estándar'])
plt.title('Regla empírica 68-95-99.7')
plt.show()
````

## **5. Ejemplificación de la Regla Empírica**

La Regla Empírica, también conocida como la regla de 68-95-99.7, es una regla empírica que describe cómo se distribuyen los datos en una distribución normal. Esta regla establece que alrededor del 68% de los datos se encuentran dentro de una desviación estándar de la media, el 95% de los datos se encuentran dentro de dos desviaciones estándar de la media, y el 99.7% de los datos se encuentran dentro de tres desviaciones estándar de la media.

A continuación se presenta un ejemplo en Python que muestra cómo usar la Regla Empírica para graficar una distribución normal y resaltar el porcentaje de datos que se encuentran dentro de cada una de las desviaciones estándar.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de una distribución normal
mu = 0
sigma = 1
x = np.random.normal(mu, sigma, size=10000)

# Graficar la distribución normal
count, bins, ignored = plt.hist(x, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
         np.exp(- (bins - mu)**2 / (2 * sigma**2)), linewidth=2, color='r')

# Calcular los porcentajes de datos dentro de cada desviación estándar
within_1std = np.sum((x >= mu - sigma) & (x <= mu + sigma)) / len(x) * 100
within_2std = np.sum((x >= mu - 2*sigma) & (x <= mu + 2*sigma)) / len(x) * 100
within_3std = np.sum((x >= mu - 3*sigma) & (x <= mu + 3*sigma)) / len(x) * 100

# Mostrar los porcentajes en la gráfica
plt.text(0.5, 0.9, f'68% de los datos\n(dentro de 1 desviación estándar)', ha='center', va='center', transform=plt.gca().transAxes)
plt.text(0.5, 0.7, f'95% de los datos\n(dentro de 2 desviaciones estándar)', ha='center', va='center', transform=plt.gca().transAxes)
plt.text(0.5, 0.5, f'99.7% de los datos\n(dentro de 3 desviaciones estándar)', ha='center', va='center', transform=plt.gca().transAxes)

plt.show()
````

Este código genera una muestra de 10000 datos de una distribución normal con media 0 y desviación estándar 1, y luego grafica la distribución normal utilizando el histograma y la curva de densidad. Luego, utiliza las máscaras de NumPy para contar los datos que se encuentran dentro de cada una de las desviaciones estándar y calcula los porcentajes correspondientes. Finalmente, muestra los porcentajes en la gráfica utilizando la función text de Matplotlib. El resultado es una gráfica que muestra la distribución normal y resalta el porcentaje de datos que se encuentran dentro de cada una de las desviaciones estándar de la media.
