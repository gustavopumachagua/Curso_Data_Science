# **11. Distribución de Datos e Histogramas**

## **Introducción a la Distribución de Datos e Histogramas**

La distribución de datos se refiere a cómo se distribuyen los valores de una variable en un conjunto de datos. Un histograma es una forma común de visualizar la distribución de datos. En un histograma, los datos se dividen en intervalos y se cuenta cuántos valores se encuentran en cada intervalo. Luego, se representa gráficamente el número de valores en cada intervalo en forma de barras.

Aquí hay un ejemplo de código y gráfica de histograma en Python:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios con distribución normal
datos = np.random.normal(loc=50, scale=10, size=1000)

# Graficar histograma con 30 intervalos
plt.hist(datos, bins=30)

# Agregar etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de datos aleatorios')

# Mostrar gráfica
plt.show()
```

## **Creación de Histogramas**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
datos = np.random.normal(loc=50, scale=10, size=1000)

# Crear histograma
plt.hist(datos, bins=30, alpha=0.5)

# Añadir etiquetas y título
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de datos aleatorios')

# Mostrar el histograma
plt.show()
```

Explicación del código:

* Importamos las librerías necesarias: numpy y matplotlib.pyplot.
* Generamos datos aleatorios utilizando la función np.random.normal, que nos permite generar datos a partir de una distribución normal.
* Creamos el histograma utilizando la función plt.hist, que toma como argumentos los datos y el número de bins que queremos utilizar para el histograma.
* Añadimos etiquetas al eje x e y, y un título al histograma utilizando las funciones plt.xlabel, plt.ylabel y plt.title.
* Finalmente, mostramos el histograma utilizando la función plt.show.

## **Datos aleatorios con cierta distribución**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar 1000 datos aleatorios con distribución normal
datos = np.random.normal(loc=50, scale=10, size=1000)

# Graficar histograma de los datos
plt.hist(datos, bins=30, alpha=0.5)
plt.title('Histograma de datos aleatorios con distribución normal')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

Este código generará 1000 datos aleatorios con una media de 50 y una desviación estándar de 10. Luego, se grafica un histograma de los datos usando 30 bins y una opacidad de 0.5. El título, las etiquetas del eje x y del eje y se agregan para hacer la gráfica más legible. El resultado será una gráfica que muestra la distribución de los datos generados con una distribución normal.

## **Distribución Normal (Gaussiana)**

La distribución normal, también conocida como distribución gaussiana, es una distribución de probabilidad continua que se utiliza en numerosas aplicaciones de estadística y análisis de datos debido a sus propiedades matemáticas y estadísticas. La distribución normal es simétrica y se describe completamente por su media (promedio) y desviación estándar.

El siguiente código en Python utiliza NumPy para generar datos aleatorios con una distribución normal y Matplotlib para graficarlos en un histograma:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios con distribución normal
mu = 0 # media
sigma = 1 # desviación estándar
datos = np.random.normal(mu, sigma, size=10000)

# Graficar histograma
plt.hist(datos, bins=50, density=True, alpha=0.7)

# Graficar función de densidad de probabilidad
x = np.linspace(-5, 5, 1000)
pdf = 1/(sigma * np.sqrt(2*np.pi)) * np.exp(-0.5*(x-mu)**2/sigma**2)
plt.plot(x, pdf, linewidth=2)

# Mostrar la gráfica
plt.show()
```

## **Distribución Normal, Uniforme y Exponencial**

La distribución normal, uniforme y exponencial son tres distribuciones de probabilidad comunes en estadística. Cada una tiene una forma diferente de distribución y se utiliza en diferentes situaciones.

* La distribución normal (también conocida como distribución gaussiana) es una distribución continua que describe la distribución de datos en una campana simétrica alrededor de la media. Esta distribución se utiliza en muchas áreas de la ciencia y la ingeniería para modelar datos que tienen una distribución aproximadamente normal.

* La distribución uniforme es una distribución continua en la que cada valor posible de la variable aleatoria tiene la misma probabilidad de ocurrir. Es una distribución simple y se utiliza a menudo para simular eventos aleatorios equitativos.

* La distribución exponencial es una distribución continua que describe la cantidad de tiempo que debe transcurrir antes de que ocurra un evento. Se utiliza en muchas aplicaciones, como en análisis de fiabilidad, donde se busca determinar la probabilidad de que una máquina falle después de un cierto número de horas de operación.

Aquí hay un ejemplo de código y gráfica que muestra estas tres distribuciones usando Numpy y Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos con distribución normal
datos_normales = np.random.normal(loc=50, scale=10, size=1000)

# Generar datos con distribución uniforme
datos_uniformes = np.random.uniform(low=0, high=100, size=1000)

# Generar datos con distribución exponencial
datos_exponenciales = np.random.exponential(scale=10, size=1000)

# Graficar histogramas de cada distribución
plt.hist(datos_normales, bins=30, alpha=0.5, label='Normal')
plt.hist(datos_uniformes, bins=30, alpha=0.5, label='Uniforme')
plt.hist(datos_exponenciales, bins=30, alpha=0.5, label='Exponencial')

# Agregar leyenda y etiquetas de eje
plt.legend(loc='upper right')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')

# Mostrar la gráfica
plt.show()
```
