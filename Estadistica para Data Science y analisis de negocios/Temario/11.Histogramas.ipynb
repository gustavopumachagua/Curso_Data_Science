## **Distribuci贸n Normal (Gaussiana)**

La distribuci贸n normal, tambi茅n conocida como distribuci贸n gaussiana, es una distribuci贸n de probabilidad continua que se utiliza en numerosas aplicaciones de estad铆stica y an谩lisis de datos debido a sus propiedades matem谩ticas y estad铆sticas. La distribuci贸n normal es sim茅trica y se describe completamente por su media (promedio) y desviaci贸n est谩ndar.

El siguiente c贸digo en Python utiliza NumPy para generar datos aleatorios con una distribuci贸n normal y Matplotlib para graficarlos en un histograma:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios con distribuci贸n normal
mu = 0 # media
sigma = 1 # desviaci贸n est谩ndar
datos = np.random.normal(mu, sigma, size=10000)

# Graficar histograma
plt.hist(datos, bins=50, density=True, alpha=0.7)

# Graficar funci贸n de densidad de probabilidad
x = np.linspace(-5, 5, 1000)
pdf = 1/(sigma * np.sqrt(2*np.pi)) * np.exp(-0.5*(x-mu)**2/sigma**2)
plt.plot(x, pdf, linewidth=2)

# Mostrar la gr谩fica
plt.show()
```

## **Distribuci贸n Normal, Uniforme y Exponencial**

La distribuci贸n normal, uniforme y exponencial son tres distribuciones de probabilidad comunes en estad铆stica. Cada una tiene una forma diferente de distribuci贸n y se utiliza en diferentes situaciones.

* La distribuci贸n normal (tambi茅n conocida como distribuci贸n gaussiana) es una distribuci贸n continua que describe la distribuci贸n de datos en una campana sim茅trica alrededor de la media. Esta distribuci贸n se utiliza en muchas 谩reas de la ciencia y la ingenier铆a para modelar datos que tienen una distribuci贸n aproximadamente normal.

* La distribuci贸n uniforme es una distribuci贸n continua en la que cada valor posible de la variable aleatoria tiene la misma probabilidad de ocurrir. Es una distribuci贸n simple y se utiliza a menudo para simular eventos aleatorios equitativos.

* La distribuci贸n exponencial es una distribuci贸n continua que describe la cantidad de tiempo que debe transcurrir antes de que ocurra un evento. Se utiliza en muchas aplicaciones, como en an谩lisis de fiabilidad, donde se busca determinar la probabilidad de que una m谩quina falle despu茅s de un cierto n煤mero de horas de operaci贸n.

Aqu铆 hay un ejemplo de c贸digo y gr谩fica que muestra estas tres distribuciones usando Numpy y Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos con distribuci贸n normal
datos_normales = np.random.normal(loc=50, scale=10, size=1000)

# Generar datos con distribuci贸n uniforme
datos_uniformes = np.random.uniform(low=0, high=100, size=1000)

# Generar datos con distribuci贸n exponencial
datos_exponenciales = np.random.exponential(scale=10, size=1000)

# Graficar histogramas de cada distribuci贸n
plt.hist(datos_normales, bins=30, alpha=0.5, label='Normal')
plt.hist(datos_uniformes, bins=30, alpha=0.5, label='Uniforme')
plt.hist(datos_exponenciales, bins=30, alpha=0.5, label='Exponencial')

# Agregar leyenda y etiquetas de eje
plt.legend(loc='upper right')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')

# Mostrar la gr谩fica
plt.show()
```
