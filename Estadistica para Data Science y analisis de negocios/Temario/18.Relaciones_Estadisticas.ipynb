# **18. Explora y Visualiza Relaciones Estad√≠sticas con Seaborn de Python**

## **Introducci√≥n a la visualizaci√≥n de datos**

La visualizaci√≥n de datos es una herramienta muy importante en el an√°lisis de datos, ya que permite mostrar de manera gr√°fica la informaci√≥n de una manera m√°s f√°cil de interpretar y comprender. En el contexto de la programaci√≥n con Python, existen varias librer√≠as para la visualizaci√≥n de datos, como por ejemplo Matplotlib, Seaborn y Plotly.

Matplotlib es una de las librer√≠as m√°s populares para la visualizaci√≥n de datos en Python, y es muy flexible y potente. Con Matplotlib, se pueden crear gr√°ficos de l√≠neas, barras, histogramas, gr√°ficos de dispersi√≥n, entre otros.

Seaborn es otra librer√≠a para la visualizaci√≥n de datos en Python, que se basa en Matplotlib. Seaborn se enfoca en gr√°ficos estad√≠sticos y es especialmente √∫til para la visualizaci√≥n de grandes conjuntos de datos.

Plotly es una librer√≠a para la visualizaci√≥n de datos en Python que permite crear gr√°ficos interactivos y din√°micos. Con Plotly, se pueden crear gr√°ficos 3D, mapas y gr√°ficos animados.

En general, la visualizaci√≥n de datos es una herramienta muy importante en el an√°lisis de datos, ya que permite obtener una mejor comprensi√≥n de la informaci√≥n que se est√° analizando. Con las librer√≠as de visualizaci√≥n de datos en Python, es posible crear gr√°ficos de alta calidad y personalizables para mostrar la informaci√≥n de la manera m√°s efectiva posible.

## **Tipos de gr√°ficas en Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un dataset de ejemplo
tips = sns.load_dataset("tips")

# Personalizar el gr√°fico de barras
plt.figure(figsize=(8, 6))  # Tama√±o de la figura
sns.barplot(x="day", y="total_bill", data=tips, errorbar=None, palette="husl")  # Quita los intervalos de confianza y personaliza la paleta de colores
plt.title('Total Bill by Day')  # T√≠tulo
plt.xlabel('Day of the Week')  # Etiqueta del eje X
plt.ylabel('Total Bill')  # Etiqueta del eje Y
plt.xticks(rotation=45)  # Rotar las etiquetas del eje X para una mejor visualizaci√≥n
plt.show()
```

```python
# Personalizar el gr√°fico de l√≠neas
plt.figure(figsize=(8, 6))
sns.lineplot(x="total_bill", y="tip", data=tips, color='green', marker='o', markersize=8, linestyle='--')
plt.title('Tip vs Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# Personalizar el gr√°fico de dispersi√≥n
plt.figure(figsize=(8, 6))
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="time", s=100)
plt.title('Tip vs Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python

# Personalizar el gr√°fico de viol√≠n
plt.figure(figsize=(8, 6))
sns.violinplot(x="day", y="total_bill", data=tips, palette="Set2")
plt.title('Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
```

```python
import warnings

# Desactivar las advertencias de disposici√≥n autom√°tica
warnings.filterwarnings("ignore", category=UserWarning)

# Crear la figura y el gr√°fico
plt.figure(figsize=(8, 6))
sns.displot(tips["total_bill"], kind="kde", fill=True, color="purple")
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.show()
```

```python
import numpy as np
# Filtrar solo las columnas num√©ricas para el c√°lculo de correlaci√≥n
numeric_columns = tips.select_dtypes(include=[np.number])

# Personalizar el gr√°fico de correlaci√≥n
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_columns.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

## **Datos a visualizar**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Crear un dataframe de ejemplo
data = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [2, 4, 6, 8, 10],
    "z": [1, 3, 5, 7, 9]
})

# Personalizar el gr√°fico de barras
sns.barplot(x="x", y="y", data=data, color='c')  # Cambiar el color de las barras
plt.title('Gr√°fico de Barras Personalizado')  # T√≠tulo
plt.xlabel('Eje X Personalizado')  # Etiqueta del eje X
plt.ylabel('Eje Y Personalizado')  # Etiqueta del eje Y
plt.show()
```

```python
# Personalizar el gr√°fico de l√≠neas
sns.lineplot(x="x", y="y", data=data, color='g', linestyle='--')  # Cambiar color y estilo de l√≠nea
plt.title('Gr√°fico de L√≠neas Personalizado')
plt.xlabel('Eje X Personalizado')
plt.ylabel('Eje Y Personalizado')
plt.show()
```

```python
# Personalizar el gr√°fico de dispersi√≥n
sns.scatterplot(x="x", y="y", data=data, color='m', s=100)  # Cambiar color y tama√±o de los puntos
plt.title('Gr√°fico de Dispersi√≥n Personalizado')
plt.xlabel('Eje X Personalizado')
plt.ylabel('Eje Y Personalizado')
plt.show()
```

```python
# Personalizar el gr√°fico de viol√≠n
sns.violinplot(data=data, palette="Set2")  # Cambiar la paleta de colores
plt.title('Gr√°fico de Viol√≠n Personalizado')
plt.xlabel('Eje X Personalizado')
plt.ylabel('Eje Y Personalizado')
plt.show()
```

```python
# Personalizar el gr√°fico de distribuci√≥n
sns.kdeplot(data["x"], color='r', fill=True)  # Cambiar color y agregar sombreado
plt.title('Gr√°fico de Distribuci√≥n Personalizado')
plt.xlabel('Eje X Personalizado')
plt.ylabel('Eje Y Personalizado')
plt.show()
```

```python
# Personalizar el mapa de calor de correlaci√≥n
sns.heatmap(data.corr(), annot=True, cmap="YlGnBu")  # Cambiar la paleta de colores
plt.title('Mapa de Calor de Correlaci√≥n Personalizado')
plt.show()
```

## **Ejemplos: gr√°ficas de l√≠neas y dispersi√≥n**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Crear un dataframe de ejemplo
data = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [2, 4, 6, 8, 10],
    "z": [1, 3, 5, 7, 9]
})
# Personalizar el gr√°fico de l√≠neas
sns.lineplot(x="x", y="y", data=data, color='blue', marker='o', linestyle='--', markersize=8)
plt.title('Gr√°fico de L√≠neas Personalizado')
plt.xlabel('Eje X Personalizado')
plt.ylabel('Eje Y Personalizado')
plt.show()
```

```python
# Personalizar el gr√°fico de dispersi√≥n
sns.scatterplot(x="z", y="y", data=data, hue=data['x'], palette='coolwarm', s=100, alpha=0.7)
plt.title('Gr√°fico de Dispersi√≥n Personalizado')
plt.xlabel('Eje X Personalizado')
plt.ylabel('Eje Y Personalizado')
plt.legend(title='Categor√≠a X', labels=data['x'])
plt.show()
```

## **Gr√°ficas de dispersi√≥n**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Crear un dataframe de ejemplo
data = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [2, 4, 6, 8, 10],
    "z": [1, 3, 5, 7, 9]
})

# Personalizar el gr√°fico de dispersi√≥n
sns.scatterplot(x="x", y="y", data=data, hue=data["z"], palette="viridis", size=data["z"], sizes=(50, 200), alpha=0.7)
plt.title('Gr√°fico de Dispersi√≥n Personalizado')
plt.xlabel('Eje X Personalizado')
plt.ylabel('Eje Y Personalizado')
plt.legend(title='Tama√±o Z', labels=data['z'])
plt.show()
```

## **Gr√°ficas de l√≠neas**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Crear un dataframe de ejemplo
data = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [2, 4, 6, 8, 10]
})

# Personalizar el gr√°fico de l√≠neas
sns.lineplot(x="x", y="y", data=data, marker="o", markersize=8, color="red", label="Datos de ejemplo")
plt.title('Gr√°fico de L√≠neas Personalizado')
plt.xlabel('Eje X Personalizado')
plt.ylabel('Eje Y Personalizado')
plt.legend()
plt.grid(True)  # Agregar una cuadr√≠cula al gr√°fico
plt.show()
```

## **Intervalo de confianza**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Crear un dataframe de ejemplo
data = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [2, 4, 6, 8, 10]
})

# Calcular el intervalo de confianza
ci = sns.utils.ci(data["y"])

# Crear el gr√°fico de l√≠neas
sns.lineplot(x="x", y="y", data=data, marker="o", markersize=8, color="blue", label="Datos de ejemplo")

# Rellenar el √°rea del intervalo de confianza
plt.fill_between(data["x"], ci[0], ci[1], alpha=0.2, color="blue")

# Personalizar el t√≠tulo y etiquetas de los ejes
plt.title('Gr√°fico de L√≠neas con Intervalo de Confianza')
plt.xlabel('Eje X Personalizado')
plt.ylabel('Eje Y Personalizado')

# Agregar leyenda
plt.legend()

# Habilitar la cuadr√≠cula
plt.grid(True)

# Mostrar el gr√°fico
plt.show()
```

## **Combinar gr√°ficas**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Crear un dataframe de ejemplo
data = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y1": [2, 4, 6, 8, 10],
    "y2": [1, 3, 5, 7, 9]
})

# Crear dos subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Gr√°fico de l√≠neas en el primer subplot
sns.lineplot(x="x", y="y1", data=data, ax=ax1, marker="o", markersize=8, color="blue", label="L√≠nea 1")
ax1.set_title('Gr√°fico de L√≠neas')
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Eje Y')
ax1.grid(True)
ax1.legend()

# Gr√°fico de dispersi√≥n en el segundo subplot
sns.scatterplot(x="x", y="y2", data=data, ax=ax2, color="red", label="Dispersi√≥n")
ax2.set_title('Gr√°fico de Dispersi√≥n')
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Eje Y')
ax2.grid(True)
ax2.legend()

# Ajustar el espacio entre los subplots
plt.tight_layout()

# Mostrar los subplots
plt.show()
```
