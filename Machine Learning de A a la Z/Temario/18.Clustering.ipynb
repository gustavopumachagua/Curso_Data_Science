# **18. Tipos de Distancias para Técnicas de Clustering (Agrupamiento): Euclidiana, Chebyshev y Manhattan**

## **Introducción**

En las técnicas de clustering (o agrupamiento), las distancias se utilizan para medir la similitud o la diferencia entre los puntos de datos. Algunos de los tipos de distancias comunes utilizados en el clustering son los siguientes:

* **Distancia Euclidiana:** es la distancia recta entre dos puntos en un espacio n-dimensional. La fórmula para la distancia euclidiana entre dos puntos en un espacio 2D es:

$d(p,q) = \sqrt{(q_{1}-p_{1})^{2} + (q_{2}-p_{2})^{2}}$

* **Distancia Manhattan:** también se conoce como distancia de la ciudad o distancia de la taxista, mide la distancia entre dos puntos como la suma de las diferencias absolutas de sus coordenadas. La fórmula para la distancia Manhattan entre dos puntos en un espacio 2D es:

$d(p,q) = |q_{1}-p_{1}| + |q_{2}-p_{2}|$

* **Distancia Chebyshev:** también se conoce como la distancia del ajedrez, mide la distancia máxima entre las coordenadas de dos puntos en un espacio n-dimensional. La fórmula para la distancia Chebyshev entre dos puntos en un espacio 2D es:

$d(p,q) = max(|q_{1}-p_{1}|, |q_{2}-p_{2}|)$

Estos tipos de distancias se utilizan comúnmente en las técnicas de `clustering` para medir la similitud entre los puntos de datos y formar grupos. Dependiendo del tipo de datos y de la aplicación, se puede elegir una distancia específica para lograr mejores resultados en el `clustering`.

## **Distancia Euclidiana**

La distancia euclidiana es una medida de distancia común utilizada en el análisis de datos y el aprendizaje automático. Es la distancia más corta entre dos puntos en un espacio euclidiano.

En un espacio euclidiano de dos dimensiones, la distancia euclidiana entre dos puntos se puede calcular utilizando la siguiente fórmula:


$d(euclidiana) = \sqrt{(x_{2}-x_{1})^{2} + (y_{2}-y_{1})^{2}}$

donde $(x_1, y_1)$ y $(x_2, y_2)$ son las coordenadas de los dos puntos.

En Python, podemos calcular la distancia euclidiana utilizando la función distance del módulo `scipy.spatial`. Aquí hay un ejemplo de cómo se puede calcular la distancia euclidiana entre dos puntos:

![euclidiana](../imagenes%20Machine_Learning/euclidiana.png "euclidiana")

````python
import numpy as np
import pandas as pd

# Crear un dataframe con algunos puntos
df = pd.DataFrame({
    'x': [2, 4, 5, 7, 8],
    'y': [6, 4, 7, 1, 3]
})

# Definir dos puntos para calcular la distancia euclidiana
p1 = np.array([2, 6])
p2 = np.array([7, 1])

# Calcular la distancia euclidiana utilizando la función norm de NumPy
distance = np.linalg.norm(p1 - p2)

# Imprimir la distancia
print(distance)
````

````python
import matplotlib.pyplot as plt

# Graficar los puntos
plt.scatter(df['x'], df['y'])

# Graficar los puntos de inicio y final para la distancia euclidiana
plt.scatter(p1[0], p1[1], color='red')
plt.scatter(p2[0], p2[1], color='red')

# Graficar la línea que conecta los puntos
plt.plot([p1[0], p2[0]], [p1[1], p2[1]], color='red')

# Agregar etiquetas
plt.title("Distancia Euclidiana")
plt.xlabel("Eje x")
plt.ylabel("Eje y")

# Mostrar la gráfica
plt.show()
````

## **Distancia Chebyshev**

La distancia Chebyshev es una medida de distancia entre dos puntos en un espacio euclidiano. Esta distancia se define como la distancia máxima entre las coordenadas de los dos puntos en cada dimensión.

En otras palabras, si tenemos dos puntos en un espacio euclidiano con coordenadas `(x1, y1)` y `(x2, y2)`, la distancia Chebyshev entre ellos se define como:

$d = max(|x1 - x2|, |y1 - y2|)$

**donde** `||` representa el valor absoluto.

Aquí hay un ejemplo de cómo calcular la distancia Chebyshev entre dos puntos usando `Pandas` y `NumPy`:

![Chebyshev](../imagenes%20Machine_Learning/Chebyshev.png "chebysev")

![chebyshev](../imagenes%20Machine_Learning/Minkowsky.webp "chebyshev")

````python
import pandas as pd
import numpy as np

# Definir dos puntos en un espacio 2D
p1 = pd.Series([1, 2])
p2 = pd.Series([4, 6])

# Calcular la distancia Chebyshev usando NumPy
dist = np.max(np.abs(p1 - p2))

print("La distancia Chebyshev entre p1 y p2 es:", dist)
````

Aquí hay un ejemplo de cómo visualizar la distancia Chebyshev en un DataFrame:

````python
import pandas as pd
import seaborn as sns
import numpy as np

# Crear un DataFrame de ejemplo
data = pd.DataFrame({
    'x1': [1, 2, 3],
    'x2': [4, 5, 6],
    'y1': [7, 8, 9],
    'y2': [10, 11, 12]
})

# Calcular la matriz de distancias Chebyshev usando NumPy
distances = np.zeros((len(data), len(data)))
for i, row1 in data.iterrows():
    for j, row2 in data.iterrows():
        distances[i, j] = np.max(np.abs(row1 - row2))

# Visualizar la matriz de distancias usando un mapa de calor de Seaborn
sns.heatmap(distances, annot=True, cmap='viridis')
````

## **Distancia Manhattan**

La distancia Manhattan (también conocida como distancia de la ciudad) es otra medida de distancia comúnmente utilizada en el clustering y otras técnicas de machine learning. La distancia Manhattan entre dos puntos se define como la suma de las diferencias absolutas de sus coordenadas.

Por ejemplo, la distancia Manhattan entre dos puntos en un espacio 2D, `(x1, y1)` y `(x2, y2)`, se calcula como:

Manhattan $distance = |x1 - x2| + |y1 - y2|$

El siguiente código muestra cómo calcular y visualizar la distancia Manhattan entre puntos de un dataframe usando Python y `Matplotlib`:

![Manhattan](../imagenes%20Machine_Learning/Manhattan.webp "Manhattan")

````python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un dataframe de ejemplo
data = {'x': [2, 3, 5, 6, 7, 8],
        'y': [4, 5, 7, 8, 9, 10]}
df = pd.DataFrame(data)

# Calcular la distancia Manhattan entre todos los pares de puntos
manhattan_dist = []
for i in range(len(df)):
    row = []
    for j in range(len(df)):
        manhattan = abs(df.iloc[i]['x'] - df.iloc[j]['x']) + abs(df.iloc[i]['y'] - df.iloc[j]['y'])
        row.append(manhattan)
    manhattan_dist.append(row)

# Visualizar la matriz de distancia Manhattan como un heatmap
fig, ax = plt.subplots()
im = ax.imshow(manhattan_dist)

# Agregar etiquetas de eje y título
ax.set_xticks(range(len(df)))
ax.set_yticks(range(len(df)))
ax.set_xticklabels(df.index)
ax.set_yticklabels(df.index)
ax.set_title("Matriz de distancia Manhattan")

# Agregar las anotaciones de los valores de la matriz de distancia
for i in range(len(df)):
    for j in range(len(df)):
        text = ax.text(j, i, manhattan_dist[i][j],
                       ha="center", va="center", color="w")

plt.show()
````

## **Comparación de las métricas usando DBSCAN**

Claro, aquí te muestro un ejemplo de cómo comparar las métricas de `DBSCAN` utilizando la distancia euclidiana, la distancia Manhattan y la distancia Chebyshev en un conjunto de datos de muestra:

Primero, importamos las bibliotecas necesarias y generamos un conjunto de datos de muestra:

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Generar datos de muestra
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
````

A continuación, realizamos el agrupamiento utilizando ```DBSCAN``` con las tres métricas de distancia y calculamos las métricas de evaluación:

````python
# Función para realizar el clustering y calcular las métricas
def cluster_and_evaluate(X, metric):
    dbscan = DBSCAN(eps=0.5, min_samples=5, metric=metric)
    y_pred = dbscan.fit_predict(X)
    n_clusters = len(set(y_pred)) - (1 if -1 in y_pred else 0)
    silhouette = silhouette_score(X, y_pred)
    chs = calinski_harabasz_score(X, y_pred)
    db = davies_bouldin_score(X, y_pred)
    return n_clusters, silhouette, chs, db

# Realizar clustering con la distancia euclidiana
n_clusters_e, silhouette_e, chs_e, db_e = cluster_and_evaluate(X, 'euclidean')

# Realizar clustering con la distancia de Manhattan
n_clusters_m, silhouette_m, chs_m, db_m = cluster_and_evaluate(X, 'manhattan')

# Realizar clustering con la distancia de Chebyshev
n_clusters_c, silhouette_c, chs_c, db_c = cluster_and_evaluate(X, 'chebyshev')
````

Finalmente, graficamos los resultados y comparamos las métricas:

````python
# Graficar los resultados
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.scatter(X[:, 0], X[:, 1], c=y_true)
plt.title("Etiquetas Verdaderas")
plt.subplot(132)
plt.scatter(X[:, 0], X[:, 1], c=DBSCAN(eps=0.5, min_samples=5, metric='euclidean').fit_predict(X))
plt.title(f"DBSCAN - Euclidean\nClusters: {n_clusters_e}\nSilhouette: {silhouette_e:.2f}\nCHS: {chs_e:.2f}\nDB: {db_e:.2f}")
plt.subplot(133)
plt.scatter(X[:, 0], X[:, 1], c=DBSCAN(eps=0.5, min_samples=5, metric='manhattan').fit_predict(X))
plt.title(f"DBSCAN - Manhattan\nClusters: {n_clusters_m}\nSilhouette: {silhouette_m:.2f}\nCHS: {chs_m:.2f}\nDB: {db_m:.2f}")
plt.show()
````

Claro, aquí te muestro un ejemplo de cómo comparar el rendimiento de las tres métricas de distancia (Euclidiana, Manhattan y Chebyshev) en el algoritmo de clustering DBSCAN:

* Primero, importamos las librerías necesarias y cargamos un conjunto de datos para realizar el clustering:

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Cargar los datos
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
````

* Luego, estandarizamos los datos y definimos los parámetros para el algoritmo DBSCAN:

````python
# Escalar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Definir los parámetros del algoritmo DBSCAN
eps = 0.3
min_samples = 5
````

* A continuación, creamos tres objetos de clustering, uno para cada métrica de distancia:

````python
# Crear objetos de clustering con diferentes métricas de distancia
dbscan_euclidean = DBSCAN(eps=eps, min_samples=min_samples, metric='euclidean')
dbscan_manhattan = DBSCAN(eps=eps, min_samples=min_samples, metric='manhattan')
dbscan_chebyshev = DBSCAN(eps=eps, min_samples=min_samples, metric='chebyshev')
````

* Finalmente, ajustamos cada modelo de clustering a los datos y visualizamos los resultados:

````python
# Ajustar los modelos de clustering
dbscan_euclidean.fit(X_scaled)
dbscan_manhattan.fit(X_scaled)
dbscan_chebyshev.fit(X_scaled)

# Obtener las etiquetas predichas
labels_euclidean = dbscan_euclidean.labels_
labels_manhattan = dbscan_manhattan.labels_
labels_chebyshev = dbscan_chebyshev.labels_

# Visualizar los resultados
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

axs[0].scatter(X[:, 0], X[:, 1], c=labels_euclidean, cmap='viridis')
axs[0].set_title("Distancia Euclidiana")

axs[1].scatter(X[:, 0], X[:, 1], c=labels_manhattan, cmap='viridis')
axs[1].set_title("Distancia Manhattan")

axs[2].scatter(X[:, 0], X[:, 1], c=labels_chebyshev, cmap='viridis')
axs[2].set_title("Distancia Chebyshev")

plt.show()
````

Este código generará una figura con tres gráficos, uno para cada métrica de distancia, donde los puntos de diferentes colores representan diferentes grupos obtenidos por el algoritmo `DBSCAN`.

Es importante tener en cuenta que el rendimiento de cada métrica de distancia puede variar según los datos y la estructura del problema en particular, por lo que es recomendable experimentar con diferentes métricas para encontrar la más adecuada para cada caso.
