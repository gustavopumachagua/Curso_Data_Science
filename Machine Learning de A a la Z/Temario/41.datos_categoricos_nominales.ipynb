Los datos categ贸ricos nominales son aquellos que no tienen un orden o jerarqu铆a natural, y se pueden clasificar en distintas categor铆as de manera arbitraria. Por ejemplo, el color de ojos o el tipo de animal son datos categ贸ricos nominales.

Aqu铆 hay un ejemplo de c贸mo trabajar con datos categ贸ricos nominales en Python utilizando la biblioteca Pandas:

````python
import pandas as pd

# Crear un conjunto de datos de ejemplo
df = pd.DataFrame({'animal': ['gato', 'perro', 'perro', 'rat贸n', 'gato'],
                   'edad': [3, 5, 2, 1, 4],
                   'peso': [4.2, 6.3, 7.1, 0.9, 3.5]})

# Convertir la columna 'animal' a una variable categ贸rica nominal
df['animal'] = pd.Categorical(df['animal'], categories=['gato', 'perro', 'rat贸n'])

# Codificar la variable categ贸rica nominal como variables dummy
dummy_df = pd.get_dummies(df['animal'], prefix='animal')
df = pd.concat([df, dummy_df], axis=1)
df.drop('animal', axis=1, inplace=True)

# Mostrar el conjunto de datos resultante
print(df)
````

Este c贸digo crea un conjunto de datos de ejemplo con tres variables: 'animal', 'edad' y 'peso'. La columna 'animal' es una variable categ贸rica nominal que puede tomar los valores 'gato', 'perro' o 'rat贸n'. Luego, se convierte la columna 'animal' a una variable categ贸rica nominal utilizando la funci贸n ```pd.Categorical```, que especifica las categor铆as permitidas. A continuaci贸n, se codifica la variable categ贸rica nominal como variables dummy utilizando la funci贸n ```pd.get_dummies```, que crea una columna para cada categor铆a con valores binarios que indican si la muestra pertenece a esa categor铆a o no. Finalmente, se concatenan las variables dummy con el conjunto de datos original y se elimina la columna original 'animal'.

El resultado es un conjunto de datos con dos variables dummy ('animal_gato' y 'animal_perro') que indican si la muestra pertenece a esa categor铆a o no, y dos variables num茅ricas ('edad' y 'peso') que se pueden utilizar en un modelo de aprendizaje autom谩tico.

## **Problem谩tica de las categor铆as**

Para ilustrar la problem谩tica de las categor铆as, podemos crear un conjunto de datos de ejemplo y aplicar un modelo de clasificaci贸n. Luego, podemos comparar los resultados de clasificaci贸n utilizando diferentes categorizaciones.

Aqu铆 est谩 el c贸digo en Python para crear los datos de ejemplo y aplicar un modelo de clasificaci贸n:

````python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Crear datos de ejemplo
np.random.seed(42)
X, y = make_classification(n_samples=1000, n_features=5, n_classes=3, n_informative=3, n_redundant=1)

# Dividir los datos en entrenamiento y prueba
n_train = 800
X_train, y_train = X[:n_train], y[:n_train]
X_test, y_test = X[n_train:], y[n_train:]

# Entrenar un modelo de clasificaci贸n
clf = LogisticRegression(multi_class='auto', solver='lbfgs')
clf.fit(X_train, y_train)

# Evaluar la precisi贸n del modelo en los datos de prueba
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f'Precisi贸n del modelo en datos de prueba: {acc:.2f}')
````

Este c贸digo crea un conjunto de datos de clasificaci贸n de ejemplo con 1000 muestras, 5 caracter铆sticas, 3 clases y 3 caracter铆sticas informativas. Luego, divide los datos en un conjunto de entrenamiento de 800 muestras y un conjunto de prueba de 200 muestras, y entrena un modelo de regresi贸n log铆stica para clasificar los datos.

A continuaci贸n, podemos comparar los resultados de clasificaci贸n utilizando diferentes categorizaciones. Por ejemplo, podemos considerar las categor铆as "Bajo", "Medio" y "Alto" para la variable objetivo, y aplicar diferentes estrategias de categorizaci贸n, como el punto de corte de la media, los cuartiles, etc.

Aqu铆 est谩 el c贸digo en Python para aplicar diferentes categorizaciones y evaluar la precisi贸n del modelo:

````python
# Categorizar los datos utilizando diferentes estrategias
y_mean = np.where(y_test >= np.mean(y_train), 2, np.where(y_test >= np.median(y_train), 1, 0))
y_q1 = np.where(y_test >= np.percentile(y_train, 25), 2, np.where(y_test >= np.percentile(y_train, 50), 1, 0))
y_q3 = np.where(y_test >= np.percentile(y_train, 75), 2, np.where(y_test >= np.percentile(y_train, 50), 1, 0))

# Evaluar la precisi贸n del modelo en los datos de prueba con diferentes categorizaciones
acc_mean = accuracy_score(y_mean, y_pred)
acc_q1 = accuracy_score(y_q1, y_pred)
acc_q3 = accuracy_score(y_q3, y_pred)
print(f'Precisi贸n del modelo en datos de prueba con categorizaci贸n de la media: {acc_mean:.2f}')
print(f'Precisi贸n del modelo en datos de prueba con categorizaci贸n del cuartil 1: {acc_q1:.2f}')
print(f'Precisi贸n del modelo en datos de prueba con categorizaci贸n del cuartil 3: {acc_q3:.2f}')
````

## **Datos categ贸ricos y Contexto**

Los datos categ贸ricos pueden ser influenciados por el contexto en el que se recopilan. Por ejemplo, la percepci贸n de un color puede variar seg煤n el contexto en el que se presente. Aqu铆 hay un ejemplo de c贸mo trabajar con datos categ贸ricos y contexto en Python utilizando la biblioteca Pandas:

````python
import pandas as pd

# Crear un conjunto de datos de ejemplo
df = pd.DataFrame({'color': ['azul', 'verde', 'rojo', 'verde', 'azul', 'rojo'],
                   'contexto': ['d铆a soleado', 'd铆a nublado', 'd铆a nublado', 'd铆a nublado', 'd铆a soleado', 'd铆a nublado']})

# Convertir la columna 'color' a una variable categ贸rica nominal
df['color'] = pd.Categorical(df['color'], categories=['rojo', 'verde', 'azul'])

# Crear una tabla de contingencia que muestre la frecuencia de cada color en cada contexto
tabla_contingencia = pd.crosstab(df['color'], df['contexto'])

# Calcular las proporciones de cada color en cada contexto
proporciones = tabla_contingencia.div(tabla_contingencia.sum(axis=0), axis=1)

# Mostrar la tabla de contingencia y las proporciones resultantes
print('Tabla de contingencia:')
print(tabla_contingencia)
print('\nProporciones:')
print(proporciones)
````

Este c贸digo crea un conjunto de datos de ejemplo con dos variables: 'color' y 'contexto'. La columna 'color' es una variable categ贸rica que puede tomar los valores 'rojo', 'verde' o 'azul', y la columna 'contexto' describe el contexto en el que se presenta cada muestra. Luego, se convierte la columna 'color' a una variable categ贸rica nominal utilizando la funci贸n ```pd.Categorical```, que especifica las categor铆as permitidas. A continuaci贸n, se crea una tabla de contingencia que muestra la frecuencia de cada color en cada contexto utilizando la funci贸n ```pd.crosstab```. Finalmente, se calculan las proporciones de cada color en cada contexto dividiendo cada valor en la tabla de contingencia por el total de la columna correspondiente utilizando el m茅todo ```.div()```.

El resultado es una tabla de contingencia que muestra la frecuencia de cada color en cada contexto y un conjunto de proporciones que muestran c贸mo var铆a la distribuci贸n de colores seg煤n el contexto. Estos resultados pueden ayudar a comprender c贸mo el contexto puede influir en la percepci贸n de un color y c贸mo se pueden ajustar los an谩lisis en consecuencia.

## **Conversi贸n de object a category**

En Pandas, a menudo se almacenan variables categ贸ricas como objetos. Sin embargo, es m谩s eficiente en t茅rminos de memoria y velocidad de procesamiento almacenarlas como variables categ贸ricas. Aqu铆 hay un ejemplo de c贸mo convertir una columna de objeto a una columna de categor铆a en Pandas:

````python
import pandas as pd

# Crear un DataFrame de ejemplo
df = pd.DataFrame({'color': ['azul', 'verde', 'rojo', 'verde', 'azul', 'rojo']})

# Convertir la columna 'color' de objeto a categor铆a
df['color'] = df['color'].astype('category')

# Mostrar la informaci贸n del DataFrame
print(df.info())
````

En este ejemplo, se crea un DataFrame de ejemplo con una columna 'color' que contiene valores de tipo objeto. Despu茅s de la conversi贸n, se muestra la informaci贸n del DataFrame utilizando el m茅todo ```info()```, que muestra el tipo de datos de cada columna.

Aqu铆, se puede observar que la columna 'color' ahora se ha convertido en una variable categ贸rica, lo que se indica mediante el tipo de datos "category". Adem谩s, el uso de memoria del DataFrame se ha reducido significativamente a solo 278.0 bytes.

## **Qu茅 no hacer con datos categ贸ricos**

Es importante tener en cuenta ciertos errores comunes que deben evitarse al trabajar con datos categ贸ricos. Aqu铆 hay un ejemplo de lo que no se debe hacer con datos categ贸ricos utilizando la biblioteca Pandas:

````python
import pandas as pd

# Crear un DataFrame de ejemplo
df = pd.DataFrame({'color': ['azul', 'verde', 'rojo', 'verde', 'azul', 'rojo'],
                   'animal': ['perro', 'gato', 'perro', 'perro', 'gato', 'perro']})

# Crear una tabla de contingencia utilizando groupby
tabla_contingencia = df.groupby(['color', 'animal']).size().unstack()

# Normalizar la tabla de contingencia
tabla_contingencia_normalizada = tabla_contingencia.div(tabla_contingencia.sum(axis=1), axis=0)

# Mostrar la tabla de contingencia y la tabla de contingencia normalizada
print('Tabla de contingencia:')
print(tabla_contingencia)
print('\nTabla de contingencia normalizada:')
print(tabla_contingencia_normalizada)
````

En este ejemplo, se crea un DataFrame de ejemplo con dos variables categ贸ricas: 'color' y 'animal'. Luego, se crea una tabla de contingencia utilizando el m茅todo ```groupby``` de Pandas. Sin embargo, este m茅todo no es adecuado para crear tablas de contingencia ya que no muestra los valores faltantes y puede dar lugar a resultados incorrectos. A continuaci贸n, se normaliza la tabla de contingencia dividiendo cada fila por la suma de los valores de la fila. Esta normalizaci贸n tambi茅n es incorrecta, ya que no tiene en cuenta los valores faltantes en la tabla de contingencia.

## **Codificaci贸n One-Hot**

La Codificaci贸n One-Hot es una t茅cnica utilizada en el an谩lisis de datos para convertir variables categ贸ricas en variables num茅ricas que puedan ser utilizadas por algoritmos de aprendizaje autom谩tico. La idea b谩sica detr谩s de la codificaci贸n One-Hot es crear una columna separada para cada valor 煤nico en una variable categ贸rica y asignar un valor binario (0 o 1) a cada columna seg煤n si ese valor 煤nico est谩 presente o no en cada fila de datos.

Aqu铆 te dejo un ejemplo de c贸mo implementar la codificaci贸n One-Hot en Python utilizando la biblioteca Pandas:

````python
import pandas as pd

# Crear un dataframe de ejemplo
df = pd.DataFrame({'Color': ['Rojo', 'Azul', 'Verde', 'Rojo', 'Amarillo']})

# Utilizar la funci贸n "get_dummies" de Pandas para codificar la columna "Color"
encoded = pd.get_dummies(df['Color'])

# Unir las columnas codificadas al dataframe original
df = pd.concat([df, encoded], axis=1)

# Imprimir el dataframe resultante
print(df)
````

En este ejemplo, se crea un dataframe con una columna llamada "Color" que contiene cinco valores categ贸ricos diferentes. Se utiliza la funci贸n ```get_dummies``` de Pandas para transformar esta columna en variables dummy codificadas en formato One-Hot. Luego, se unen estas columnas codificadas al dataframe original mediante la funci贸n ```concat```. Como se puede ver en el resultado, cada valor 煤nico de la columna "Color" se convierte en una columna propia en el dataframe codificado, donde el valor 1 indica que la fila pertenece a esa categor铆a y 0 indica que no.
