## **Qu茅 es un modelo de regresi贸n lineal**

![regresion lineal](../imagenes%20Machine_Learning/regresion%20lineal.jpg "regresion lineal")

Un modelo de regresi贸n lineal es un tipo de modelo de aprendizaje autom谩tico que se utiliza para predecir el valor de una variable continua (denominada variable dependiente) en funci贸n de una o m谩s variables predictoras (denominadas variables independientes). En el caso de la regresi贸n lineal simple, el modelo utiliza una ecuaci贸n de la forma $y = mx + b$ para modelar la relaci贸n entre la variable dependiente y una 煤nica variable independiente.

Aqu铆 te dejo un ejemplo de c贸mo implementar un modelo de regresi贸n lineal simple en Python utilizando la biblioteca Scikit-learn y Matplotlib para generar una gr谩fica del resultado:

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Crear un conjunto de datos de ejemplo
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 5, 7, 9, 11])

# Crear una instancia del modelo de regresi贸n lineal
model = LinearRegression()

# Ajustar el modelo a los datos de entrenamiento
model.fit(x.reshape(-1, 1), y)

# Generar una gr谩fica del modelo ajustado
plt.scatter(x, y)
plt.plot(x, model.predict(x.reshape(-1, 1)), color='red')
plt.title('Regresi贸n lineal simple')
plt.xlabel('Variable independiente (x)')
plt.ylabel('Variable dependiente (y)')
plt.show()
````

En este ejemplo, se crea un conjunto de datos de ejemplo con una variable independiente (x) y una variable dependiente (y) que siguen una relaci贸n lineal. Se utiliza la clase ```LinearRegression``` de Scikit-learn para ajustar un modelo de regresi贸n lineal a estos datos. Luego, se utiliza Matplotlib para generar una gr谩fica que muestra los datos de entrenamiento junto con la l铆nea ajustada por el modelo. Como se puede ver en la gr谩fica, la l铆nea ajustada por el modelo sigue aproximadamente la relaci贸n lineal entre las variables x e y.

## **Fundamentos**

La regresi贸n lineal es uno de los algoritmos m谩s b谩sicos y ampliamente utilizados en el aprendizaje autom谩tico supervisado para predecir una variable num茅rica continua a partir de un conjunto de variables explicativas. A continuaci贸n, te mostrar茅 un ejemplo de c贸mo implementar una regresi贸n lineal simple utilizando Scikit-learn en Python.

Finalmente, se realiza una predicci贸n para una nueva muestra utilizando el m茅todo predict del modelo.

Estos son algunos de los fundamentos de la regresi贸n lineal que se utilizan en este ejemplo, pero hay muchos otros conceptos importantes a tener en cuenta dependiendo del problema espec铆fico que se est茅 abordando.

## **Suposiciones**

La regresi贸n lineal se basa en varias suposiciones sobre los datos, que deben ser verificadas antes de aplicar el modelo. Algunas de las suposiciones m谩s comunes son:

* Linealidad: la relaci贸n entre las variables explicativas y la variable objetivo debe ser lineal.
* Homocedasticidad: la varianza de los errores debe ser constante en todos los valores de las variables explicativas.
* Normalidad: los errores deben tener una distribuci贸n normal con media cero.
* Independencia: los errores deben ser independientes entre s铆.

A continuaci贸n, se presenta un ejemplo de c贸mo verificar algunas de estas suposiciones utilizando el modelo de regresi贸n lineal simple de Scikit-learn y Python.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression

# Generar datos sint茅ticos de entrenamiento
X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=42)

# Visualizar los datos
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()

# Entrenar el modelo de regresi贸n lineal
reg = LinearRegression()
reg.fit(X, y)

# Obtener la pendiente y el intercepto del modelo
coef = reg.coef_
intercept = reg.intercept_

# Verificar la linealidad
x_line = np.linspace(np.min(X), np.max(X), 100).reshape(-1, 1)
y_line = reg.predict(x_line)
plt.plot(x_line, y_line, c='r')
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()

# Verificar la homocedasticidad
y_pred = reg.predict(X)
residuos = y - y_pred
plt.scatter(y_pred, residuos)
plt.xlabel('y_pred')
plt.ylabel('Residuos')
plt.show()

# Verificar la normalidad
import scipy.stats as stats
stats.probplot(residuos.reshape(-1), dist='norm', plot=plt)
plt.show()
````

En este ejemplo, se generan datos sint茅ticos de entrenamiento utilizando la funci贸n ```make_regression``` de Scikit-learn. Se visualizan los datos utilizando la funci贸n ```scatter``` de Matplotlib. A continuaci贸n, se entrena un modelo de regresi贸n lineal utilizando la clase ```LinearRegression``` de Scikit-learn y se obtienen la pendiente y el intercepto del modelo.

Para verificar la linealidad, se visualiza la l铆nea de regresi贸n junto con los datos utilizando las funciones ```plot``` y ```scatter``` de Matplotlib. Si la relaci贸n entre las variables explicativas y la variable objetivo no es lineal, la l铆nea de regresi贸n no ser谩 una buena aproximaci贸n a los datos.

Para verificar la homocedasticidad, se grafica la relaci贸n entre los valores predichos y los residuos (es decir, las diferencias entre los valores predichos y los valores reales). Si la varianza de los residuos cambia a medida que aumentan los valores predichos, se dice que hay heterocedasticidad.

## **Contexto y datos**

La regresi贸n lineal es una t茅cnica de modelado estad铆stico utilizada para predecir la relaci贸n entre una variable dependiente y una o m谩s variables independientes. En este ejemplo, utilizaremos un conjunto de datos de viviendas para predecir el precio de una casa en funci贸n de su tama帽o.

El conjunto de datos contiene informaci贸n sobre el tama帽o de la casa (en pies cuadrados) y el precio de venta (en miles de d贸lares) para 20 casas.

````python
import numpy as np
import matplotlib.pyplot as plt

# Datos de entrenamiento
X = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700, 
              1600, 880, 900, 1200, 600, 1100, 1550, 1400, 1700, 2450])
y = np.array([245, 312, 279, 308, 199, 219, 405, 324, 319, 255, 
              242, 126, 139, 200, 50, 120, 249, 185, 230, 540])

# Visualizaci贸n de los datos
plt.scatter(X, y)
plt.xlabel('Tama帽o de la casa (pies cuadrados)')
plt.ylabel('Precio de venta (miles de d贸lares)')
plt.show()
````

En este ejemplo, los datos de entrenamiento se definen como dos matrices NumPy: ```X```, que contiene los tama帽os de las casas en pies cuadrados, y ```y```, que contiene los precios de venta en miles de d贸lares.

````python
from sklearn.linear_model import LinearRegression

# Entrenamiento del modelo de regresi贸n lineal
reg = LinearRegression().fit(X.reshape(-1, 1), y)

# Coeficiente de determinaci贸n (R^2)
r2 = reg.score(X.reshape(-1, 1), y)
print("Coeficiente de determinaci贸n (R^2): {:.3f}".format(r2))

# Visualizaci贸n de la l铆nea de regresi贸n
x_line = np.linspace(np.min(X), np.max(X), 100).reshape(-1, 1)
y_line = reg.predict(x_line)
plt.scatter(X, y)
plt.plot(x_line, y_line, c='r')
plt.xlabel('Tama帽o de la casa (pies cuadrados)')
plt.ylabel('Precio de venta (miles de d贸lares)')
plt.show()
````

## **Creaci贸n del modelo con Scikit-Learn**

Para crear un modelo de regresi贸n lineal con Scikit-Learn en Python, podemos utilizar la clase ```LinearRegression```. En este ejemplo, ajustaremos un modelo de regresi贸n lineal a los datos de entrenamiento de tama帽o de casa y precio de venta que hemos estado utilizando, y graficaremos la l铆nea de regresi贸n resultante en el mismo gr谩fico de dispersi贸n que en el ejemplo anterior.



````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de entrenamiento
X = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700, 
              1600, 880, 900, 1200, 600, 1100, 1550, 1400, 1700, 2450])
y = np.array([245, 312, 279, 308, 199, 219, 405, 324, 319, 255, 
              242, 126, 139, 200, 50, 120, 249, 185, 230, 540])

# Ajuste del modelo de regresi贸n lineal
reg = LinearRegression().fit(X.reshape(-1, 1), y)

# Predicci贸n de los precios de venta para los tama帽os de casa dados
X_test = np.array([[1000], [1500], [2000], [2500]])
y_pred = reg.predict(X_test)

# Visualizaci贸n de los datos y la l铆nea de regresi贸n
plt.scatter(X, y)
plt.plot(X_test, y_pred, color='red')
plt.xlabel('Tama帽o de la casa (pies cuadrados)')
plt.ylabel('Precio de venta (miles de d贸lares)')
plt.show()
````

En este c贸digo, primero importamos la clase ```LinearRegression``` de Scikit-Learn. Luego, definimos las matrices `X` e `y` como los datos de entrenamiento, al igual que en el ejemplo anterior.

Para ajustar el modelo de regresi贸n lineal, creamos una instancia de la clase `LinearRegression` y llamamos al m茅todo `fit` con `X` y `y` como argumentos. Luego, creamos una matriz `X_test` de tama帽os de casa para los que queremos predecir los precios de venta, y usamos el m茅todo predict del modelo para obtener las predicciones correspondientes.

Finalmente, utilizamos la funci贸n plot de Matplotlib para graficar la l铆nea de regresi贸n resultante en el mismo gr谩fico de dispersi贸n que en el ejemplo anterior. La l铆nea de regresi贸n se traza en rojo, mientras que los puntos de datos se muestran como un gr谩fico de dispersi贸n.
