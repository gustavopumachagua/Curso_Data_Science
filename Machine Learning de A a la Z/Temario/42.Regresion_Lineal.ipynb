# **42. Regresión Lineal con Python**

## **Qué es un modelo de regresión lineal**

![regresion lineal](../imagenes%20Machine_Learning/regresion%20lineal.jpg "regresion lineal")

Un modelo de regresión lineal es un tipo de modelo de aprendizaje automático que se utiliza para predecir el valor de una variable continua (denominada variable dependiente) en función de una o más variables predictoras (denominadas variables independientes). En el caso de la regresión lineal simple, el modelo utiliza una ecuación de la forma $y = mx + b$ para modelar la relación entre la variable dependiente y una única variable independiente.

Aquí te dejo un ejemplo de cómo implementar un modelo de regresión lineal simple en Python utilizando la biblioteca Scikit-learn y Matplotlib para generar una gráfica del resultado:

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Crear un conjunto de datos de ejemplo
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 5, 7, 9, 11])

# Crear una instancia del modelo de regresión lineal
model = LinearRegression()

# Ajustar el modelo a los datos de entrenamiento
model.fit(x.reshape(-1, 1), y)

# Generar una gráfica del modelo ajustado
plt.scatter(x, y)
plt.plot(x, model.predict(x.reshape(-1, 1)), color='red')
plt.title('Regresión lineal simple')
plt.xlabel('Variable independiente (x)')
plt.ylabel('Variable dependiente (y)')
plt.show()
````

En este ejemplo, se crea un conjunto de datos de ejemplo con una variable independiente (x) y una variable dependiente (y) que siguen una relación lineal. Se utiliza la clase ```LinearRegression``` de Scikit-learn para ajustar un modelo de regresión lineal a estos datos. Luego, se utiliza Matplotlib para generar una gráfica que muestra los datos de entrenamiento junto con la línea ajustada por el modelo. Como se puede ver en la gráfica, la línea ajustada por el modelo sigue aproximadamente la relación lineal entre las variables x e y.

## **Fundamentos**

La regresión lineal es uno de los algoritmos más básicos y ampliamente utilizados en el aprendizaje automático supervisado para predecir una variable numérica continua a partir de un conjunto de variables explicativas. A continuación, te mostraré un ejemplo de cómo implementar una regresión lineal simple utilizando Scikit-learn en Python.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generar datos sintéticos de entrenamiento
np.random.seed(42)
X = np.random.rand(50, 1)
y = 2 + 3 * X + np.random.randn(50, 1)

# Visualizar los datos
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()

# Entrenar el modelo de regresión lineal
reg = LinearRegression()
reg.fit(X, y)

# Obtener la pendiente y el intercepto del modelo
coef = reg.coef_
intercept = reg.intercept_

# Visualizar la línea de regresión
x_line = np.linspace(0, 1, 100).reshape(-1, 1)
y_line = reg.predict(x_line)
plt.plot(x_line, y_line, c='r')
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()

# Predecir una nueva muestra
x_new = [[0.5]]
y_new = reg.predict(x_new)
print('Predicción para x = {}: {}'.format(x_new[0][0], y_new[0][0]))
````

En este ejemplo, se generan datos sintéticos de entrenamiento utilizando la función ```rand``` de NumPy. Se visualizan los datos utilizando la función ```scatter``` de Matplotlib. A continuación, se entrena un modelo de regresión lineal utilizando la clase ```LinearRegression``` de Scikit-learn y se obtienen la pendiente y el intercepto del modelo. Se visualiza la línea de regresión junto con los datos utilizando las funciones ```plot``` y ```scatter``` de Matplotlib. Finalmente, se realiza una predicción para una nueva muestra utilizando el método predict del modelo.

Estos son algunos de los fundamentos de la regresión lineal que se utilizan en este ejemplo, pero hay muchos otros conceptos importantes a tener en cuenta dependiendo del problema específico que se esté abordando.

## **Suposiciones**

La regresión lineal se basa en varias suposiciones sobre los datos, que deben ser verificadas antes de aplicar el modelo. Algunas de las suposiciones más comunes son:

* Linealidad: la relación entre las variables explicativas y la variable objetivo debe ser lineal.
* Homocedasticidad: la varianza de los errores debe ser constante en todos los valores de las variables explicativas.
* Normalidad: los errores deben tener una distribución normal con media cero.
* Independencia: los errores deben ser independientes entre sí.

A continuación, se presenta un ejemplo de cómo verificar algunas de estas suposiciones utilizando el modelo de regresión lineal simple de Scikit-learn y Python.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression

# Generar datos sintéticos de entrenamiento
X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=42)

# Visualizar los datos
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()

# Entrenar el modelo de regresión lineal
reg = LinearRegression()
reg.fit(X, y)

# Obtener la pendiente y el intercepto del modelo
coef = reg.coef_
intercept = reg.intercept_

# Verificar la linealidad
x_line = np.linspace(np.min(X), np.max(X), 100).reshape(-1, 1)
y_line = reg.predict(x_line)
plt.plot(x_line, y_line, c='r')
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()

# Verificar la homocedasticidad
y_pred = reg.predict(X)
residuos = y - y_pred
plt.scatter(y_pred, residuos)
plt.xlabel('y_pred')
plt.ylabel('Residuos')
plt.show()

# Verificar la normalidad
import scipy.stats as stats
stats.probplot(residuos.reshape(-1), dist='norm', plot=plt)
plt.show()
````

A continuación, se entrena un modelo de regresión lineal utilizando la clase ```LinearRegression``` de Scikit-learn y se obtienen la pendiente y el intercepto del modelo.

Para verificar la linealidad, se visualiza la línea de regresión junto con los datos utilizando las funciones ```plot``` y ```scatter``` de Matplotlib. Si la relación entre las variables explicativas y la variable objetivo no es lineal, la línea de regresión no será una buena aproximación a los datos.

Para verificar la homocedasticidad, se grafica la relación entre los valores predichos y los residuos (es decir, las diferencias entre los valores predichos y los valores reales). Si la varianza de los residuos cambia a medida que aumentan los valores predichos, se dice que hay heterocedasticidad.

## **Contexto y datos**

La regresión lineal es una técnica de modelado estadístico utilizada para predecir la relación entre una variable dependiente y una o más variables independientes. En este ejemplo, utilizaremos un conjunto de datos de viviendas para predecir el precio de una casa en función de su tamaño.

El conjunto de datos contiene información sobre el tamaño de la casa (en pies cuadrados) y el precio de venta (en miles de dólares) para 20 casas.

````python
import numpy as np
import matplotlib.pyplot as plt

# Datos de entrenamiento
X = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700, 
              1600, 880, 900, 1200, 600, 1100, 1550, 1400, 1700, 2450])
y = np.array([245, 312, 279, 308, 199, 219, 405, 324, 319, 255, 
              242, 126, 139, 200, 50, 120, 249, 185, 230, 540])

# Visualización de los datos
plt.scatter(X, y)
plt.xlabel('Tamaño de la casa (pies cuadrados)')
plt.ylabel('Precio de venta (miles de dólares)')
plt.show()
````

En este ejemplo, los datos de entrenamiento se definen como dos matrices NumPy: ```X```, que contiene los tamaños de las casas en pies cuadrados, y ```y```, que contiene los precios de venta en miles de dólares.

Para visualizar los datos, utilizamos la función ```scatter``` de Matplotlib para crear un gráfico de dispersión, donde el tamaño de la casa se muestra en el eje x y el precio de venta en el eje y.

````python
from sklearn.linear_model import LinearRegression

# Entrenamiento del modelo de regresión lineal
reg = LinearRegression().fit(X.reshape(-1, 1), y)

# Coeficiente de determinación (R^2)
r2 = reg.score(X.reshape(-1, 1), y)
print("Coeficiente de determinación (R^2): {:.3f}".format(r2))

# Visualización de la línea de regresión
x_line = np.linspace(np.min(X), np.max(X), 100).reshape(-1, 1)
y_line = reg.predict(x_line)
plt.scatter(X, y)
plt.plot(x_line, y_line, c='r')
plt.xlabel('Tamaño de la casa (pies cuadrados)')
plt.ylabel('Precio de venta (miles de dólares)')
plt.show()
````

A continuación, utilizamos la clase LinearRegression de Scikit-learn para entrenar un modelo de regresión lineal. La función fit se utiliza para ajustar el modelo a los datos de entrenamiento.

Una vez que se entrena el modelo, podemos calcular el coeficiente de determinación (R^2) para evaluar qué tan bien se ajusta el modelo a los datos. El R^2 varía de 0 a 1, donde 1 indica un ajuste perfecto. En este caso, el R^2 es de 0.678, lo que indica que el modelo explica el 67.8% de la variación en el precio de venta.

## **Creación del modelo con Scikit-Learn**

Para crear un modelo de regresión lineal con Scikit-Learn en Python, podemos utilizar la clase ```LinearRegression```. En este ejemplo, ajustaremos un modelo de regresión lineal a los datos de entrenamiento de tamaño de casa y precio de venta que hemos estado utilizando, y graficaremos la línea de regresión resultante en el mismo gráfico de dispersión que en el ejemplo anterior.



````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de entrenamiento
X = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700, 
              1600, 880, 900, 1200, 600, 1100, 1550, 1400, 1700, 2450])
y = np.array([245, 312, 279, 308, 199, 219, 405, 324, 319, 255, 
              242, 126, 139, 200, 50, 120, 249, 185, 230, 540])

# Ajuste del modelo de regresión lineal
reg = LinearRegression().fit(X.reshape(-1, 1), y)

# Predicción de los precios de venta para los tamaños de casa dados
X_test = np.array([[1000], [1500], [2000], [2500]])
y_pred = reg.predict(X_test)

# Visualización de los datos y la línea de regresión
plt.scatter(X, y)
plt.plot(X_test, y_pred, color='red')
plt.xlabel('Tamaño de la casa (pies cuadrados)')
plt.ylabel('Precio de venta (miles de dólares)')
plt.show()
````

En este código, primero importamos la clase ```LinearRegression``` de Scikit-Learn. Luego, definimos las matrices `X` e `y` como los datos de entrenamiento, al igual que en el ejemplo anterior.

Para ajustar el modelo de regresión lineal, creamos una instancia de la clase `LinearRegression` y llamamos al método `fit` con `X` y `y` como argumentos. Luego, creamos una matriz `X_test` de tamaños de casa para los que queremos predecir los precios de venta, y usamos el método predict del modelo para obtener las predicciones correspondientes.

Finalmente, utilizamos la función plot de Matplotlib para graficar la línea de regresión resultante en el mismo gráfico de dispersión que en el ejemplo anterior. La línea de regresión se traza en rojo, mientras que los puntos de datos se muestran como un gráfico de dispersión. Al ejecutar este código, se mostrará el gráfico resultante con la línea de regresión ajustada a los datos.
