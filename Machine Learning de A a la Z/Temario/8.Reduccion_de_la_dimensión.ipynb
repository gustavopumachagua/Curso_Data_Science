# **8. Reducción de la dimensión**

## **Análisis de Componentes Principales (ACP)**

El Análisis de Componentes Principales `(ACP)` es una técnica de reducción de dimensionalidad que se utiliza para encontrar patrones en los datos y resumir la información en un conjunto más pequeño de variables llamadas componentes principales. Es una técnica no supervisada que se utiliza comúnmente en la exploración de datos y el análisis de datos.

El objetivo del `ACP` es reducir la dimensionalidad de los datos al proyectarlos en un nuevo espacio de menor dimensión mientras se mantiene la mayor cantidad de información posible. Esto se logra mediante la identificación de las variables que tienen la mayor varianza en los datos originales y la creación de nuevas variables que combinan estas variables originales en componentes principales.

Veamos un ejemplo en Python de cómo aplicar el `ACP` a un conjunto de datos utilizando la biblioteca `scikit-learn`:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Cargar el conjunto de datos iris
iris = load_iris()
X = iris.data
y = iris.target

# Crear un objeto PCA y ajustarlo a los datos
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualizar los datos en el espacio de componentes principales
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```

En este ejemplo, cargamos el conjunto de datos `iris`, que contiene información sobre las medidas de las flores de `iris`. Luego creamos un objeto `PCA` y lo ajustamos a los datos. Especificamos que queremos reducir la dimensionalidad a dos componentes principales y luego usamos el método `fit_transform()` para ajustar y transformar los datos originales en el espacio de componentes principales.

Finalmente, visualizamos los datos en el espacio de componentes principales utilizando un gráfico de dispersión. Podemos ver que las tres especies de `iris` están bien separadas en el espacio de dos componentes principales.

En resumen, el `ACP` es una técnica de reducción de dimensionalidad que se utiliza para encontrar patrones en los datos y resumir la información en un conjunto más pequeño de variables llamadas componentes principales. El `ACP` se puede implementar en Python utilizando la biblioteca `scikit-learn`.

## **Análisis discriminante lineal (LDA)**

El Análisis Discriminante Lineal `(LDA)` es una técnica de aprendizaje supervisado utilizada para clasificar datos en diferentes grupos o categorías. Se basa en encontrar una combinación lineal de características que maximice la separación entre diferentes grupos.

En el `LDA`, se busca una proyección lineal de los datos originales en un espacio de menor dimensión para que la separación entre las clases sea máxima. Es decir, se busca un conjunto de características que describa la mayor cantidad de variabilidad posible entre las clases y la menor cantidad de variabilidad dentro de las clases.

Veamos un ejemplo en Python de cómo aplicar el `LDA` a un conjunto de datos utilizando la biblioteca `scikit-learn`:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Cargar el conjunto de datos iris
iris = load_iris()
X = iris.data
y = iris.target

# Crear un objeto LDA y ajustarlo a los datos
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X, y)

# Visualizar los datos en el espacio de características discriminantes
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y)
plt.xlabel('Característica Discriminante 1')
plt.ylabel('Característica Discriminante 2')
plt.show()
```

En este ejemplo, cargamos el conjunto de datos `iris`, que contiene información sobre las medidas de las flores de `iris`. Luego creamos un objeto `LDA` y lo ajustamos a los datos. Especificamos que queremos reducir la dimensionalidad a dos características discriminantes y luego usamos el método `fit_transform()` para ajustar y transformar los datos originales en el espacio de características discriminantes.

Finalmente, visualizamos los datos en el espacio de características discriminantes utilizando un gráfico de dispersión. Podemos ver que las tres especies de `iris` están bien separadas en el espacio de dos características discriminantes.

En resumen, el `LDA` es una técnica de aprendizaje supervisado utilizada para clasificar datos en diferentes grupos o categorías. El `LDA` se puede implementar en Python utilizando la biblioteca `scikit-learn`.

## **Kernel ACP**

El Kernel ACP (Kernel Principal Component Analysis) es una técnica de aprendizaje no supervisado utilizada para reducir la dimensionalidad de los datos y encontrar patrones subyacentes. A diferencia del `ACP` tradicional, el Kernel ACP utiliza una función kernel para mapear los datos a un espacio de características de mayor dimensión donde los patrones pueden ser más fácilmente separados.

En el Kernel ACP, se busca una proyección no lineal de los datos originales en un espacio de mayor dimensión para que la separación entre las clases sea máxima. Es decir, se busca un conjunto de características que describa la mayor cantidad de variabilidad posible entre las clases y la menor cantidad de variabilidad dentro de las clases, pero utilizando una función kernel para mapear los datos a un espacio de características de mayor dimensión.

Veamos un ejemplo en Python de cómo aplicar el Kernel ACP a un conjunto de datos utilizando la biblioteca `scikit-learn`:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.decomposition import KernelPCA

# Crear un conjunto de datos de ejemplo en forma de círculos
X, y = make_circles(n_samples=1000, noise=0.1, factor=0.4)

# Crear un objeto Kernel ACP y ajustarlo a los datos
kpca = KernelPCA(n_components=2, kernel='rbf')
X_kpca = kpca.fit_transform(X)

# Visualizar los datos en el espacio de características del Kernel ACP
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y)
plt.xlabel('Característica del Kernel ACP 1')
plt.ylabel('Característica del Kernel ACP 2')
plt.show()
```

En este ejemplo, creamos un conjunto de datos de ejemplo en forma de círculos y luego creamos un objeto Kernel ACP y lo ajustamos a los datos. Especificamos que queremos reducir la dimensionalidad a dos características del Kernel ACP y luego usamos el método `fit_transform()` para ajustar y transformar los datos originales en el espacio de características del Kernel ACP.

Finalmente, visualizamos los datos en el espacio de características del Kernel ACP utilizando un gráfico de dispersión. Podemos ver que los datos en forma de círculos se han transformado en un espacio de características del Kernel ACP donde pueden ser más fácilmente separados en dos grupos.

En resumen, el Kernel ACP es una técnica de aprendizaje no supervisado utilizada para reducir la dimensionalidad de los datos y encontrar patrones subyacentes. El Kernel ACP se puede implementar en Python utilizando la biblioteca `scikit-learn`.
