# **23. K-means (o K-medias) para detecci√≥n de clusters**

## **Introducci√≥n**

`K-means` (o K-medias) es un algoritmo de aprendizaje no supervisado que se utiliza para clasificar un conjunto de datos en `K` grupos diferentes. El objetivo del algoritmo es agrupar los datos de tal manera que los puntos dentro de un grupo sean similares entre s√≠, mientras que los puntos de diferentes grupos sean diferentes entre s√≠.

Por ejemplo, si tuvi√©ramos un conjunto de datos que representara las caracter√≠sticas de diferentes tipos de frutas, podr√≠amos utilizar `K-means` para clasificar las frutas en diferentes grupos en funci√≥n de sus caracter√≠sticas. Podr√≠amos tener un grupo para las frutas c√≠tricas, otro para las frutas de pepita, y otro para las frutas tropicales, por ejemplo.

## **Contexto y algoritmo**

`K-means` (o K-medias) es un algoritmo de agrupamiento no supervisado utilizado en el an√°lisis de datos. Su objetivo es agrupar un conjunto de observaciones en `k` grupos, donde `k` es un n√∫mero predefinido de grupos.

El algoritmo funciona de la siguiente manera:

* Seleccionar el n√∫mero `k` de grupos que se desean crear.
* Seleccionar `k` puntos aleatorios del conjunto de observaciones como centros iniciales de los grupos.
* Asignar cada observaci√≥n al grupo cuyo centro est√© m√°s cercano.
* Calcular los nuevos centros de los grupos como la media de las observaciones asignadas a cada grupo.
* Repetir los pasos `3` y `4` hasta que los centros de los grupos no cambien de manera significativa.

El resultado final del algoritmo es una partici√≥n de las observaciones en k grupos, donde cada observaci√≥n pertenece a un solo grupo.

El algoritmo `K-means` es ampliamente utilizado en la industria y la investigaci√≥n para clasificar grandes conjuntos de datos en grupos homog√©neos. Tambi√©n es un algoritmo relativamente simple y r√°pido, lo que lo hace muy √∫til en situaciones donde se necesitan resultados r√°pidos. Sin embargo, es importante tener en cuenta que el resultado del algoritmo `K-means` puede depender de la selecci√≥n inicial de los centros de los grupos y puede no ser el √≥ptimo en algunos casos.

## **Creaci√≥n de modelo K-means**

Claro, te mostrar√© c√≥mo crear un modelo `K-means` en Python utilizando la biblioteca `Scikit-learn` y c√≥mo visualizar los resultados con una gr√°fica. En este ejemplo, utilizaremos un conjunto de datos de `iris`.

```python
# importar las bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

# cargar el conjunto de datos de iris
iris = load_iris()
X = iris.data

# crear el modelo K-means con 3 grupos
kmeans = KMeans(n_clusters=3)

# ajustar el modelo a los datos
kmeans.fit(X)

# obtener las etiquetas de los grupos
labels = kmeans.labels_

# obtener las coordenadas de los centroides
centroids = kmeans.cluster_centers_

# visualizar los resultados en una gr√°fica
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=300, c='r')
plt.show()
```

En este c√≥digo, primero importamos las bibliotecas necesarias y cargamos el conjunto de datos de `iris`. Luego, creamos el modelo `K-means` con `3` grupos y lo ajustamos a los datos.

A continuaci√≥n, obtenemos las etiquetas de los grupos y las coordenadas de los centroides. Finalmente, visualizamos los resultados en una gr√°fica, donde cada punto representa una observaci√≥n en el conjunto de datos y el color indica a qu√© grupo pertenece. Los centroides se muestran como estrellas rojas en la gr√°fica.

## **Visualizaci√≥n de clusters**

La visualizaci√≥n de clusters es una t√©cnica importante para entender la estructura de los grupos formados por un algoritmo de clustering, como `K-means`. Aqu√≠ te muestro algunos m√©todos comunes para visualizar clusters:

* **Gr√°fico de dispersi√≥n:** se puede utilizar para visualizar dos caracter√≠sticas del conjunto de datos y colorear los puntos seg√∫n su asignaci√≥n de grupo. En `K-means`, los centroides tambi√©n se pueden mostrar en la gr√°fica para visualizar su posici√≥n relativa a los puntos del conjunto de datos.

* **Mapa de calor:** se puede utilizar para visualizar la densidad de puntos de un conjunto de datos y destacar las √°reas de alta densidad. Los puntos se pueden colorear seg√∫n su asignaci√≥n de grupo.

* **Gr√°fico de barras:** se puede utilizar para visualizar la distribuci√≥n de los grupos en una caracter√≠stica espec√≠fica del conjunto de datos. Por ejemplo, si el conjunto de datos contiene informaci√≥n sobre el g√©nero de los participantes, se puede crear un gr√°fico de barras para mostrar cu√°ntos participantes hay en cada grupo y en cada g√©nero.

* **Gr√°fico de l√≠neas:** se puede utilizar para visualizar c√≥mo cambian las caracter√≠sticas del conjunto de datos en funci√≥n de la posici√≥n o el tiempo. Los grupos se pueden colorear para visualizar la asignaci√≥n de grupo en diferentes puntos o tiempos.

```python
# importar las bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

# cargar el conjunto de datos de iris
iris = load_iris()
X = iris.data

# crear el modelo K-means con 3 grupos
kmeans = KMeans(n_clusters=3)

# ajustar el modelo a los datos
kmeans.fit(X)

# obtener las etiquetas de los grupos
labels = kmeans.labels_

# obtener las coordenadas de los centroides
centroids = kmeans.cluster_centers_

# visualizar los resultados en una gr√°fica de dispersi√≥n
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=300, c='r')
plt.xlabel('Longitud del s√©palo')
plt.ylabel('Anchura del s√©palo')
plt.title('Gr√°fico de dispersi√≥n de clusters')
plt.show()
```

En este ejemplo, utilizamos el conjunto de datos de `iris` y creamos un modelo `K-means` con 3 grupos. Luego, utilizamos el gr√°fico de dispersi√≥n para visualizar los clusters, donde cada punto representa una observaci√≥n en el conjunto de datos y el color indica a qu√© grupo pertenece. Los centroides se muestran como estrellas rojas en la gr√°fica.

## **Determina k con el m√©todo del codo**

```python
# importar las bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

# cargar el conjunto de datos de iris
iris = load_iris()
X = iris.data

# crear una lista vac√≠a para almacenar las inercias
inertias = []

# crear una lista de valores de k para probar
k_values = range(1, 11)

# ajustar el modelo K-means para cada valor de k y obtener su inercia
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# trazar la curva del codo
plt.plot(k_values, inertias, 'bx-')
plt.xlabel('N√∫mero de clusters (k)')
plt.ylabel('Inercia')
plt.title('M√©todo del codo')
plt.show()
```

Este c√≥digo ajusta el modelo `K-means` para cada valor de `k` de `1` a `10` y almacena la inercia del modelo en una lista. Luego, traza la curva del codo utilizando los valores de `k` y las inercias correspondientes. La curva del codo es una representaci√≥n gr√°fica de la relaci√≥n entre el n√∫mero de clusters y la inercia. Se espera que la inercia disminuya a medida que aumenta el n√∫mero de clusters. Sin embargo, a partir de un cierto punto, los beneficios de dividir los datos en m√°s clusters disminuyen, lo que resulta en una disminuci√≥n m√°s lenta de la inercia. El "codo" en la curva representa el punto √≥ptimo de clusters, despu√©s del cual los beneficios de agregar m√°s clusters son menores.
