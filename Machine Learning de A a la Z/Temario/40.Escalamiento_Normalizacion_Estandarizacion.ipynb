# **40. Escalamiento, Normalizaci√≥n y Estandarizaci√≥n de Datos con Python**

## **Problem√°tica**

La problem√°tica de escalamiento, normalizaci√≥n y estandarizaci√≥n de datos se refiere a la necesidad de transformar los datos antes de aplicar ciertos algoritmos de an√°lisis de datos o aprendizaje autom√°tico.

* Escalamiento: implica cambiar la escala de los datos para que est√©n en un rango similar, lo que ayuda a evitar que algunos atributos con valores muy grandes dominen el an√°lisis.

* Normalizaci√≥n: se utiliza para escalar los datos a un rango de 0 a 1. Esto es √∫til cuando los datos tienen diferentes unidades o escalas.

* Estandarizaci√≥n: implica transformar los datos para que tengan una media de cero y una desviaci√≥n est√°ndar de uno. Este m√©todo es √∫til cuando los datos tienen diferentes medias y desviaciones est√°ndar, lo que puede dificultar la comparaci√≥n de diferentes variables.

La problem√°tica surge porque la elecci√≥n del m√©todo de transformaci√≥n adecuado depende de la distribuci√≥n y caracter√≠sticas de los datos, y aplicar el m√©todo incorrecto puede afectar negativamente el rendimiento del modelo o an√°lisis. Adem√°s, el proceso de transformaci√≥n puede ser computacionalmente costoso y aumentar el tiempo de procesamiento.

Por lo tanto, es importante comprender la naturaleza de los datos y seleccionar el m√©todo adecuado de transformaci√≥n de datos para garantizar una correcta interpretaci√≥n y an√°lisis de los resultados.

## **Contexto y datos**

La problem√°tica de escalamiento, normalizaci√≥n y estandarizaci√≥n de datos es com√∫n en el campo del an√°lisis de datos y el aprendizaje autom√°tico. En muchos casos, los datos utilizados en estos campos provienen de diferentes fuentes y pueden tener diferentes unidades, escalas y distribuciones. Si estos datos no se transforman adecuadamente, pueden afectar negativamente el rendimiento del modelo o an√°lisis.

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Creamos un conjunto de datos aleatorios con diferentes escalas
data = np.array([[10, 2], [20, 4], [30, 6], [40, 8], [50, 10]])

# Escalamos los datos utilizando MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
print("Datos escalados con MinMaxScaler:\n", scaled_data)

# Normalizamos los datos utilizando L2 norm
normalized_data = data / np.linalg.norm(data, ord=2, axis=1, keepdims=True)
print("Datos normalizados con L2 norm:\n", normalized_data)

# Estandarizamos los datos utilizando StandardScaler
standard_scaler = StandardScaler()
standardized_data = standard_scaler.fit_transform(data)
print("Datos estandarizados con StandardScaler:\n", standardized_data)
```

## **Datos graficados**

La problem√°tica de escalamiento, normalizaci√≥n y estandarizaci√≥n de datos surge cuando se tienen datos que var√≠an en magnitudes y escalas diferentes, lo que puede afectar el rendimiento de los modelos de aprendizaje autom√°tico que se entrenan con ellos.

El escalamiento, la normalizaci√≥n y la estandarizaci√≥n son t√©cnicas que se utilizan para transformar los datos de manera que se ajusten a una escala com√∫n, lo que puede mejorar el rendimiento de los modelos de aprendizaje autom√°tico.

A continuaci√≥n, se presentar√° un ejemplo de c√≥mo aplicar estas t√©cnicas utilizando Python y se mostrar√°n gr√°ficas para ilustrar los resultados.

Primero, se importar√°n las bibliotecas necesarias:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Luego, se generar√° un conjunto de datos de muestra con dos caracter√≠sticas (variables) que var√≠an en magnitud y escala diferentes:

```python
np.random.seed(42)

# Generar datos de muestra con dos caracter√≠sticas
X = np.random.normal(loc=10, scale=5, size=(100, 2))
```

A continuaci√≥n, se mostrar√° un diagrama de dispersi√≥n de los datos de muestra:

```python
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')
plt.show()
```

Como se puede observar en el diagrama de dispersi√≥n, las dos caracter√≠sticas var√≠an en magnitud y escala diferentes, lo que puede afectar el rendimiento de los modelos de aprendizaje autom√°tico que se entrenan con estos datos.

A continuaci√≥n, se aplicar√°n las t√©cnicas de escalamiento, normalizaci√≥n y estandarizaci√≥n para transformar los datos de muestra:

```python
# Escalar los datos
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Normalizar los datos
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
X_normalized = normalizer.fit_transform(X)

# Estandarizar los datos
from sklearn.preprocessing import StandardScaler
standardizer = StandardScaler()
X_standardized = standardizer.fit_transform(X)
```

Una vez que se han transformado los datos de muestra, se mostrar√°n diagramas de dispersi√≥n para cada t√©cnica de transformaci√≥n:

```python
# Escalamiento
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')
plt.show()

# Normalizaci√≥n
plt.scatter(X_normalized[:, 0], X_normalized[:, 1])
plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')
plt.show()

# Estandarizaci√≥n
plt.scatter(X_standardized[:, 0], X_standardized[:, 1])
plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')
plt.show()
```

## **Distribuci√≥n de los datos**

```python
# Escalar los datos
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Normalizar los datos
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
X_normalized = normalizer.fit_transform(X)

# Estandarizar los datos
from sklearn.preprocessing import StandardScaler
standardizer = StandardScaler()
X_standardized = standardizer.fit_transform(X)
```

Es importante destacar que cada t√©cnica de transformaci√≥n tiene sus propias caracter√≠sticas y se deben seleccionar seg√∫n el problema en particular y los requisitos del modelo a entrenar.

## **Escalamiento**

El escalamiento es una t√©cnica de transformaci√≥n de datos que se utiliza para llevar los valores de las caracter√≠sticas de un conjunto de datos a un rango espec√≠fico. Esta t√©cnica es √∫til para cuando los valores de las caracter√≠sticas tienen escalas muy diferentes y se quiere asegurar que todas las caracter√≠sticas tengan un rango similar.

A continuaci√≥n, se muestra un ejemplo de c√≥mo aplicar el escalamiento en Python utilizando la biblioteca scikit-learn:

```python
# Importar la biblioteca scikit-learn
from sklearn.preprocessing import MinMaxScaler

# Crear un conjunto de datos de muestra
X = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]

# Crear un objeto scaler
scaler = MinMaxScaler()

# Escalar los datos
X_scaled = scaler.fit_transform(X)

# Imprimir los datos escalados
print(X_scaled)
```

Como se puede observar, los datos han sido escalados para que los valores de cada caracter√≠stica est√©n dentro del rango de 0 a 1. Ahora, se puede graficar estos datos escalados para observar c√≥mo han sido transformados:

```python
# Importar la biblioteca matplotlib
import matplotlib.pyplot as plt

# Graficar los datos escalados
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')
plt.show()
```

Como se puede observar, la gr√°fica muestra que todos los valores de las caracter√≠sticas est√°n dentro del rango de 0 a 1, lo que indica que el escalamiento ha sido efectivo en la transformaci√≥n de los datos.

## **Normalizaci√≥n**

La normalizaci√≥n es otra t√©cnica de transformaci√≥n de datos que se utiliza para ajustar la escala de los datos, pero a diferencia del escalamiento, la normalizaci√≥n se utiliza para asegurar que los datos tengan una norma o magnitud igual a 1. Esta t√©cnica es √∫til cuando se quiere comparar la magnitud relativa de las caracter√≠sticas.

A continuaci√≥n, se muestra un ejemplo de c√≥mo aplicar la normalizaci√≥n en Python utilizando la biblioteca scikit-learn:

```python
# Importar la biblioteca scikit-learn
from sklearn.preprocessing import Normalizer

# Crear un conjunto de datos de muestra
X = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]

# Crear un objeto normalizer
normalizer = Normalizer()

# Normalizar los datos
X_normalized = normalizer.fit_transform(X)

# Imprimir los datos normalizados
print(X_normalized)
```

Como se puede observar, los datos han sido normalizados para que la norma de cada fila sea igual a 1. Ahora, se puede graficar estos datos normalizados para observar c√≥mo han sido transformados:

```python
# Graficar los datos normalizados
plt.scatter(X_normalized[:, 0], X_normalized[:, 1])
plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')
plt.show()
```

Como se puede observar, la gr√°fica muestra que la magnitud de los datos ha sido normalizada a una norma de 1, lo que indica que la normalizaci√≥n ha sido efectiva en la transformaci√≥n de los datos.

## **Estandarizaci√≥n**

La estandarizaci√≥n es otra t√©cnica de transformaci√≥n de datos que se utiliza para ajustar la escala de los datos, pero a diferencia del escalamiento y la normalizaci√≥n, la estandarizaci√≥n se utiliza para transformar los datos de manera que tengan una media de cero y una desviaci√≥n est√°ndar de uno. Esta t√©cnica es √∫til cuando se quiere que los datos tengan una distribuci√≥n normal y cuando se utilizan algoritmos que asumen que los datos est√°n normalmente distribuidos.

A continuaci√≥n, se muestra un ejemplo de c√≥mo aplicar la estandarizaci√≥n en Python utilizando la biblioteca scikit-learn:

```python
# Importar la biblioteca scikit-learn
from sklearn.preprocessing import StandardScaler

# Crear un conjunto de datos de muestra
X = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]

# Crear un objeto scaler
scaler = StandardScaler()

# Estandarizar los datos
X_std = scaler.fit_transform(X)

# Imprimir los datos estandarizados
print(X_std)
```

Como se puede observar, los datos han sido estandarizados para que tengan una media de cero y una desviaci√≥n est√°ndar de uno. Ahora, se puede graficar estos datos estandarizados para observar c√≥mo han sido transformados:

```python
# Graficar los datos estandarizados
plt.scatter(X_std[:, 0], X_std[:, 1])
plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')
plt.show()
```

Como se puede observar, la gr√°fica muestra que los datos han sido transformados de manera que tengan una media de cero y una desviaci√≥n est√°ndar de uno, lo que indica que la estandarizaci√≥n ha sido efectiva en la transformaci√≥n de los datos.

## **Comparaci√≥n de los m√©todos**

Para comparar los m√©todos de escalamiento, normalizaci√≥n y estandarizaci√≥n, se puede utilizar el mismo conjunto de datos y graficar los datos transformados para observar c√≥mo cambian en cada m√©todo.

A continuaci√≥n, se muestra un ejemplo de c√≥mo aplicar los tres m√©todos de transformaci√≥n de datos en Python utilizando la biblioteca scikit-learn y c√≥mo graficar los datos transformados:

```python
# Importar la biblioteca scikit-learn y matplotlib
from sklearn.preprocessing import MinMaxScaler, Normalizer, StandardScaler
import matplotlib.pyplot as plt
import numpy as np

X = np.array([[-1, 2], [-0.5, 6], [0, 10], [1, 18]])


# Crear objetos para los tres m√©todos de transformaci√≥n de datos
scaler = MinMaxScaler()
normalizer = Normalizer()
std_scaler = StandardScaler()

# Aplicar los tres m√©todos de transformaci√≥n de datos
X_scaled = scaler.fit_transform(X)
X_normalized = normalizer.fit_transform(X)
X_std = std_scaler.fit_transform(X)

# Graficar los datos originales y transformados
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].scatter(X[:, 0], X[:, 1])
axs[0, 0].set_title('Datos originales')
axs[0, 1].scatter(X_scaled[:, 0], X_scaled[:, 1])
axs[0, 1].set_title('Escalamiento')
axs[1, 0].scatter(X_normalized[:, 0], X_normalized[:, 1])
axs[1, 0].set_title('Normalizaci√≥n')
axs[1, 1].scatter(X_std[:, 0], X_std[:, 1])
axs[1, 1].set_title('Estandarizaci√≥n')

for ax in axs.flat:
    ax.set(xlabel='Caracter√≠stica 1', ylabel='Caracter√≠stica 2')

plt.show()
```

Como se puede observar en la gr√°fica, cada m√©todo de transformaci√≥n de datos produce un resultado diferente. El escalamiento ajusta la escala de los datos para que est√©n en el rango [0, 1], la normalizaci√≥n ajusta la magnitud de los datos para que tengan una norma de 1 y la estandarizaci√≥n ajusta la media y la desviaci√≥n est√°ndar de los datos para que tengan una distribuci√≥n normal con media cero y desviaci√≥n est√°ndar de uno. La elecci√≥n del m√©todo de transformaci√≥n de datos depende del problema espec√≠fico y del tipo de algoritmo que se utilice para analizar los datos.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

# Generamos un conjunto de datos de ejemplo con dos caracter√≠sticas
X = np.array([[10, 5], [20, 10], [30, 15], [40, 20]])

# Escalamos los datos utilizando MinMaxScaler
scaler_minmax = MinMaxScaler()
X_minmax = scaler_minmax.fit_transform(X)

# Escalamos los datos utilizando StandardScaler
scaler_std = StandardScaler()
X_std = scaler_std.fit_transform(X)

# Escalamos los datos utilizando RobustScaler
scaler_robust = RobustScaler()
X_robust = scaler_robust.fit_transform(X)

# Graficamos los datos originales y los escalados por cada m√©todo
fig, axs = plt.subplots(2, 2, figsize=(8, 8))
axs[0, 0].scatter(X[:, 0], X[:, 1])
axs[0, 0].set_title("Datos originales")
axs[0, 1].scatter(X_minmax[:, 0], X_minmax[:, 1])
axs[0, 1].set_title("MinMaxScaler")
axs[1, 0].scatter(X_std[:, 0], X_std[:, 1])
axs[1, 0].set_title("StandardScaler")
axs[1, 1].scatter(X_robust[:, 0], X_robust[:, 1])
axs[1, 1].set_title("RobustScaler")
plt.show()
```

Este c√≥digo genera un conjunto de datos de ejemplo con dos caracter√≠sticas y luego utiliza tres m√©todos diferentes de escalado: MinMaxScaler, StandardScaler y RobustScaler. A continuaci√≥n, se grafican los datos originales y los escalados por cada m√©todo.

Como puedes ver en la imagen, los datos originales (arriba a la izquierda) tienen una escala diferente en cada caracter√≠stica. Sin embargo, despu√©s de aplicar los diferentes m√©todos de escalado, los datos tienen una escala similar y se encuentran en rangos similares. En este ejemplo, los tres m√©todos de escalado producen resultados similares, pero en otros conjuntos de datos puede haber diferencias significativas en los resultados. Es importante probar diferentes m√©todos de escalado en cada conjunto de datos para encontrar el que mejor se adapte a las necesidades del problema que se est√° resolviendo.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Crear datos de ejemplo
np.random.seed(42)
X = np.random.normal(loc=50, scale=10, size=(100, 2))

# Escalamiento
scaler = MinMaxScaler()
X_scaled_minmax = scaler.fit_transform(X)

# Normalizaci√≥n
X_normalized = X / np.linalg.norm(X, axis=1, keepdims=True)

# Estandarizaci√≥n
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Crear gr√°fico
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].scatter(X[:, 0], X[:, 1])
axs[0, 0].set_title('Datos originales')
axs[0, 1].scatter(X_scaled_minmax[:, 0], X_scaled_minmax[:, 1])
axs[0, 1].set_title('Escalamiento (min-max)')
axs[1, 0].scatter(X_normalized[:, 0], X_normalized[:, 1])
axs[1, 0].set_title('Normalizaci√≥n')
axs[1, 1].scatter(X_standardized[:, 0], X_standardized[:, 1])
axs[1, 1].set_title('Estandarizaci√≥n')
plt.show()
```

Este c√≥digo crea un conjunto de datos de 100 puntos, cada uno con dos dimensiones, distribuidos normalmente con media 50 y desviaci√≥n est√°ndar 10. Luego, aplica los tres m√©todos de preprocesamiento: Escalamiento (min-max), Normalizaci√≥n y Estandarizaci√≥n, y crea un gr√°fico de dispersi√≥n para comparar los resultados.

En el gr√°fico, el panel superior izquierdo muestra los datos originales. El panel superior derecho muestra los datos despu√©s del Escalamiento (min-max), donde los datos se han escalado en el rango [0, 1] para cada dimensi√≥n. El panel inferior izquierdo muestra los datos despu√©s de la Normalizaci√≥n, donde los datos se han normalizado para que cada punto tenga una norma igual a 1. El panel inferior derecho muestra los datos despu√©s de la Estandarizaci√≥n, donde los datos se han transformado para tener media 0 y desviaci√≥n est√°ndar 1 para cada dimensi√≥n.

Podemos observar que los tres m√©todos de preprocesamiento producen resultados diferentes, pero √∫tiles para diferentes situaciones. El Escalamiento (min-max) es √∫til cuando queremos que todas las caracter√≠sticas tengan el mismo rango de valores, la Normalizaci√≥n es √∫til cuando queremos que todas las muestras tengan la misma escala y la Estandarizaci√≥n es √∫til cuando queremos que los datos est√©n centrados en torno a la media y tengan una desviaci√≥n est√°ndar conocida.
