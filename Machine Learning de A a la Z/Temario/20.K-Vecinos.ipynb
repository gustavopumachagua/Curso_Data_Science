# **20. K-Vecinos Más Cercanos (KNN) para Regresión con Python**

## **Interpretación básica**

![k-Vecinos](../imagenes%20Machine_Learning/k-vecinos.webp "k-Vecinos")

Los K-Vecinos Más Cercanos `(KNN)` es un algoritmo de aprendizaje supervisado utilizado para clasificación y regresión. La idea básica detrás de `KNN` es encontrar los `k` ejemplos más cercanos a un punto de consulta y clasificar el punto de acuerdo a la mayoría de las clases encontradas entre los `k` ejemplos.

A continuación, se muestra un ejemplo de cómo implementar `KNN` en Python utilizando la biblioteca `Scikit-Learn`:

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np

# Cargar el conjunto de datos Iris
iris = load_iris()

# Dividir el conjunto de datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3)

# Crear un clasificador KNN con k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Entrenar el clasificador con el conjunto de entrenamiento
knn.fit(X_train, y_train)

# Predecir las etiquetas de clase para el conjunto de prueba
y_pred = knn.predict(X_test)

# Calcular la precisión del clasificador en el conjunto de prueba
accuracy = np.mean(y_pred == y_test)
print(f"Precisión del clasificador KNN: {accuracy:.2f}")

# Crear una gráfica de dispersión para visualizar los datos
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.xlabel('Longitud del sépalo')
plt.ylabel('Anchura del sépalo')
plt.title('Conjunto de entrenamiento Iris')
plt.show()
````

El código anterior entrena un clasificador `KNN` con `k=3` utilizando el conjunto de datos `Iris` y muestra la precisión del clasificador en el conjunto de prueba. Además, se crea una gráfica de dispersión para visualizar los datos de entrenamiento.

La gráfica muestra la longitud y la anchura del sépalo para cada flor del conjunto de entrenamiento `Iris`, y las clases se representan con diferentes colores.

## **Datos y contexto**

Aquí hay un ejemplo de cómo se puede utilizar `K-Vecinos` Más Cercanos `(KNN)` para clasificar imágenes de dígitos escritos a mano utilizando el conjunto de datos `MNIST`. Este conjunto de datos contiene `70,000` imágenes en escala de grises de 28x28 píxeles de dígitos del `0` al `9` escritos a mano.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Cargar el conjunto de datos MNIST
mnist = fetch_openml('mnist_784')

# Convertir las imágenes de cadena a números enteros
X = mnist.data.astype('int')
y = mnist.target.astype('int')

# Dividir el conjunto de datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Crear un clasificador KNN con k=5
knn = KNeighborsClassifier(n_neighbors=5)

# Entrenar el clasificador con el conjunto de entrenamiento
knn.fit(X_train, y_train)

# Calcular la precisión del clasificador en el conjunto de prueba
accuracy = knn.score(X_test, y_test)
print(f"Precisión del clasificador KNN: {accuracy:.2f}")
````

El código anterior carga el conjunto de datos `MNIST`, divide el conjunto de datos en conjunto de entrenamiento y prueba, crea un clasificador `KNN` con `k=5` y lo entrena con el conjunto de entrenamiento. Luego, calcula la precisión del clasificador en el conjunto de prueba y muestra algunos ejemplos de imágenes del conjunto de prueba junto con las etiquetas reales y predichas.

La precisión del clasificador `KNN` en el conjunto de prueba es del `96.8%`, lo que indica que es capaz de clasificar correctamente la mayoría de las imágenes.

## **Explicación de KNN para regresión**

K-Vecinos Más Cercanos `(KNN)` también puede ser utilizado para la regresión, donde se trata de predecir un valor numérico en lugar de una etiqueta de clase. La idea básica sigue siendo la misma: encontrar los `k` ejemplos más cercanos a un punto de consulta y predecir el valor numérico promedio de los `k` ejemplos.

Aquí hay un ejemplo de cómo se puede utilizar `KNN` para la regresión utilizando un conjunto de datos de precios de viviendas en Boston:

````python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
import numpy as np

# Cargar el conjunto de datos Boston Housing
boston = fetch_openml(name='boston')

# Dividir el conjunto de datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.3)

# Crear un regresor KNN con k=5
knn = KNeighborsRegressor(n_neighbors=5)

# Entrenar el regresor con el conjunto de entrenamiento
knn.fit(X_train, y_train)

# Predecir los valores de precio para el conjunto de prueba
y_pred = knn.predict(X_test)

# Calcular el error cuadrático medio del regresor en el conjunto de prueba
mse = np.mean((y_pred - y_test) ** 2)
print(f"Error cuadrático medio del regresor KNN: {mse:.2f}")

# Crear una gráfica de dispersión para visualizar las predicciones del regresor
plt.scatter(y_test, y_pred)
plt.xlabel('Valores reales')
plt.ylabel('Valores predichos')
plt.title('Predicciones del regresor KNN')
plt.show()
````

El código anterior carga el conjunto de datos Boston Housing, divide el conjunto de datos en conjunto de entrenamiento y prueba, crea un regresor `KNN` con `k=5` y lo entrena con el conjunto de entrenamiento. Luego, se predice los valores de precio para el conjunto de prueba y se calcula el error cuadrático medio del regresor en el conjunto de prueba. Finalmente, se muestra una gráfica de dispersión para visualizar las predicciones del regresor.

La gráfica muestra los valores reales en el eje `x` y los valores predichos por el regresor en el eje `y`. La línea diagonal representa una predicción perfecta, y se puede ver que las predicciones del regresor están bastante cerca de la línea diagonal, lo que indica que el modelo funciona bien en la tarea de regresión.

## **Preprocesamiento de datos**

El preprocesamiento de datos es una parte importante del proceso de aprendizaje automático y puede afectar significativamente el rendimiento de los modelos, incluido el modelo de `K-Vecinos` Más Cercanos `(KNN)`. A continuación, se presenta un ejemplo de cómo se puede realizar el preprocesamiento de datos antes de aplicar `KNN`.

El conjunto de datos que se utilizará es el conjunto de datos de flores `Iris`:

````python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Cargar el conjunto de datos Iris
iris = load_iris()

# Convertir los datos en un DataFrame de pandas
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(df.drop('target', axis=1), df['target'], test_size=0.3, random_state=42)

# Escalar los datos de entrenamiento y prueba
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear un clasificador KNN con k=5
knn = KNeighborsClassifier(n_neighbors=5)

# Entrenar el clasificador con los datos escalados de entrenamiento
knn.fit(X_train_scaled, y_train)

# Calcular la precisión del clasificador en los datos escalados de prueba
accuracy = knn.score(X_test_scaled, y_test)
print(f"Precisión del clasificador KNN después del preprocesamiento de datos: {accuracy:.2f}")
````

El código anterior carga el conjunto de datos `Iris`, lo convierte en un DataFrame de `pandas` y divide los datos en conjunto de entrenamiento y prueba. Luego, escala los datos utilizando ```StandardScaler```, un método comúnmente utilizado para escalar características numéricas. Finalmente, crea un clasificador `KNN` con `k=5`, lo entrena con los datos escalados de entrenamiento y calcula la precisión del clasificador en los datos escalados de prueba.

Aquí hay un ejemplo de cómo se puede visualizar el efecto del preprocesamiento de datos en la precisión del clasificador `KNN`:

````python
import matplotlib.pyplot as plt

# Calcular la precisión del clasificador para diferentes valores de k
accuracies = []
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    accuracy = knn.score(X_test_scaled, y_test)
    accuracies.append(accuracy)

# Visualizar los resultados
plt.plot(range(1, 21), accuracies)
plt.xlabel('Valor de k')
plt.ylabel('Precisión')
plt.title('Efecto del preprocesamiento de datos en la precisión del clasificador KNN')
plt.show()
````

El código anterior crea un clasificador `KNN` para diferentes valores de `k` y calcula la precisión del clasificador en los datos escalados de prueba. Luego, visualiza cómo cambia la precisión del clasificador para diferentes valores de `k`.

## **Construcción del modelo**

````python
# Importar librerías
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.datasets import fetch_openml
import numpy as np

# Cargar el conjunto de datos Boston Housing
boston = fetch_openml(name='boston')

# Dividir el conjunto de datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.3)

# Crear un regresor KNN con k=5
knn = KNeighborsRegressor(n_neighbors=5)

# Entrenar el regresor con el conjunto de entrenamiento
knn.fit(X_train, y_train)

# Predecir los valores de precio para el conjunto de prueba
y_pred = knn.predict(X_test)

# Calcular el error cuadrático medio del regresor en el conjunto de prueba
mse = np.mean((y_pred - y_test) ** 2)
print(f"Error cuadrático medio del regresor KNN: {mse:.2f}")
````

En este ejemplo, cargamos el conjunto de datos Boston Housing y lo dividimos en un conjunto de entrenamiento y un conjunto de prueba. Luego, creamos un regresor KNN con ```k=5``` y lo entrenamos con el conjunto de entrenamiento. A continuación, utilizamos el regresor entrenado para predecir los valores de precio para el conjunto de prueba y calculamos el error cuadrático medio del regresor en el conjunto de prueba.

## **Predicción de un rango de precios**

````python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

diamonds = sns.load_dataset('diamonds')

# Convertir variables categóricas en variables ficticias
diamonds = pd.get_dummies(diamonds, columns=['cut', 'color', 'clarity'], drop_first=True)

X = diamonds.drop(['price'], axis=1)
y = diamonds['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

plt.scatter(y_test, y_pred)
plt.plot([0, np.max(y)], [0, np.max(y)], '--k')
plt.xlabel('Precio Real')
plt.ylabel('Precio Predicho')
plt.show()
````

## **KNN con diferentes valores de K**

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Generar datos de clasificación aleatorios
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0,
                           n_informative=2, random_state=1, n_clusters_per_class=1)

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Crear una lista de valores de K para probar
k_values = [1, 3, 5, 7, 9]

# Crear una figura para visualizar los resultados
fig, axs = plt.subplots(nrows=1, ncols=len(k_values), figsize=(20, 5))

# Iterar sobre los valores de K y ajustar un modelo para cada uno
for i, k in enumerate(k_values):
    # Crear un clasificador KNN con k vecinos
    knn = KNeighborsClassifier(n_neighbors=k)
    # Entrenar el clasificador con los datos de entrenamiento
    knn.fit(X_train, y_train)
    # Calcular la precisión del clasificador en los datos de prueba
    accuracy = knn.score(X_test, y_test)
    # Visualizar la frontera de decisión del clasificador en la figura
    xx, yy = np.meshgrid(np.linspace(-4, 4, 100), np.linspace(-4, 4, 100))
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    axs[i].contourf(xx, yy, Z, cmap=plt.cm.RdBu, alpha=0.5)
    # Visualizar los datos de entrenamiento y prueba en la figura
    axs[i].scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.RdBu, edgecolor='black')
    axs[i].scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=plt.cm.RdBu, alpha=0.6, edgecolor='black')
    axs[i].set_title(f'K = {k}, Precisión = {accuracy:.2f}')

plt.show()
````
