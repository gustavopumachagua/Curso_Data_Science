# **36. Descubre c√≥mo Visualizar Datos Categ√≥ricos con Seaborn y Python: gr√°ficas de viol√≠n, enjambre y ‚ûï**

## **Visualizaci√≥n con datos categ√≥ricos**

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
frutas = ['manzanas', 'manzanas', 'pl√°tanos', 'naranjas', 'manzanas', 'pl√°tanos', 'peras', 'naranjas', 'naranjas']

# Calcular la frecuencia de cada tipo de fruta
frecuencias = {}
for fruta in frutas:
    if fruta in frecuencias:
        frecuencias[fruta] += 1
    else:
        frecuencias[fruta] = 1

# Visualizar los resultados en un gr√°fico de barras
nombres_frutas = sorted(frecuencias.keys())
frecuencias_ordenadas = [frecuencias[fruta] for fruta in nombres_frutas]

plt.bar(nombres_frutas, frecuencias_ordenadas)
plt.title('Frecuencia de Frutas')
plt.xlabel('Fruta')
plt.ylabel('Frecuencia')
plt.show()
```

## **Tipos de gr√°ficas categ√≥ricas**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos de ejemplo
datos = {'Bebida': ['Caf√©', 'T√©', 'Refresco', 'Agua', 'Jugo'],
         'Frecuencia': [20, 15, 10, 30, 25]}

# Convertir los datos a un DataFrame de pandas
df = pd.DataFrame(datos)

# Gr√°fico de barras
plt.bar(df['Bebida'], df['Frecuencia'])
plt.title('Preferencias de Bebidas')
plt.xlabel('Bebida')
plt.ylabel('Frecuencia')
plt.show()

# Gr√°fico de pastel
plt.pie(df['Frecuencia'], labels=df['Bebida'])
plt.title('Preferencias de Bebidas')
plt.show()

# Gr√°fico de barras horizontales
plt.barh(df['Bebida'], df['Frecuencia'])
plt.title('Preferencias de Bebidas')
plt.xlabel('Frecuencia')
plt.ylabel('Bebida')
plt.show()
```

En este ejemplo, los datos se representan en un DataFrame de pandas con dos columnas: 'Bebida' y 'Frecuencia'. Para visualizar los datos, se utilizan tres tipos de gr√°ficas categ√≥ricas diferentes: un gr√°fico de barras, un gr√°fico de pastel y un gr√°fico de barras horizontales.

En el primer gr√°fico de barras, se traza la frecuencia de cada bebida en un eje vertical. El segundo gr√°fico es un gr√°fico de pastel que muestra la proporci√≥n de cada bebida en el total de las preferencias. Por √∫ltimo, el gr√°fico de barras horizontales muestra la frecuencia de cada bebida en un eje horizontal.

Cada tipo de gr√°fico tiene sus ventajas y desventajas, por lo que es importante elegir el tipo adecuado para los datos que se desean visualizar.

## **Gr√°ficas de bandas (stripplot)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar un conjunto de datos de ejemplo de Seaborn
tips = sns.load_dataset("tips")

# Crear una gr√°fica de bandas con Seaborn
sns.stripplot(x="day", y="total_bill", data=tips)

# Personalizar la gr√°fica
plt.title("Distribuci√≥n del total de la cuenta por d√≠a")
plt.xlabel("D√≠a de la semana")
plt.ylabel("Total de la cuenta ($)")
plt.show()
```

## **Gr√°ficas de enjambre (swarmplot)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar un conjunto de datos de ejemplo de Seaborn
tips = sns.load_dataset("tips")

# Crear una gr√°fica de enjambre con Seaborn
sns.swarmplot(x="day", y="total_bill", data=tips)

# Personalizar la gr√°fica
plt.title("Distribuci√≥n del total de la cuenta por d√≠a")
plt.xlabel("D√≠a de la semana")
plt.ylabel("Total de la cuenta ($)")
plt.show()
```

## **Diagramas de cajas y bigotes (boxplot)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar un conjunto de datos de ejemplo de Seaborn
tips = sns.load_dataset("tips")

# Crear un diagrama de cajas y bigotes con Seaborn
sns.boxplot(x="day", y="total_bill", data=tips)

# Personalizar la gr√°fica
plt.title("Distribuci√≥n del total de la cuenta por d√≠a")
plt.xlabel("D√≠a de la semana")
plt.ylabel("Total de la cuenta ($)")
plt.show()
```

## **Gr√°ficas de puntos (pointplot)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar un conjunto de datos de ejemplo de Seaborn
tips = sns.load_dataset("tips")

# Crear una gr√°fica de puntos con Seaborn
sns.pointplot(x="day", y="tip", data=tips)

# Personalizar la gr√°fica
plt.title("Propina promedio por d√≠a de la semana")
plt.xlabel("D√≠a de la semana")
plt.ylabel("Propina promedio ($)")
plt.show()
```

## **Gr√°ficas de barras (barplots)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar un conjunto de datos de ejemplo de Seaborn
titanic = sns.load_dataset("titanic")

# Crear una gr√°fica de barras con Seaborn
sns.barplot(x="class", y="survived", data=titanic)

# Personalizar la gr√°fica
plt.title("Porcentaje de supervivencia por clase en el Titanic")
plt.xlabel("Clase")
plt.ylabel("Porcentaje de supervivencia")
plt.show()
```

## **Gr√°ficas de conteo (countplot)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar un conjunto de datos de ejemplo de Seaborn
titanic = sns.load_dataset("titanic")

# Crear una gr√°fica de conteo con Seaborn
sns.countplot(x="class", hue="who", data=titanic)

# Personalizar la gr√°fica
plt.title("Distribuci√≥n de pasajeros del Titanic por clase y g√©nero")
plt.xlabel("Clase")
plt.ylabel("N√∫mero de pasajeros")
plt.show()
```
