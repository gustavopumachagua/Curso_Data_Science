# **10. Gradiente Descendente: Un Algoritmo de Optimización para Machine Learning**

## **Introducción**

El Gradiente Descendente `(GD)` es un algoritmo de optimización utilizado en Machine Learning para encontrar el mínimo de una función de coste, lo que ayuda a ajustar los parámetros del modelo para que se ajusten mejor a los datos.

En términos simples, el Gradiente Descendente implica ajustar iterativamente los parámetros de un modelo en la dirección que maximiza la reducción del error de predicción. El `GD` se basa en el cálculo de la derivada de la función de coste con respecto a los parámetros del modelo, y luego ajusta los parámetros en la dirección de la derivada más negativa (es decir, la dirección en la que la función de coste disminuye más rápidamente).

Algunos ejemplos de algoritmos de Machine Learning que utilizan el Gradiente Descendente son:

* **Regresión lineal:**

 En la regresión lineal, el Gradiente Descendente se utiliza para ajustar los parámetros de la línea que mejor se ajusta a los datos.

* **Redes Neuronales:**

 En las redes neuronales, el Gradiente Descendente se utiliza para ajustar los pesos y sesgos de cada neurona en la dirección que minimiza el error de predicción.

* **Regresión logística:**

 En la regresión logística, el Gradiente Descendente se utiliza para ajustar los parámetros que definen la curva sigmoidea que mejor se ajusta a los datos.

* **Máquinas de Vectores de Soporte (SVM):**

 En las `SVM`, el Gradiente Descendente se utiliza para ajustar los pesos que definen el hiperplano que mejor separa las clases de datos.

En resumen, el Gradiente Descendente es un algoritmo clave en Machine Learning que se utiliza en una amplia variedad de aplicaciones.

## **Optimizando un modelo de regresión lineal**

Optimizar un modelo de regresión lineal implica ajustar los parámetros del modelo para que se ajusten mejor a los datos de entrenamiento y, por lo tanto, puedan realizar mejores predicciones sobre nuevos datos.

Aquí hay algunos ejemplos de cómo se puede optimizar un modelo de regresión lineal:

* **Selección de características:**

 En lugar de usar todas las características disponibles en los datos, es posible que algunas de ellas no sean relevantes para la predicción y pueden incluso empeorar el rendimiento del modelo. Por lo tanto, es importante seleccionar cuidadosamente las características que se utilizarán en el modelo. Esto se puede hacer mediante métodos de selección de características como la eliminación recursiva de características `(Recursive Feature Elimination - RFE)` o la selección basada en árboles `(Tree-based feature selection)`.

* **Normalización de características:**

 Las características pueden tener diferentes escalas y esto puede afectar la precisión del modelo. Por lo tanto, es importante normalizar las características para que tengan una escala similar. Esto se puede hacer mediante técnicas de normalización como la estandarización (restar la media y dividir por la desviación estándar) o la normalización de rango (escalar las características en un rango común, por ejemplo, entre 0 y 1).

* **Regularización:**

 La regularización se utiliza para evitar el sobreajuste `(overfitting)` del modelo. Esto se logra al agregar un término de penalización a la función de costo del modelo, que favorece soluciones más simples. Las técnicas comunes de regularización incluyen la regresión de `Ridge` y la regresión de `Lasso`.

* **Optimización de hiperparámetros:**

 Los hiperparámetros son parámetros del modelo que no se ajustan durante el entrenamiento, como la tasa de aprendizaje o la cantidad de regularización. La optimización de hiperparámetros implica encontrar los valores óptimos para estos parámetros que maximizan el rendimiento del modelo. Esto se puede hacer utilizando métodos como la validación cruzada y la búsqueda de cuadrícula `(grid search)`.

* **Ajuste de modelos:**

 El ajuste de modelos implica experimentar con diferentes modelos de regresión lineal, como la regresión lineal simple, la regresión lineal múltiple, la regresión polinómica, etc. Se puede utilizar una técnica de selección de modelos para determinar qué modelo se ajusta mejor a los datos.

En general, la optimización de un modelo de regresión lineal implica una combinación de estas técnicas para mejorar el rendimiento del modelo y hacer predicciones más precisas.

### **Regresión lineal Descendiente**

La Regresión Lineal Descendente (Gradient Descent en inglés) es un algoritmo de optimización que se utiliza comúnmente para ajustar los parámetros de un modelo de Regresión Lineal. El objetivo es minimizar la función de costo que mide la diferencia entre las predicciones del modelo y los valores reales.

A continuación se presenta un ejemplo de cómo implementar la Regresión Lineal Descendente en Python usando la biblioteca `NumPy`:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Agregar término de sesgo al conjunto de datos X
X_b = np.c_[np.ones((100, 1)), X]

# Definir tasa de aprendizaje, número de iteraciones y tamaño del lote
eta = 0.1
n_iterations = 1000
m = 100

# Inicializar los parámetros del modelo de manera aleatoria
theta = np.random.randn(2,1)

# Implementar la Regresión Lineal Descendente
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients

# Imprimir los parámetros del modelo
print("Parámetros del modelo: ")
print(theta)

# Graficar los datos y la línea de regresión
plt.plot(X, y, "b.")
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta)
plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predicciones")
plt.xlabel("X", fontsize=14)
plt.ylabel("y", rotation=0, fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.show()
````

En este ejemplo, se generan datos aleatorios para simular un problema de regresión lineal. El algoritmo de Regresión Lineal Descendente se implementa en un bucle de iteraciones que actualiza los parámetros del modelo en la dirección de los gradientes de la función de costo. Finalmente, se grafica la línea de regresión junto con los datos.

Es importante señalar que existen variaciones de la Regresión Lineal Descendente, como la Regresión Lineal Descendente Estocástica `(Stochastic Gradient Descent en inglés)` y la Regresión Lineal Descendente en Lotes `(Batch Gradient Descent en inglés)`. Cada variación tiene sus propias ventajas y desventajas en términos de velocidad y precisión.

````python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from sklearn.linear_model import LinearRegression

#Variable independiente
experiencia = np.array([[5], [6], [13]])

#Variable objetivo
salario = np.array([6.85, 16.83, 26.84])

#Modelo linear creado con el metodo de minimos cuadrados
modelo = LinearRegression().fit(experiencia.reshape(3, -1), salario)

#Visualizacion de datos y modelo
print("Interseccion con eje y (b): %0.2f" % modelo.intercept_)
print("Pendiente (m): %0.2f" % modelo.coef_[0])
print("Suma de cuadrados de los residuos (RSS): %0.2f" %
((salario - modelo.predict(experiencia))**2).sum())

plt.figure(figsize=(8, 6))

plt.scatter(experiencia, salario, color="gold", s=250,
marker="o", label="Valor verdadero")

plt.scatter(experiencia, modelo.predict(experiencia),
color="blue", s=250, marker="P", label="Valor Predicho")

plt.plot(experiencia, modelo.predict(experiencia),
linewidth=4, color="deeppink", label="Modelo lineal")

experiencia=experiencia.reshape(3)

plt.ylabel("Salario en miles de Pesos ($)", size=16)
plt.xlabel("Años de Experiencia", size=16)
plt.title("Regresión lineal simple: salario en función de la experiencia", size=18)
plt.legend(bbox_to_anchor=(1.3, 0.5), prop={'size': 16})
plt.grid()
plt.box(False)
plt.show()
````

## **Diferentes modelos con diferentes pendientes**

Los diferentes modelos con diferentes pendientes se refieren a la variación de la pendiente de la recta de regresión lineal para ajustar a los datos de manera más precisa. A continuación, se muestra un ejemplo de cómo generar diferentes modelos con diferentes pendientes utilizando la biblioteca `NumPy` y `matplotlib` en Python:

````python
# Creacion de multiples pendientes para exploracion
pendientes = np.arange(2.5, 1.6, -0.1)

# Vector para almacenar el error de los diferentes modelos
errores = np.array([])

# Visualizacion de modelos
plt.figure(figsize=(8, 6))

for pendiente in pendientes:
  # Error del modelo (suma de cuadrados de los residuos)
  error = ((pendiente*experiencia - salario)**2).sum()
  
  # Visualizacion de un modelo para una pendiente dada
  plt.plot(experiencia, pendiente*experiencia, linewidth=4,
           label="m: %0.2ff | error: %0.2f" %
           (pendiente, error))
  errores = np.append(errores, error)

plt.scatter(experiencia, salario, color="gold", s=250,
            marker="o", label="Valor verdadero")

plt.ylabel("Salario en miles de Pesos ($)", size=16)
plt.xlabel("Años de Experiencia", size=16)
plt.legend(bbox_to_anchor=(1, 0.5))
plt.grid()
plt.box(False)
plt.show()
````

## **Visualización del error (RSS)**

La visualización del error `RSS` `(Residual Sum of Squares)` nos permite evaluar el rendimiento de nuestro modelo de regresión lineal al graficar el valor del `RSS` en función del número de iteraciones. Esto nos permite identificar si el modelo está convergiendo o no, y si es necesario ajustar los hiperparámetros para mejorar su desempeño.

Aquí hay un ejemplo de cómo visualizar el error `RSS` en función del número de iteraciones:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Agregar término de sesgo al conjunto de datos X
X_b = np.c_[np.ones((100, 1)), X]

# Definir tasa de aprendizaje, número de iteraciones y tamaño del lote
eta = 0.1
n_iterations = 1000
m = 100

# Inicializar los parámetros del modelo de manera aleatoria
theta = np.random.randn(2,1)

# Crear una lista para almacenar los valores del RSS en cada iteración
rss_list = []

# Implementar la Regresión Lineal Descendente y calcular el RSS en cada iteración
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
    y_predict = X_b.dot(theta)
    rss = np.sum((y_predict - y)**2)
    rss_list.append(rss)

# Graficar el error RSS en función del número de iteraciones
plt.plot(range(1, n_iterations+1), rss_list)
plt.xlabel('Número de iteraciones')
plt.ylabel('RSS')
plt.show()
````

En este ejemplo, creamos una lista llamada ` rss_list ` para almacenar los valores del `RSS` en cada iteración del bucle. Luego, calculamos el valor del `RSS` utilizando la fórmula `RSS = sum((y_predict - y)^2)` en cada iteración y lo agregamos a la lista. Finalmente, graficamos el valor del `RSS` en función del número de iteraciones.

Al graficar el error `RSS` en función del número de iteraciones, podemos ver cómo disminuye el error a medida que el modelo se ajusta a los datos. En general, esperamos que el error disminuya a medida que el número de iteraciones aumenta, y que eventualmente se estabilice en un valor mínimo. Si el error no disminuye o fluctúa mucho, es posible que necesitemos ajustar los hiperparámetros del modelo (como la tasa de aprendizaje) para mejorar su desempeño.

````python
plt.figure(figsize=(7.5, 6))
plt.title("Suma de cuadrados de los residuos (RSS)", size=16)
plt.ylabel("Error (RSS)", size=16)
plt.xlabel("Pendiente (m)", size=16)
plt.scatter(pendientes, errores, color="purple", marker="D", s=99)
plt.grid()
plt.box(False)
plt.show()
````

## **Creación y derivada del error E(m)**

### **Funcion de Error a ser Minimizada**

La función de error a ser minimizada en la regresión lineal es la suma de los errores cuadrados (RSS, por sus siglas en inglés), que mide la diferencia entre los valores predichos por el modelo y los valores reales de la variable objetivo. La función `RSS` se define como:

`RSS = ∑(y - ŷ)²`

donde `y` son los valores reales de la variable objetivo, ```ŷ``` son los valores predichos por el modelo y la sumatoria se extiende a través de todas las observaciones en el conjunto de datos. El objetivo de la regresión lineal es encontrar los valores óptimos de los parámetros del modelo (intercepto y pendiente) que minimizan esta función `RSS`.

**Suma de los cuadrados de los residuos (RSS)**

La suma de los cuadrados de los residuos `(RSS)` es una medida del error residual de un modelo de regresión lineal. Se calcula sumando los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo para cada punto de datos en el conjunto de entrenamiento.

Matemáticamente, se puede expresar como:

`RSS = Σ(yᵢ - ŷᵢ)²`

**Donde:**

* `RSS` es la suma de los cuadrados de los residuos
* `yᵢ` es el valor observado en el punto `i`
* `ŷᵢ` es el valor predicho por el modelo en el punto `i`

El objetivo del modelo de regresión lineal es minimizar la `RSS`. Esto se logra ajustando los valores de los parámetros del modelo para que la línea de regresión se ajuste lo mejor posible a los datos. La `RSS` se puede utilizar para evaluar la bondad de ajuste del modelo, es decir, cuánto se ajusta el modelo a los datos. Un valor más bajo de `RSS` indica un mejor ajuste del modelo a los datos.

Es importante destacar que la `RSS` es solo una medida del error residual en el conjunto de entrenamiento y no debe utilizarse como única medida de rendimiento del modelo. Para evaluar el rendimiento del modelo de regresión lineal, se deben utilizar otras métricas, como el coeficiente de determinación `(R²)` o el error de predicción.

````python
# persona1: (6.85, 5) persona2: (16.83, 6) persona3: (26.84, 13)
m= sym.Symbol("m")

# Funcion de error
error=(6.85 - m*5)**2 + (16.83 - m*6)**2 + (26.84 - m*13)**2

# Derivada de la funcion de error
derivada = sym.diff(error, m)

for pendiente in pendientes:
  print(derivada, "Evaluacion %0.2f" % derivada.evalf(subs={m : pendiente}))
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Agregar término de sesgo al conjunto de datos X
X_b = np.c_[np.ones((100, 1)), X]

# Definir diferentes valores para la pendiente de la recta de regresión lineal
theta_range = np.linspace(-5, 5, 100)

# Calcular el RSS para cada valor de theta
rss_values = []
for theta in theta_range:
    y_predict = X_b.dot(np.array([[4], [theta]]))
    rss = ((y_predict - y) ** 2).sum()
    rss_values.append(rss)

# Graficar el valor de RSS para cada valor de theta
plt.plot(theta_range, rss_values, "b-")
plt.xlabel("Pendiente (theta_1)", fontsize=14)
plt.ylabel("RSS", fontsize=14)
plt.show()
````

Este código calcula el `RSS` para diferentes valores de la pendiente de la recta de regresión lineal, y grafica el valor de `RSS` en función de la pendiente. Esto permite visualizar el valor de `RSS` para diferentes valores de la pendiente, lo que ayuda a identificar el valor óptimo de la pendiente que minimiza el error. En este ejemplo, podemos ver que el valor óptimo de la pendiente parece estar cerca de `3`.

### **Visualizando -Derivada de la funcion de Error**

La derivada de la función de error en relación a los parámetros del modelo es una herramienta importante en la regresión lineal. Permite encontrar los valores de los parámetros que minimizan la función de error, lo cual es esencial para encontrar una buena aproximación lineal a los datos.

Aquí hay un ejemplo de visualización de la derivada de la función de error para una regresión lineal simple:

````python
import numpy as np
import matplotlib.pyplot as plt

# Datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Añadir término de sesgo
X_b = np.c_[np.ones((100, 1)), X]

# Definir función de error y su derivada
def mse(theta, X, y):
    m = len(y)
    predictions = X.dot(theta)
    error = predictions - y
    return (1/m) * np.sum(np.square(error))

def mse_derivative(theta, X, y):
    m = len(y)
    predictions = X.dot(theta)
    error = predictions - y
    return (2/m) * X.T.dot(error)

# Calcular valores de error y su derivada para diferentes valores de theta
theta_range = np.linspace(-5, 5, 100)
errors = []
derivatives = []
for theta in theta_range:
    errors.append(mse(np.array([[4], [theta]]), X_b, y))
    derivatives.append(mse_derivative(np.array([[4], [theta]]), X_b, y)[1])

# Graficar la función de error y su derivada
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Pendiente')
ax1.set_ylabel('Error', color=color)
ax1.plot(theta_range, errors, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('Derivada del Error', color=color)
ax2.plot(theta_range, derivatives, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
````

Este código calcula los valores de error y su derivada para diferentes valores de pendiente y los grafica en dos ejes diferentes. La curva `roja` muestra la función de error y la curva `azul` muestra su derivada. El punto donde la derivada es cero indica el valor óptimo de la pendiente que minimiza el error. En este ejemplo, ese valor es aproximadamente `3`.

````python
plt.figure(figsize=(7.5, 6))

for i in range(0, len(errores), 1):
  # Error de un modelo dado
  plt.scatter(pendientes[i], errores[i],
              label="%0.2f" % derivada.evalf(subs={m: pendientes[i]}),
              marker="D", s=200)
  
  # Evalucaion de la derivada para un error dado
  pendiente= derivada.evalf(subs={m: pendientes[i]})
  
  plt.plot(np.array([1.7, 2.5]),
           np.array([1.7, 2.5])*pendiente + 
           pendiente*(-2.3 + i*0.05) + errores.min())

plt.title("Derivada de la funcion de Error", size=16)
plt.ylabel("Error (RSS)", size=16)
plt.xlabel("Pendiente (m)", size=16)
plt.ylim(30, 70)
plt.grid()
plt.box(False)
plt.show()
````

## **Generalización para dos o más parámetros**

La generalización para dos o más parámetros es una extensión del modelo de regresión lineal simple a situaciones en las que hay más de una variable independiente o predictor. En este caso, se utiliza una función lineal de la forma:

```y = θ0 + θ1x1 + θ2x2 + ... + θnxn```

Donde `y` es la variable dependiente, ```θ0``` es el término de sesgo, ```θ1, θ2, ..., θn``` son los parámetros que se deben estimar y ```x1, x2, ..., xn``` son las variables independientes o predictoras.

El objetivo es encontrar los valores de los parámetros ```θ0, θ1, θ2, ..., θn``` que minimizan la función de error `RSS`. Esto se hace utilizando técnicas de optimización, como el método de descenso de gradiente o la ecuación normal.

Una vez que se han encontrado los valores de los parámetros, se pueden utilizar para hacer predicciones sobre nuevos datos utilizando la misma función lineal.

El proceso de encontrar los valores de los parámetros se puede realizar utilizando bibliotecas de aprendizaje automático, como `Scikit-learn` en Python, o mediante programación manual utilizando librerías de álgebra lineal como `Numpy`.

El resultado final es un modelo de regresión lineal múltiple que puede ser utilizado para hacer predicciones sobre nuevos datos.

### **Generalizando para 2 o mas parametros: m y b**

La función de error de la regresión lineal con dos parámetros (pendiente y término de intersección) está dada por:

$J(\theta_0, \theta_1) = \frac{1}{2m} \sum\limits_{i=1}^{m} (h_{\theta}(x^{(i)}) - y^{(i)})^2$

Donde $h_{\theta}(x^{(i)})$ es la predicción para el i-ésimo ejemplo de entrenamiento dada por $h_{\theta}(x^{(i)}) = \theta_0 + \theta_1 x^{(i)}$, $y^{(i)}$ es la etiqueta verdadera para el i-ésimo ejemplo de entrenamiento y $m$ es el número de ejemplos de entrenamiento.

Aquí está el código para la visualización de la función de error con dos parámetros:

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de error para la regresión lineal con dos parámetros
def J(theta0, theta1, X, y):
    m = len(y)
    h = X.dot(np.array([[theta0], [theta1]]))
    return (1/(2*m)) * np.sum((h-y)**2)

# Generar datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Agregar término de sesgo al conjunto de datos X
X_b = np.c_[np.ones((100, 1)), X]

# Definir rangos de valores para los parámetros theta0 y theta1
theta0_range = np.linspace(-10, 10, 100)
theta1_range = np.linspace(-5, 5, 100)

# Crear una malla de valores de theta0 y theta1
theta0, theta1 = np.meshgrid(theta0_range, theta1_range)

# Calcular los valores de J para cada par de valores de theta0 y theta1
J_vals = np.zeros((len(theta0_range), len(theta1_range)))
for i, theta0_val in enumerate(theta0_range):
    for j, theta1_val in enumerate(theta1_range):
        J_vals[i, j] = J(theta0_val, theta1_val, X_b, y)

# Graficar la función de error como una superficie
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(theta0, theta1, J_vals, cmap='viridis')
ax.set_xlabel(r'$\theta_0$')
ax.set_ylabel(r'$\theta_1$')
ax.set_zlabel(r'$J(\theta_0, \theta_1)$')
plt.show()
````

La gráfica resultante muestra la superficie de la función de error en el espacio de parámetros $\theta_0$ y $\theta_1$. Se puede ver que hay un mínimo global en el punto $(4, 3)$, que corresponde a los valores verdaderos de los parámetros en los datos generados aleatoriamente.

Generalizando para 2 o más parámetros, se considera una función de error que depende tanto de la pendiente como de la intersección con el eje `y`:

$J(\theta) = \frac{1}{2m} \sum\limits_{i=1}^{m} (h_{\theta}(x^{(i)}) - y^{(i)})^2$
 
donde $\theta$ es un vector de parámetros que incluye tanto la pendiente $m$ como la intersección $b$, y $h_{\theta}(x)$ es la función de predicción lineal:

$h_{\theta}(x)=\theta^{(T)}x = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + ... + \theta_n x_n$
 
La idea es encontrar los valores de $\theta$ que minimizan la función de error $J(\theta)$. Para esto se pueden aplicar algoritmos de optimización, como el descenso de gradiente.

A continuación se muestra un ejemplo de cómo aplicar el descenso de gradiente para encontrar los valores óptimos de los parámetros $m$ y $b$ para un conjunto de datos bidimensionales.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Agregar término de sesgo al conjunto de datos X
X_b = np.c_[np.ones((100, 1)), X]

# Definir tasa de aprendizaje, número de iteraciones y tamaño del lote
eta = 0.1
n_iterations = 1000
m = 100

# Inicializar valores de los parámetros
theta = np.random.randn(2,1)

# Aplicar descenso de gradiente para encontrar valores óptimos de los parámetros
for iteration in range(n_iterations):
    gradients = 1/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients

# Imprimir valores de los parámetros encontrados
print("Valor óptimo de b (intercepto):", theta[0][0])
print("Valor óptimo de m (pendiente):", theta[1][0])

# Graficar los datos y la recta de regresión lineal resultante
plt.plot(X, y, "b.")
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta)
plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predicción")
plt.xlabel("X", fontsize=14)
plt.ylabel("y", rotation=0, fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.show()
````

En la gráfica resultante se puede observar la nube de puntos correspondiente al conjunto de datos y la recta de regresión lineal resultante, encontrada mediante el descenso de gradiente.

## **Creación y derivada parcial del error E(m, b)**

La creación y derivada parcial del error `E(m, b)` se refiere a la creación de una función de error que tenga en cuenta tanto el parámetro de la pendiente `m` como el término de intersección `b` en un modelo de regresión lineal. Luego, se derivan parcialmente la función de error con respecto a cada parámetro para encontrar los valores óptimos que minimizan la función de error.

La función de error para el modelo de regresión lineal con `m` y `b` puede definirse como:

```E(m, b) = 1/2m * sum((y_i - (mx_i + b))^2)```

donde `m` es la pendiente, `b` es la intersección, `y_i` es el valor objetivo correspondiente al dato de entrada `x_i`.

Para encontrar los valores óptimos de `m` y `b`, se puede tomar la derivada parcial de `E(m, b)` con respecto a cada parámetro, y luego igualar a cero. De esta manera, se pueden obtener las siguientes ecuaciones para `m` y `b`:

```∂E/∂m = -1/m * sum(x_i(y_i - (mx_i + b)))```

```∂E/∂b = -1/m * sum(y_i - (mx_i + b))```

A continuación se presenta un ejemplo de código que crea y deriva parcialmente la función de error `E(m, b)` para un conjunto de datos de regresión lineal:

````python
import numpy as np

# Crear conjunto de datos de regresión lineal
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Definir función de error
def error_function(X, y, m, b):
    m = np.array([m]).reshape(1, 1)
    b = np.array([b]).reshape(1, 1)
    y_hat = np.dot(X, m) + b
    error = 1 / (2 * len(X)) * np.sum(np.square(y_hat - y))
    return error

# Calcular derivadas parciales
def partial_derivative_m(X, y, m, b):
    m = np.array([m]).reshape(1, 1)
    b = np.array([b]).reshape(1, 1)
    y_hat = np.dot(X, m) + b
    derivative = -1 / len(X) * np.sum(np.multiply(X, y - y_hat))
    return derivative

def partial_derivative_b(X, y, m, b):
    m = np.array([m]).reshape(1, 1)
    b = np.array([b]).reshape(1, 1)
    y_hat = np.dot(X, m) + b
    derivative = -1 / len(X) * np.sum(y - y_hat)
    return derivative

# Inicializar valores para m y b
m = 0
b = 0
learning_rate = 0.01
n_iterations = 1000

# Gradiente descendente
for i in range(n_iterations):
    m -= learning_rate * partial_derivative_m(X, y, m, b)
    b -= learning_rate * partial_derivative_b(X, y, m, b)

# Imprimir valores de m y b
print("Valor de m:", m)
print("Valor de b:", b)
````

La salida de este código será los valores de `m` y `b` óptimos que minimizan la función de error. Estos valores se pueden utilizar para trazar la línea.

### **Derivadas parciales de la Funcion de Error**

Las derivadas parciales de la función de error nos permiten calcular el gradiente de la función de error con respecto a los parámetros del modelo. Esto es útil para actualizar los valores de los parámetros en cada iteración del algoritmo de optimización.

A continuación se presenta un ejemplo de cómo calcular las derivadas parciales de la función de error con respecto a los parámetros `m` y `b` en Python:

````python
import numpy as np

# Definir función de error
def error_function(x, y, m, b):
    y_pred = m * x + b
    error = np.mean((y - y_pred) ** 2)
    return error

# Calcular derivadas parciales
def partial_derivatives(x, y, m, b):
    dm = np.mean(-2 * x * (y - (m * x + b)))
    db = np.mean(-2 * (y - (m * x + b)))
    return dm, db

# Ejemplo de uso
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Inicializar parámetros
m = 1
b = 0

# Calcular error y derivadas parciales
error = error_function(x, y, m, b)
dm, db = partial_derivatives(x, y, m, b)

# Imprimir resultados
print("Error:", error)
print("Derivada parcial con respecto a m:", dm)
print("Derivada parcial con respecto a b:", db)
````

En este ejemplo, la función ```error_function``` calcula el error cuadrático medio entre las predicciones del modelo y los valores reales de la variable objetivo. La función ```partial_derivatives``` calcula las derivadas parciales de la función de error con respecto a los parámetros `m` y `b`.

## **Pseudocódigo del gradiente descendente**

El pseudocódigo del algoritmo de gradiente descendente se puede describir de la siguiente manera:

1. Inicializar los pesos y el umbral con valores aleatorios cercanos a cero.
2. Seleccionar un tamaño de paso (también conocido como tasa de aprendizaje).
3. Repetir hasta que se alcance un criterio de parada (como un número máximo de iteraciones o una precisión deseada):
a. Calcular la salida de la red para un conjunto de entrada.

b. Calcular el error cometido por la red en la salida.

c. Calcular la tasa de cambio del error con respecto a los pesos y el umbral.

d. Actualizar los pesos y el umbral utilizando la tasa de cambio del error y la tasa de aprendizaje.

4. Devolver los pesos y el umbral final.

A continuación se muestra un ejemplo de código en Python que implementa el algoritmo de gradiente descendente para ajustar una línea a un conjunto de datos:

````python
# Importar bibliotecas necesarias
import numpy as np

# Definir la función de costo
def costo(X, y, theta):
    m = len(y)
    J = 1/(2*m) * np.sum((X.dot(theta) - y)**2)
    return J

# Definir el algoritmo de gradiente descendente
def gradiente_descendente(X, y, theta_inicial, tasa_aprendizaje, num_iteraciones):
    m = len(y)
    theta = theta_inicial.copy()
    costo_historico = np.zeros(num_iteraciones)
    
    for iteracion in range(num_iteraciones):
        error = X.dot(theta) - y
        gradiente = 1/m * X.T.dot(error)
        theta -= tasa_aprendizaje * gradiente
        costo_historico[iteracion] = costo(X, y, theta)
    
    return theta, costo_historico
````

En este ejemplo, ```X``` es una matriz de características de entrada, ```y``` es un vector de valores de salida, ```theta``` es un vector de pesos y el umbral, ```tasa_aprendizaje``` es la tasa de aprendizaje y ```num_iteraciones``` es el número de iteraciones que se desea realizar. La función ```costo``` calcula el costo o error de la red en función de los valores actuales de ```X```, ```y``` y ```theta```. La función ```gradiente_descendente``` implementa el algoritmo de gradiente descendente y devuelve los pesos finales y el historial de costos a lo largo de las iteraciones.

## **Programación del gradiente descendente**

### **Implementacion del Gradiente Descendiente**

El gradiente descendiente es un algoritmo de optimización utilizado en el aprendizaje automático para encontrar los valores óptimos de los parámetros de un modelo. El objetivo es minimizar una función de error, como la suma de los cuadrados de los residuos `(RSS)`, ajustando los parámetros de un modelo de regresión lineal.

La implementación del gradiente descendiente implica la actualización iterativa de los valores de los parámetros utilizando las derivadas parciales de la función de error en relación a cada parámetro. Los valores de los parámetros se ajustan en la dirección opuesta del gradiente de la función de error, es decir, se reduce la función de error en cada iteración.

A continuación se muestra un ejemplo de implementación del gradiente descendiente en Python para ajustar los parámetros de un modelo de regresión lineal.

````python
import numpy as np

# Definir la función de error RSS
def RSS(X, y, theta):
    return np.sum((X.dot(theta) - y) ** 2) / (2 * len(y))

# Implementación del gradiente descendiente
def gradient_descent(X, y, theta, eta, n_iterations):
    m = len(y)
    for iteration in range(n_iterations):
        gradients = 2 / m * X.T.dot(X.dot(theta) - y)
        theta = theta - eta * gradients
        if iteration % 100 == 0:
            print(f"Iteration: {iteration}, Error: {RSS(X, y, theta)}")
    return theta

# Generar datos de ejemplo
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Agregar término de sesgo al conjunto de datos X
X_b = np.c_[np.ones((100, 1)), X]

# Definir valores iniciales para los parámetros
theta = np.random.randn(2, 1)

# Definir tasa de aprendizaje y número de iteraciones
eta = 0.1
n_iterations = 1000

# Aplicar el gradiente descendiente para ajustar los parámetros del modelo
theta_best = gradient_descent(X_b, y, theta, eta, n_iterations)
print(f"Theta Best: {theta_best}")

````

En este ejemplo, se define la función de error `RSS` y se implementa el gradiente descendiente en la función ```gradient_descent```. La función toma como entrada los datos de entrenamiento `X` y `y`, los valores iniciales para los parámetros `theta`, la tasa de aprendizaje `eta` y el número de iteraciones `n_iterations`. En cada iteración, se calculan las derivadas parciales de la función de error en relación a los parámetros y se actualizan los valores de los parámetros en la dirección opuesta del gradiente de la función de error. El resultado final es el conjunto óptimo de valores de los parámetros del modelo.

La salida de este ejemplo incluye la evolución del error en cada iteración y el conjunto óptimo de valores de los parámetros del modelo.

## **Visualización del descenso del error**

Aquí te muestro un ejemplo de cómo visualizar el descenso del error durante el proceso de optimización utilizando el algoritmo del Gradiente Descendente:

Supongamos que tenemos un conjunto de datos ```X``` y ```y``` y queremos encontrar la recta de regresión que mejor se ajuste a estos datos utilizando el algoritmo del Gradiente Descendente. Primero, definimos una función que calcule el error cuadrático medio `(MSE)` para un conjunto de datos y unos parámetros dados:

````python
def mse(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    error = predictions - y
    mse = 1/(2*m) * np.sum(error**2)
    return mse
````

Luego, definimos una función que implemente el algoritmo del Gradiente Descendente:

````python
def gradient_descent(X, y, theta, learning_rate, n_iterations):
    m = len(y)
    history = []
    for i in range(n_iterations):
        predictions = X.dot(theta)
        error = predictions - y
        gradient = 1/m * X.T.dot(error)
        theta = theta - learning_rate * gradient
        history.append(mse(X, y, theta))
    return theta, history
````

Finalmente, utilizamos estas funciones para encontrar la recta de regresión que mejor se ajuste a los datos y visualizar el descenso del error durante el proceso de optimización:

````python
# Generar datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Agregar término de sesgo al conjunto de datos X
X_b = np.c_[np.ones((100, 1)), X]

# Definir tasa de aprendizaje y número de iteraciones
learning_rate = 0.1
n_iterations = 1000

# Definir valor inicial para los parámetros theta
theta = np.random.randn(2,1)

# Aplicar el algoritmo del Gradiente Descendente
theta_final, error_history = gradient_descent(X_b, y, theta, learning_rate, n_iterations)

# Visualizar el descenso del error
plt.plot(range(n_iterations), error_history)
plt.xlabel("Número de Iteraciones")
plt.ylabel("Error Cuadrático Medio")
plt.title("Descenso del Error durante el Gradiente Descendente")
plt.show()
````
