# **10. Gradiente Descendente: Un Algoritmo de Optimizaci贸n para Machine Learning**

## **Introducci贸n**

El Gradiente Descendente `(GD)` es un algoritmo de optimizaci贸n utilizado en Machine Learning para encontrar el m铆nimo de una funci贸n de coste, lo que ayuda a ajustar los par谩metros del modelo para que se ajusten mejor a los datos.

En t茅rminos simples, el Gradiente Descendente implica ajustar iterativamente los par谩metros de un modelo en la direcci贸n que maximiza la reducci贸n del error de predicci贸n. El `GD` se basa en el c谩lculo de la derivada de la funci贸n de coste con respecto a los par谩metros del modelo, y luego ajusta los par谩metros en la direcci贸n de la derivada m谩s negativa (es decir, la direcci贸n en la que la funci贸n de coste disminuye m谩s r谩pidamente).

Algunos ejemplos de algoritmos de Machine Learning que utilizan el Gradiente Descendente son:

* **Regresi贸n lineal:**

 En la regresi贸n lineal, el Gradiente Descendente se utiliza para ajustar los par谩metros de la l铆nea que mejor se ajusta a los datos.

* **Redes Neuronales:**

 En las redes neuronales, el Gradiente Descendente se utiliza para ajustar los pesos y sesgos de cada neurona en la direcci贸n que minimiza el error de predicci贸n.

* **Regresi贸n log铆stica:**

 En la regresi贸n log铆stica, el Gradiente Descendente se utiliza para ajustar los par谩metros que definen la curva sigmoidea que mejor se ajusta a los datos.

* **M谩quinas de Vectores de Soporte (SVM):**

 En las `SVM`, el Gradiente Descendente se utiliza para ajustar los pesos que definen el hiperplano que mejor separa las clases de datos.

En resumen, el Gradiente Descendente es un algoritmo clave en Machine Learning que se utiliza en una amplia variedad de aplicaciones.

## **Optimizando un modelo de regresi贸n lineal**

Optimizar un modelo de regresi贸n lineal implica ajustar los par谩metros del modelo para que se ajusten mejor a los datos de entrenamiento y, por lo tanto, puedan realizar mejores predicciones sobre nuevos datos.

Aqu铆 hay algunos ejemplos de c贸mo se puede optimizar un modelo de regresi贸n lineal:

* **Selecci贸n de caracter铆sticas:**

 En lugar de usar todas las caracter铆sticas disponibles en los datos, es posible que algunas de ellas no sean relevantes para la predicci贸n y pueden incluso empeorar el rendimiento del modelo. Por lo tanto, es importante seleccionar cuidadosamente las caracter铆sticas que se utilizar谩n en el modelo. Esto se puede hacer mediante m茅todos de selecci贸n de caracter铆sticas como la eliminaci贸n recursiva de caracter铆sticas `(Recursive Feature Elimination - RFE)` o la selecci贸n basada en 谩rboles `(Tree-based feature selection)`.

* **Normalizaci贸n de caracter铆sticas:**

 Las caracter铆sticas pueden tener diferentes escalas y esto puede afectar la precisi贸n del modelo. Por lo tanto, es importante normalizar las caracter铆sticas para que tengan una escala similar. Esto se puede hacer mediante t茅cnicas de normalizaci贸n como la estandarizaci贸n (restar la media y dividir por la desviaci贸n est谩ndar) o la normalizaci贸n de rango (escalar las caracter铆sticas en un rango com煤n, por ejemplo, entre 0 y 1).

* **Regularizaci贸n:**

 La regularizaci贸n se utiliza para evitar el sobreajuste `(overfitting)` del modelo. Esto se logra al agregar un t茅rmino de penalizaci贸n a la funci贸n de costo del modelo, que favorece soluciones m谩s simples. Las t茅cnicas comunes de regularizaci贸n incluyen la regresi贸n de `Ridge` y la regresi贸n de `Lasso`.

* **Optimizaci贸n de hiperpar谩metros:**

 Los hiperpar谩metros son par谩metros del modelo que no se ajustan durante el entrenamiento, como la tasa de aprendizaje o la cantidad de regularizaci贸n. La optimizaci贸n de hiperpar谩metros implica encontrar los valores 贸ptimos para estos par谩metros que maximizan el rendimiento del modelo. Esto se puede hacer utilizando m茅todos como la validaci贸n cruzada y la b煤squeda de cuadr铆cula `(grid search)`.

* **Ajuste de modelos:**

 El ajuste de modelos implica experimentar con diferentes modelos de regresi贸n lineal, como la regresi贸n lineal simple, la regresi贸n lineal m煤ltiple, la regresi贸n polin贸mica, etc. Se puede utilizar una t茅cnica de selecci贸n de modelos para determinar qu茅 modelo se ajusta mejor a los datos.

En general, la optimizaci贸n de un modelo de regresi贸n lineal implica una combinaci贸n de estas t茅cnicas para mejorar el rendimiento del modelo y hacer predicciones m谩s precisas.

### **Regresi贸n lineal Descendiente**

La Regresi贸n Lineal Descendente (Gradient Descent en ingl茅s) es un algoritmo de optimizaci贸n que se utiliza com煤nmente para ajustar los par谩metros de un modelo de Regresi贸n Lineal. El objetivo es minimizar la funci贸n de costo que mide la diferencia entre las predicciones del modelo y los valores reales.

A continuaci贸n se presenta un ejemplo de c贸mo implementar la Regresi贸n Lineal Descendente en Python usando la biblioteca `NumPy`:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Agregar t茅rmino de sesgo al conjunto de datos X
X_b = np.c_[np.ones((100, 1)), X]

# Definir tasa de aprendizaje, n煤mero de iteraciones y tama帽o del lote
eta = 0.1
n_iterations = 1000
m = 100

# Inicializar los par谩metros del modelo de manera aleatoria
theta = np.random.randn(2,1)

# Implementar la Regresi贸n Lineal Descendente
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients

# Imprimir los par谩metros del modelo
print("Par谩metros del modelo: ")
print(theta)

# Graficar los datos y la l铆nea de regresi贸n
plt.plot(X, y, "b.")
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta)
plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predicciones")
plt.xlabel("X", fontsize=14)
plt.ylabel("y", rotation=0, fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.show()
````

En este ejemplo, se generan datos aleatorios para simular un problema de regresi贸n lineal. El algoritmo de Regresi贸n Lineal Descendente se implementa en un bucle de iteraciones que actualiza los par谩metros del modelo en la direcci贸n de los gradientes de la funci贸n de costo. Finalmente, se grafica la l铆nea de regresi贸n junto con los datos.

Es importante se帽alar que existen variaciones de la Regresi贸n Lineal Descendente, como la Regresi贸n Lineal Descendente Estoc谩stica `(Stochastic Gradient Descent en ingl茅s)` y la Regresi贸n Lineal Descendente en Lotes `(Batch Gradient Descent en ingl茅s)`. Cada variaci贸n tiene sus propias ventajas y desventajas en t茅rminos de velocidad y precisi贸n.

````python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from sklearn.linear_model import LinearRegression

#Variable independiente
experiencia = np.array([[5], [6], [13]])

#Variable objetivo
salario = np.array([6.85, 16.83, 26.84])

#Modelo linear creado con el metodo de minimos cuadrados
modelo = LinearRegression().fit(experiencia.reshape(3, -1), salario)

#Visualizacion de datos y modelo
print("Interseccion con eje y (b): %0.2f" % modelo.intercept_)
print("Pendiente (m): %0.2f" % modelo.coef_[0])
print("Suma de cuadrados de los residuos (RSS): %0.2f" %
((salario - modelo.predict(experiencia))**2).sum())

plt.figure(figsize=(8, 6))

plt.scatter(experiencia, salario, color="gold", s=250,
marker="o", label="Valor verdadero")

plt.scatter(experiencia, modelo.predict(experiencia),
color="blue", s=250, marker="P", label="Valor Predicho")

plt.plot(experiencia, modelo.predict(experiencia),
linewidth=4, color="deeppink", label="Modelo lineal")

experiencia=experiencia.reshape(3)

plt.ylabel("Salario en miles de Pesos ($)", size=16)
plt.xlabel("A帽os de Experiencia", size=16)
plt.title("Regresi贸n lineal simple: salario en funci贸n de la experiencia", size=18)
plt.legend(bbox_to_anchor=(1.3, 0.5), prop={'size': 16})
plt.grid()
plt.box(False)
plt.show()
````

## **Visualizaci贸n del error (RSS)**

La visualizaci贸n del error `RSS` `(Residual Sum of Squares)` nos permite evaluar el rendimiento de nuestro modelo de regresi贸n lineal al graficar el valor del `RSS` en funci贸n del n煤mero de iteraciones. Esto nos permite identificar si el modelo est谩 convergiendo o no, y si es necesario ajustar los hiperpar谩metros para mejorar su desempe帽o.

Aqu铆 hay un ejemplo de c贸mo visualizar el error `RSS` en funci贸n del n煤mero de iteraciones:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Agregar t茅rmino de sesgo al conjunto de datos X
X_b = np.c_[np.ones((100, 1)), X]

# Definir tasa de aprendizaje, n煤mero de iteraciones y tama帽o del lote
eta = 0.1
n_iterations = 1000
m = 100

# Inicializar los par谩metros del modelo de manera aleatoria
theta = np.random.randn(2,1)

# Crear una lista para almacenar los valores del RSS en cada iteraci贸n
rss_list = []

# Implementar la Regresi贸n Lineal Descendente y calcular el RSS en cada iteraci贸n
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
    y_predict = X_b.dot(theta)
    rss = np.sum((y_predict - y)**2)
    rss_list.append(rss)

# Graficar el error RSS en funci贸n del n煤mero de iteraciones
plt.plot(range(1, n_iterations+1), rss_list)
plt.xlabel('N煤mero de iteraciones')
plt.ylabel('RSS')
plt.show()
````

En este ejemplo, creamos una lista llamada ` rss_list ` para almacenar los valores del `RSS` en cada iteraci贸n del bucle. Luego, calculamos el valor del `RSS` utilizando la f贸rmula `RSS = sum((y_predict - y)^2)` en cada iteraci贸n y lo agregamos a la lista. Finalmente, graficamos el valor del `RSS` en funci贸n del n煤mero de iteraciones.

Al graficar el error `RSS` en funci贸n del n煤mero de iteraciones, podemos ver c贸mo disminuye el error a medida que el modelo se ajusta a los datos. En general, esperamos que el error disminuya a medida que el n煤mero de iteraciones aumenta, y que eventualmente se estabilice en un valor m铆nimo. Si el error no disminuye o fluct煤a mucho, es posible que necesitemos ajustar los hiperpar谩metros del modelo (como la tasa de aprendizaje) para mejorar su desempe帽o.

### **Funcion de Error a ser Minimizada**

La funci贸n de error a ser minimizada en la regresi贸n lineal es la suma de los errores cuadrados (RSS, por sus siglas en ingl茅s), que mide la diferencia entre los valores predichos por el modelo y los valores reales de la variable objetivo. La funci贸n `RSS` se define como:

`RSS = (y - 欧)虏`

donde `y` son los valores reales de la variable objetivo, ```欧``` son los valores predichos por el modelo y la sumatoria se extiende a trav茅s de todas las observaciones en el conjunto de datos. El objetivo de la regresi贸n lineal es encontrar los valores 贸ptimos de los par谩metros del modelo (intercepto y pendiente) que minimizan esta funci贸n `RSS`.

**Suma de los cuadrados de los residuos (RSS)**

La suma de los cuadrados de los residuos `(RSS)` es una medida del error residual de un modelo de regresi贸n lineal. Se calcula sumando los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo para cada punto de datos en el conjunto de entrenamiento.

Matem谩ticamente, se puede expresar como:

`RSS = 危(y岬 - 欧岬)虏`

**Donde:**

* `RSS` es la suma de los cuadrados de los residuos
* `y岬` es el valor observado en el punto `i`
* `欧岬` es el valor predicho por el modelo en el punto `i`

El objetivo del modelo de regresi贸n lineal es minimizar la `RSS`. Esto se logra ajustando los valores de los par谩metros del modelo para que la l铆nea de regresi贸n se ajuste lo mejor posible a los datos. La `RSS` se puede utilizar para evaluar la bondad de ajuste del modelo, es decir, cu谩nto se ajusta el modelo a los datos. Un valor m谩s bajo de `RSS` indica un mejor ajuste del modelo a los datos.

Es importante destacar que la `RSS` es solo una medida del error residual en el conjunto de entrenamiento y no debe utilizarse como 煤nica medida de rendimiento del modelo. Para evaluar el rendimiento del modelo de regresi贸n lineal, se deben utilizar otras m茅tricas, como el coeficiente de determinaci贸n `(R虏)` o el error de predicci贸n.

### **Visualizando -Derivada de la funcion de Error**

La derivada de la funci贸n de error en relaci贸n a los par谩metros del modelo es una herramienta importante en la regresi贸n lineal. Permite encontrar los valores de los par谩metros que minimizan la funci贸n de error, lo cual es esencial para encontrar una buena aproximaci贸n lineal a los datos.

Aqu铆 hay un ejemplo de visualizaci贸n de la derivada de la funci贸n de error para una regresi贸n lineal simple:

````python
import numpy as np
import matplotlib.pyplot as plt

# Datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# A帽adir t茅rmino de sesgo
X_b = np.c_[np.ones((100, 1)), X]

# Definir funci贸n de error y su derivada
def mse(theta, X, y):
    m = len(y)
    predictions = X.dot(theta)
    error = predictions - y
    return (1/m) * np.sum(np.square(error))

def mse_derivative(theta, X, y):
    m = len(y)
    predictions = X.dot(theta)
    error = predictions - y
    return (2/m) * X.T.dot(error)

# Calcular valores de error y su derivada para diferentes valores de theta
theta_range = np.linspace(-5, 5, 100)
errors = []
derivatives = []
for theta in theta_range:
    errors.append(mse(np.array([[4], [theta]]), X_b, y))
    derivatives.append(mse_derivative(np.array([[4], [theta]]), X_b, y)[1])

# Graficar la funci贸n de error y su derivada
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Pendiente')
ax1.set_ylabel('Error', color=color)
ax1.plot(theta_range, errors, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('Derivada del Error', color=color)
ax2.plot(theta_range, derivatives, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
````

Este c贸digo calcula los valores de error y su derivada para diferentes valores de pendiente y los grafica en dos ejes diferentes. La curva `roja` muestra la funci贸n de error y la curva `azul` muestra su derivada. El punto donde la derivada es cero indica el valor 贸ptimo de la pendiente que minimiza el error. En este ejemplo, ese valor es aproximadamente `3`.

````python
plt.figure(figsize=(7.5, 6))

for i in range(0, len(errores), 1):
  # Error de un modelo dado
  plt.scatter(pendientes[i], errores[i],
              label="%0.2f" % derivada.evalf(subs={m: pendientes[i]}),
              marker="D", s=200)
  
  # Evalucaion de la derivada para un error dado
  pendiente= derivada.evalf(subs={m: pendientes[i]})
  
  plt.plot(np.array([1.7, 2.5]),
           np.array([1.7, 2.5])*pendiente + 
           pendiente*(-2.3 + i*0.05) + errores.min())

plt.title("Derivada de la funcion de Error", size=16)
plt.ylabel("Error (RSS)", size=16)
plt.xlabel("Pendiente (m)", size=16)
plt.ylim(30, 70)
plt.grid()
plt.box(False)
plt.show()
````

## **Generalizaci贸n para dos o m谩s par谩metros**

El objetivo es encontrar los valores de los par谩metros ```胃0, 胃1, 胃2, ..., 胃n``` que minimizan la funci贸n de error `RSS`. Esto se hace utilizando t茅cnicas de optimizaci贸n, como el m茅todo de descenso de gradiente o la ecuaci贸n normal.

Una vez que se han encontrado los valores de los par谩metros, se pueden utilizar para hacer predicciones sobre nuevos datos utilizando la misma funci贸n lineal.

El proceso de encontrar los valores de los par谩metros se puede realizar utilizando bibliotecas de aprendizaje autom谩tico, como `Scikit-learn` en Python, o mediante programaci贸n manual utilizando librer铆as de 谩lgebra lineal como `Numpy`.

El resultado final es un modelo de regresi贸n lineal m煤ltiple que puede ser utilizado para hacer predicciones sobre nuevos datos.

### **Generalizando para 2 o mas parametros: m y b**

La funci贸n de error de la regresi贸n lineal con dos par谩metros (pendiente y t茅rmino de intersecci贸n) est谩 dada por:

$J(\theta_0, \theta_1) = \frac{1}{2m} \sum\limits_{i=1}^{m} (h_{\theta}(x^{(i)}) - y^{(i)})^2$

La creaci贸n y derivada parcial del error `E(m, b)` se refiere a la creaci贸n de una funci贸n de error que tenga en cuenta tanto el par谩metro de la pendiente `m` como el t茅rmino de intersecci贸n `b` en un modelo de regresi贸n lineal. La funci贸n de error para el modelo de regresi贸n lineal con `m` y `b` puede definirse como:

```E(m, b) = 1/2m * sum((y_i - (mx_i + b))^2)```

donde `m` es la pendiente, `b` es la intersecci贸n, `y_i` es el valor objetivo correspondiente al dato de entrada `x_i`.

Para encontrar los valores 贸ptimos de `m` y `b`, se puede tomar la derivada parcial de `E(m, b)` con respecto a cada par谩metro, y luego igualar a cero. De esta manera, se pueden obtener las siguientes ecuaciones para `m` y `b`:

```E/m = -1/m * sum(x_i(y_i - (mx_i + b)))```

```E/b = -1/m * sum(y_i - (mx_i + b))```

A continuaci贸n se presenta un ejemplo de c贸digo que crea y deriva parcialmente la funci贸n de error `E(m, b)` para un conjunto de datos de regresi贸n lineal:

````python
import numpy as np

# Crear conjunto de datos de regresi贸n lineal
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Definir funci贸n de error
def error_function(X, y, m, b):
    m = np.array([m]).reshape(1, 1)
    b = np.array([b]).reshape(1, 1)
    y_hat = np.dot(X, m) + b
    error = 1 / (2 * len(X)) * np.sum(np.square(y_hat - y))
    return error

# Calcular derivadas parciales
def partial_derivative_m(X, y, m, b):
    m = np.array([m]).reshape(1, 1)
    b = np.array([b]).reshape(1, 1)
    y_hat = np.dot(X, m) + b
    derivative = -1 / len(X) * np.sum(np.multiply(X, y - y_hat))
    return derivative

def partial_derivative_b(X, y, m, b):
    m = np.array([m]).reshape(1, 1)
    b = np.array([b]).reshape(1, 1)
    y_hat = np.dot(X, m) + b
    derivative = -1 / len(X) * np.sum(y - y_hat)
    return derivative

# Inicializar valores para m y b
m = 0
b = 0
learning_rate = 0.01
n_iterations = 1000

# Gradiente descendente
for i in range(n_iterations):
    m -= learning_rate * partial_derivative_m(X, y, m, b)
    b -= learning_rate * partial_derivative_b(X, y, m, b)

# Imprimir valores de m y b
print("Valor de m:", m)
print("Valor de b:", b)
````

## **Pseudoc贸digo del gradiente descendente**

El pseudoc贸digo del algoritmo de gradiente descendente se puede describir de la siguiente manera:

1. Inicializar los pesos y el umbral con valores aleatorios cercanos a cero.
2. Seleccionar un tama帽o de paso (tambi茅n conocido como tasa de aprendizaje).
3. Repetir hasta que se alcance un criterio de parada (como un n煤mero m谩ximo de iteraciones o una precisi贸n deseada):
a. Calcular la salida de la red para un conjunto de entrada.

b. Calcular el error cometido por la red en la salida.

c. Calcular la tasa de cambio del error con respecto a los pesos y el umbral.

d. Actualizar los pesos y el umbral utilizando la tasa de cambio del error y la tasa de aprendizaje.

4. Devolver los pesos y el umbral final.

A continuaci贸n se muestra un ejemplo de c贸digo en Python que implementa el algoritmo de gradiente descendente para ajustar una l铆nea a un conjunto de datos:

````python
# Importar bibliotecas necesarias
import numpy as np

# Definir la funci贸n de costo
def costo(X, y, theta):
    m = len(y)
    J = 1/(2*m) * np.sum((X.dot(theta) - y)**2)
    return J

# Definir el algoritmo de gradiente descendente
def gradiente_descendente(X, y, theta_inicial, tasa_aprendizaje, num_iteraciones):
    m = len(y)
    theta = theta_inicial.copy()
    costo_historico = np.zeros(num_iteraciones)
    
    for iteracion in range(num_iteraciones):
        error = X.dot(theta) - y
        gradiente = 1/m * X.T.dot(error)
        theta -= tasa_aprendizaje * gradiente
        costo_historico[iteracion] = costo(X, y, theta)
    
    return theta, costo_historico
````

En este ejemplo, ```X``` es una matriz de caracter铆sticas de entrada, ```y``` es un vector de valores de salida, ```theta``` es un vector de pesos y el umbral, ```tasa_aprendizaje``` es la tasa de aprendizaje y ```num_iteraciones``` es el n煤mero de iteraciones que se desea realizar. La funci贸n ```costo``` calcula el costo o error de la red en funci贸n de los valores actuales de ```X```, ```y``` y ```theta```. La funci贸n ```gradiente_descendente``` implementa el algoritmo de gradiente descendente y devuelve los pesos finales y el historial de costos a lo largo de las iteraciones.

### **Implementacion del Gradiente Descendiente**

El gradiente descendiente es un algoritmo de optimizaci贸n utilizado en el aprendizaje autom谩tico para encontrar los valores 贸ptimos de los par谩metros de un modelo. El objetivo es minimizar una funci贸n de error, como la suma de los cuadrados de los residuos `(RSS)`, ajustando los par谩metros de un modelo de regresi贸n lineal.

La implementaci贸n del gradiente descendiente implica la actualizaci贸n iterativa de los valores de los par谩metros utilizando las derivadas parciales de la funci贸n de error en relaci贸n a cada par谩metro. Los valores de los par谩metros se ajustan en la direcci贸n opuesta del gradiente de la funci贸n de error, es decir, se reduce la funci贸n de error en cada iteraci贸n.

A continuaci贸n se muestra un ejemplo de implementaci贸n del gradiente descendiente en Python para ajustar los par谩metros de un modelo de regresi贸n lineal.

````python
import numpy as np

# Definir la funci贸n de error RSS
def RSS(X, y, theta):
    return np.sum((X.dot(theta) - y) ** 2) / (2 * len(y))

# Implementaci贸n del gradiente descendiente
def gradient_descent(X, y, theta, eta, n_iterations):
    m = len(y)
    for iteration in range(n_iterations):
        gradients = 2 / m * X.T.dot(X.dot(theta) - y)
        theta = theta - eta * gradients
        if iteration % 100 == 0:
            print(f"Iteration: {iteration}, Error: {RSS(X, y, theta)}")
    return theta

# Generar datos de ejemplo
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Agregar t茅rmino de sesgo al conjunto de datos X
X_b = np.c_[np.ones((100, 1)), X]

# Definir valores iniciales para los par谩metros
theta = np.random.randn(2, 1)

# Definir tasa de aprendizaje y n煤mero de iteraciones
eta = 0.1
n_iterations = 1000

# Aplicar el gradiente descendiente para ajustar los par谩metros del modelo
theta_best = gradient_descent(X_b, y, theta, eta, n_iterations)
print(f"Theta Best: {theta_best}")

````

En este ejemplo, se define la funci贸n de error `RSS` y se implementa el gradiente descendiente en la funci贸n ```gradient_descent```. La funci贸n toma como entrada los datos de entrenamiento `X` y `y`, los valores iniciales para los par谩metros `theta`, la tasa de aprendizaje `eta` y el n煤mero de iteraciones `n_iterations`. En cada iteraci贸n, se calculan las derivadas parciales de la funci贸n de error en relaci贸n a los par谩metros y se actualizan los valores de los par谩metros en la direcci贸n opuesta del gradiente de la funci贸n de error. El resultado final es el conjunto 贸ptimo de valores de los par谩metros del modelo.

La salida de este ejemplo incluye la evoluci贸n del error en cada iteraci贸n y el conjunto 贸ptimo de valores de los par谩metros del modelo.

## **Visualizaci贸n del descenso del error**

Aqu铆 te muestro un ejemplo de c贸mo visualizar el descenso del error durante el proceso de optimizaci贸n utilizando el algoritmo del Gradiente Descendente:

Supongamos que tenemos un conjunto de datos ```X``` y ```y``` y queremos encontrar la recta de regresi贸n que mejor se ajuste a estos datos utilizando el algoritmo del Gradiente Descendente. Primero, definimos una funci贸n que calcule el error cuadr谩tico medio `(MSE)` para un conjunto de datos y unos par谩metros dados:

````python
def mse(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    error = predictions - y
    mse = 1/(2*m) * np.sum(error**2)
    return mse
````

Luego, definimos una funci贸n que implemente el algoritmo del Gradiente Descendente:

````python
def gradient_descent(X, y, theta, learning_rate, n_iterations):
    m = len(y)
    history = []
    for i in range(n_iterations):
        predictions = X.dot(theta)
        error = predictions - y
        gradient = 1/m * X.T.dot(error)
        theta = theta - learning_rate * gradient
        history.append(mse(X, y, theta))
    return theta, history
````

Finalmente, utilizamos estas funciones para encontrar la recta de regresi贸n que mejor se ajuste a los datos y visualizar el descenso del error durante el proceso de optimizaci贸n:

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Agregar t茅rmino de sesgo al conjunto de datos X
X_b = np.c_[np.ones((100, 1)), X]

# Definir tasa de aprendizaje, n煤mero de iteraciones y tama帽o del lote
eta = 0.1
n_iterations = 1000
m = 100

# Inicializar los par谩metros theta aleatoriamente
theta = np.random.randn(2,1)

# Implementar el Gradiente Descendiente
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients

    # Calcular el error para cada iteraci贸n
    if iteration % 10 == 0:
        y_predict = X_b.dot(theta)
        error = np.mean((y_predict - y) ** 2)
        print(f"Iteraci贸n {iteration}: error = {error}")

# Visualizar los datos y la recta de regresi贸n lineal
plt.plot(X, y, "b.")
y_predict = X_b.dot(theta)
plt.plot(X, y_predict, "r-", linewidth=2, label="Predicciones")
plt.xlabel("X", fontsize=14)
plt.ylabel("y", rotation=0, fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.show()
````
