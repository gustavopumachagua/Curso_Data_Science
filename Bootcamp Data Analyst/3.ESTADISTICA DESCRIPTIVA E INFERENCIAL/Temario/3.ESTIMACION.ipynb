## **Universo, población y muestra**

En estadística, el universo, la población y la muestra son conceptos relacionados con el estudio y análisis de datos. A continuación, te daré una explicación detallada de cada uno, junto con ejemplos en código Python y sus gráficas correspondientes.

* **Universo:**

El universo se refiere al conjunto completo de elementos o individuos que cumplen con ciertas características y que son objeto de estudio en un análisis estadístico. Es el conjunto más amplio y completo que se considera en un estudio.

* **Población:**

La población es un subconjunto del universo y se define como el conjunto de todos los elementos o individuos que comparten las mismas características o propiedades que se están investigando. La población puede ser finita o infinita, dependiendo del tamaño del conjunto.

* **Muestra:**

La muestra es un subconjunto seleccionado de la población y se utiliza para obtener información y hacer inferencias sobre la población en su conjunto. La muestra se elige de manera representativa, lo que significa que debe reflejar adecuadamente las características y la variabilidad de la población.

A continuación, se muestra un ejemplo de cómo generar una muestra aleatoria utilizando código Python y visualizarla en un gráfico:

```python
import numpy as np
import matplotlib.pyplot as plt

# Población de datos
poblacion = np.random.normal(loc=50, scale=10, size=1000) # Distribución normal con media 50 y desviación estándar 10

# Tamaño de la muestra
tam_muestra = 100

# Generar una muestra aleatoria
muestra = np.random.choice(poblacion, size=tam_muestra, replace=False)

# Histograma de la población
plt.hist(poblacion, bins=30, alpha=0.5, label='Población')

# Histograma de la muestra
plt.hist(muestra, bins=30, alpha=0.5, label='Muestra')

# Agregar leyendas y título
plt.legend()
plt.title('Histograma de Población y Muestra')

# Mostrar el gráfico
plt.show()
```

En este ejemplo, se utiliza la biblioteca NumPy para generar una población de datos siguiendo una distribución normal con media 50 y desviación estándar 10. Luego, se selecciona una muestra aleatoria de tamaño 100 utilizando la función `random.choice`. Finalmente, se comparan los histogramas de la población y la muestra utilizando la función `hist` de Matplotlib.

El gráfico muestra la distribución de la población y la muestra en forma de histogramas. La muestra debe ser representativa de la población en términos de sus características y variabilidad. Al seleccionar una muestra adecuada, podemos inferir conclusiones sobre la población en base a los datos de la muestra.

Es importante tener en cuenta que el tamaño de la muestra es un factor importante a considerar. Muestras más grandes tienden a proporcionar estimaciones más precisas de las características de la población.

## **Distribución conjunta, marginal, condicional**

La distribución conjunta, marginal y condicional son conceptos utilizados en estadística para describir la relación entre variables aleatorias. A continuación, te daré una explicación detallada de cada una de ellas, junto con ejemplos en código Python y sus gráficas correspondientes.

* **Distribución conjunta:**

La distribución conjunta se refiere a la distribución de probabilidad conjunta de dos o más variables aleatorias. Proporciona información sobre la probabilidad conjunta de ocurrencia de diferentes combinaciones de valores de las variables aleatorias.

Para ilustrar esto, consideremos un ejemplo de dos variables aleatorias, `X` y `Y`, que representan el tiempo de espera de llegada de dos eventos diferentes:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(0)
X = np.random.normal(loc=10, scale=2, size=1000)
Y = np.random.normal(loc=20, scale=3, size=1000)

# Graficar la distribución conjunta
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Distribución Conjunta')
plt.show()
```

En este ejemplo, se generan datos aleatorios para las variables `X` e `Y` siguiendo distribuciones normales con medias y desviaciones estándar especificadas. Luego, se grafica la distribución conjunta utilizando un gráfico de dispersión, donde cada punto representa una observación de `X` e `Y`.

* **Distribución marginal:**

La distribución marginal se obtiene considerando la distribución de una variable aleatoria en particular, ignorando la influencia de las demás variables. Es la distribución de probabilidad de una variable aleatoria individual.

Continuando con el ejemplo anterior, supongamos que queremos calcular la distribución marginal de la variable `X`. Podemos hacer esto simplemente observando los valores de `X` y calculando su distribución de probabilidad.

```python
# Graficar la distribución marginal de X
plt.hist(X, bins=30)
plt.xlabel('X')
plt.ylabel('Frecuencia')
plt.title('Distribución Marginal de X')
plt.show()
```

En este caso, se utiliza la función `hist` de Matplotlib para graficar el histograma de los valores de `X`. El histograma muestra la distribución de probabilidad de la variable `X` de forma individual, sin tener en cuenta la variable `Y`.

* **Distribución condicional:**

La distribución condicional se refiere a la distribución de probabilidad de una variable aleatoria dada cierto valor o condición de otra variable aleatoria. En otras palabras, es la distribución de una variable aleatoria bajo una restricción o condición específica.
Siguiendo con el ejemplo anterior, supongamos que queremos calcular la distribución condicional de la variable `Y` dado un valor específico de `X`. Por ejemplo, queremos calcular la distribución de `Y` cuando `X` es igual a 12.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(0)
X = np.random.normal(loc=10, scale=2, size=1000)
Y = np.random.normal(loc=20, scale=3, size=1000)

# Establecer una condición basada en una tolerancia
tolerancia = 0.1
condicion = np.isclose(X, 12, atol=tolerancia)
Y_condicional = Y[condicion]

# Graficar la distribución condicional de Y dado X cerca de 12
plt.hist(Y_condicional, bins=30)
plt.xlabel('Y')
plt.ylabel('Frecuencia')
plt.title('Distribución Condicional de Y dado X cerca de 12')
plt.show()
```

En este caso, se seleccionan las observaciones donde `X` es igual a 12 utilizando una condición booleana. Luego, se calcula la distribución condicional de `Y` para estas observaciones y se grafica su distribución utilizando un histograma.

Estos ejemplos ilustran la diferencia entre la distribución conjunta, marginal y condicional. La distribución conjunta muestra la relación entre dos variables aleatorias, la distribución marginal se centra en una variable aleatoria en particular y la distribución condicional se refiere a la distribución de una variable aleatoria bajo una restricción específica. Estos conceptos son fundamentales en el análisis estadístico y son útiles para comprender la dependencia y relación entre variables.

## **Vector de esperanzas, matriz de covarianzas, correlaciones**

El vector de esperanzas, la matriz de covarianzas y las correlaciones son conceptos relacionados que se utilizan para describir la relación entre variables aleatorias en un conjunto de datos. Estas medidas son comúnmente utilizadas en estadística y análisis de datos para comprender la estructura y las interacciones entre las variables.

1. **Vector de esperanzas:**

El vector de esperanzas, también conocido como vector de medias o valor esperado, es un vector que contiene las medias de las variables aleatorias en un conjunto de datos. Cada componente del vector de esperanzas representa la media de una variable. Se calcula tomando la media de cada variable por separado.

**Ejemplo:**

Supongamos que tenemos dos variables aleatorias `X` y `Y`, y queremos calcular el vector de esperanzas. Podemos utilizar la biblioteca NumPy para calcular las medias de las variables:

```python
import numpy as np

# Datos de ejemplo
X = [1, 2, 3, 4, 5]
Y = [2, 4, 6, 8, 10]

# Calcular el vector de esperanzas
esperanzas = np.mean(X), np.mean(Y)
print("Vector de Esperanzas:", esperanzas)
```

Esto indica que la media de la variable `X` es 3.0 y la media de la variable `Y` es 6.0.

2. **Matriz de covarianzas:**

La matriz de covarianzas describe la relación lineal entre pares de variables aleatorias. Es una matriz cuadrada simétrica que contiene las covarianzas entre todas las combinaciones posibles de variables en un conjunto de datos. La covarianza mide cómo varían conjuntamente dos variables.

**Ejemplo:**

Supongamos que tenemos tres variables aleatorias `X`, `Y` y `Z`, y queremos calcular la matriz de covarianzas. Podemos utilizar la función `np.cov()` de NumPy para calcular la matriz de covarianzas:

```python
import numpy as np

# Datos de ejemplo
X = [1, 2, 3, 4, 5]
Y = [2, 4, 6, 8, 10]
Z = [3, 6, 9, 12, 15]

# Crear la matriz de covarianzas
datos = np.stack((X, Y, Z))
covarianzas = np.cov(datos)

print("Matriz de Covarianzas:")
print(covarianzas)
```

Esto indica que la covarianza entre `X` y `X` es 2.5, la covarianza entre `X` y `Y` es 5.0, la covarianza entre `X` y `Z` es 7.5, y así sucesivamente.

3. **Correlaciones:**

La correlación es una medida estandarizada de la relación lineal entre dos variables aleatorias. La correlación varía entre -1 y 1, donde -1 indica una correlación negativa perfecta, 1 indica una correlación positiva perfecta y 0 indica una falta de correlación lineal.

**Ejemplo:**

Supongamos que tenemos dos variables aleatorias `X` y `Y`, y queremos calcular la correlación. Podemos utilizar la función `np.corrcoef()` de NumPy para calcular la matriz de correlaciones:

```python
import numpy as np

# Datos de ejemplo
X = [1, 2, 3, 4, 5]
Y = [2, 4, 6, 8, 10]

# Calcular la matriz de correlaciones
datos = np.stack((X, Y))
correlaciones = np.corrcoef(datos)

print("Matriz de Correlaciones:")
print(correlaciones)
```

Esto indica que la correlación entre `X` y `X` es 1.0, y la correlación entre `X` y `Y` también es 1.0.

Las matrices de covarianzas y correlaciones nos proporcionan información sobre la relación y la fuerza de la asociación entre las variables en un conjunto de datos. También nos permiten identificar patrones y dependencias entre las variables, lo que puede ser útil en análisis posteriores o en la construcción de modelos predictivos.

## **Muestreo aleatorio**

El muestreo aleatorio es una técnica utilizada en estadística para seleccionar una muestra representativa de una población más grande. Consiste en seleccionar aleatoriamente individuos de la población de manera que todos tengan la misma probabilidad de ser seleccionados. Esto ayuda a reducir el sesgo y proporciona una base sólida para realizar inferencias sobre la población en su conjunto.

A continuación, te mostraré un ejemplo en código Python para realizar un muestreo aleatorio utilizando la biblioteca NumPy y graficar los resultados:

```python
import numpy as np
import matplotlib.pyplot as plt

# Población de ejemplo (valores de 1 a 100)
poblacion = np.arange(1, 101)

# Tamaño de la muestra
tam_muestra = 20

# Realizar el muestreo aleatorio
muestra = np.random.choice(poblacion, size=tam_muestra, replace=False)

# Graficar la población y la muestra
plt.figure(figsize=(10, 4))

# Población
plt.subplot(1, 2, 1)
plt.hist(poblacion, bins=10)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Población')

# Muestra
plt.subplot(1, 2, 2)
plt.hist(muestra, bins=10)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Muestra Aleatoria')

plt.tight_layout()
plt.show()
```

En este ejemplo, tenemos una población que consiste en valores del 1 al 100. Queremos obtener una muestra aleatoria de tamaño 20. Utilizamos la función `np.random.choice()` de NumPy para realizar la selección aleatoria sin reemplazo.

Luego, graficamos la población y la muestra utilizando histogramas. En el gráfico de la población, se muestra la distribución de los valores del 1 al 100. En el gráfico de la muestra, se muestra la distribución de los valores seleccionados aleatoriamente.

El muestreo aleatorio nos permite obtener una muestra representativa de una población más grande, lo que nos permite hacer inferencias y generalizaciones sobre la población en general. Además, al seleccionar al azar, reducimos la posibilidad de sesgos y aseguramos la imparcialidad en nuestras conclusiones.

## **Función de verosimilitud**

La función de verosimilitud es una herramienta fundamental en inferencia estadística y aprendizaje automático. Representa la probabilidad de obtener los datos observados para un conjunto de parámetros específicos en un modelo estadístico.

En términos simples, la función de verosimilitud nos permite evaluar qué tan bien se ajustan los datos a un modelo y nos proporciona una medida de la "`verosimilitud`" de que los parámetros del modelo sean los correctos.

A continuación, te mostraré un ejemplo en código Python que ilustra el cálculo de la función de verosimilitud utilizando una distribución normal:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos observados
datos = np.array([1.2, 1.5, 1.8, 2.1, 2.4])

# Parámetros del modelo
media = 2.0
desviacion = 0.5

# Función de densidad de probabilidad (PDF) de la distribución normal
def distribucion_normal(x, media, desviacion):
    return (1 / (desviacion * np.sqrt(2 * np.pi))) * np.exp(-(x - media)**2 / (2 * desviacion**2))

# Calcular la función de verosimilitud
verosimilitud = np.prod(distribucion_normal(datos, media, desviacion))

# Graficar la distribución normal y los datos observados
x = np.linspace(0, 3, 100)
y = distribucion_normal(x, media, desviacion)

plt.plot(x, y, label='Distribución normal')
plt.scatter(datos, np.zeros_like(datos), color='red', label='Datos observados')
plt.xlabel('Valores')
plt.ylabel('Densidad de probabilidad')
plt.title('Función de Verosimilitud')
plt.legend()
plt.show()

print(f"Verosimilitud: {verosimilitud}")
```

En este ejemplo, tenemos un conjunto de datos observados [1.2, 1.5, 1.8, 2.1, 2.4] y queremos evaluar qué tan bien se ajustan a una distribución normal con una media de 2.0 y una desviación estándar de 0.5.

La función `distribucion_normal` define la función de densidad de probabilidad (`PDF`) de la distribución normal. Utilizamos esta función para calcular la probabilidad de obtener cada dato observado en el modelo.

Luego, calculamos la función de verosimilitud multiplicando las probabilidades individuales de cada dato observado. En este caso, utilizamos el producto de las probabilidades porque asumimos que los datos son independientes y se distribuyen normalmente.

Finalmente, graficamos la distribución normal junto con los datos observados y mostramos el valor de la verosimilitud calculada.

La función de verosimilitud es una herramienta crucial en la estimación de parámetros en modelos estadísticos y en el ajuste de modelos en el aprendizaje automático. Nos permite evaluar la calidad del ajuste del modelo a los datos observados y tomar decisiones basadas en la probabilidad de los parámetros estimados.

## **Teorema del límite central**

El Teorema del Límite Central (TLC) es uno de los conceptos fundamentales en estadística y establece que, bajo ciertas condiciones, la suma o media de un gran número de variables aleatorias independientes y idénticamente distribuidas (iid) tiende a seguir una distribución normal, independientemente de la distribución original de las variables.

En términos más simples, el TLC afirma que cuando se toma una muestra lo suficientemente grande de una población, la distribución de las medias de esa muestra se aproximará a una distribución normal, incluso si la distribución de la población original no es normal.

A continuación, te mostraré un ejemplo en código Python que ilustra el Teorema del Límite Central:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar 1000 muestras de tamaño n
n = 100
num_muestras = 1000
muestras = np.random.rand(num_muestras, n)

# Calcular la media de cada muestra
medias_muestras = np.mean(muestras, axis=1)

# Graficar el histograma de las medias muestrales
plt.hist(medias_muestras, bins=30, density=True)
plt.xlabel('Media Muestral')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución de las Medias Muestrales')
plt.show()
```

En este ejemplo, generamos 1000 muestras, cada una de tamaño 100, a partir de una distribución uniforme utilizando la función `np.random.rand()`. Luego, calculamos la media de cada muestra utilizando la función `np.mean()`.

Finalmente, graficamos el histograma de las medias muestrales utilizando `plt.hist()`. Verás que a medida que aumenta el tamaño de las muestras, la distribución de las medias muestrales se aproxima a una distribución normal.

El Teorema del Límite Central es importante porque permite realizar inferencias estadísticas sobre la media de una población, incluso si no conocemos la distribución de la población original. También justifica el uso de técnicas basadas en la distribución normal en muchos análisis estadísticos y pruebas de hipótesis.

## **Distribuciones derivadas de la normal**

Las distribuciones derivadas de la normal son distribuciones que se obtienen a partir de la distribución normal estándar mediante transformaciones matemáticas. Estas distribuciones derivadas tienen propiedades y características específicas que las hacen útiles en diferentes situaciones.

A continuación, te mostraré algunos ejemplos de distribuciones derivadas de la normal junto con su código Python y gráficas correspondientes:

1. **Distribución normal (gaussiana):**

Es la distribución original de la cual se derivan las demás distribuciones. Tiene forma de campana y se caracteriza por su media (`μ`) y su desviación estándar (`σ`).

**Ejemplo:**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = (1 / (np.sqrt(2 * np.pi) * 1)) * np.exp(-(x - 0) ** 2 / (2 * 1 ** 2))

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Normal')
plt.show()
```

2. **Distribución t de Student:**

Es utilizada cuando el tamaño de la muestra es pequeño y la varianza de la población es desconocida. Tiene forma de campana, pero con colas más pesadas que la distribución normal.

**Ejemplo:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

x = np.linspace(-5, 5, 100)
y = t.pdf(x, df=10)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución t de Student')
plt.show()
```

3. **Distribución chi-cuadrado:**

Se utiliza en pruebas de hipótesis y análisis de varianza. Es asimétrica y toma solo valores positivos.

**Ejemplo:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

x = np.linspace(0, 10, 100)
y = chi2.pdf(x, df=5)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Chi-cuadrado')
plt.show()
```

Estos son solo algunos ejemplos de las distribuciones derivadas de la normal. Cada distribución tiene sus propias características y se utiliza en diferentes contextos según las necesidades del análisis estadístico. La elección de la distribución adecuada depende de las características de los datos y del objetivo del estudio.

## **Estimación de parámetros**

La estimación de parámetros es un proceso estadístico que consiste en determinar los valores desconocidos de los parámetros de una distribución de probabilidad utilizando información obtenida de una muestra de datos. Los parámetros son medidas numéricas que resumen características importantes de la distribución, como la media, la varianza, la proporción, entre otros.

A continuación, te mostraré un ejemplo de estimación de parámetros utilizando Python y gráficas correspondientes:

Supongamos que tenemos una muestra de datos que sigue una distribución normal y queremos estimar los parámetros de media y varianza de esta distribución.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de datos
np.random.seed(0)
datos = np.random.normal(loc=10, scale=2, size=100)

# Estimar los parámetros de media y varianza
media_estimada = np.mean(datos)
varianza_estimada = np.var(datos)

# Graficar el histograma de los datos
plt.hist(datos, bins=30, density=True, alpha=0.7)
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.title('Histograma de la muestra de datos')
plt.show()

# Imprimir los valores estimados de media y varianza
print("Media estimada:", media_estimada)
print("Varianza estimada:", varianza_estimada)
```

En el código anterior, generamos una muestra de datos utilizando la función `np.random.normal()` que sigue una distribución normal con media 10 y desviación estándar 2. Luego, utilizamos las funciones `np.mean()` y `np.var()` para estimar los parámetros de media y varianza de la distribución, respectivamente.

Posteriormente, graficamos el histograma de los datos utilizando `plt.hist()` para visualizar la distribución de los datos.

Finalmente, imprimimos los valores estimados de media y varianza.

Es importante destacar que la estimación de parámetros está sujeta a incertidumbre y error debido a la variabilidad inherente en la muestra de datos. Por lo tanto, es útil proporcionar intervalos de confianza que indiquen la precisión de las estimaciones.

En resumen, la estimación de parámetros es un proceso fundamental en la inferencia estadística que nos permite obtener valores aproximados de los parámetros de una distribución de probabilidad a partir de una muestra de datos. Esto nos brinda información sobre las características de la población a partir de la cual se obtuvo la muestra.






## **Intervalos de confianza**

Los intervalos de confianza son herramientas estadísticas que nos permiten estimar un rango de valores dentro del cual es probable que se encuentre un parámetro desconocido de una población, con un nivel de confianza determinado. Proporcionan una medida de la incertidumbre asociada a una estimación puntual.

A continuación, te mostraré un ejemplo de cómo calcular e interpretar un intervalo de confianza utilizando Python y gráficas correspondientes:

Supongamos que queremos estimar el intervalo de confianza para la media de una población utilizando una muestra de datos.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generar una muestra de datos
np.random.seed(0)
datos = np.random.normal(loc=10, scale=2, size=100)

# Calcular el intervalo de confianza para la media
media_muestra = np.mean(datos)
error_estandar = stats.sem(datos)  # Error estándar de la media
nivel_confianza = 0.95
z_score = stats.norm.ppf((1 + nivel_confianza) / 2)  # Valor crítico Z
intervalo_confianza = (media_muestra - z_score * error_estandar,
                       media_muestra + z_score * error_estandar)

# Graficar el histograma de los datos
plt.hist(datos, bins=30, density=True, alpha=0.7)
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.title('Histograma de la muestra de datos')

# Graficar el intervalo de confianza
plt.axvline(intervalo_confianza[0], color='red', linestyle='--', label='Intervalo de Confianza')
plt.axvline(intervalo_confianza[1], color='red', linestyle='--')
plt.legend()

plt.show()

# Imprimir el intervalo de confianza
print("Intervalo de confianza para la media:", intervalo_confianza)
```

En el código anterior, generamos una muestra de datos utilizando la función `np.random.normal()` que sigue una distribución normal con media 10 y desviación estándar 2.

Luego, utilizamos la función `stats.sem()` para calcular el error estándar de la media de la muestra. Posteriormente, especificamos el nivel de confianza deseado, en este caso, 0.95, y utilizamos la función `stats.norm.ppf()` para obtener el valor crítico `Z` correspondiente al nivel de confianza.

A continuación, calculamos el intervalo de confianza utilizando la fórmula: $media de la muestra ± valor crítico Z * error estándar de la media$.

Graficamos el histograma de los datos y agregamos líneas verticales para representar el intervalo de confianza.

Finalmente, imprimimos el intervalo de confianza calculado.
