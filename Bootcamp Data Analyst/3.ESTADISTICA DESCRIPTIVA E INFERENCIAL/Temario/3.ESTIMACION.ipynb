## **Distribuci贸n conjunta, marginal, condicional**

La distribuci贸n conjunta, marginal y condicional son conceptos utilizados en estad铆stica para describir la relaci贸n entre variables aleatorias. A continuaci贸n, te dar茅 una explicaci贸n detallada de cada una de ellas, junto con ejemplos en c贸digo Python y sus gr谩ficas correspondientes.

* **Distribuci贸n conjunta:**

La distribuci贸n conjunta se refiere a la distribuci贸n de probabilidad conjunta de dos o m谩s variables aleatorias. Proporciona informaci贸n sobre la probabilidad conjunta de ocurrencia de diferentes combinaciones de valores de las variables aleatorias.

Para ilustrar esto, consideremos un ejemplo de dos variables aleatorias, `X` y `Y`, que representan el tiempo de espera de llegada de dos eventos diferentes:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(0)
X = np.random.normal(loc=10, scale=2, size=1000)
Y = np.random.normal(loc=20, scale=3, size=1000)

# Graficar la distribuci贸n conjunta
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Distribuci贸n Conjunta')
plt.show()
```

En este ejemplo, se generan datos aleatorios para las variables `X` e `Y` siguiendo distribuciones normales con medias y desviaciones est谩ndar especificadas. Luego, se grafica la distribuci贸n conjunta utilizando un gr谩fico de dispersi贸n, donde cada punto representa una observaci贸n de `X` e `Y`.

* **Distribuci贸n marginal:**

La distribuci贸n marginal se obtiene considerando la distribuci贸n de una variable aleatoria en particular, ignorando la influencia de las dem谩s variables. Es la distribuci贸n de probabilidad de una variable aleatoria individual.

Continuando con el ejemplo anterior, supongamos que queremos calcular la distribuci贸n marginal de la variable `X`. Podemos hacer esto simplemente observando los valores de `X` y calculando su distribuci贸n de probabilidad.

```python
# Graficar la distribuci贸n marginal de X
plt.hist(X, bins=30)
plt.xlabel('X')
plt.ylabel('Frecuencia')
plt.title('Distribuci贸n Marginal de X')
plt.show()
```

En este caso, se utiliza la funci贸n `hist` de Matplotlib para graficar el histograma de los valores de `X`. El histograma muestra la distribuci贸n de probabilidad de la variable `X` de forma individual, sin tener en cuenta la variable `Y`.

* **Distribuci贸n condicional:**

La distribuci贸n condicional se refiere a la distribuci贸n de probabilidad de una variable aleatoria dada cierto valor o condici贸n de otra variable aleatoria. En otras palabras, es la distribuci贸n de una variable aleatoria bajo una restricci贸n o condici贸n espec铆fica.
Siguiendo con el ejemplo anterior, supongamos que queremos calcular la distribuci贸n condicional de la variable `Y` dado un valor espec铆fico de `X`. Por ejemplo, queremos calcular la distribuci贸n de `Y` cuando `X` es igual a 12.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(0)
X = np.random.normal(loc=10, scale=2, size=1000)
Y = np.random.normal(loc=20, scale=3, size=1000)

# Establecer una condici贸n basada en una tolerancia
tolerancia = 0.1
condicion = np.isclose(X, 12, atol=tolerancia)
Y_condicional = Y[condicion]

# Graficar la distribuci贸n condicional de Y dado X cerca de 12
plt.hist(Y_condicional, bins=30)
plt.xlabel('Y')
plt.ylabel('Frecuencia')
plt.title('Distribuci贸n Condicional de Y dado X cerca de 12')
plt.show()
```

En este caso, se seleccionan las observaciones donde `X` es igual a 12 utilizando una condici贸n booleana. Luego, se calcula la distribuci贸n condicional de `Y` para estas observaciones y se grafica su distribuci贸n utilizando un histograma.

Estos ejemplos ilustran la diferencia entre la distribuci贸n conjunta, marginal y condicional. La distribuci贸n conjunta muestra la relaci贸n entre dos variables aleatorias, la distribuci贸n marginal se centra en una variable aleatoria en particular y la distribuci贸n condicional se refiere a la distribuci贸n de una variable aleatoria bajo una restricci贸n espec铆fica. Estos conceptos son fundamentales en el an谩lisis estad铆stico y son 煤tiles para comprender la dependencia y relaci贸n entre variables.

## **Vector de esperanzas, matriz de covarianzas, correlaciones**

El vector de esperanzas, la matriz de covarianzas y las correlaciones son conceptos relacionados que se utilizan para describir la relaci贸n entre variables aleatorias en un conjunto de datos. Estas medidas son com煤nmente utilizadas en estad铆stica y an谩lisis de datos para comprender la estructura y las interacciones entre las variables.

1. **Vector de esperanzas:**

El vector de esperanzas, tambi茅n conocido como vector de medias o valor esperado, es un vector que contiene las medias de las variables aleatorias en un conjunto de datos. Cada componente del vector de esperanzas representa la media de una variable. Se calcula tomando la media de cada variable por separado.

**Ejemplo:**

Supongamos que tenemos dos variables aleatorias `X` y `Y`, y queremos calcular el vector de esperanzas. Podemos utilizar la biblioteca NumPy para calcular las medias de las variables:

```python
import numpy as np

# Datos de ejemplo
X = [1, 2, 3, 4, 5]
Y = [2, 4, 6, 8, 10]

# Calcular el vector de esperanzas
esperanzas = np.mean(X), np.mean(Y)
print("Vector de Esperanzas:", esperanzas)
```

Esto indica que la media de la variable `X` es 3.0 y la media de la variable `Y` es 6.0.

2. **Matriz de covarianzas:**

La matriz de covarianzas describe la relaci贸n lineal entre pares de variables aleatorias. Es una matriz cuadrada sim茅trica que contiene las covarianzas entre todas las combinaciones posibles de variables en un conjunto de datos. La covarianza mide c贸mo var铆an conjuntamente dos variables.

**Ejemplo:**

Supongamos que tenemos tres variables aleatorias `X`, `Y` y `Z`, y queremos calcular la matriz de covarianzas. Podemos utilizar la funci贸n `np.cov()` de NumPy para calcular la matriz de covarianzas:

```python
import numpy as np

# Datos de ejemplo
X = [1, 2, 3, 4, 5]
Y = [2, 4, 6, 8, 10]
Z = [3, 6, 9, 12, 15]

# Crear la matriz de covarianzas
datos = np.stack((X, Y, Z))
covarianzas = np.cov(datos)

print("Matriz de Covarianzas:")
print(covarianzas)
```

Esto indica que la covarianza entre `X` y `X` es 2.5, la covarianza entre `X` y `Y` es 5.0, la covarianza entre `X` y `Z` es 7.5, y as铆 sucesivamente.

3. **Correlaciones:**

La correlaci贸n es una medida estandarizada de la relaci贸n lineal entre dos variables aleatorias. La correlaci贸n var铆a entre -1 y 1, donde -1 indica una correlaci贸n negativa perfecta, 1 indica una correlaci贸n positiva perfecta y 0 indica una falta de correlaci贸n lineal.

**Ejemplo:**

Supongamos que tenemos dos variables aleatorias `X` y `Y`, y queremos calcular la correlaci贸n. Podemos utilizar la funci贸n `np.corrcoef()` de NumPy para calcular la matriz de correlaciones:

```python
import numpy as np

# Datos de ejemplo
X = [1, 2, 3, 4, 5]
Y = [2, 4, 6, 8, 10]

# Calcular la matriz de correlaciones
datos = np.stack((X, Y))
correlaciones = np.corrcoef(datos)

print("Matriz de Correlaciones:")
print(correlaciones)
```

Esto indica que la correlaci贸n entre `X` y `X` es 1.0, y la correlaci贸n entre `X` y `Y` tambi茅n es 1.0.

Las matrices de covarianzas y correlaciones nos proporcionan informaci贸n sobre la relaci贸n y la fuerza de la asociaci贸n entre las variables en un conjunto de datos. Tambi茅n nos permiten identificar patrones y dependencias entre las variables, lo que puede ser 煤til en an谩lisis posteriores o en la construcci贸n de modelos predictivos.

## **Funci贸n de verosimilitud**

La funci贸n de verosimilitud es una herramienta fundamental en inferencia estad铆stica y aprendizaje autom谩tico. Representa la probabilidad de obtener los datos observados para un conjunto de par谩metros espec铆ficos en un modelo estad铆stico.

En t茅rminos simples, la funci贸n de verosimilitud nos permite evaluar qu茅 tan bien se ajustan los datos a un modelo y nos proporciona una medida de la "`verosimilitud`" de que los par谩metros del modelo sean los correctos.

A continuaci贸n, te mostrar茅 un ejemplo en c贸digo Python que ilustra el c谩lculo de la funci贸n de verosimilitud utilizando una distribuci贸n normal:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos observados
datos = np.array([1.2, 1.5, 1.8, 2.1, 2.4])

# Par谩metros del modelo
media = 2.0
desviacion = 0.5

# Funci贸n de densidad de probabilidad (PDF) de la distribuci贸n normal
def distribucion_normal(x, media, desviacion):
    return (1 / (desviacion * np.sqrt(2 * np.pi))) * np.exp(-(x - media)**2 / (2 * desviacion**2))

# Calcular la funci贸n de verosimilitud
verosimilitud = np.prod(distribucion_normal(datos, media, desviacion))

# Graficar la distribuci贸n normal y los datos observados
x = np.linspace(0, 3, 100)
y = distribucion_normal(x, media, desviacion)

plt.plot(x, y, label='Distribuci贸n normal')
plt.scatter(datos, np.zeros_like(datos), color='red', label='Datos observados')
plt.xlabel('Valores')
plt.ylabel('Densidad de probabilidad')
plt.title('Funci贸n de Verosimilitud')
plt.legend()
plt.show()

print(f"Verosimilitud: {verosimilitud}")
```

En este ejemplo, tenemos un conjunto de datos observados [1.2, 1.5, 1.8, 2.1, 2.4] y queremos evaluar qu茅 tan bien se ajustan a una distribuci贸n normal con una media de 2.0 y una desviaci贸n est谩ndar de 0.5.

La funci贸n `distribucion_normal` define la funci贸n de densidad de probabilidad (`PDF`) de la distribuci贸n normal. Utilizamos esta funci贸n para calcular la probabilidad de obtener cada dato observado en el modelo.

Luego, calculamos la funci贸n de verosimilitud multiplicando las probabilidades individuales de cada dato observado. En este caso, utilizamos el producto de las probabilidades porque asumimos que los datos son independientes y se distribuyen normalmente.

Finalmente, graficamos la distribuci贸n normal junto con los datos observados y mostramos el valor de la verosimilitud calculada.

La funci贸n de verosimilitud es una herramienta crucial en la estimaci贸n de par谩metros en modelos estad铆sticos y en el ajuste de modelos en el aprendizaje autom谩tico. Nos permite evaluar la calidad del ajuste del modelo a los datos observados y tomar decisiones basadas en la probabilidad de los par谩metros estimados.

## **Teorema del l铆mite central**

El Teorema del L铆mite Central (TLC) es uno de los conceptos fundamentales en estad铆stica y establece que, bajo ciertas condiciones, la suma o media de un gran n煤mero de variables aleatorias independientes y id茅nticamente distribuidas (iid) tiende a seguir una distribuci贸n normal, independientemente de la distribuci贸n original de las variables.

A continuaci贸n, te mostrar茅 un ejemplo en c贸digo Python que ilustra el Teorema del L铆mite Central:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar 1000 muestras de tama帽o n
n = 100
num_muestras = 1000
muestras = np.random.rand(num_muestras, n)

# Calcular la media de cada muestra
medias_muestras = np.mean(muestras, axis=1)

# Graficar el histograma de las medias muestrales
plt.hist(medias_muestras, bins=30, density=True)
plt.xlabel('Media Muestral')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribuci贸n de las Medias Muestrales')
plt.show()
```

En este ejemplo, generamos 1000 muestras, cada una de tama帽o 100, a partir de una distribuci贸n uniforme utilizando la funci贸n `np.random.rand()`. Luego, calculamos la media de cada muestra utilizando la funci贸n `np.mean()`.

El Teorema del L铆mite Central es importante porque permite realizar inferencias estad铆sticas sobre la media de una poblaci贸n, incluso si no conocemos la distribuci贸n de la poblaci贸n original. ## **Estimaci贸n de par谩metros**

La estimaci贸n de par谩metros es un proceso estad铆stico que consiste en determinar los valores desconocidos de los par谩metros de una distribuci贸n de probabilidad utilizando informaci贸n obtenida de una muestra de datos. Los par谩metros son medidas num茅ricas que resumen caracter铆sticas importantes de la distribuci贸n, como la media, la varianza, la proporci贸n, entre otros.

A continuaci贸n, te mostrar茅 un ejemplo de estimaci贸n de par谩metros utilizando Python y gr谩ficas correspondientes:

Supongamos que tenemos una muestra de datos que sigue una distribuci贸n normal y queremos estimar los par谩metros de media y varianza de esta distribuci贸n.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de datos
np.random.seed(0)
datos = np.random.normal(loc=10, scale=2, size=100)

# Estimar los par谩metros de media y varianza
media_estimada = np.mean(datos)
varianza_estimada = np.var(datos)

# Graficar el histograma de los datos
plt.hist(datos, bins=30, density=True, alpha=0.7)
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.title('Histograma de la muestra de datos')
plt.show()

# Imprimir los valores estimados de media y varianza
print("Media estimada:", media_estimada)
print("Varianza estimada:", varianza_estimada)
```

En el c贸digo anterior, generamos una muestra de datos utilizando la funci贸n `np.random.normal()` que sigue una distribuci贸n normal con media 10 y desviaci贸n est谩ndar 2. Luego, utilizamos las funciones `np.mean()` y `np.var()` para estimar los par谩metros de media y varianza de la distribuci贸n, respectivamente.

Posteriormente, graficamos el histograma de los datos utilizando `plt.hist()` para visualizar la distribuci贸n de los datos.

## **Intervalos de confianza**

Los intervalos de confianza son herramientas estad铆sticas que nos permiten estimar un rango de valores dentro del cual es probable que se encuentre un par谩metro desconocido de una poblaci贸n, con un nivel de confianza determinado. Proporcionan una medida de la incertidumbre asociada a una estimaci贸n puntual.

A continuaci贸n, te mostrar茅 un ejemplo de c贸mo calcular e interpretar un intervalo de confianza utilizando Python y gr谩ficas correspondientes:

Supongamos que queremos estimar el intervalo de confianza para la media de una poblaci贸n utilizando una muestra de datos.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generar una muestra de datos
np.random.seed(0)
datos = np.random.normal(loc=10, scale=2, size=100)

# Calcular el intervalo de confianza para la media
media_muestra = np.mean(datos)
error_estandar = stats.sem(datos)  # Error est谩ndar de la media
nivel_confianza = 0.95
z_score = stats.norm.ppf((1 + nivel_confianza) / 2)  # Valor cr铆tico Z
intervalo_confianza = (media_muestra - z_score * error_estandar,
                       media_muestra + z_score * error_estandar)

# Graficar el histograma de los datos
plt.hist(datos, bins=30, density=True, alpha=0.7)
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.title('Histograma de la muestra de datos')

# Graficar el intervalo de confianza
plt.axvline(intervalo_confianza[0], color='red', linestyle='--', label='Intervalo de Confianza')
plt.axvline(intervalo_confianza[1], color='red', linestyle='--')
plt.legend()

plt.show()

# Imprimir el intervalo de confianza
print("Intervalo de confianza para la media:", intervalo_confianza)
```

En el c贸digo anterior, generamos una muestra de datos utilizando la funci贸n `np.random.normal()` que sigue una distribuci贸n normal con media 10 y desviaci贸n est谩ndar 2.

Luego, utilizamos la funci贸n `stats.sem()` para calcular el error est谩ndar de la media de la muestra. Posteriormente, especificamos el nivel de confianza deseado, en este caso, 0.95, y utilizamos la funci贸n `stats.norm.ppf()` para obtener el valor cr铆tico `Z` correspondiente al nivel de confianza.

A continuaci贸n, calculamos el intervalo de confianza utilizando la f贸rmula: $media de la muestra 卤 valor cr铆tico Z * error est谩ndar de la media$.

Graficamos el histograma de los datos y agregamos l铆neas verticales para representar el intervalo de confianza.

Finalmente, imprimimos el intervalo de confianza calculado.

El intervalo de confianza indica que, en base a la muestra de datos y el nivel de confianza seleccionado, es probable que el par谩metro de la poblaci贸n (en este caso, la media) se encuentre dentro del intervalo calculado. Cuanto mayor sea el nivel de confianza, mayor ser谩 la amplitud del intervalo de confianza.

Es importante tener en cuenta que el intervalo de confianza proporciona informaci贸n sobre la precisi贸n de la estimaci贸n, pero no garantiza que el valor real del par谩metro se encuentre dentro del intervalo. Adem谩s, el c谩lculo del intervalo de confianza asume que los datos siguen una distribuci贸n normal o que la muestra es lo suficientemente grande para aplicar el Teorema del L铆mite Central.

En resumen, los intervalos de confianza son una herramienta 煤til para estimar rangos de valores probables de un par谩metro desconocido de una poblaci贸n. Nos permiten comprender la incertidumbre asociada a una estimaci贸n y tomar decisiones informadas en base a la variabilidad de los datos muestrales.
