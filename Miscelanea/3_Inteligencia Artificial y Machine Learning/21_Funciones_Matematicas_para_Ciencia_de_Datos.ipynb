### **C√≥mo leer las matem√°ticas: Conjuntos**

* Los `Diagramas de Venn` son otra forma de representar un conjunto de manera gr√°fica. Los diagramas nos permiten comprender y analizar f√°cilmente los conjuntos.

* conjunto vaci√≥ $(œï)$

**Nota:** El conjunto que contiene el elemento `0` o el $œï$ no est√° vac√≠o.

* La `uni√≥n` de dos conjuntos. $A$ y $B$ es el conjunto formado por todos los elementos de ambos conjuntos sin repetir ninguno y se representa as√≠:

$A ‚à™ B = { x ‚à£ x ‚àà A oh X ‚àà B }$

* La `intersecci√≥n` de dos conjuntos. $A$ y $B$ son el conjunto de elementos que forman parte tanto de $A$ como de $B$, es decir:

$A ‚à© B = { x ‚à£ x ‚àà A y x ‚àà B }$

* La diferencia de dos conjuntos. $A$ y $B$, es un conjunto formado por todos los elementos de $A$ que no est√°n en $B$ y se escribe as√≠:

$A / B = A ‚àí B = { x ‚àà U n ‚à£ x ‚àâ B }$

* Conjunto universo $(Œ©)$

Es el conjunto que incluye a todos los elementos y un conjunto.
$A$ `subconjunto` de $Œ©$.

* El `complemento` de $A$ son todos aquellos elementos del conjunto universo que no pertenecen a $A$, es decir:

$A^‚Ä≤ = { x ‚àà Œ© ‚à£ x ‚àâ U n }$

* Seg√∫n su n√∫mero de elementos, los conjuntos se pueden clasificar como `finitos o infinitos`.

* La cardinalidad de un conjunto $A$ es la cantidad de elementos diferentes que tiene y se representa con $‚à£ U n ‚à£$

Un conjunto es finito si se conoce su cardinalidad (podemos contar sus elementos) Un conjunto es infinito si se desconoce su cardinalidad (no podemos contar sus elementos).

**Funci√≥n lineal**

Tiene la forma de

$f(x) = mx + b$

**Donde:** $m$ y $b$ $‚àà$ $R$.

$m$ puede ser calculada por:

$m = \frac{y_2 - y_1}{x_2 - x_1}$

* $m$ se le conoce como `pendiente de la recta`.
* $b$ es `el punto de corte` con el eje $y$.
* Su `dominio` es $Domf = (‚àí‚àû,‚àû)$
* Su `imagen` es $Imf = (‚àí‚àû,‚àû)$

```python
# Ejemplo Funcion Lineal
def graficar(x,y):
    plt.plot(x,y)
    plt.grid()
    plt.axhline(y=0, color='r', linewidth=0.8)
    plt.axvline()
    plt.show()

N = 100 # N es la resolucion de una funcion

m = 1 # Pendiente de la recta

b = 0 # Punto de corte con el eje Y

# Funcion lineal de primer grado
def f(x):
    return m*x+b

x = np.linspace(-10,10, num=N)

y = f(x)

graficar(x,y)
```

**Funciones seccionadas**

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la funci√≥n escal√≥n de Heaviside:

$0, para, x < 0$

$1, para, x ‚â• 0$

```python
import numpy as np
import matplotlib.pyplot as plt

def H(x):
    Y = np.zeros(len(x))

    for idx, val in enumerate(x):
        if val >= 0:
            Y[idx] = 1
    return Y

N = 1000
x = np.linspace(-10, 10, N)  # Correcci√≥n: especificar el n√∫mero de puntos N

# Calcular la funci√≥n escal√≥n de Heaviside
y = H(x)

# Graficar la funci√≥n
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('H(x)')
plt.title('Funci√≥n Escal√≥n de Heaviside')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def A(x):
    Y = np.zeros(len(x))

    for idx, val in enumerate(x):
        if val >= 0:
            Y[idx] = val
        else:
            Y[idx] = -val

    return Y

def graficar(x, y):
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('|x|')
    plt.title('Funci√≥n Valor Absoluto')
    plt.grid(True)
    plt.show()

N = 1000
x = np.linspace(-120, 120, N)

# Calcular la funci√≥n valor absoluto
y = A(x)

# Graficar la funci√≥n
graficar(x, y)
```

### **¬øC√≥mo manipular funciones?**

**Desplazamientos verticales y horizontales**

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

* $y = f(x) + c$ se desplaza $c$ unidades hacia arriba.
* $y = f(x) ‚àí c$ se desplaza $c$ unidades hacia abajo.
* $y = f(x‚àíc)$ se desplaza $c$ unidades hacia la derecha.
* $y = f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100

def f(x):
    return x**2

c = 5

x = np.linspace(-10,10, num=N)

y2 = f(x) + (c+5)
y3 = f(x) - (c+5)
y4 = f(x - c)
y5 = f(x + c)

fig, ax = plt.subplots(2,2, figsize=(18,8))

def graficar2(pos, y, title):
    ax[pos].plot(x, y)
    ax[pos].set_title(title)
    ax[pos].grid()
    ax[pos].axhline(y=0, color='r', linewidth=0.8)
    ax[pos].axvline(x=0, color='r', linewidth=0.8)

graficar2((0,0), y2, '$y = f(x) + (c+5)$')
graficar2((0,1), y3, '$y = f(x) - (c+5)$')
graficar2((1,0), y4, '$y = f(x - c)$')
graficar2((1,1), y5, '$y = f(x + c)$')
plt.tight_layout()
plt.show()
```

**Alargamientos y compresiones**

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

* $y = c‚ãÖf(x)$ alarga la gr√°fica verticalmente en un factor de $c$.
* $y = \frac{1}{c}.f(x)$ comprime la gr√°fica verticalmente en un factor de $c$.
* $y = f(c‚ãÖx)$ comprime la gr√°fica horizontelmente en un factor de $c$.
* $y = f(\frac{1}{c}‚ãÖx)$ alarga la gr√°fica horizontelmente en un factor de $c$.

```python
# Ejemplos Alargamientos y Compresion
N = 1000

def f(x):
    return np.sin(x)

c = 2
y = f(x)
y1 = c*f(x)
y2 = (1/c) * f(x)
y3 = f(c * x)
y4 = f((1/c) * x)

graficar(x,y) # Funcion Original


fig, ax = plt.subplots(2,2, figsize=(18,8))

# graficar2(())
graficar2((0,0),y1,'$y = c*f(x)$')
graficar2((0,1),y2,'$y = (1/c) * f(x)$')
graficar2((1,0),y3,'$y = f(c * x)$')
graficar2((1,1),y4,'$y = f((1/c) * x)$')
plt.tight_layout()
```

**Reflexiones**

* $y = ‚àíf(x)$ refleja la gr√°fica respecto al eje $x$.
* $y = f(‚àíx)$ refleja la gr√°fica respecto al eje $y$.

```python
# Ejemplos Reflexion
N = 1000

def f(x):
  return x**3

x = np.linspace(-10,10, num=N)

y1 = f(x)
y2 = -f(x)
y3 = f(-x)

fig, ax = plt.subplots(1,3, figsize=(18,8))
graficar2((0),y1,'$y=f(x)$')
graficar2((1),y2,'$y=f(-x)$')
graficar2((2),y3,'$y=-f(x)$')
fig.tight_layout()
```

**Caracter√≠sticas de las funciones**

**¬øPor qu√© se llaman funciones reales?**

Se llaman funciones reales porque tanto su `dominio` como el `codominio` (recuerda que al `codominio tambi√©n se llama rango o imagen`) est√°n contenidos en el conjunto de los `n√∫meros reales`. Es decir, el `conjunto que contiene` a los `n√∫meros racionales e irracionales`. En otras palabras cualquier n√∫mero que no sea imaginario.

**Funci√≥n par**

Una funci√≥n es par si cumple la siguiente relaci√≥n a lo largo de su dominio:

$f(‚àíx) = f(x)$

Si lo notaste, esta relaci√≥n nos dice que una funci√≥n es par si es sim√©trica al eje vertical (eje Y). Por ejemplo, una par√°bola es una funci√≥n es par.

**Funci√≥n impar**

Una funci√≥n es impar si cumple la siguiente relaci√≥n a lo largo de su dominio:

$f(‚àíx) = ‚àíf(x)$

Desde un punto de vista geom√©trico, una funci√≥n impar posee una simetr√≠a rotacional con respecto al origen de coordenadas, lo que quiere decir que su gr√°fica no se altera luego de una rotaci√≥n de 180 grados alrededor del origen.

**Funci√≥n acotada**

Una funci√≥n es acotada si su codominio (tambi√©n conocido como rango o imagen) se encuentra entre dos valores, es decir, est√° acotado. Esta definici√≥n se define como que hay un n√∫mero $m$ que para todo valor del dominio de la funci√≥n se cumple que:

$‚àím ‚â§ f(x) ‚â§ m$

Por ejemplo, la funci√≥n seno o coseno est√°n acotadas en el intervalo [-1, 1] dentro de su co-dominio.

**Funciones mon√≥tonas**

Estas funciones son √∫tiles de reconocer o analizar debido a que nos permiten saber si una funci√≥n crece o decrece en alguno de sus intervalos. Que algo sea mon√≥tono significa que no tiene variaciones. Entonces las funciones mon√≥tonas son aquellas que dentro de un intervalo $I$, perteneciente a los n√∫meros reales, cumple alguna de estas propiedades:

**La funci√≥n es mon√≥tona y estrictamente creciente:**

si para todo $x_1$, $x_2$ $‚àà I : x_1 < x_2 ‚áî f(x_1) < f(x_2)$


La forma correcta de leer esto es ‚Äúsi para todo $x_1$ y $x_2$ que pertenecen al intervalo $I$, tal que $x_1$ sea menor a $x_2$, si y solo si $f(x_1)$ sea menor a $f(x_2)$‚Äù. En palabras mucho m√°s sencillas, lo que nos dice esta definici√≥n es que $x_1$ siempre tiene que ser menor que $x_2$ en nuestro intervalo $I$, y que al evaluar $x_2$ en la funci√≥n el resultado de esto siempre ser√° mayor que si evaluamos la funci√≥n en $x_1$. Para las siguientes tres definiciones restantes no cambia mucho la forma en la que se interpretan.

**La funci√≥n es mon√≥tona y estrictamente decreciente:**

si para todo $x_1, x_2 ‚àà I : x_1 < x_2 ‚áî f(x-1) > f(x_2)$


**La funci√≥n es mon√≥tona y creciente:**


si para todo $x_1, x_2 ‚àà I : x_1 < x_2 ‚áî f(x_1) ‚â§ f(x_2)$


**La funci√≥n es mon√≥tona y decreciente:**


si para todo $x_1, x_2 ‚àà I : x_1 < x_2 ‚áî f(x_1) ‚â• f(x_2)$

**Funciones peri√≥dicas**

Las funciones peri√≥dicas son aquellas que se repiten cada cierto periodo, este periodo se denomina con la letra $T$. La relaci√≥n que debe cumplir la funci√≥n para ser peri√≥dica es la siguiente.

$f(x) = f(x+T), T ‚â† 0$

Por ejemplo, la funci√≥n seno y coseno son funciones peri√≥dicas con un periodo $T = 2œÄ$. Es decir que si nosotros calculamos $f(x)$ y calculamos $f(x + 2œÄ)$ en la funci√≥n seno el valor que nos den ambas expresiones es el mismo.

**Funciones c√≥ncavas y convexas**

La forma de demostrar la concavidad de una funci√≥n se puede hacer a trav√©s del an√°lisis de derivadas consecutivas, pero a√∫n no llegamos a eso, as√≠ que no te preocupes. A continuaci√≥n, te dejo una forma s√∫per intuitiva de ver si una funci√≥n es c√≥ncava o convexa.

Se dice que una funci√≥n dentro de un intervalo es convexa si la funci√≥n ‚Äúabre hacia arriba‚Äù. Es decir si se ve la siguiente manera:

![convexa](../img/convexa.png "convexa")

**Resolucion**

La regresi√≥n lineal permite explicar la relaci√≥n de una variable dependiente ($y$), con respecto a otras variables independiente ($x$). Permite explicar como se afecta la variable dependiente ($y$) por los cambios que tenga la variable independiente ($x$).

$m = \frac{n‚àëxy - n‚àëx * n‚àëy}{n‚àëx^2 - n‚àëx^2}$


$b = \frac{‚àëy ‚àí a‚àó‚àëx}{n}$


$y=m‚àóx+b$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x = np.array([1.2, 2, 3.2, 2.5, 5, 6, 4, 8])
y = np.array([2, 3, 3.4, 3.1, 4, 4.7, 3.8, 7])
n = len(x)

xy = np.sum((x * y), axis=0)

x_2 = np.sum((x**2), axis=0)

sum_x = np.sum(x, axis=0)
sum_y = np.sum(y, axis=0)

def m():
    numerador = (n * xy) - (sum_x * sum_y)
    denominador = (n * x_2) - (sum_x**2)
    return np.divide(numerador, denominador)
m = m()

def b():
    numerador = sum_y - (m * sum_x)
    denominador = n
    return np.divide(numerador, denominador)
b = b()

def f():
    return (m * x) + b

y_prima = f()

df = pd.DataFrame.from_dict({
    "xy"      : xy,
    "x_2"     : x_2,
    "sum_x"   : sum_x,
    "sum_y"   : sum_y,
    "m"       : m,
    "b"       : b,
    "y_prima" : y_prima,
})
df.head(1)
```

```python
# Graficar
plt.scatter(x,y)
plt.plot(x,y_prima, '--')

plt.xlabel('x1000\npublicidad')
plt.ylabel('ganancias')

plt.grid()

plt.xlim(0,10)
plt.ylim(0,10)
```
