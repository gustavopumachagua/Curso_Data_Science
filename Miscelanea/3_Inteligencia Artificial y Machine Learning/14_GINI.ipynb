# **14. Qué es la Impureza GINI y por qué es tan relevante: Implementación paso a paso con Python**

## **Introducción**

La impureza Gini es una medida utilizada en algoritmos de aprendizaje automático para evaluar la calidad de una partición de datos en un conjunto de clases. Es especialmente relevante en el contexto de árboles de decisión y bosques aleatorios, ya que ayuda a determinar cómo dividir los datos de manera efectiva para crear nodos que sean lo más homogéneos posible en términos de clases.

La impureza Gini se calcula en función de la proporción de las clases en un nodo. Se mide en una escala de 0 a 1, donde 0 representa un nodo puro (todos los ejemplos pertenecen a la misma clase) y 1 representa un nodo impuro (las clases están distribuidas uniformemente). Un valor más bajo de impureza Gini indica que el nodo está más homogéneo en términos de clases.

La fórmula para calcular la impureza Gini en un nodo con respecto a las clases `C` es:

$$
\text{Gini}(N) = 1 - \sum_{i=1}^{C} (p_i)^2
$$

Donde:

- ($N$) es el nodo en consideración.
- ($C$) es el número de clases.
- ($p_i$) es la proporción de ejemplos en la clase ($i$) en el nodo ($N$).

Ahora, vamos a ver una implementación paso a paso de cómo calcular la impureza Gini utilizando Python y generaremos gráficos para comprender su comportamiento.

```python
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular la impureza Gini
def gini_impurity(classes):
    total_samples = np.sum(classes)
    if total_samples == 0:
        return 0
    class_probabilities = classes / total_samples
    gini = 1 - np.sum(class_probabilities ** 2)
    return gini

# Crear ejemplos de dos clases para demostración
class_1 = np.array([1, 1, 0, 1, 0, 0, 1, 0, 0, 1])
class_2 = np.array([0, 0, 1, 0, 1, 1, 0, 1, 1, 0])

# Calcular la impureza Gini para cada clase
gini_class_1 = gini_impurity(np.bincount(class_1))
gini_class_2 = gini_impurity(np.bincount(class_2))

print(f"Impureza Gini de la Clase 1: {gini_class_1:.2f}")
print(f"Impureza Gini de la Clase 2: {gini_class_2:.2f}")

# Graficar impureza Gini para diferentes proporciones de clases
proportions = np.linspace(0, 1, num=100)
gini_values = 1 - (proportions**2) - ((1 - proportions)**2)

plt.plot(proportions, gini_values)
plt.xlabel('Proporción de Clase 1')
plt.ylabel('Impureza Gini')
plt.title('Impureza Gini en función de la Proporción de Clase 1')
plt.grid()
plt.show()
```

En este ejemplo, hemos creado una función `gini_impurity` que calcula la impureza Gini para un conjunto de clases. Luego, hemos calculado la impureza Gini para dos conjuntos de clases de ejemplo y trazado cómo varía la impureza Gini en función de la proporción de una de las clases. Esto nos permite visualizar cómo la impureza Gini cambia a medida que cambia la distribución de las clases.

## **Interpretación de la Impureza Gini**

La impureza Gini es una medida importante en el contexto de los algoritmos de clasificación, especialmente en árboles de decisión y bosques aleatorios. Su valor proporciona información sobre la homogeneidad de las clases en un nodo o partición, lo que ayuda a determinar qué división de los datos resulta en nodos más puros.

Cuando se trabaja con la impureza Gini en el contexto de árboles de decisión, su interpretación se puede resumir de la siguiente manera:

- Un valor de impureza Gini igual a 0 indica que todos los ejemplos en el nodo pertenecen a la misma clase. Es decir, el nodo está puro y no necesita ser dividido más.

- Un valor de impureza Gini cercano a 1 indica que los ejemplos en el nodo están distribuidos de manera uniforme entre las clases. Esto significa que el nodo es impuro y contiene una mezcla de diferentes clases.

- Al tomar una decisión para dividir un nodo en un árbol de decisión, se elige la división que minimiza la impureza Gini ponderada de los nodos hijos. Esto significa que se busca la división que separa mejor las clases y crea nodos más puros.

- La impureza Gini se utiliza en el proceso de selección de características y para decidir qué atributos usar para dividir los nodos en un árbol. Los atributos que reducen la impureza Gini de manera efectiva son preferidos.

En resumen, la impureza Gini es una medida crucial para evaluar y tomar decisiones en algoritmos de clasificación. Proporciona información sobre la calidad de la división de datos en términos de la distribución de clases y ayuda a crear árboles de decisión más eficaces y precisos.

## **Impureza Gini paso a paso**

La impureza Gini es una medida utilizada para evaluar la homogeneidad de las clases en un conjunto de datos. Se aplica comúnmente en árboles de decisión y otros algoritmos de clasificación. Vamos a explorar cómo se calcula la impureza Gini paso a paso utilizando un ejemplo y código Python.

Supongamos que tenemos un conjunto de datos con dos clases, A y B. A continuación, vamos a calcular la impureza Gini para este conjunto de datos.

### Paso 1: Calcula la proporción de cada clase en el nodo
Supongamos que en nuestro conjunto de datos tenemos 40 ejemplos de clase A y 60 ejemplos de clase B.

- Proporción de clase A = 40 / (40 + 60) = 0.4
- Proporción de clase B = 60 / (40 + 60) = 0.6

### Paso 2: Calcula la impureza Gini para el nodo
La fórmula para calcular la impureza Gini en un nodo es:

$$Gini = 1 - (P_A^2 + P_B^2)$$

Donde $P_A$ es la proporción de ejemplos de clase A en el nodo y $P_B$ es la proporción de ejemplos de clase B en el nodo.

Sustituyendo los valores:

$$Gini = 1 - (0.4^2 + 0.6^2) = 0.48$$

### Interpretación
Un valor de impureza Gini más cercano a 0 indica que el nodo está más puro, es decir, contiene ejemplos principalmente de una sola clase. Un valor de impureza Gini más cercano a 0.5 indica que el nodo está menos puro, con una mezcla equilibrada de ejemplos de diferentes clases.

Ahora, vamos a realizar la implementación de este cálculo utilizando Python:

```python
# Proporciones de clases en el nodo
p_a = 0.4
p_b = 0.6

# Cálculo de la impureza Gini
gini = 1 - (p_a**2 + p_b**2)
print("Impureza Gini:", gini)
```

Este código calculará la impureza Gini y mostrará el resultado.

Recuerda que en árboles de decisión, la impureza Gini se utiliza para evaluar las posibles divisiones de nodos y se elige la división que reduzca la impureza Gini en los nodos hijos. De esta manera, se busca crear un árbol que separe de manera efectiva las clases y haga predicciones precisas.
