# **16. Datos de Entrenamiento, Validación y Prueba: ¿Cómo crearlos y qué objetivos tienen? Machine Learning**

## **Justificación de cada Conjunto**

Los conjuntos de datos de entrenamiento, validación y prueba son componentes fundamentales en el proceso de desarrollo y evaluación de modelos de machine learning. Estos conjuntos de datos se utilizan para entrenar, ajustar y evaluar el rendimiento de los modelos de machine learning. Cada conjunto de datos tiene un propósito específico y contribuye a la construcción de modelos más efectivos y generalizables.

**Conjunto de Datos de Entrenamiento:**

El conjunto de datos de entrenamiento es el conjunto inicial de ejemplos que se utiliza para entrenar un modelo de machine learning. Estos datos son utilizados por el modelo para aprender patrones y relaciones entre las características de entrada y las salidas esperadas. El modelo ajusta sus parámetros durante el proceso de entrenamiento para minimizar el error en la predicción de los datos de entrenamiento. El objetivo del conjunto de datos de entrenamiento es permitir que el modelo adquiera un entendimiento de los patrones en los datos para realizar predicciones precisas en nuevos datos.

**Conjunto de Datos de Validación:**

El conjunto de datos de validación se utiliza para ajustar los hiperparámetros del modelo y evaluar su rendimiento durante el proceso de desarrollo. Después de entrenar un modelo en el conjunto de entrenamiento, se evalúa en el conjunto de validación para ajustar hiperparámetros como la tasa de aprendizaje, la profundidad máxima de un árbol, etc. Esto se hace para evitar el sobreajuste (overfitting) y encontrar una configuración óptima del modelo. La retroalimentación del conjunto de validación permite afinar el modelo antes de probarlo en datos no vistos.

**Conjunto de Datos de Prueba:**

El conjunto de datos de prueba se utiliza para evaluar el rendimiento final y la generalización del modelo después de haber sido completamente ajustado y validado. Estos datos son independientes de los datos de entrenamiento y validación, y proporcionan una evaluación objetiva del rendimiento del modelo en datos no vistos. Evaluar el modelo en un conjunto de datos de prueba no utilizado previamente ayuda a estimar cómo se comportará en el mundo real.

**Creación de los Conjuntos de Datos:**

Los conjuntos de datos se crean dividiendo el conjunto de datos total disponible en tres partes: entrenamiento, validación y prueba. Por lo general, se utiliza una proporción como 70-15-15 o 80-10-10, pero esto puede variar según el tamaño del conjunto de datos y la naturaleza del problema.

**Objetivos de los Conjuntos de Datos:**

1. **Entrenamiento:** Ayudar al modelo a aprender patrones y relaciones en los datos para realizar predicciones precisas.
2. **Validación:** Ajustar los hiperparámetros del modelo y evaluar su rendimiento para evitar el sobreajuste y encontrar la mejor configuración.
3. **Prueba:** Evaluar el rendimiento final del modelo en datos no vistos para estimar su generalización y capacidad de predicción en el mundo real.

**Ejemplo:**
Supongamos que estamos construyendo un modelo para predecir el precio de las casas. Dividimos nuestro conjunto de datos en entrenamiento (70%), validación (15%) y prueba (15%) de la siguiente manera:

- Entrenamiento: 70% de los datos
- Validación: 15% de los datos
- Prueba: 15% de los datos

Luego, utilizamos el conjunto de entrenamiento para entrenar el modelo, el conjunto de validación para ajustar hiperparámetros y el conjunto de prueba para evaluar el rendimiento final del modelo.

**Gráfica:**
En una gráfica, la división de los conjuntos de datos se vería así:

```
+-------------------------------------+
|          Datos Totales              |
+-------------------------------------+
               /           |          \
+-----------+ +---------+ +---------+
| Entrenam. | | Valid.  | | Prueba  |
+-----------+ +---------+ +---------+
```

Cada conjunto de datos tiene un papel importante en el proceso de desarrollo y evaluación del modelo de machine learning.

## **Datos de Entrenamiento**

Los datos de entrenamiento son un conjunto de ejemplos que se utilizan para enseñar a un modelo de machine learning a realizar tareas específicas, como clasificación o regresión. Estos datos consisten en pares de entradas (características) y salidas (etiquetas o valores esperados). El modelo utiliza los datos de entrenamiento para aprender patrones y relaciones entre las características y las salidas, lo que le permite hacer predicciones precisas en datos no vistos.

**Ejemplo de Datos de Entrenamiento:**

Supongamos que estamos construyendo un modelo de clasificación para determinar si un correo electrónico es spam o no spam. Nuestro conjunto de datos de entrenamiento podría verse así:


|   Características (Entradas)   | Etiqueta (Salida) |
|-------------------------|------------------|
| Contenido del Correo 1  |      No Spam     |
| Contenido del Correo 2  |      Spam        |
| Contenido del Correo 3  |      No Spam     |
| ...                     |      ...         |


En este ejemplo, las características son los contenidos de los correos electrónicos y la etiqueta es si el correo es spam o no spam.

**Creación de Datos de Entrenamiento:**

Los datos de entrenamiento se crean dividiendo el conjunto de datos total disponible en una proporción específica, como 70-80% del total. Esto implica seleccionar un subconjunto de ejemplos y sus respectivas etiquetas para ser utilizados en la fase de entrenamiento del modelo.

**Gráfica en Código Python:**

A continuación se muestra un ejemplo de cómo crear y visualizar un conjunto de datos de entrenamiento utilizando Python y la biblioteca Matplotlib:

````python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de entrenamiento sintéticos
np.random.seed(0)
num_samples = 100
X_train = np.random.rand(num_samples, 1)  # Característica
y_train = (X_train > 0.5).astype(int)      # Etiqueta (0 o 1)

# Graficar los datos de entrenamiento
plt.scatter(X_train, y_train, c=y_train, cmap='viridis')
plt.xlabel('Característica')
plt.ylabel('Etiqueta')
plt.title('Datos de Entrenamiento')
plt.show()
````

En este ejemplo, se generan datos de entrenamiento sintéticos en una dimensión. Los puntos se distribuyen aleatoriamente y se asignan etiquetas 0 o 1 según si la característica es mayor que 0.5. Luego, los puntos se grafican con colores diferentes según sus etiquetas.

Los datos de entrenamiento son esenciales para el proceso de construcción de modelos de machine learning, ya que permiten al modelo aprender patrones y relaciones en los datos para realizar predicciones precisas en nuevos ejemplos no vistos.

## **Datos de Prueba**

Los datos de prueba son un conjunto de ejemplos que se utilizan para evaluar el rendimiento y la generalización de un modelo de machine learning después de que haya sido entrenado. Estos datos son independientes de los datos de entrenamiento y no se utilizan durante el proceso de entrenamiento del modelo. En lugar de eso, se usan para medir cuán bien el modelo puede hacer predicciones en datos que no ha visto previamente.

**Ejemplo de Datos de Prueba:**

Siguiendo el ejemplo anterior de clasificación de correos electrónicos como spam o no spam, los datos de prueba podrían verse así:


|   Características (Entradas)   | Etiqueta (Salida) |
|-------------------------|------------------|
| Contenido del Correo A  |      No Spam     |
| Contenido del Correo B  |      Spam        |
| Contenido del Correo C  |      Spam        |
| ...                     |      ...         |


Estos ejemplos se utilizan después de entrenar el modelo para verificar su capacidad de generalización en datos no vistos.

**Creación de Datos de Prueba:**

Los datos de prueba se seleccionan de manera similar a los datos de entrenamiento, pero son independientes de ellos. Los datos de prueba no deben utilizarse en ninguna etapa del entrenamiento o ajuste del modelo, para que la evaluación sea imparcial.

**Gráfica en Código Python:**

A continuación, se presenta un ejemplo de cómo crear y visualizar un conjunto de datos de prueba utilizando Python y la biblioteca Matplotlib:

````python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de prueba sintéticos
np.random.seed(1)
num_samples = 50
X_test = np.random.rand(num_samples, 1)  # Característica
y_test = (X_test > 0.5).astype(int)      # Etiqueta (0 o 1)

# Graficar los datos de prueba
plt.scatter(X_test, y_test, c=y_test, cmap='viridis')
plt.xlabel('Característica')
plt.ylabel('Etiqueta')
plt.title('Datos de Prueba')
plt.show()
````

En este ejemplo, se generan datos de prueba sintéticos en una dimensión de manera similar al ejemplo anterior. Los puntos se distribuyen aleatoriamente y se asignan etiquetas 0 o 1 según si la característica es mayor que 0.5. Luego, los puntos se grafican con colores diferentes según sus etiquetas.

Los datos de prueba son esenciales para evaluar la capacidad de generalización de un modelo y determinar su rendimiento en datos no vistos durante el entrenamiento.

## **Datos de Validación**

Los datos de validación son un subconjunto de los datos de entrenamiento que se utilizan para ajustar los hiperparámetros y tomar decisiones sobre la configuración del modelo de machine learning. Estos datos son diferentes de los datos de prueba y se usan para evaluar y mejorar el rendimiento del modelo antes de que se pruebe en los datos de prueba.

**Importancia de los Datos de Validación:**

La validación es una etapa crucial en el proceso de entrenamiento de un modelo. Permite ajustar los hiperparámetros, que son configuraciones que no se aprenden durante el entrenamiento, sino que se establecen antes del proceso de entrenamiento. Los hiperparámetros pueden influir en la capacidad del modelo para generalizar correctamente y evitar el sobreajuste (overfitting). Utilizar datos de prueba para ajustar hiperparámetros puede llevar a una optimización excesiva para un conjunto específico de datos de prueba, lo que puede resultar en un rendimiento pobre en datos no vistos.

**Ejemplo de Datos de Validación:**

Siguiendo el ejemplo anterior de clasificación de correos electrónicos como spam o no spam, los datos de validación podrían verse así:


|   Características (Entradas)   | Etiqueta (Salida) |
|-------------------------|------------------|
| Contenido del Correo X  |      Spam        |
| Contenido del Correo Y  |      No Spam     |
| Contenido del Correo Z  |      Spam        |
| ...                     |      ...         |


**Creación de Datos de Validación:**

Los datos de validación se toman del conjunto de datos de entrenamiento y se reservan para ajustar hiperparámetros. Una práctica común es dividir los datos de entrenamiento en dos partes: una parte para entrenamiento y otra para validación. La relación entre los datos de entrenamiento y los de validación puede variar dependiendo de la cantidad total de datos disponibles.

**Gráfica en Código Python:**

A continuación, se presenta un ejemplo de cómo crear y visualizar un conjunto de datos de validación utilizando Python y la biblioteca Matplotlib:

````python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de validación sintéticos
np.random.seed(2)
num_samples = 50
X_validation = np.random.rand(num_samples, 1)  # Característica
y_validation = (X_validation > 0.6).astype(int)  # Etiqueta (0 o 1)

# Graficar los datos de validación
plt.scatter(X_validation, y_validation, c=y_validation, cmap='viridis')
plt.xlabel('Característica')
plt.ylabel('Etiqueta')
plt.title('Datos de Validación')
plt.show()
````

En este ejemplo, se generan datos de validación sintéticos en una dimensión. Los puntos se distribuyen aleatoriamente y se asignan etiquetas 0 o 1 según si la característica es mayor que 0.6. Luego, los puntos se grafican con colores diferentes según sus etiquetas.

Los datos de validación se utilizan para afinar y seleccionar los hiperparámetros del modelo antes de realizar pruebas en los datos de prueba.
