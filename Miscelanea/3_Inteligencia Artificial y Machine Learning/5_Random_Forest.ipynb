**Conceptos B谩sicos:**

**Ejemplo con Gr谩ficas:**

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Datos de ejemplo
horas_estudio = np.array([2, 3, 5, 7, 8, 10])
calificaciones = np.array([65, 70, 80, 85, 90, 95])

# Divisi贸n de datos
X_train, X_test, y_train, y_test = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=42)

# Crear y entrenar el modelo
model = LinearRegression()
model.fit(X_train.reshape(-1, 1), y_train)

# Predicciones en el conjunto de prueba
y_pred = model.predict(X_test.reshape(-1, 1))

# Evaluaci贸n del modelo
mse = mean_squared_error(y_test, y_pred)
print(f"Error cuadrado medio: {mse:.2f}")
```

**Interpretaci贸n del Ejemplo:**

En este ejemplo, hemos utilizado Regresi贸n Lineal para crear un modelo que predice las calificaciones en funci贸n de las horas de estudio. Hemos entrenado el modelo, realizado predicciones en el conjunto de prueba y evaluado su rendimiento utilizando el error cuadrado medio.

En resumen, la creaci贸n de un modelo implica seleccionar el algoritmo adecuado, preparar los datos, entrenar el modelo, ajustar sus hiperpar谩metros, evaluar su rendimiento y utilizarlo para hacer predicciones en nuevos datos.
