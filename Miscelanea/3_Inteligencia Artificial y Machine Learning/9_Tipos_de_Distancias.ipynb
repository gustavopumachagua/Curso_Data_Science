# **9. 3 Tipos de Distancias para T√©cnicas de Clustering con Python: Euclidiana, Chebyshev y Manhattan**

## **Introducci√≥n**

Las distancias son una parte esencial de las t√©cnicas de clustering, ya que determinan c√≥mo se calcula la similitud o la diferencia entre los puntos de datos. Hay varias medidas de distancia que se pueden utilizar, y aqu√≠ te presento tres tipos comunes: Euclidiana, Chebyshev y Manhattan. Usaremos la biblioteca Scikit-learn en Python para ilustrar cada una de estas distancias en un ejemplo de clustering.

**1. Distancia Euclidiana:**

La distancia Euclidiana es la distancia m√°s com√∫nmente utilizada y se calcula como la ra√≠z cuadrada de la suma de los cuadrados de las diferencias entre las coordenadas de dos puntos. En otras palabras, mide la longitud del segmento de l√≠nea que conecta dos puntos en un espacio n-dimensional.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances

# Generar datos sint√©ticos
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Calcular distancias Euclidianas entre puntos
euclidean_distances = pairwise_distances(X, metric='euclidean')

# Visualizar los datos y las distancias
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o', s=50)
plt.title('Datos y Distancias Euclidianas')
plt.xlabel('Caracter√≠stica 1')
plt.ylabel('Caracter√≠stica 2')

plt.figure(figsize=(10, 6))
plt.imshow(euclidean_distances, cmap='viridis', aspect='auto')
plt.colorbar(label='Distancia Euclidiana')
plt.title('Matriz de Distancias Euclidianas')
plt.show()
```

**2. Distancia de Chebyshev:**

La distancia de Chebyshev se calcula tomando la m√°xima diferencia entre las coordenadas de dos puntos a lo largo de cualquier dimensi√≥n.

```python
# Calcular distancias de Chebyshev entre puntos
chebyshev_distances = pairwise_distances(X, metric='chebyshev')

# Visualizar las distancias
plt.figure(figsize=(10, 6))
plt.imshow(chebyshev_distances, cmap='viridis', aspect='auto')
plt.colorbar(label='Distancia de Chebyshev')
plt.title('Matriz de Distancias de Chebyshev')
plt.show()
```

**3. Distancia de Manhattan:**

La distancia de Manhattan (tambi√©n conocida como distancia de la ciudad) se calcula como la suma de las diferencias absolutas entre las coordenadas de dos puntos a lo largo de todas las dimensiones.

```python
# Calcular distancias de Manhattan entre puntos
manhattan_distances = pairwise_distances(X, metric='manhattan')

# Visualizar las distancias
plt.figure(figsize=(10, 6))
plt.imshow(manhattan_distances, cmap='viridis', aspect='auto')
plt.colorbar(label='Distancia de Manhattan')
plt.title('Matriz de Distancias de Manhattan')
plt.show()
```

En estos ejemplos, hemos generado datos sint√©ticos y luego calculado las matrices de distancias correspondientes utilizando las funciones `pairwise_distances` de Scikit-learn. Luego, hemos visualizado las matrices de distancias para cada tipo de distancia. Cada celda de la matriz de distancias representa la distancia entre dos puntos en el espacio de caracter√≠sticas.

Estas tres distancias (Euclidiana, Chebyshev y Manhattan) son √∫tiles en diferentes situaciones y contextos. La elecci√≥n de la distancia depende del problema y los datos espec√≠ficos que est√©s tratando de abordar.

## **Distancia Euclidiana**

La distancia Euclidiana es una medida com√∫nmente utilizada para calcular la distancia entre dos puntos en un espacio euclidiano de m√∫ltiples dimensiones. Esta distancia se basa en el teorema de Pit√°goras y mide la longitud de la l√≠nea recta que conecta dos puntos en un espacio n-dimensional. La f√≥rmula para calcular la distancia Euclidiana entre dos puntos ($P$) y ($Q$) en un espacio ($n$)-dimensional es la siguiente:

$d(P, Q) = \sqrt{(q_1 - p_1)^2 + (q_2 - p_2)^2 + \ldots + (q_n - p_n)^2}$

Donde ($p_i$) y ($q_i$) son las coordenadas del punto ($P$) y ($Q$) en la dimensi√≥n ($i$), respectivamente.

Ahora, veamos un ejemplo de c√≥mo calcular la distancia Euclidiana entre dos puntos en Python utilizando la biblioteca NumPy y visualizarla con una gr√°fica:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir dos puntos en un espacio bidimensional
point_p = np.array([2, 3])
point_q = np.array([5, 7])

# Calcular la distancia Euclidiana entre los dos puntos
euclidean_distance = np.linalg.norm(point_q - point_p)

# Visualizar los puntos y la distancia Euclidiana
plt.figure(figsize=(8, 6))
plt.scatter([point_p[0], point_q[0]], [point_p[1], point_q[1]], color='blue', marker='o', label='Puntos')
plt.plot([point_p[0], point_q[0]], [point_p[1], point_q[1]], color='red', linestyle='--', label='Distancia Euclidiana')
plt.annotate('P', (point_p[0], point_p[1]), textcoords="offset points", xytext=(-15,10), ha='center', fontsize=12)
plt.annotate('Q', (point_q[0], point_q[1]), textcoords="offset points", xytext=(15,10), ha='center', fontsize=12)
plt.xlabel('Dimensi√≥n 1')
plt.ylabel('Dimensi√≥n 2')
plt.title('Distancia Euclidiana entre dos puntos')
plt.legend()
plt.grid()
plt.show()

print(f"Distancia Euclidiana entre P y Q: {euclidean_distance:.2f}")
```

En este ejemplo, hemos definido dos puntos \(P\) y \(Q\) en un espacio bidimensional y calculado la distancia Euclidiana entre ellos utilizando la funci√≥n `np.linalg.norm()` de NumPy. Luego, hemos visualizado los puntos y la distancia Euclidiana en una gr√°fica. La l√≠nea roja en la gr√°fica representa la distancia Euclidiana entre los dos puntos.

La distancia Euclidiana es ampliamente utilizada en diversas aplicaciones, incluyendo machine learning y an√°lisis de datos, para medir la similitud o diferencia entre puntos en un espacio multidimensional.

## **Distancia Chebyshev**

La distancia Chebyshev, tambi√©n conocida como la distancia del ajedrez, es una medida de distancia que calcula la mayor diferencia entre las coordenadas de dos puntos en un espacio ($n$)-dimensional. A diferencia de la distancia Euclidiana que considera las diferencias cuadr√°ticas en todas las dimensiones, la distancia Chebyshev considera √∫nicamente la mayor diferencia en cualquiera de las dimensiones.

La f√≥rmula para calcular la distancia Chebyshev entre dos puntos ($P$) y ($Q$) en un espacio ($n$)-dimensional es:

$d(P, Q) = \max\left(|q_1 - p_1|, |q_2 - p_2|, \ldots, |q_n - p_n|\right)$

Donde ($p_i$) y ($q_i$) son las coordenadas del punto ($P$) y ($Q$) en la dimensi√≥n ($i$), respectivamente.

Ahora, veamos un ejemplo de c√≥mo calcular la distancia Chebyshev entre dos puntos en Python utilizando la biblioteca NumPy y visualizarla con una gr√°fica:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir dos puntos en un espacio bidimensional
point_p = np.array([2, 3])
point_q = np.array([5, 7])

# Calcular la distancia Chebyshev entre los dos puntos
chebyshev_distance = np.max(np.abs(point_q - point_p))

# Visualizar los puntos y la distancia Chebyshev
plt.figure(figsize=(8, 6))
plt.scatter([point_p[0], point_q[0]], [point_p[1], point_q[1]], color='blue', marker='o', label='Puntos')
plt.plot([point_p[0], point_q[0]], [point_p[1], point_q[1]], color='green', linestyle='--', label='Distancia Chebyshev')
plt.annotate('P', (point_p[0], point_p[1]), textcoords="offset points", xytext=(-15,10), ha='center', fontsize=12)
plt.annotate('Q', (point_q[0], point_q[1]), textcoords="offset points", xytext=(15,10), ha='center', fontsize=12)
plt.xlabel('Dimensi√≥n 1')
plt.ylabel('Dimensi√≥n 2')
plt.title('Distancia Chebyshev entre dos puntos')
plt.legend()
plt.grid()
plt.show()

print(f"Distancia Chebyshev entre P y Q: {chebyshev_distance}")
```

En este ejemplo, hemos definido dos puntos ($P$) y ($Q$) en un espacio bidimensional y calculado la distancia Chebyshev entre ellos utilizando la funci√≥n `np.max()` y `np.abs()` de NumPy para obtener la mayor diferencia absoluta entre las coordenadas. Luego, hemos visualizado los puntos y la distancia Chebyshev en una gr√°fica. La l√≠nea verde en la gr√°fica representa la distancia Chebyshev entre los dos puntos.

La distancia Chebyshev es especialmente √∫til cuando se desea medir la distancia entre puntos considerando √∫nicamente la mayor diferencia en cualquiera de las dimensiones, lo que puede ser relevante en ciertos escenarios de clustering y an√°lisis de datos.

## **Distancia Manhattan**

La distancia Manhattan, tambi√©n conocida como distancia de la ciudad o distancia de taxicab, es una medida de distancia que calcula la suma de las diferencias absolutas entre las coordenadas de dos puntos en un espacio ($n$)-dimensional. A diferencia de la distancia Euclidiana que considera las diferencias cuadr√°ticas en todas las dimensiones, la distancia Manhattan suma las diferencias absolutas en cada dimensi√≥n.

La f√≥rmula para calcular la distancia Manhattan entre dos puntos ($P$) y ($Q$) en un espacio ($n$)-dimensional es:

$d(P, Q) = \sum_{i=1}^{n} |q_i - p_i|$

Donde ($p_i$) y ($q_i$) son las coordenadas del punto ($P$) y ($Q$) en la dimensi√≥n ($i$), respectivamente.

Ahora, veamos un ejemplo de c√≥mo calcular la distancia Manhattan entre dos puntos en Python utilizando la biblioteca NumPy y visualizarla con una gr√°fica:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir dos puntos en un espacio bidimensional
point_p = np.array([2, 3])
point_q = np.array([5, 7])

# Calcular la distancia Manhattan entre los dos puntos
manhattan_distance = np.sum(np.abs(point_q - point_p))

# Visualizar los puntos y la distancia Manhattan
plt.figure(figsize=(8, 6))
plt.scatter([point_p[0], point_q[0]], [point_p[1], point_q[1]], color='blue', marker='o', label='Puntos')
plt.plot([point_p[0], point_q[0]], [point_p[1], point_q[1]], color='green', linestyle='--', label='Distancia Manhattan')
plt.annotate('P', (point_p[0], point_p[1]), textcoords="offset points", xytext=(-15,10), ha='center', fontsize=12)
plt.annotate('Q', (point_q[0], point_q[1]), textcoords="offset points", xytext=(15,10), ha='center', fontsize=12)
plt.xlabel('Dimensi√≥n 1')
plt.ylabel('Dimensi√≥n 2')
plt.title('Distancia Manhattan entre dos puntos')
plt.legend()
plt.grid()
plt.show()

print(f"Distancia Manhattan entre P y Q: {manhattan_distance}")
```

En este ejemplo, hemos definido dos puntos ($P$) y ($Q$) en un espacio bidimensional y calculado la distancia Manhattan entre ellos utilizando la funci√≥n `np.sum()` y `np.abs()` de NumPy para obtener la suma de las diferencias absolutas entre las coordenadas en cada dimensi√≥n. Luego, hemos visualizado los puntos y la distancia Manhattan en una gr√°fica. La l√≠nea verde en la gr√°fica representa la distancia Manhattan entre los dos puntos.

La distancia Manhattan es especialmente √∫til cuando se desea medir la distancia entre puntos considerando las diferencias absolutas en cada dimensi√≥n, lo que puede ser relevante en ciertos escenarios de clustering y an√°lisis de datos.

## **Comparaci√≥n de las m√©tricas usando DBSCAN**

La comparaci√≥n de m√©tricas utilizando DBSCAN (Density-Based Spatial Clustering of Applications with Noise) es una forma de evaluar el rendimiento de diferentes m√©tricas de distancia al aplicar el algoritmo DBSCAN en un conjunto de datos. DBSCAN es un algoritmo de clustering que agrupa puntos cercanos en funci√≥n de su densidad en lugar de la distancia euclidiana. Las m√©tricas de distancia juegan un papel crucial en la formaci√≥n de los grupos en DBSCAN.

Vamos a realizar una comparaci√≥n de m√©tricas utilizando el conjunto de datos de ejemplo "blobs" de la biblioteca `sklearn.datasets`, aplicando DBSCAN con diferentes m√©tricas de distancia y visualizando los resultados. En este ejemplo, compararemos las m√©tricas "euclidean" (euclidiana), "manhattan" (Manhattan) y "chebyshev" (Chebyshev).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Crear un conjunto de datos de ejemplo "blobs"
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Escalar los datos para una mejor comparaci√≥n de las m√©tricas
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Definir las m√©tricas de distancia a comparar
metrics = ['euclidean', 'manhattan', 'chebyshev']

# Crear subplots para visualizar los resultados
fig, axes = plt.subplots(1, len(metrics), figsize=(15, 5))

for i, metric in enumerate(metrics):
    # Crear instancia de DBSCAN con la m√©trica actual
    dbscan = DBSCAN(eps=0.5, min_samples=5, metric=metric)
    
    # Realizar clustering con DBSCAN
    labels = dbscan.fit_predict(X_scaled)
    
    # Ignorar puntos ruidosos para el c√°lculo de la silueta
    if len(set(labels)) > 1:  # Asegurarse de que hay m√°s de un cluster
        silhouette_avg = silhouette_score(X_scaled, labels)
    else:
        silhouette_avg = 0.0
    
    # Visualizar los resultados
    axes[i].scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis')
    axes[i].set_title(f'M√©trica: {metric}\nSilueta: {silhouette_avg:.2f}')
    axes[i].set_xlabel('Feature 1')
    axes[i].set_ylabel('Feature 2')

plt.tight_layout()
plt.show()
```

En este c√≥digo, primero creamos un conjunto de datos de ejemplo "blobs" y lo escalamos para normalizar los valores. Luego, definimos una lista de m√©tricas de distancia a comparar: "euclidean", "manhattan" y "chebyshev". Utilizamos un bucle para iterar a trav√©s de estas m√©tricas y, en cada iteraci√≥n, creamos una instancia de DBSCAN con la m√©trica actual y realizamos el clustering en los datos escalados.

Luego, calculamos el n√∫mero de clusters formados, as√≠ como el √≠ndice de silueta para evaluar la calidad de los clusters. Finalmente, visualizamos los resultados de clustering coloreando los puntos seg√∫n sus etiquetas de cluster y mostrando informaci√≥n sobre la m√©trica, el n√∫mero de clusters y el √≠ndice de silueta en cada gr√°fico.

La comparaci√≥n de m√©tricas en este ejemplo nos ayuda a entender c√≥mo diferentes m√©tricas afectan la formaci√≥n de clusters utilizando DBSCAN en un conjunto de datos espec√≠fico.
