# **11. K-Vecinos M谩s Cercanos (KNN) para Regresi贸n con Python**

## **Interpretaci贸n b谩sica**

El algoritmo de los K-Vecinos M谩s Cercanos (KNN) tambi茅n se puede utilizar para problemas de regresi贸n, no solo para clasificaci贸n. En la regresi贸n KNN, en lugar de predecir una clase, se trata de predecir un valor num茅rico continuo. El proceso general del KNN para regresi贸n es similar al de KNN para clasificaci贸n, pero en lugar de votar por la clase mayoritaria, se promedian los valores de los k vecinos m谩s cercanos para obtener la predicci贸n final.

Aqu铆 tienes una explicaci贸n detallada con un ejemplo y c贸digo en Python:

**Paso 1: Importar bibliotecas y cargar los datos**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Cargar el conjunto de datos Diabetes
data = load_diabetes()
X = data.data
y = data.target

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Paso 2: Crear y entrenar el modelo KNN para regresi贸n**

```python
# Crear y entrenar el modelo KNN para regresi贸n
k = 5  # N煤mero de vecinos
regressor = KNeighborsRegressor(n_neighbors=k)
regressor.fit(X_train_scaled, y_train)
```

**Paso 3: Realizar predicciones y evaluar el modelo**

```python
# Realizar predicciones en el conjunto de prueba
y_pred = regressor.predict(X_test_scaled)

# Calcular el error cuadr谩tico medio (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Error cuadr谩tico medio: {mse:.2f}")
```

**Paso 4: Visualizaci贸n de resultados**

```python
# Graficar las predicciones vs. valores reales
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([0, 350], [0, 350], color='red', linestyle='--')
plt.xlabel('Valor Real')
plt.ylabel('Predicci贸n')
plt.title('Predicciones vs. Valores Reales en Regresi贸n KNN')
plt.grid()
plt.show()
```

En este ejemplo, hemos utilizado el conjunto de datos Diabetes para realizar regresi贸n KNN. El modelo KNN para regresi贸n se crea y entrena de manera similar al modelo de clasificaci贸n KNN. Luego, se realizan predicciones en el conjunto de prueba y se calcula el error cuadr谩tico medio (MSE) para evaluar el rendimiento del modelo.

La gr谩fica muestra las predicciones en funci贸n de los valores reales. Idealmente, las predicciones deber铆an estar cerca de la l铆nea de referencia (l铆nea roja punteada), lo que indicar铆a una buena correlaci贸n entre las predicciones y los valores reales.

Este ejemplo ilustra c贸mo aplicar KNN para regresi贸n en Python. El modelo busca patrones en los datos de entrenamiento y luego estima los valores continuos en funci贸n de los vecinos m谩s cercanos en el espacio de caracter铆sticas.

## **Construcci贸n del modelo**

Por supuesto, la construcci贸n del modelo de regresi贸n KNN para el conjunto de datos de diabetes se puede hacer utilizando el c贸digo proporcionado en la explicaci贸n anterior. Aqu铆 est谩 el c贸digo completo para construir el modelo:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Cargar el conjunto de datos Diabetes
data = load_diabetes()
X = data.data
y = data.target

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear y entrenar el modelo KNN para regresi贸n
k = 5  # N煤mero de vecinos
regressor = KNeighborsRegressor(n_neighbors=k)
regressor.fit(X_train_scaled, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = regressor.predict(X_test_scaled)

# Calcular el error cuadr谩tico medio (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Error cuadr谩tico medio: {mse:.2f}")

# Graficar las predicciones vs. valores reales
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([0, 350], [0, 350], color='red', linestyle='--')
plt.xlabel('Valor Real')
plt.ylabel('Predicci贸n')
plt.title('Predicciones vs. Valores Reales en Regresi贸n KNN')
plt.grid()
plt.show()
```

Este c贸digo realiza todas las etapas necesarias para construir y evaluar un modelo de regresi贸n KNN utilizando el conjunto de datos de diabetes. Desde la carga de datos y el escalado hasta la predicci贸n y la visualizaci贸n de los resultados, cada paso se detalla en el c贸digo. Puedes ejecutar este c贸digo en tu entorno de Python para construir y evaluar el modelo de regresi贸n KNN.

## **Predicci贸n de un rango de precios**

Si deseas realizar predicciones de un rango de precios utilizando el modelo de regresi贸n KNN que construimos previamente, aqu铆 tienes un ejemplo de c贸mo podr铆as hacerlo:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor

# Cargar el conjunto de datos Diabetes
data = load_diabetes()
X = data.data
y = data.target

# Utilizar solo una caracter铆stica (por ejemplo, edad)
X = X[:, np.newaxis, 0]

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear y entrenar el modelo KNN para regresi贸n
k = 5  # N煤mero de vecinos
regressor = KNeighborsRegressor(n_neighbors=k)
regressor.fit(X_train_scaled, y_train)

# Definir el rango de valores para la predicci贸n
precio_min = 50
precio_max = 300
num_puntos = 100
precios = np.linspace(precio_min, precio_max, num_puntos).reshape(-1, 1)

# Realizar predicciones en el rango de precios
precios_scaled = scaler.transform(precios)
predicciones = regressor.predict(precios_scaled)

# Graficar las predicciones en funci贸n del precio
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, label='Valores Reales')
plt.plot(precios, predicciones, color='red', label='Predicciones')
plt.xlabel('Edad')
plt.ylabel('Nivel de Az煤car en la Sangre')
plt.title('Predicciones en un Rango de Precios con Regresi贸n KNN')
plt.legend()
plt.grid()
plt.show()
```

En este ejemplo, hemos definido un rango de precios (`precio_min` a `precio_max`) y hemos generado una serie de valores de precios (`precios`) dentro de ese rango. Luego, escalamos los valores de precios utilizando el mismo escalador que se utiliz贸 para el conjunto de entrenamiento. Despu茅s de eso, realizamos predicciones para los valores de precios escalados y graficamos tanto los valores reales como las predicciones en funci贸n del precio. Esto te dar谩 una idea de c贸mo se comporta el modelo de regresi贸n KNN en el rango de precios que has especificado.

## **KNN con diferentes valores de K**

El algoritmo de los k vecinos m谩s cercanos (KNN) es un m茅todo de aprendizaje supervisado utilizado tanto para clasificaci贸n como para regresi贸n. KNN clasifica un punto de datos seg煤n la mayor铆a de las etiquetas de sus k vecinos m谩s cercanos en el espacio de caracter铆sticas. Aqu铆, exploraremos c贸mo KNN se comporta con diferentes valores de k en un problema de clasificaci贸n utilizando Python y la librer铆a Scikit-Learn.

**Pasos a seguir**:

1. **Cargar datos y dividirlos en conjuntos de entrenamiento y prueba**:
   Empezaremos cargando un conjunto de datos de ejemplo y dividi茅ndolo en conjuntos de entrenamiento y prueba.

2. **Entrenar modelos KNN con diferentes valores de k**:
   Entrenaremos varios modelos KNN utilizando diferentes valores de k y evaluaremos su rendimiento en el conjunto de prueba.

3. **Visualizaci贸n de resultados**:
   Visualizaremos c贸mo cambia la precisi贸n en el conjunto de prueba a medida que variamos el valor de k.

**Ejemplo de c贸digo**:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data
y = iris.target

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Definir diferentes valores de k
k_values = [1, 3, 5, 7, 9]

# Entrenar modelos KNN con diferentes valores de k
accuracy_scores = []
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracy_scores, marker='o')
plt.title('Precisi贸n en el Conjunto de Prueba para diferentes valores de k')
plt.xlabel('Valor de k')
plt.ylabel('Precisi贸n')
plt.xticks(k_values)
plt.grid()
plt.show()
```

En este ejemplo, estamos utilizando el conjunto de datos Iris y evaluando el rendimiento del modelo KNN con diferentes valores de k. La gr谩fica resultante mostrar谩 c贸mo cambia la precisi贸n en el conjunto de prueba a medida que variamos el valor de k. Es importante encontrar un equilibrio entre valores peque帽os y grandes de k para evitar sobreajuste (k peque帽o) y subajuste (k grande).
