# **11. K-Vecinos Más Cercanos (KNN) para Regresión con Python**

## **Interpretación básica**

El algoritmo de los K-Vecinos Más Cercanos (KNN) también se puede utilizar para problemas de regresión, no solo para clasificación. En la regresión KNN, en lugar de predecir una clase, se trata de predecir un valor numérico continuo. El proceso general del KNN para regresión es similar al de KNN para clasificación, pero en lugar de votar por la clase mayoritaria, se promedian los valores de los k vecinos más cercanos para obtener la predicción final.

Aquí tienes una explicación detallada con un ejemplo y código en Python:

**Paso 1: Importar bibliotecas y cargar los datos**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Cargar el conjunto de datos Diabetes
data = load_diabetes()
X = data.data
y = data.target

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Paso 2: Crear y entrenar el modelo KNN para regresión**

```python
# Crear y entrenar el modelo KNN para regresión
k = 5  # Número de vecinos
regressor = KNeighborsRegressor(n_neighbors=k)
regressor.fit(X_train_scaled, y_train)
```

**Paso 3: Realizar predicciones y evaluar el modelo**

```python
# Realizar predicciones en el conjunto de prueba
y_pred = regressor.predict(X_test_scaled)

# Calcular el error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Error cuadrático medio: {mse:.2f}")
```

**Paso 4: Visualización de resultados**

```python
# Graficar las predicciones vs. valores reales
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([0, 350], [0, 350], color='red', linestyle='--')
plt.xlabel('Valor Real')
plt.ylabel('Predicción')
plt.title('Predicciones vs. Valores Reales en Regresión KNN')
plt.grid()
plt.show()
```

En este ejemplo, hemos utilizado el conjunto de datos Diabetes para realizar regresión KNN. El modelo KNN para regresión se crea y entrena de manera similar al modelo de clasificación KNN. Luego, se realizan predicciones en el conjunto de prueba y se calcula el error cuadrático medio (MSE) para evaluar el rendimiento del modelo.

La gráfica muestra las predicciones en función de los valores reales. Idealmente, las predicciones deberían estar cerca de la línea de referencia (línea roja punteada), lo que indicaría una buena correlación entre las predicciones y los valores reales.

## **Explicación de KNN para regresión**

**Explicación de K-Vecinos Más Cercanos (KNN) para Regresión con Python**

El algoritmo K-Vecinos Más Cercanos (KNN) no se limita a la clasificación, sino que también se puede aplicar a problemas de regresión. En la regresión KNN, en lugar de predecir una clase, se estima un valor numérico continuo. La idea central es similar: encontrar los k ejemplos más cercanos en el conjunto de entrenamiento a un punto de prueba y luego promediar los valores de destino correspondientes a esos ejemplos para predecir el valor del punto de prueba.

Aquí tienes una explicación detallada junto con un ejemplo y su código en Python:

**Paso 1: Importar bibliotecas y cargar los datos**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Cargar el conjunto de datos Diabetes
data = load_diabetes()
X = data.data
y = data.target

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Paso 2: Crear y entrenar el modelo KNN para regresión**

```python
# Crear y entrenar el modelo KNN para regresión
k = 5  # Número de vecinos
regressor = KNeighborsRegressor(n_neighbors=k)
regressor.fit(X_train_scaled, y_train)
```

**Paso 3: Realizar predicciones y evaluar el modelo**

```python
# Realizar predicciones en el conjunto de prueba
y_pred = regressor.predict(X_test_scaled)

# Calcular el error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Error cuadrático medio: {mse:.2f}")
```

**Paso 4: Visualización de resultados**

```python
# Graficar las predicciones vs. valores reales
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([0, 350], [0, 350], color='red', linestyle='--')
plt.xlabel('Valor Real')
plt.ylabel('Predicción')
plt.title('Predicciones vs. Valores Reales en Regresión KNN')
plt.grid()
plt.show()
```

**Explicación detallada:**

1. **Cargar y dividir datos:** Comenzamos importando las bibliotecas necesarias y cargando el conjunto de datos Diabetes. Dividimos los datos en conjuntos de entrenamiento y prueba utilizando la función `train_test_split`.

2. **Escalamiento de características:** Para asegurar que todas las características tengan la misma escala, aplicamos el escalado utilizando `StandardScaler`.

3. **Creación y entrenamiento del modelo:** Creamos un modelo de regresión KNN utilizando `KNeighborsRegressor` y especificamos el número de vecinos `k`. Luego, ajustamos el modelo utilizando los datos de entrenamiento escalados.

4. **Predicciones y evaluación:** Realizamos predicciones en el conjunto de prueba utilizando el modelo entrenado. Luego, calculamos el error cuadrático medio (MSE) para evaluar el rendimiento del modelo. Un MSE más bajo indica un mejor ajuste.

5. **Visualización:** Graficamos las predicciones en función de los valores reales. La línea roja punteada representa la línea de referencia ideal donde las predicciones serían idénticas a los valores reales. Una dispersión cercana a esta línea indica un buen rendimiento del modelo.

Este ejemplo ilustra cómo aplicar KNN para regresión en Python. El modelo busca patrones en los datos de entrenamiento y luego estima los valores continuos en función de los vecinos más cercanos en el espacio de características.

## **Construcción del modelo**

Por supuesto, la construcción del modelo de regresión KNN para el conjunto de datos de diabetes se puede hacer utilizando el código proporcionado en la explicación anterior. Aquí está el código completo para construir el modelo:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Cargar el conjunto de datos Diabetes
data = load_diabetes()
X = data.data
y = data.target

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear y entrenar el modelo KNN para regresión
k = 5  # Número de vecinos
regressor = KNeighborsRegressor(n_neighbors=k)
regressor.fit(X_train_scaled, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = regressor.predict(X_test_scaled)

# Calcular el error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Error cuadrático medio: {mse:.2f}")

# Graficar las predicciones vs. valores reales
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([0, 350], [0, 350], color='red', linestyle='--')
plt.xlabel('Valor Real')
plt.ylabel('Predicción')
plt.title('Predicciones vs. Valores Reales en Regresión KNN')
plt.grid()
plt.show()
```

Este código realiza todas las etapas necesarias para construir y evaluar un modelo de regresión KNN utilizando el conjunto de datos de diabetes. Desde la carga de datos y el escalado hasta la predicción y la visualización de los resultados, cada paso se detalla en el código. Puedes ejecutar este código en tu entorno de Python para construir y evaluar el modelo de regresión KNN.

## **Predicción de un rango de precios**

Si deseas realizar predicciones de un rango de precios utilizando el modelo de regresión KNN que construimos previamente, aquí tienes un ejemplo de cómo podrías hacerlo:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor

# Cargar el conjunto de datos Diabetes
data = load_diabetes()
X = data.data
y = data.target

# Utilizar solo una característica (por ejemplo, edad)
X = X[:, np.newaxis, 0]

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear y entrenar el modelo KNN para regresión
k = 5  # Número de vecinos
regressor = KNeighborsRegressor(n_neighbors=k)
regressor.fit(X_train_scaled, y_train)

# Definir el rango de valores para la predicción
precio_min = 50
precio_max = 300
num_puntos = 100
precios = np.linspace(precio_min, precio_max, num_puntos).reshape(-1, 1)

# Realizar predicciones en el rango de precios
precios_scaled = scaler.transform(precios)
predicciones = regressor.predict(precios_scaled)

# Graficar las predicciones en función del precio
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, label='Valores Reales')
plt.plot(precios, predicciones, color='red', label='Predicciones')
plt.xlabel('Edad')
plt.ylabel('Nivel de Azúcar en la Sangre')
plt.title('Predicciones en un Rango de Precios con Regresión KNN')
plt.legend()
plt.grid()
plt.show()
```

En este ejemplo, hemos definido un rango de precios (`precio_min` a `precio_max`) y hemos generado una serie de valores de precios (`precios`) dentro de ese rango. Luego, escalamos los valores de precios utilizando el mismo escalador que se utilizó para el conjunto de entrenamiento. Después de eso, realizamos predicciones para los valores de precios escalados y graficamos tanto los valores reales como las predicciones en función del precio. Esto te dará una idea de cómo se comporta el modelo de regresión KNN en el rango de precios que has especificado.

## **KNN con diferentes valores de K**

El algoritmo de los k vecinos más cercanos (KNN) es un método de aprendizaje supervisado utilizado tanto para clasificación como para regresión. KNN clasifica un punto de datos según la mayoría de las etiquetas de sus k vecinos más cercanos en el espacio de características. Aquí, exploraremos cómo KNN se comporta con diferentes valores de k en un problema de clasificación utilizando Python y la librería Scikit-Learn.

**Pasos a seguir**:

1. **Cargar datos y dividirlos en conjuntos de entrenamiento y prueba**:
   Empezaremos cargando un conjunto de datos de ejemplo y dividiéndolo en conjuntos de entrenamiento y prueba.

2. **Entrenar modelos KNN con diferentes valores de k**:
   Entrenaremos varios modelos KNN utilizando diferentes valores de k y evaluaremos su rendimiento en el conjunto de prueba.

3. **Visualización de resultados**:
   Visualizaremos cómo cambia la precisión en el conjunto de prueba a medida que variamos el valor de k.

**Ejemplo de código**:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data
y = iris.target

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Definir diferentes valores de k
k_values = [1, 3, 5, 7, 9]

# Entrenar modelos KNN con diferentes valores de k
accuracy_scores = []
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracy_scores, marker='o')
plt.title('Precisión en el Conjunto de Prueba para diferentes valores de k')
plt.xlabel('Valor de k')
plt.ylabel('Precisión')
plt.xticks(k_values)
plt.grid()
plt.show()
```

En este ejemplo, estamos utilizando el conjunto de datos Iris y evaluando el rendimiento del modelo KNN con diferentes valores de k. La gráfica resultante mostrará cómo cambia la precisión en el conjunto de prueba a medida que variamos el valor de k. Es importante encontrar un equilibrio entre valores pequeños y grandes de k para evitar sobreajuste (k pequeño) y subajuste (k grande).
