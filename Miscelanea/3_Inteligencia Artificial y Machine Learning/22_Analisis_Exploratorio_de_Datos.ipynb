### **¬øC√≥mo hacer un an√°lisis exploratorio de datos?**

**Organizar y entender las variables**

Lograremos identificar los distintos tipos de variables, las categor√≠as a las que pertenecen y por consiguiente el tipo de an√°lisis que podemos realizar con ellas.

### **Tipos de an√°lisis de datos**

Sin importar el tama√±o del conjunto de datos que tengamos, podemos dividir el conjunto de datos en 4 partes evocando a una pregunta esencial por cada etapa

**Anal√≠tica Descriptiva**

Es el an√°lisis m√°s b√°sico. Permite dar una vista general de las propiedades, dimensiones y comportamiento de un fen√≥meno desde la mera descripci√≥n de los datos.

**Anal√≠tica Diagnostica**

Es aplicar unos criterios diferenciales a los datos descriptivos para identificar una situaci√≥n problem√°tica. Permite establecer relaciones entre variables y crear hip√≥tesis sobre las posibles causas de su comportamiento actual.

**Anal√≠tica Predictiva**

Utiliza datos para explicar un posible comportamiento de las variables de inter√©s si se siguen las tendencias actuales o si estas cambian. Permite determinar si la situaci√≥n problem√°tica identificada puede representar mayor o menor riesgo en el futuro.

**Anal√≠tica Prescriptiva**

Emplea datos para apoyar la toma de decisiones, con base en eventos del pasado. Propone soluciones basadas en evidencia. Igualmente, la etapa prescriptiva tambi√©n puede hacer iniciar el ciclo de nuevo, ya que monitorear la soluci√≥n tambi√©n da pie a procesos nuevos de descripci√≥n y diagn√≥stico de ese monitoreo, volviendo a iniciar el ciclo una vez m√°s.

![Analisis](../img/analisis.png "Analisis")

**Ejemplo**

1. **Anal√≠tica Descriptiva (Que sucedi√≥?)**

Fuiste a dormir por la noche como cualquier otra, sinti√©ndote bien, pero al despertar te das cuenta de que te sientes fatigado, con dolor muscular, flema y fiebre. Al preguntarte que est√° sucediendo hiciste una descripci√≥n de los s√≠ntomas que te hacen sentir mal.
Al tratar de averiguar porque te sientes mal recuerdas que estuviste junto a un compa√±ero de trabajo que se sent√≠a enfermo hace dos d√≠as, por lo que decides ir al m√©dico para poder recibir un diagn√≥stico

2. **Anal√≠tica Diagnostica (Porque sucedi√≥?)**

Cuando llegas al hospital, el doctor har√° un breve interrogatorio para tener una descripci√≥n general de lo que te est√° ocurriendo, pidi√©ndote que le repitas los s√≠ntomas que tienes en ese momento para luego darte un diagn√≥stico y concluir de que tienes un resfriado com√∫n tas la descripci√≥n de tus s√≠ntomas.

3. **Anal√≠tica Predictiva (Que podr√≠a pasar si?)**

Antes de irte del hospital, una enfermera te detiene y te menciona que quiere hacerte unos an√°lisis, ya que puede ocurrir el caso de lo que ser√≠a una enfermedad com√∫n para la normalidad de las personas, a ti te trate completamente diferente, por lo cual la enfermera realiza los an√°lisis necesarios para saber si puede ocurrir una situaci√≥n inesperada ante esta enfermedad por cierta condici√≥n, tras los an√°lisis se concluye que est√°s bien y pasas a la siguiente etapa

Anal√≠tica Prescriptiva (Que deber√≠a hacerse?)
La enfermera al llegar a la conclusi√≥n de que te encuentras bien, te da el alta y previamente te prescribe los medicamentos y tipo de descanso que tendr√°s que realizar.

### **Tipos de datos y an√°lisis de variables**

**Tipos de variables**

Las variables categ√≥ricas pueden clasificarse en:

**Variables categ√≥ricas**

* **Variable nominal**

Es aquella variable cualitativa cuya categor√≠a no sigue ning√∫n orden, se agrupa sin ninguna jerarqu√≠a entre s√≠.

**Ejemplos:**

* Lateralidad (zurdo, diestro)
* Estado civil (soltero, casado, divorciado, uni√≥n libre)
* Grupo sangu√≠neo (Grupo A-Grupo B-Grupo AB-Grupo 0)

* **Variable ordinal**

Son aquellas variables categ√≥ricas con orden secuencial o progresi√≥n natural esperable o jerarqu√≠a.

**Ejemplos:**

* Medalla deportiva (oro, plata, bronce)
* Nivel de clase educativa (√∫ltimo a√±o, primer a√±o, etc.)

**Variables num√©ricas**

* **Variables continuas**

Son aquellas caracter√≠sticas que se miden dentro de un rango continuo infinito de valores num√©ricos y se registran con n√∫meros reales. Pueden presentar cualquier valor dentro de cierto intervalo.

**Ejemplos:**

* Estatura (1,76543 m)
* Peso (55,6625 kg)
* Tama√±o de lesi√≥n de leishmaniasis (6,0458 mm)

* **Variables discretas**

Tambi√©n llamadas discontinuas, est√°n asociadas a conteos o enumeraciones, raz√≥n por la cual solo pueden registrarse con n√∫meros enteros (0,1,2,3, etc.)

**Ejemplos:**

* N√∫mero de hijos en una familia (0, 1, 2, 3, etc.)
* N√∫mero de familias residentes en una manzana (20, 25, 45)
* N√∫mero de insectos atrapados en una red (0, 1, 2, 5, 10)

### **Tipos de an√°lisis**

**An√°lisis Univariado**

El An√°lisis Univariado busca analizar cada variable por separado, entender sus caracter√≠sticas y como se definen por s√≠ misma.

**An√°lisis Bivariado**

El An√°lisis Bi variado busca analizar la relaci√≥n de cada par de variables. Permite buscar relaciones intr√≠nsecas entre los datos

**An√°lisis Multivariado**

El An√°lisis Multivariado busca analizar el efecto simult√°neo de m√∫ltiples variables. Analiza la variable como un conjunto

En el comienzo de la EDA, el an√°lisis univariado te va a dar informaci√≥n mucho m√°s valiosa que comenzar con todas las variables al mismo tiempo. Te ayuda a analizar la contribuci√≥n individual de una variable sobre las dem√°s para luego incrementar la cantidad de variables y analizar los efectos en conjunto.

**Tipos de recolecci√≥n de datos**

**Primaria**

Datos Recolectados de primera mano a trav√©s de encuestas, entrevistas, experimentos y otros. Puedes decidir que recolectas y que quiere medir. Tienes total control de los datos.

$2\frac{IQR(x)}{\sqrt{n}}$

```python
def freedman_diaconis_bindwith(x: pd.Series) -> float:
    """Find optimal bindwith using Freedman-Diaconis rule."""

    IQR = x.quantile(.075) - x.quantile(0.25)
    N = x.size

    return np.abs(2 * IQR / N * 1/3)
freedman_diaconis_bindwith(penguins_df.bill_length_mm)
```

Estas preguntas evocan a las funciones de probabilidad:

* Funci√≥n de Probabilidad de masas (PMF, Probability Mass Function)
* Funci√≥n de Distribuci√≥n Acumulada (CDF, Cumulative Distribution Function)
* Funci√≥n de Probabilidad de Densidad (PDF, Probability Density Function)

**Funci√≥n de Probabilidad de masas (PMFs)**

Nos indica la probabilidad de que una variable aleatoria discreta tome un valor determinado.

**Ejemplo:** Cu√°l es la probabilidad de que en un sal√≥n haya personas de 23 a√±os.

**Funci√≥n de Distribuci√≥n Acumulada (CDFs)**

Devuelve la probabilidad de que una variable sea menor o igual a un valor determinado.

**Ejemplo:** Cu√°l es la probabilidad de que en un sal√≥n haya personas de 23 o menos a√±os.

**Funci√≥n de Probabilidad de Densidad (PDFs)**

Determina la probabilidad de que una variable continua tome un valor determinado entre un intervalo de valores.

**Ejemplo:** Cu√°l es la probabilidad de que en un sal√≥n haya una persona de 1.845 metros

**Bonus: Ley de los Grandes N√∫meros y Teorema del L√≠mite Central**

**¬øQu√© es la Ley de los Grandes N√∫meros?**

La probabilidad experimental tiende a la probabilidad te√≥rica a medida que aumenta el n√∫mero de repeticiones del experimento.

Es muy √∫til cuando no se conoce la probabilidad te√≥rica de un evento y tenemos la capacidad de aumentar la muestra. Mediante este teorema, al incrementar la muestra, la probabilidad experimental al final se convertir√° en la probabilidad te√≥rica.

**Ejemplo:**

Imaginemos que tenemos un dado de 6 caras que se encuentra en perfecto equilibrio, o sea, que la probabilidad de que caiga cada cara es exactamente 1/6.

Si tiramos el dado 10 veces, el dado cay√≥ m√°s veces de alguna cara que sus otras caras, pero si el experimento lo repetimos a un n√∫mero lo suficiente grande para que las probabilidades se emparejen, tendremos una distribuci√≥n casi exacta de 1/6 por cara

**Teorema del l√≠mite central**

La media de las muestras tiende aproximadamente a una distribuci√≥n normal

La suma de $n$ variables aleatorias independientes con medias y varianzas finitas converge en distribuci√≥n a una variable aleatoria normal

Fijamos el n√∫mero de muestras a tomar y el tama√±o de cada muestra, luego calculamos la media de los machos de cada muestra y a todas ellas le sacamos la media observando as√≠ que se aproxima a la proporci√≥n de machos dada por los datos

```python
import pandas as pd
import numpy as np

# Supongamos que sex_numeric es una lista con "Male" y "Female"
sex_numeric = ["Male", "Female"] * 20  # Ejemplo de datos

number_sample = 1000  # Cantidad de muestras a tomar
sample_size = 35  # Tama√±o de la muestra

sample_proportions = []  # Lista para almacenar las proporciones de "Male" en cada muestra

for i in range(number_sample):
    # Tomar una muestra aleatoria de 'sex_numeric'
    sex_numeric_sample = np.random.choice(sex_numeric, sample_size)
    # Calcular la proporci√≥n de "Male" en la muestra
    male_proportion = np.mean(sex_numeric_sample == "Male")
    # Agregar la proporci√≥n al lista
    sample_proportions.append(male_proportion)

# Calcular la media de las proporciones
male_population_mean = np.mean(sample_proportions)
print(f'Estimated percentage of male penguins in population is: {male_population_mean * 100:.2f}%')
```

**Limitantes de los coeficientes de correlaci√≥n lineal**

Solo nos ayuda a determinar la posible existencia de una correlaci√≥n lineal; sin embargo, su ausencia no significa que no exista otro tipo de correlaci√≥n.

Para demostrarlo de una manera gr√°fica, pasaremos a generar un conjunto de datos aleatorios.

```python
from sklearn.metrics import r2_score


plt.figure(figsize=(5,5))

x = np.linspace(-100,120,100)
y = x**2
y += np.random.normal(0,1000,x.size)

plt.scatter(x=x,y=y)
plt.show()

np.corrcoef(x,y)
```

Para este caso podemos ver la correlaci√≥n cuadr√°tica de nuestros datos.

En el caso de que aplic√°ramos un Coeficiente de Pearson (`np.corrcoef`) ver√≠amos que la correlaci√≥n entre variables seria de 0.07, lo cual indicar√≠a una correlaci√≥n lineal inexistente.
Pero con este resultado no podemos afirmar que no existe una correlaci√≥n entre nuestros datos, ya que en el gr√°fico podemos ver una correlaci√≥n cuadr√°tica.

Cuando la distribuci√≥n tiene un comportamiento c√∫bico, el coeficiente detecta la correlaci√≥n simulando que es una l√≠nea recta donde se compensa los valores, por lo que hace una lectura err√≥nea del panorama.

Recordando nuestra matriz de correlaci√≥n, ve√≠amos que la correlaci√≥n entre el largo y ancho del pico era cercana al $0$, por lo tanto, pod√≠amos concluir que nuestras variables no se encontraban relacionadas

**El coeficiente de correlaci√≥n no nos habla del impacto de la relaci√≥n**

Con impacto nos referimos al hecho de que podemos tener un coeficiente de correlaci√≥n que es m√°s alto que otro, pero esto no necesariamente nos indica que este coeficiente sea mejor para predecir que otro coeficiente con menor valor.

**Ejemplo:**

```python
np.random.seed(42)
x_1= np.linspace(0,100,100)
y_1= 0.1 * x_1 + 3 + np.random.uniform(-2,2, size=x_1.size)

sns.scatterplot(
    x=x_1,
    y=y_1
)

x_2= np.linspace(0,100,100)
y_2= 0.5 * x_2 + 1 + np.random.uniform(0,60, size=x_2.size)

sns.scatterplot(
    x=x_2,
    y=y_2
)

plt.legend(labels=['1','2'])

print(np.corrcoef(x_1,y_1))
print(np.corrcoef(x_2,y_2))
```

Como vemos, tenemos dos series, los azules ($x_1$) y los naranjas ($x_2$), y vemos que el coeficiente de correlaci√≥n de $x_1$ es de 0.92 lo cual nos indica una correlaci√≥n casi perfecta, pero en nuestra $x_2$ vemos que su coeficiente de correlaci√≥n es menor 0.67.

En este caso, si solo hubi√©ramos visto el coeficiente de correlaci√≥n de $x_1$ por ser m√°s alto, nos hubi√©ramos perdido de la serie $x_2$, que tiene un mayor impacto (la pendiente o inclinaci√≥n de $Y$)

Imagina el caso en que la $X$ es algo que represente alg√∫n esfuerzo y la $Y$ es lo que te represente el dinero generado por el esfuerzo.

Efectivamente, en la serie 1 tenemos una correlaci√≥n alta, que se encuentra bien definida su trayectoria, pero a cambio su impacto es bajo, por lo tanto, crece poco en referencia a la serie 2.

Pero analizando a la serie 2, vemos que la correlaci√≥n es menos marcada que la serie 1, pero el impacto que genera este en el eje Y es mucho mayor que el de la serie 1, representando una cantidad mayor de dinero.

Este tipo de ejemplo es muy √∫til para entender por qu√© visualizar el impacto de la relaci√≥n, ya que:

**¬øComo calcular la fuerza del efecto?**

**Limitaciones del an√°lisis de regresi√≥n simple**

A misma manera que vimos con las matrices de correlaciones, la regresi√≥n simple posee ventajas y desventajas, entenderlas nos ayudaran a saber cuando usarlas en el momento correcto.

**Limitaciones:**

* La regresi√≥n lineal simple de $A ‚àí B$ no es lo mismo que la de $B ‚àí A$
* Si dos variables crecen o decrecen siguiendo las mismas pautas, no implica necesariamente que una cause la otra
* Solo puede manejar relaciones lineales.

Pongamos en pr√°ctica estas limitantes con Python

**La regresi√≥n lineal simple no es sim√©trica**

```python
# Definimos nuestras variables de interes
x = penguins_df.bill_length_mm
y = penguins_df.bill_depth_mm
```

```python
# Ajustemos un modelo de regresion lineal para cada variable

# X contra Y
res_x_y = scipy.stats.linregress(x=x,y=y)

# Y contra X
res_y_x = scipy.stats.linregress(x=y, y=x)
```

```python
print(f'''
{res_x_y}

{res_y_x}''')
```

Visualizaci√≥n de la Asimetr√≠a de la Regresi√≥n Lineal

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
sns.scatterplot(
    ax=ax[0],
    x=y,
    y=x
)

fx_2 = np.array([y.min(),y.max()])
fy_2 = res_y_x.intercept + res_y_x.slope * fx_2

ax[0].plot(fx_2,fy_2)
sns.scatterplot(
    ax=ax[1],
    x=x,
    y=y,
)

fx_1 = np.array([x.min(), x.max()])
fy_1 = res_x_y.intercept + res_x_y.slope * fx_1

ax[1].plot(fx_1, fy_1)
```

A pesar de que los gr√°ficos parecen similares, no lo son, ya que uno es el transpuesto del otro y esto cambia la forma que interpretamos la pendiente o `Slope` entre las variables.

```python
# Grafico de ambos slopes
sns.scatterplot(
    x=x,
    y=y
)

plt.plot(fx_1, fy_1)
plt.plot(fy_2, fx_2)
plt.legend({f'x:{x.name}',f'x:{y.name}'})
```

**La regresi√≥n no nos dice nada sobre la causalidad, pero existen herramientas para separar las relaciones entre m√∫ltiples variables**

Ahora que sabemos que una limitante de las regresiones lineales simples, veremos el concepto en Python con la librer√≠a statsmodels para separar sus relaciones

* statsmodels.formula.api.ols
* Fitting models using R-style formulas

```python
(
    smf.ols(
         formula='bill_length_mm ~ bill_depth_mm',
        data=penguins_df
    )
    .fit()
    .params
)
```

```python
(
    smf.ols(
        formula='bill_depth_mm ~ bill_length_mm',
        data=penguins_df
    )
    .fit()
    .params
)
```

De esta forma logramos ver √©l intercepto o el punto de corte del eje $Y$ y como por cada unidad que se mueva en el eje $X$ se modificaria el `slope`.

En el ejemplo vemos que cuando tenemos a `bill_length_mm` en el eje $X$ y a `bill_depth_mm` en el eje $Y$, el punto de corte del eje $Y$ se daria en el punto 54.89 y el `slope` o pendiente seria de -0.634.

Y cuando transponemos las variables el punto de corte se da en el punto 20.78 y el slope o pendiente es de -0.082

En la f√≥rmula podr√≠amos agregar distintas variables para explicar el fen√≥meno de nuestro inter√©s, pero esto involucrar√≠a trabajar con m√°s de dos variables, que es lo que veremos con el an√°lisis de regresi√≥n m√∫ltiple

**An√°lisis de regresi√≥n m√∫ltiple**

Ya vimos las ventajas, desventajas de usar una regresi√≥n lineal simple, pero ya ha llegado el momento de que implementemos m√°s de dos variables a la vez, ya sea porque:

* Buscamos encontrar la relaci√≥n entre dos o m√°s variables
* Creemos que la interacci√≥n de m√∫ltiple variables pueden afectar a nuestra variable objetivo

Para poder explorar y determinar estos intereses, un ejemplo es el an√°lisis de regresi√≥n m√∫ltiple.

Anteriormente, vimos que una de las limitaciones del an√°lisis de regresi√≥n simple era que es un modelo para regresiones lineales, donde no pod√≠amos incorporar comportamientos no lineales.
Pero con el an√°lisis de regresi√≥n m√∫ltiple si podemos analizar distintos comportamientos.

Otra de las limitaciones que vimos de la regresi√≥n simple era que no necesariamente nos indicaba la relaci√≥n entre la causa y efecto de las variables.
Pero el an√°lisis de regresi√≥n m√∫ltiple al dejarnos incorporar distintas variables nos permitir√° explicar si la relaci√≥n entre nuestras variables realmente est√° afectando o no el comportamiento de nuestra variable objetivo

**An√°lisis de regresi√≥n m√∫ltiple en Python:**

Para realizar este ejercicio, imaginaremos un escenario hipot√©tico, donde somos cient√≠ficos y nos encontr√°bamos en la Ant√°rtida haciendo mediciones sobre los ping√ºinos, donde mediamos la longitud y grosor de sus alas, sus picos, si son machos o hembras y su peso. Pero resulta que te olvidaste tu balanza donde pesabas a los ping√ºinos, entonces tomaste todas las otras medidas, pero te falto registrar cuanto pesaba cada ping√ºino, por lo tanto, se te planteo este problema:

Olvidaste tu b√°scula para pesar a los ping√ºinos.¬øCu√°l ser√≠a la mejor forma de capturar ese dato?

¬øTe preguntaste si exist√≠a alguna manera de poder asociar las medidas que ten√≠as y lograr inferir cuanto pesaba cada ping√ºino?

Por lo tanto, para resolver este problema usaremos los modelos de regresi√≥n m√∫ltiple

**Creando modelos**

```python
import pandas as pd
import numpy as np

# Supongamos que tienes un DataFrame llamado penguins_df con varias columnas, incluyendo algunas no num√©ricas
# Vamos a crear un DataFrame de ejemplo para demostraci√≥n

# Crear un DataFrame de ejemplo
data = {
    'height': [172, 165, 180, 168, 175],
    'weight': [70, 62, 78, 65, 72],
    'species': ['Adelie', 'Gentoo', 'Chinstrap', 'Adelie', 'Gentoo']
}

penguins_df = pd.DataFrame(data)

# Aseg√∫rate de que solo estamos considerando columnas num√©ricas
numeric_penguins_df = penguins_df[['height', 'weight']]

# Calcula la matriz de correlaci√≥n
correlation_matrix = numeric_penguins_df.corr()

# Muestra la matriz de correlaci√≥n
print("Matriz de correlaci√≥n:")
print(correlation_matrix)
```
