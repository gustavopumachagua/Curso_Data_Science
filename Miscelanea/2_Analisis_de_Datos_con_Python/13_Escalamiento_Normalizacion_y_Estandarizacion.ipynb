# **13. Escalamiento, Normalización y Estandarización de Datos con Python para Ciencia de Datos**

## **Problemática**

Por supuesto, te explicaré los conceptos de escalamiento, normalización y estandarización de datos en el contexto de la ciencia de datos utilizando NumPy en Python. Estos son procesos comunes para preprocesar datos y asegurarse de que estén en una escala adecuada antes de aplicar algoritmos de aprendizaje automático.

**Escalamiento, Normalización y Estandarización:**

1. **Escalamiento:** El escalamiento se refiere a cambiar el rango de los valores en los datos. Puede implicar cambiar el rango de valores de una variable para que estén dentro de un cierto rango, como entre 0 y 1.

2. **Normalización:** La normalización es un proceso en el que se ajustan los valores de las variables para que tengan una media de 0 y una desviación estándar de 1.

3. **Estandarización:** La estandarización también ajusta los valores para que tengan una media de 0 y una desviación estándar de 1, pero en este caso, no necesariamente los valores estarán en un rango específico.

**Ejemplos de Escalamiento, Normalización y Estandarización:**

Supongamos que tienes un conjunto de datos con una variable llamada "edad" y otra variable llamada "ingreso".

````python
import numpy as np

# Crear un conjunto de datos de ejemplo
edad = np.array([25, 30, 35, 40, 45])
ingreso = np.array([50000, 75000, 100000, 85000, 90000])

# Escalamiento (entre 0 y 1)
edad_escala = (edad - edad.min()) / (edad.max() - edad.min())
ingreso_escala = (ingreso - ingreso.min()) / (ingreso.max() - ingreso.min())

print("Edad escalada:", edad_escala)
print("Ingreso escalado:", ingreso_escala)

# Normalización (media 0, desviación estándar 1)
edad_norm = (edad - edad.mean()) / edad.std()
ingreso_norm = (ingreso - ingreso.mean()) / ingreso.std()

print("Edad normalizada:", edad_norm)
print("Ingreso normalizado:", ingreso_norm)

# Estandarización (media 0, desviación estándar 1)
edad_estand = (edad - edad.mean()) / edad.std()
ingreso_estand = (ingreso - ingreso.mean()) / ingreso.std()

print("Edad estandarizada:", edad_estand)
print("Ingreso estandarizado:", ingreso_estand)
````

**Explicación:**

En los ejemplos anteriores:

- Realizamos el escalamiento dividiendo cada valor por la diferencia entre el valor máximo y el valor mínimo.
- Para la normalización y estandarización, restamos la media y dividimos por la desviación estándar para centrar los datos alrededor de 0 y ajustar su escala.
- Los resultados muestran las variables escaladas, normalizadas y estandarizadas respectivamente.

**Importancia:**

Escalamiento, normalización y estandarización son importantes para garantizar que los algoritmos de aprendizaje automático funcionen correctamente y para evitar que las variables con rangos muy diferentes tengan un peso desproporcionado en los modelos.

En resumen, el escalamiento, la normalización y la estandarización son procesos cruciales para preprocesar datos en ciencia de datos. Estos métodos aseguran que las variables estén en una escala adecuada para el análisis y los algoritmos de aprendizaje automático.

**Código para Crear las Gráficas:**

````python
import numpy as np
import matplotlib.pyplot as plt

# Crear un conjunto de datos de ejemplo
edad = np.array([25, 30, 35, 40, 45])
ingreso = np.array([50000, 75000, 100000, 85000, 90000])

# Escalamiento (entre 0 y 1)
edad_escala = (edad - edad.min()) / (edad.max() - edad.min())
ingreso_escala = (ingreso - ingreso.min()) / (ingreso.max() - ingreso.min())

# Normalización (media 0, desviación estándar 1)
edad_norm = (edad - edad.mean()) / edad.std()
ingreso_norm = (ingreso - ingreso.mean()) / ingreso.std()

# Estandarización (media 0, desviación estándar 1)
edad_estand = (edad - edad.mean()) / edad.std()
ingreso_estand = (ingreso - ingreso.mean()) / ingreso.std()

# Crear subplots
fig, axs = plt.subplots(3, 2, figsize=(10, 12))

# Gráficas para Escalamiento
axs[0, 0].plot(edad, ingreso, 'bo', label='Datos originales')
axs[0, 0].set_title('Datos Originales')
axs[0, 1].plot(edad_escala, ingreso_escala, 'ro', label='Datos escalados')
axs[0, 1].set_title('Escalamiento')

# Gráficas para Normalización
axs[1, 0].plot(edad, ingreso, 'bo', label='Datos originales')
axs[1, 0].set_title('Datos Originales')
axs[1, 1].plot(edad_norm, ingreso_norm, 'go', label='Datos normalizados')
axs[1, 1].set_title('Normalización')

# Gráficas para Estandarización
axs[2, 0].plot(edad, ingreso, 'bo', label='Datos originales')
axs[2, 0].set_title('Datos Originales')
axs[2, 1].plot(edad_estand, ingreso_estand, 'co', label='Datos estandarizados')
axs[2, 1].set_title('Estandarización')

# Agregar leyendas
for ax in axs.flat:
    ax.legend()

# Ajustar espacios entre subplots
plt.tight_layout()

# Mostrar las gráficas
plt.show()
````

**Explicación:**

Este código crea una figura con subplots que representan los datos originales junto con los datos escalados, normalizados y estandarizados en diferentes subplots. Utilizo puntos de colores para representar los datos en las gráficas.

Cada conjunto de subplots muestra:
- Datos originales en la izquierda.
- Datos escalados, normalizados y estandarizados en la derecha.

Las gráficas te permitirán visualizar cómo cambian los datos después de aplicar los procesos de escalamiento, normalización y estandarización.

## **Distribución de los datos**

Por supuesto, entender la distribución de los datos es fundamental en ciencia de datos para comprender cómo están distribuidos los valores en un conjunto de datos. La visualización de la distribución te ayuda a tomar decisiones informadas sobre qué métodos de análisis y modelos son apropiados para tus datos. Usaremos la biblioteca Matplotlib para crear visualizaciones de la distribución de datos.

**Código para Visualizar la Distribución de Datos:**

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Crear un conjunto de datos de ejemplo
datos = np.random.normal(0, 1, size=1000)  # Datos normalmente distribuidos

# Histograma de la distribución
plt.figure(figsize=(8, 6))
plt.hist(datos, bins=30, color='blue', alpha=0.7)
plt.title('Histograma de Distribución')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico de densidad
plt.figure(figsize=(8, 6))
sns.kdeplot(datos, color='green', shade=True)
plt.title('Gráfico de Densidad de Distribución')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.show()

# Gráfico de caja (boxplot)
plt.figure(figsize=(8, 6))
sns.boxplot(y=datos, color='purple')
plt.title('Gráfico de Caja (Boxplot)')
plt.ylabel('Valor')
plt.show()
````

**Explicación:**

- Generamos un conjunto de datos de ejemplo utilizando la función `np.random.normal` para crear datos distribuidos normalmente con media 0 y desviación estándar 1.
- Creamos tres tipos de gráficos para visualizar la distribución de los datos:
  1. **Histograma:** Un gráfico que muestra la frecuencia de los valores en intervalos (bins).
  2. **Gráfico de Densidad:** Un gráfico suave que estima la función de densidad de probabilidad.
  3. **Gráfico de Caja (Boxplot):** Un gráfico que muestra la mediana, los cuartiles y los valores atípicos.

**Resultado:**

Al ejecutar el código, se mostrarán tres gráficos que te ayudarán a entender la distribución de los datos generados. En el histograma, podrás observar cómo están agrupados los valores en intervalos. El gráfico de densidad muestra una representación suavizada de la distribución. El gráfico de caja proporciona información sobre la dispersión y presencia de valores atípicos.

Estos gráficos te ayudarán a identificar características importantes de la distribución de tus datos, como su forma, dispersión y posibles valores atípicos. Esto a su vez te ayudará a tomar decisiones informadas sobre cómo manipular y analizar los datos en tu proyecto de ciencia de datos.

## **Escalamiento**

**Escalamiento en NumPy:**

El escalamiento es un proceso de transformación en el cual se ajusta el rango de valores de una variable para que estén dentro de un rango específico. En el contexto de la ciencia de datos y el aprendizaje automático, el escalamiento es importante para garantizar que las variables tengan valores comparables y se encuentren en una escala adecuada antes de aplicar algoritmos.

El escalamiento es especialmente útil cuando se trabaja con algoritmos sensibles a la escala, como las máquinas de soporte vectorial (SVM) y los métodos de descenso de gradiente.

**Ejemplo de Escalamiento en NumPy:**

Supongamos que tenemos un conjunto de datos que representa el tiempo en segundos que tarda un grupo de personas en completar una tarea:

````python
import numpy as np

# Crear un array de ejemplo con los tiempos en segundos
tiempos = np.array([35, 50, 20, 60, 45, 30])

# Escalar los tiempos entre 0 y 1
tiempos_escala = (tiempos - tiempos.min()) / (tiempos.max() - tiempos.min())

print("Tiempos originales:", tiempos)
print("Tiempos escalados:", tiempos_escala)
````

**Explicación:**

En este ejemplo, hemos creado un array `tiempos` que representa los tiempos en segundos que tarda un grupo de personas en completar una tarea. Luego, hemos aplicado el proceso de escalamiento para ajustar los valores de `tiempos` entre 0 y 1.

El proceso de escalamiento se realiza mediante la fórmula:

$\text{Valor Escalado} = \frac{\text{Valor Original} - \text{Valor Mínimo}}{\text{Valor Máximo} - \text{Valor Mínimo}}$

En este caso, estamos ajustando los tiempos para que estén en la escala entre 0 y 1. Esto puede ser útil cuando queremos que los valores de diferentes magnitudes estén en la misma escala para el análisis o la modelización.

**Importancia del Escalamiento:**

El escalamiento es importante para garantizar que las variables tengan un impacto similar en los algoritmos de aprendizaje automático. Sin escalamiento, las variables con rangos de valores diferentes pueden influir de manera desproporcionada en el resultado de los modelos. Al escalar las variables, aseguramos que todas tengan un rango similar, lo que puede conducir a un mejor rendimiento del modelo y resultados más confiables.

## **Normalización**

**Normalización en NumPy:**

La normalización es un proceso en el que se ajustan los valores de una variable para que tengan una media de 0 y una desviación estándar de 1. Este proceso es útil para transformar variables en una escala común, lo que puede ser beneficioso para algunos algoritmos de aprendizaje automático que funcionan mejor cuando las variables tienen valores comparables y una distribución cercana a una distribución normal.

La normalización también es útil cuando las variables tienen unidades diferentes y queremos tratarlas de manera uniforme en el análisis.

**Ejemplo de Normalización en NumPy:**

Supongamos que tenemos un conjunto de datos que representa los puntajes de dos exámenes para un grupo de estudiantes:

````python
import numpy as np

# Crear un array de ejemplo con los puntajes de los exámenes
examen1 = np.array([85, 90, 78, 88, 92])
examen2 = np.array([70, 75, 68, 82, 78])

# Normalizar los puntajes de los exámenes
examen1_norm = (examen1 - np.mean(examen1)) / np.std(examen1)
examen2_norm = (examen2 - np.mean(examen2)) / np.std(examen2)

print("Puntajes originales de Examen 1:", examen1)
print("Puntajes originales de Examen 2:", examen2)
print("Puntajes normalizados de Examen 1:", examen1_norm)
print("Puntajes normalizados de Examen 2:", examen2_norm)
````

**Explicación:**

En este ejemplo, tenemos dos arrays `examen1` y `examen2`, que representan los puntajes de dos exámenes diferentes para un grupo de estudiantes. Hemos calculado los puntajes normalizados utilizando la fórmula:

$\text{Puntaje Normalizado} = \frac{\text{Puntaje Original} - \text{Media}}{\text{Desviación Estándar}}$

Aplicamos esta fórmula a cada uno de los arrays de puntajes para normalizarlos.

**Importancia de la Normalización:**

La normalización es importante porque puede ayudar a mejorar el rendimiento de los algoritmos de aprendizaje automático que son sensibles a la escala de las variables. Al normalizar las variables, estamos ajustando sus valores para que tengan una distribución similar a una distribución normal con media 0 y desviación estándar 1. Esto puede resultar en un mejor rendimiento del modelo y en la interpretación más confiable de los resultados.

## **Estandarización**

**Estandarización en NumPy:**

La estandarización es un proceso en el cual los valores de una variable son ajustados para tener una media de 0 y una desviación estándar de 1. A diferencia de la normalización, la estandarización no necesariamente restringe los valores dentro de un rango específico, pero sí los transforma para que tengan una distribución similar a una distribución normal con media 0 y desviación estándar 1.

La estandarización es particularmente útil cuando se trabaja con algoritmos que asumen que los datos están distribuidos normalmente o cuando se quieren tratar diferentes variables en una escala uniforme.

**Ejemplo de Estandarización en NumPy:**

Supongamos que tenemos un conjunto de datos que representa las alturas en centímetros de un grupo de personas:

````python
import numpy as np

# Crear un array de ejemplo con las alturas en centímetros
alturas = np.array([160, 175, 168, 155, 182])

# Estandarizar las alturas
alturas_estand = (alturas - np.mean(alturas)) / np.std(alturas)

print("Alturas originales:", alturas)
print("Alturas estandarizadas:", alturas_estand)
````

**Explicación:**

En este ejemplo, hemos creado un array `alturas` que representa las alturas en centímetros de un grupo de personas. Luego, hemos aplicado el proceso de estandarización utilizando la fórmula:

$\text{Altura Estandarizada} = \frac{\text{Altura Original} - \text{Media}}{\text{Desviación Estándar}}$

Calculamos esta fórmula para cada valor en el array `alturas` para estandarizar los valores.

**Importancia de la Estandarización:**

La estandarización es importante porque puede mejorar el rendimiento de los algoritmos de aprendizaje automático que asumen una distribución normal de los datos. También es útil cuando queremos tratar diferentes variables en una escala uniforme, ya que al estandarizarlas, aseguramos que todas tengan una media de 0 y una desviación estándar de 1.

La estandarización puede ser especialmente útil en algoritmos sensibles a la escala, como el descenso de gradiente, la regresión logística y las máquinas de soporte vectorial (SVM). Al estandarizar las variables, mejoramos la convergencia del algoritmo y facilitamos la interpretación de los coeficientes o pesos asociados a cada variable.

## **Comparación de los métodos**

¡Por supuesto! Vamos a comparar los métodos de Escalamiento, Normalización y Estandarización utilizando NumPy. Además, crearé gráficas para visualizar cómo se ven los datos antes y después de cada método.

**Escalamiento, Normalización y Estandarización: Comparación con Ejemplos y Gráficas**

Supongamos que tenemos un conjunto de datos que representa el rendimiento de estudiantes en dos exámenes: "Examen 1" y "Examen 2".

````python
import numpy as np
import matplotlib.pyplot as plt

# Crear un conjunto de datos de ejemplo
examen1 = np.array([85, 90, 78, 88, 92])
examen2 = np.array([70, 75, 68, 82, 78])

# Escalamiento
examen1_escala = (examen1 - examen1.min()) / (examen1.max() - examen1.min())
examen2_escala = (examen2 - examen2.min()) / (examen2.max() - examen2.min())

# Normalización
examen1_norm = (examen1 - np.mean(examen1)) / np.std(examen1)
examen2_norm = (examen2 - np.mean(examen2)) / np.std(examen2)

# Estandarización
examen1_estand = (examen1 - np.mean(examen1)) / np.std(examen1)
examen2_estand = (examen2 - np.mean(examen2)) / np.std(examen2)

# Crear subplots
fig, axs = plt.subplots(3, 2, figsize=(10, 12))

# Gráficas para Escalamiento
axs[0, 0].plot(examen1, examen2, 'bo', label='Datos originales')
axs[0, 0].set_title('Datos Originales')
axs[0, 1].plot(examen1_escala, examen2_escala, 'ro', label='Datos escalados')
axs[0, 1].set_title('Escalamiento')

# Gráficas para Normalización
axs[1, 0].plot(examen1, examen2, 'bo', label='Datos originales')
axs[1, 0].set_title('Datos Originales')
axs[1, 1].plot(examen1_norm, examen2_norm, 'go', label='Datos normalizados')
axs[1, 1].set_title('Normalización')

# Gráficas para Estandarización
axs[2, 0].plot(examen1, examen2, 'bo', label='Datos originales')
axs[2, 0].set_title('Datos Originales')
axs[2, 1].plot(examen1_estand, examen2_estand, 'co', label='Datos estandarizados')
axs[2, 1].set_title('Estandarización')

# Agregar leyendas
for ax in axs.flat:
    ax.legend()

# Ajustar espacios entre subplots
plt.tight_layout()

# Mostrar las gráficas
plt.show()
````

**Explicación:**

Este código compara los tres métodos: escalamiento, normalización y estandarización, en el contexto de un conjunto de datos con dos variables: "Examen 1" y "Examen 2". Utilizamos puntos de colores para representar los datos en las gráficas.

Cada conjunto de subplots muestra:
- Datos originales en la izquierda.
- Datos transformados en la derecha utilizando cada uno de los métodos.

Las gráficas te permitirán visualizar cómo cambian los datos después de aplicar cada método.

**Resultado:**

Al ejecutar el código, se mostrarán tres conjuntos de gráficas, cada uno representando un método diferente de transformación de datos. Compararás cómo se ven los datos antes y después de aplicar cada método, lo que te ayudará a comprender mejor cómo afectan los métodos a la escala y distribución de los datos.
