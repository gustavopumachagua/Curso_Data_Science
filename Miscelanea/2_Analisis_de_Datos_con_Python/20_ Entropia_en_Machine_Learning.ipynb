# **20. Entendiendo la Entrop√≠a en Machine Learning y Ciencia de Datos con Python**

## **Introducci√≥n**

La entrop√≠a es un concepto fundamental en el campo de la teor√≠a de la informaci√≥n y se utiliza en machine learning y ciencia de datos para medir la impureza o incertidumbre en un conjunto de datos. En el contexto de la toma de decisiones, la entrop√≠a se utiliza para cuantificar la falta de homogeneidad en un conjunto de datos y es com√∫nmente empleada en algoritmos de clasificaci√≥n y √°rboles de decisi√≥n para determinar c√≥mo dividir los datos en subconjuntos m√°s puros.

La entrop√≠a se calcula a trav√©s de la f√≥rmula:

$Entrop√≠a(S) = - p_1 \cdot log_2(p_1) - p_2 \cdot log_2(p_2) - \ldots - p_n \cdot log_2(p_n)$

donde ( $p_1$, $p_2$, $\ldots$, $p_n$) son las proporciones de las clases diferentes en el conjunto de datos.

Un valor de entrop√≠a cercano a cero indica que el conjunto de datos est√° altamente ordenado y homog√©neo en t√©rminos de clases, mientras que un valor de entrop√≠a cercano a 1 indica que el conjunto de datos es m√°s diverso y desordenado.

Ahora veamos un ejemplo con datos ficticios y graficamos la entrop√≠a para una comprensi√≥n m√°s clara:



```python
import numpy as np
import matplotlib.pyplot as plt

# Funci√≥n para calcular la entrop√≠a
def entropy(p):
    return -p * np.log2(p)

# Valores de proporciones p
proportions = np.linspace(0.01, 0.99, 100)

# Calcular entrop√≠a para cada proporci√≥n
entropies = [entropy(p) for p in proportions]

# Graficar
plt.figure(figsize=(8, 6))
plt.plot(proportions, entropies, label='Entrop√≠a', color='blue')
plt.title('Gr√°fico de Entrop√≠a')
plt.xlabel('Proporci√≥n de Clase')
plt.ylabel('Entrop√≠a')
plt.legend()
plt.show()
```

En este ejemplo, estamos calculando y graficando la entrop√≠a en funci√≥n de las proporciones de clase. Notar√°s que a medida que la proporci√≥n se acerca a 0 o 1, la entrop√≠a tiende a 0, lo que significa que la impureza disminuye y el conjunto de datos se vuelve m√°s homog√©neo. Por otro lado, cuando la proporci√≥n est√° cerca de 0.5, la entrop√≠a alcanza su valor m√°ximo, lo que indica que hay m√°s incertidumbre y el conjunto de datos es menos homog√©neo.

En resumen, la entrop√≠a es una medida crucial para evaluar la pureza de los datos y es ampliamente utilizada en algoritmos de machine learning y toma de decisiones para encontrar las divisiones m√°s informativas en los datos.

## **Explicaci√≥n conceptual**

La entrop√≠a es un concepto tomado de la teor√≠a de la informaci√≥n y se utiliza en el √°mbito de la ciencia de datos y el machine learning para medir la incertidumbre o la impureza en un conjunto de datos. En otras palabras, la entrop√≠a nos proporciona una medida de cu√°nto caos o desorden hay en los datos en t√©rminos de la distribuci√≥n de las clases.

En el contexto de la toma de decisiones, como en los algoritmos de clasificaci√≥n y √°rboles de decisi√≥n, la entrop√≠a se usa para decidir c√≥mo dividir los datos en subconjuntos m√°s homog√©neos o puros. La idea principal es que deseamos dividir los datos de manera que cada subconjunto resultante contenga principalmente una sola clase, lo que simplifica la tarea de clasificaci√≥n.

La entrop√≠a se calcula a partir de las proporciones de las diferentes clases presentes en el conjunto de datos. Si todas las instancias en el conjunto de datos pertenecen a una sola clase, la entrop√≠a es 0, lo que significa que no hay incertidumbre. Por otro lado, si las instancias se distribuyen uniformemente entre todas las clases, la entrop√≠a es m√°xima, indicando una mayor incertidumbre.

Para calcular la entrop√≠a, se utiliza la f√≥rmula que involucra los logaritmos de las proporciones de clases. A medida que las proporciones de clases se acercan a 0 o 1, los t√©rminos en la f√≥rmula tienden a cero, lo que reduce la entrop√≠a, mientras que una distribuci√≥n uniforme con proporciones cercanas a 0.5 resulta en una entrop√≠a m√°s alta.

En resumen, la entrop√≠a es una m√©trica valiosa para evaluar la pureza o impureza de un conjunto de datos en t√©rminos de distribuci√≥n de clases. Ayuda a los algoritmos de machine learning a tomar decisiones informadas sobre c√≥mo dividir los datos para lograr una clasificaci√≥n m√°s precisa y eficiente.

## **Definici√≥n formal y matem√°tica**

La entrop√≠a en el contexto de la teor√≠a de la informaci√≥n y la ciencia de datos se define formalmente y matem√°ticamente de la siguiente manera:

Dado un conjunto de datos ($S$) que contiene ($n$) ejemplos de entrenamiento pertenecientes a diferentes clases ($C_1$, $C_2$, $\ldots$, $C_k$), donde ($p_i$) es la proporci√≥n de ejemplos en ($S$) que pertenecen a la clase ($C_i$), la entrop√≠a ($H(S)$) se calcula utilizando la siguiente f√≥rmula:

$H(S) = -p_1 \cdot \log_2(p_1) - p_2 \cdot \log_2(p_2) - \ldots - p_k \cdot \log_2(p_k)$

Donde:

- ($p_i$) es la proporci√≥n de ejemplos en ($S$) que pertenecen a la clase ($C_i$).

- ($k$) es el n√∫mero total de clases en el conjunto de datos.

- ($\log_2$) es el logaritmo en base 2.

La entrop√≠a se mide en bits (o unidades de informaci√≥n) y proporciona una medida de la incertidumbre en la distribuci√≥n de clases en el conjunto de datos. Si la entrop√≠a es baja, significa que el conjunto de datos est√° altamente ordenado y contiene principalmente ejemplos de una sola clase. Si la entrop√≠a es alta, el conjunto de datos est√° desordenado y los ejemplos est√°n distribuidos entre m√∫ltiples clases de manera m√°s uniforme.

Un conjunto de datos perfectamente puro, donde todos los ejemplos pertenecen a una sola clase, tendr√≠a una entrop√≠a de 0. Por otro lado, si el conjunto de datos est√° distribuido uniformemente entre todas las clases, la entrop√≠a ser√° m√°xima y alcanzar√° ($\log_2(k)$), donde ($k$) es el n√∫mero de clases.

En resumen, la entrop√≠a proporciona una medida cuantitativa de la incertidumbre o impureza en la distribuci√≥n de clases de un conjunto de datos y es fundamental en la toma de decisiones en algoritmos de clasificaci√≥n y √°rboles de decisi√≥n.

## **C√≥mo se mide la cantidad de informaci√≥n en bits**

La cantidad de informaci√≥n se mide en bits, que es una unidad b√°sica de la teor√≠a de la informaci√≥n. Los bits representan la cantidad de elecciones binarias (0 o 1) necesarias para transmitir un mensaje o representar un evento. La idea es que, si hay dos opciones igualmente probables, se necesita un bit para distinguirlas; si hay cuatro opciones, se necesitan dos bits, y as√≠ sucesivamente.

Matem√°ticamente, la cantidad de informaci√≥n ($I$) en bits asociada a un evento con probabilidad ($P$) se calcula utilizando la f√≥rmula:

$I = -\log_2(P)$

Donde:

- ( $\log_2$) es el logaritmo en base 2.

- ( $P$) es la probabilidad del evento.

Esta f√≥rmula significa que cuanto menor sea la probabilidad del evento, mayor ser√° la cantidad de informaci√≥n transmitida al observar ese evento. En otras palabras, eventos raros llevan m√°s informaci√≥n porque son inesperados.

Por ejemplo, si tienes una moneda justa (probabilidad de cara = 0.5, probabilidad de cruz = 0.5), el resultado de lanzarla tiene una cantidad de informaci√≥n de:

$I = -\log_2(0.5) = 1 \text{ bit}$

Esto significa que lanzar una moneda justa nos brinda 1 bit de informaci√≥n. Si tuvieras una moneda cargada en la que una cara es tres veces m√°s probable que la cruz (probabilidad de cara = 0.75, probabilidad de cruz = 0.25), la cantidad de informaci√≥n ser√≠a:

$I = -\log_2(0.75) \approx 0.415 \text{ bits}$

En resumen, la cantidad de informaci√≥n en bits mide cu√°nta sorpresa o incertidumbre hay en un evento o mensaje, y se calcula utilizando la f√≥rmula ($I = -\log_2(P)$), donde ($P$) es la probabilidad del evento. Cuanto menos probable sea el evento, m√°s informaci√≥n llevar√°.

## **Ejemplo de Entrop√≠a: dado (in)justo**

Supongamos que tenemos un dado de seis caras y queremos calcular la entrop√≠a de los resultados al lanzarlo. Si el dado es justo, todas las caras tienen la misma probabilidad de aparecer, que es ($1/6$) para cada cara.

La entrop√≠a ($H$) se calcula utilizando la f√≥rmula:

$H(S) = -p_1 \cdot \log_2(p_1) - p_2 \cdot \log_2(p_2) - \ldots - p_k \cdot \log_2(p_k)$

En este caso, ($k = 6$) (n√∫mero de caras del dado) y ($p_i = 1/6$) para ($i = 1$, $2$, $\ldots$, $6$).

Calculamos la entrop√≠a:

$H(S) = - \left( \frac{1}{6} \cdot \log_2\left(\frac{1}{6}\right) \right) \times 6 = - \log_2\left(\frac{1}{6}\right) \approx 2.585 \text{ bits}$

Esto significa que, en promedio, lanzar un dado justo de seis caras nos proporciona alrededor de ($2.585$) bits de informaci√≥n en t√©rminos de qu√© cara saldr√°. Dado que todas las caras son igualmente probables, no hay ninguna cara que sea m√°s predecible que otra, y la entrop√≠a es relativamente alta.

Ahora, si tuvi√©ramos un dado cargado de manera injusta, donde una cara tiene una probabilidad ($p$) mayor de aparecer y las otras caras tienen igual probabilidad ($q = (1-p)/5$), podr√≠amos calcular la entrop√≠a de manera similar para ver c√≥mo cambia con la distribuci√≥n de probabilidad desigual.

## **Entrop√≠a con Python**

```python
import numpy as np

# Probabilidad de cada cara en un dado justo de seis caras
probabilidad_cara = 1/6

# Calcular la entrop√≠a
entropia = -probabilidad_cara * np.log2(probabilidad_cara) * 6

print("Entrop√≠a del dado justo de seis caras:", entropia, "bits")
```

En este ejemplo, estamos utilizando la f√≥rmula $H(S) = -p_i \cdot \log_2(p_i)$  para cada cara del dado, y multiplicamos por ( $6$) para considerar todas las caras.

Si deseas calcular la entrop√≠a para un dado injusto con distribuciones de probabilidad diferentes, simplemente cambia el valor de `probabilidad_cara` en el c√≥digo anterior y realiza los c√°lculos correspondientes.
