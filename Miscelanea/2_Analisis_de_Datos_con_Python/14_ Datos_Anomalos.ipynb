# **14. Detección de Datos Anómalos (outliers) con Bosques de Aislamiento (iForests) y Python**

## **Qué son los datos anómalos**

La detección de datos anómalos es una tarea importante en el análisis de datos, donde buscamos identificar puntos que se desvían significativamente del comportamiento general de los datos. Los outliers pueden ser indicativos de errores en la recopilación de datos, eventos raros o incluso problemas en los sistemas. Los Bosques de Aislamiento (Isolation Forests o iForests) son un método de detección de outliers que se basa en el concepto de que los outliers son más fáciles de aislar que los datos normales.

Voy a guiarte a través de una explicación detallada junto con un ejemplo en Python usando la biblioteca scikit-learn y gráficos para ilustrar el proceso.

**Instalación:**
Asegúrate de tener la biblioteca scikit-learn instalada:
```
pip install scikit-learn
```

**Detección de Outliers con iForests: Ejemplo y Gráficos**

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Crear datos de ejemplo
np.random.seed(42)
datos_normales = np.random.normal(0, 1, size=1000)
outliers = np.array([10, -10])  # Introducir outliers

# Combinar datos normales y outliers
datos_totales = np.concatenate((datos_normales, outliers), axis=0)
datos_totales = datos_totales.reshape(-1, 1)

# Crear y entrenar el modelo de iForests
modelo_iforest = IsolationForest(contamination=0.02, random_state=42)
modelo_iforest.fit(datos_totales)

# Predecir si los datos son outliers o no
predicciones = modelo_iforest.predict(datos_totales)

# Visualizar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(range(len(datos_totales)), datos_totales, c=predicciones, cmap='coolwarm')
plt.xlabel('Índice de Datos')
plt.ylabel('Valor')
plt.title('Detección de Outliers con iForests')
plt.colorbar(label='Predicción de Outlier')
plt.show()
````

**Explicación:**

En este ejemplo, estamos generando datos normales con una distribución normal y luego introducimos deliberadamente dos valores anómalos (outliers) en la lista `outliers`. Luego, combinamos los datos normales y los outliers para crear el conjunto de datos total.

Luego creamos y entrenamos un modelo iForests utilizando `IsolationForest` de scikit-learn. El parámetro `contamination` establece la proporción esperada de outliers en los datos. En este caso, hemos establecido el valor en 0.02 para indicar que esperamos que alrededor del 2% de los datos sean outliers.

Finalmente, usamos las predicciones del modelo para colorear los puntos en un gráfico, donde los puntos azules representan datos normales y los puntos rojos representan outliers.

**Resultado:**

Al ejecutar el código, verás un gráfico que muestra cómo iForests ha identificado los outliers en el conjunto de datos. Los puntos rojos en el gráfico representan los valores que el modelo considera como outliers, mientras que los puntos azules son los datos considerados como normales.

iForests es eficaz para detectar outliers, especialmente en conjuntos de datos multidimensionales. Puedes ajustar el parámetro `contamination` según tus necesidades y explorar más sobre cómo iForests puede ser útil en la detección de datos anómalos.

## **Explicación de iForests**

**Bosques de Aislamiento (Isolation Forests - iForests):**

Los Bosques de Aislamiento (iForests) son una técnica eficiente y efectiva para la detección de outliers en conjuntos de datos. Esta técnica se basa en el principio de que los outliers son más fáciles de aislar que los datos normales. iForests logra esto construyendo múltiples árboles de decisión de manera aleatoria y midiendo cuántas divisiones se necesitan para aislar un punto de datos específico. Los puntos que se pueden aislar con menos divisiones se consideran más anómalos.

**Funcionamiento de iForests:**

1. Selecciona aleatoriamente un atributo y un valor en el conjunto de datos.
2. Divide los datos en dos subconjuntos según el valor seleccionado.
3. Repite el proceso de división hasta que los datos queden completamente aislados o hasta que se alcance una altura máxima predefinida.
4. Cuanto menos divisiones se necesiten para aislar un dato, más anómalo se considera.

**Ejemplo de iForests con Gráficas:**

Para ilustrar cómo funciona iForests, generaremos datos aleatorios y los combinaremos con valores anómalos para crear un conjunto de datos mixto. Luego, construiremos el modelo de iForests y visualizaremos cómo clasifica los datos normales y los outliers.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Crear datos de ejemplo
np.random.seed(42)
datos_normales = np.random.normal(0, 1, size=300)
outliers = np.array([10, -10])  # Introducir outliers
datos_totales = np.concatenate((datos_normales, outliers), axis=0)
datos_totales = datos_totales.reshape(-1, 1)

# Crear y entrenar el modelo de iForests
modelo_iforest = IsolationForest(contamination=0.02, random_state=42)
modelo_iforest.fit(datos_totales)

# Predecir si los datos son outliers o no
predicciones = modelo_iforest.predict(datos_totales)

# Visualizar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(range(len(datos_totales)), datos_totales, c=predicciones, cmap='coolwarm')
plt.xlabel('Índice de Datos')
plt.ylabel('Valor')
plt.title('Detección de Outliers con iForests')
plt.colorbar(label='Predicción de Outlier')
plt.show()
````

**Explicación y Resultado:**

En este ejemplo, generamos un conjunto de datos que consiste en datos normales y dos valores anómalos (`outliers`). Usamos la biblioteca scikit-learn para construir y entrenar un modelo de iForests. Luego, predecimos si los datos son outliers o no utilizando el modelo entrenado.

El gráfico resultante muestra cómo iForests ha etiquetado los puntos como normales (azul) o outliers (rojo). Los puntos rojos son los que iForests ha identificado como outliers, mientras que los puntos azules son considerados como datos normales.

La efectividad de iForests en la detección de outliers es evidente en el gráfico, donde los outliers están claramente separados de los datos normales. iForests es particularmente útil para detectar valores anómalos en conjuntos de datos grandes y complejos, donde puede ser difícil aplicar enfoques tradicionales.

## **iForests con Scikit-Learn**

**Bosques de Aislamiento (Isolation Forests - iForests) con Scikit-Learn:**

**Ejemplo de iForests con Scikit-Learn y Gráficas:**

Vamos a generar un ejemplo de detección de outliers utilizando iForests con scikit-learn. Generaremos datos aleatorios y luego agregaremos valores anómalos para demostrar cómo iForests puede identificarlos.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Crear datos de ejemplo
np.random.seed(42)
datos_normales = np.random.normal(0, 1, size=300)
outliers = np.array([10, -10])  # Introducir outliers
datos_totales = np.concatenate((datos_normales, outliers), axis=0)
datos_totales = datos_totales.reshape(-1, 1)

# Crear y entrenar el modelo de iForests
modelo_iforest = IsolationForest(contamination=0.02, random_state=42)
modelo_iforest.fit(datos_totales)

# Predecir si los datos son outliers o no
predicciones = modelo_iforest.predict(datos_totales)

# Visualizar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(range(len(datos_totales)), datos_totales, c=predicciones, cmap='coolwarm')
plt.xlabel('Índice de Datos')
plt.ylabel('Valor')
plt.title('Detección de Outliers con iForests (Scikit-Learn)')
plt.colorbar(label='Predicción de Outlier')
plt.show()
````

7. Visualizamos los resultados en un gráfico utilizando `matplotlib.pyplot`. Los puntos rojos representan outliers, mientras que los puntos azules representan datos normales.

**Resultado:**

El gráfico muestra cómo iForests ha clasificado los puntos como outliers o datos normales. Los puntos rojos son los que el modelo considera outliers, mientras que los puntos azules son datos normales. iForests es efectivo para detectar valores anómalos y proporciona una herramienta útil para la detección de outliers en conjuntos de datos complejos.
