# **14. Detecci贸n de Datos An贸malos (outliers) con Bosques de Aislamiento (iForests) y Python**

## **Qu茅 son los datos an贸malos**

La detecci贸n de datos an贸malos es una tarea importante en el an谩lisis de datos, donde buscamos identificar puntos que se desv铆an significativamente del comportamiento general de los datos. Los outliers pueden ser indicativos de errores en la recopilaci贸n de datos, eventos raros o incluso problemas en los sistemas. Los Bosques de Aislamiento (Isolation Forests o iForests) son un m茅todo de detecci贸n de outliers que se basa en el concepto de que los outliers son m谩s f谩ciles de aislar que los datos normales.

Voy a guiarte a trav茅s de una explicaci贸n detallada junto con un ejemplo en Python usando la biblioteca scikit-learn y gr谩ficos para ilustrar el proceso.

**Detecci贸n de Outliers con iForests: Ejemplo y Gr谩ficos**

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Crear datos de ejemplo
np.random.seed(42)
datos_normales = np.random.normal(0, 1, size=1000)
outliers = np.array([10, -10])  # Introducir outliers

# Combinar datos normales y outliers
datos_totales = np.concatenate((datos_normales, outliers), axis=0)
datos_totales = datos_totales.reshape(-1, 1)

# Crear y entrenar el modelo de iForests
modelo_iforest = IsolationForest(contamination=0.02, random_state=42)
modelo_iforest.fit(datos_totales)

# Predecir si los datos son outliers o no
predicciones = modelo_iforest.predict(datos_totales)

# Visualizar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(range(len(datos_totales)), datos_totales, c=predicciones, cmap='coolwarm')
plt.xlabel('ndice de Datos')
plt.ylabel('Valor')
plt.title('Detecci贸n de Outliers con iForests')
plt.colorbar(label='Predicci贸n de Outlier')
plt.show()
````

**Explicaci贸n:**

En este ejemplo, estamos generando datos normales con una distribuci贸n normal y luego introducimos deliberadamente dos valores an贸malos (outliers) en la lista `outliers`. Luego, combinamos los datos normales y los outliers para crear el conjunto de datos total.

Luego creamos y entrenamos un modelo iForests utilizando `IsolationForest` de scikit-learn. El par谩metro `contamination` establece la proporci贸n esperada de outliers en los datos. En este caso, hemos establecido el valor en 0.02 para indicar que esperamos que alrededor del 2% de los datos sean outliers.

Finalmente, usamos las predicciones del modelo para colorear los puntos en un gr谩fico, donde los puntos azules representan datos normales y los puntos rojos representan outliers.

**Resultado:**

Al ejecutar el c贸digo, ver谩s un gr谩fico que muestra c贸mo iForests ha identificado los outliers en el conjunto de datos. Los puntos rojos en el gr谩fico representan los valores que el modelo considera como outliers, mientras que los puntos azules son los datos considerados como normales.

iForests es eficaz para detectar outliers, especialmente en conjuntos de datos multidimensionales. Puedes ajustar el par谩metro `contamination` seg煤n tus necesidades y explorar m谩s sobre c贸mo iForests puede ser 煤til en la detecci贸n de datos an贸malos.

## **Explicaci贸n de iForests**

**Bosques de Aislamiento (Isolation Forests - iForests):**

Los Bosques de Aislamiento (iForests) son una t茅cnica eficiente y efectiva para la detecci贸n de outliers en conjuntos de datos. Esta t茅cnica se basa en el principio de que los outliers son m谩s f谩ciles de aislar que los datos normales. iForests logra esto construyendo m煤ltiples 谩rboles de decisi贸n de manera aleatoria y midiendo cu谩ntas divisiones se necesitan para aislar un punto de datos espec铆fico. Los puntos que se pueden aislar con menos divisiones se consideran m谩s an贸malos.

**Funcionamiento de iForests:**

1. Selecciona aleatoriamente un atributo y un valor en el conjunto de datos.
2. Divide los datos en dos subconjuntos seg煤n el valor seleccionado.
3. Repite el proceso de divisi贸n hasta que los datos queden completamente aislados o hasta que se alcance una altura m谩xima predefinida.
4. Cuanto menos divisiones se necesiten para aislar un dato, m谩s an贸malo se considera.

**Ejemplo de iForests con Gr谩ficas:**

Para ilustrar c贸mo funciona iForests, generaremos datos aleatorios y los combinaremos con valores an贸malos para crear un conjunto de datos mixto. Luego, construiremos el modelo de iForests y visualizaremos c贸mo clasifica los datos normales y los outliers.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Crear datos de ejemplo
np.random.seed(42)
datos_normales = np.random.normal(0, 1, size=300)
outliers = np.array([10, -10])  # Introducir outliers
datos_totales = np.concatenate((datos_normales, outliers), axis=0)
datos_totales = datos_totales.reshape(-1, 1)

# Crear y entrenar el modelo de iForests
modelo_iforest = IsolationForest(contamination=0.02, random_state=42)
modelo_iforest.fit(datos_totales)

# Predecir si los datos son outliers o no
predicciones = modelo_iforest.predict(datos_totales)

# Visualizar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(range(len(datos_totales)), datos_totales, c=predicciones, cmap='coolwarm')
plt.xlabel('ndice de Datos')
plt.ylabel('Valor')
plt.title('Detecci贸n de Outliers con iForests')
plt.colorbar(label='Predicci贸n de Outlier')
plt.show()
````

**Explicaci贸n y Resultado:**

En este ejemplo, generamos un conjunto de datos que consiste en datos normales y dos valores an贸malos (`outliers`). Usamos la biblioteca scikit-learn para construir y entrenar un modelo de iForests. Luego, predecimos si los datos son outliers o no utilizando el modelo entrenado.

El gr谩fico resultante muestra c贸mo iForests ha etiquetado los puntos como normales (azul) o outliers (rojo). Los puntos rojos son los que iForests ha identificado como outliers, mientras que los puntos azules son considerados como datos normales.

La efectividad de iForests en la detecci贸n de outliers es evidente en el gr谩fico, donde los outliers est谩n claramente separados de los datos normales. iForests es particularmente 煤til para detectar valores an贸malos en conjuntos de datos grandes y complejos, donde puede ser dif铆cil aplicar enfoques tradicionales.

**Ejemplo de iForests con Scikit-Learn y Gr谩ficas:**
