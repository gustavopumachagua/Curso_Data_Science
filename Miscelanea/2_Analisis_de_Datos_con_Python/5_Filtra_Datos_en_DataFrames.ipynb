# **5. Filtra Datos en DataFrames de Pandas para un mejor An√°lisis de Datos con Python**

## **Introducci√≥n a los Filtros en Pandas**

¬°Por supuesto! Trabajar con Pandas en Python es una excelente manera de filtrar y analizar datos. Aqu√≠ tienes una explicaci√≥n detallada sobre c√≥mo filtrar datos en DataFrames de Pandas para un mejor an√°lisis, acompa√±ada de ejemplos y visualizaciones gr√°ficas.

**Paso 1: Importar Pandas y cargar los datos**

Primero, necesitamos importar la biblioteca Pandas y cargar los datos en un DataFrame. Supongamos que tenemos un conjunto de datos de ventas mensuales:

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    'Ventas': [1000, 1200, 800, 1500, 1100]
}

df = pd.DataFrame(data)
```

**Paso 2: Filtrar datos seg√∫n condiciones**

Ahora, vamos a aprender c√≥mo filtrar los datos en funci√≥n de ciertas condiciones. Supongamos que queremos analizar los meses en los que las ventas fueron superiores a 1000:

```python
ventas_altas = df[df['Ventas'] > 1000]
```

**Paso 3: Visualizar los datos filtrados**

Una vez que hemos filtrado los datos, podemos visualizarlos para un an√°lisis m√°s claro. Aqu√≠ vamos a usar la biblioteca Matplotlib para crear gr√°ficos simples:

```python
import matplotlib.pyplot as plt

# Crear un gr√°fico de barras para las ventas altas
plt.bar(ventas_altas['Mes'], ventas_altas['Ventas'])
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Ventas Mensuales Superiores a 1000')
plt.show()
```

**Paso 4: Filtrar usando m√∫ltiples condiciones**

Pandas tambi√©n permite filtrar utilizando m√∫ltiples condiciones. Supongamos que queremos analizar los meses en los que las ventas fueron superiores a 1000 y tambi√©n inferiores a 1500:

```python
ventas_interesantes = df[(df['Ventas'] > 1000) & (df['Ventas'] < 1500)]
```

**Paso 5: Visualizar datos filtrados con m√∫ltiples condiciones**

Vamos a visualizar estos datos filtrados nuevamente utilizando un gr√°fico de barras:

```python
plt.bar(ventas_interesantes['Mes'], ventas_interesantes['Ventas'])
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Ventas Mensuales entre 1000 y 1500')
plt.show()
```

Recuerda que estos son ejemplos b√°sicos para ilustrar c√≥mo filtrar y visualizar datos con Pandas y Matplotlib. En situaciones reales, tus conjuntos de datos y an√°lisis pueden ser mucho m√°s complejos, pero estos pasos te proporcionan una base s√≥lida para comenzar a trabajar con tus propios datos.

## **C√≥mo Filtrar Informaci√≥n en DataFrames**

**Paso 1: Importar Pandas y cargar los datos**

Primero, importamos la biblioteca Pandas y creamos un DataFrame con datos de ejemplo:

```python
import pandas as pd

# Crear un DataFrame con datos de ejemplo
data = {
    'Producto': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Ventas': [1000, 1200, 800, 1500, 1100, 900]
}

df = pd.DataFrame(data)
```

**Paso 2: Filtrar datos utilizando condiciones**

Pandas nos permite filtrar datos utilizando condiciones l√≥gicas. Supongamos que queremos filtrar los productos con ventas superiores a 1000:

```python
ventas_altas = df[df['Ventas'] > 1000]
```

**Paso 3: Visualizar datos filtrados**

Ahora, vamos a visualizar los datos filtrados mediante un gr√°fico de barras. Utilizaremos la biblioteca Matplotlib para crear la visualizaci√≥n:

```python
import matplotlib.pyplot as plt

# Crear un gr√°fico de barras para productos con ventas altas
plt.bar(ventas_altas['Producto'], ventas_altas['Ventas'])
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.title('Productos con Ventas Altas')
plt.show()
```

**Paso 4: Filtrar utilizando m√∫ltiples condiciones**

Pandas tambi√©n nos permite combinar m√∫ltiples condiciones para realizar filtrados m√°s complejos. Supongamos que queremos filtrar los productos con ventas superiores a 1000 y que pertenecen a la categor√≠a 'A':

```python
productos_interesantes = df[(df['Ventas'] > 1000) & (df['Producto'] == 'A')]
```

**Paso 5: Visualizar datos filtrados con m√∫ltiples condiciones**

Vamos a visualizar nuevamente estos datos filtrados utilizando un gr√°fico de barras:

```python
plt.bar(productos_interesantes['Producto'], productos_interesantes['Ventas'])
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.title('Productos Interesantes (Ventas > 1000 y Producto A)')
plt.show()
```

**Contexto de los Ejemplos**

En estos ejemplos, estamos trabajando con un DataFrame que contiene informaci√≥n sobre productos y sus ventas. Cada fila representa una venta de un producto espec√≠fico. Los pasos que hemos seguido ilustran c√≥mo filtrar los datos para analizar productos con ciertas condiciones de ventas y c√≥mo visualizar esta informaci√≥n utilizando gr√°ficos de barras.

Es importante recordar que los ejemplos anteriores son simples y se utilizan para ilustrar los conceptos de filtrado y visualizaci√≥n. En situaciones reales, podr√≠as tener conjuntos de datos m√°s grandes y aplicar filtros m√°s complejos para obtener informaci√≥n significativa sobre tus datos.

## **Ejercicios B√°sicos de Filtrado**

**Ejercicio 1: Filtrar datos por una columna num√©rica**

Supongamos que tienes un DataFrame con informaci√≥n sobre estudiantes y sus calificaciones en un curso. Quieres filtrar los estudiantes que han obtenido una calificaci√≥n superior a 90.

```python
import pandas as pd

data = {
    'Estudiante': ['Ana', 'Juan', 'Mar√≠a', 'Carlos'],
    'Calificacion': [85, 92, 88, 95]
}

df = pd.DataFrame(data)

# Filtrar estudiantes con calificaci√≥n superior a 90
calif_altas = df[df['Calificacion'] > 90]
print(calif_altas)
```

**Ejercicio 2: Filtrar datos por una columna categ√≥rica**

Ahora, imagina que tienes un DataFrame con informaci√≥n sobre empleados y sus departamentos. Quieres filtrar los empleados que trabajan en el departamento de "Ventas".

```python
data = {
    'Empleado': ['Elena', 'Pedro', 'Laura', 'Miguel'],
    'Departamento': ['Ventas', 'Finanzas', 'Ventas', 'Recursos Humanos']
}

df = pd.DataFrame(data)

# Filtrar empleados del departamento de "Ventas"
ventas_empleados = df[df['Departamento'] == 'Ventas']
print(ventas_empleados)
```

**Ejercicio 3: Filtrar datos usando m√∫ltiples condiciones**

Supongamos que tienes un DataFrame con informaci√≥n sobre productos y sus precios. Quieres filtrar los productos que tienen un precio superior a 50 y pertenecen a la categor√≠a "Electr√≥nicos".

```python
data = {
    'Producto': ['Laptop', 'TV', 'Smartphone', 'Tablet'],
    'Precio': [800, 600, 1000, 40],
    'Categoria': ['Electr√≥nicos', 'Electr√≥nicos', 'Electr√≥nicos', 'Electr√≥nicos']
}

df = pd.DataFrame(data)

# Filtrar productos con precio superior a 50 y categor√≠a "Electr√≥nicos"
productos_interesantes = df[(df['Precio'] > 50) & (df['Categoria'] == 'Electr√≥nicos')]
print(productos_interesantes)
```

**Ejercicio 4: Filtrar datos por un rango de fechas**

Imagina que tienes un DataFrame con informaci√≥n sobre transacciones financieras y sus fechas. Quieres filtrar las transacciones que ocurrieron en el segundo trimestre del a√±o.

```python
data = {
    'Fecha': ['2023-04-15', '2023-05-20', '2023-06-10', '2023-07-05'],
    'Monto': [150, 200, 300, 50]
}

df = pd.DataFrame(data)
df['Fecha'] = pd.to_datetime(df['Fecha'])

# Filtrar transacciones del segundo trimestre
segundo_trimestre = df[(df['Fecha'].dt.month >= 4) & (df['Fecha'].dt.month <= 6)]
print(segundo_trimestre)
```

Estos ejercicios b√°sicos te ayudar√°n a comprender c√≥mo aplicar diferentes tipos de filtros en DataFrames de Pandas. A medida que ganes experiencia, podr√°s realizar filtrados m√°s complejos y adaptarlos a situaciones de an√°lisis de datos del mundo real.

## **Cargar archivo con datos de 229 pa√≠ses**

¬°Por supuesto! Para cargar un archivo con datos de 229 pa√≠ses en un DataFrame de Pandas, necesitar√°s tener el archivo en un formato compatible, como CSV (Comma-Separated Values) o Excel. Aqu√≠ te proporcionar√© ejemplos para ambos formatos.

**Cargar datos desde un archivo CSV:**

Supongamos que tienes un archivo CSV llamado "paises.csv" que contiene datos de los 229 pa√≠ses. Aseg√∫rate de que el archivo est√© en el mismo directorio que tu script de Python o proporciona la ruta correcta al archivo.

```python
import pandas as pd

# Cargar datos desde un archivo CSV
df = pd.read_csv('paises.csv')

# Mostrar las primeras filas del DataFrame
print(df.head())
```

**Cargar datos desde un archivo Excel:**

## **Ejercicios Avanzados con Condiciones Booleanas**

¬°Por supuesto! Aqu√≠ tienes algunos ejercicios avanzados que involucran el uso de condiciones booleanas para filtrar y analizar datos en DataFrames de Pandas.

**Ejercicio 1: Filtrar pa√≠ses con poblaci√≥n y √°rea grandes**

Supongamos que tienes un DataFrame llamado `paises_df` con informaci√≥n sobre diferentes pa√≠ses, incluyendo su nombre, poblaci√≥n y √°rea en kil√≥metros cuadrados. Quieres filtrar los pa√≠ses que tienen una poblaci√≥n superior a 100 millones y un √°rea mayor a 500,000 kil√≥metros cuadrados.

```python
import pandas as pd

# Crear un DataFrame con datos ficticios sobre pa√≠ses
data = {
    'Pais': ['Estados Unidos', 'China', 'India', 'Brasil', 'Rusia'],
    'Continente': ['Am√©rica', 'Asia', 'Asia', 'Am√©rica', 'Europa'],
    'Poblacion': [331002651, 1444216107, 1380004385, 212559417, 145934462],
    'Area': [9525067, 9596960, 3287263, 8515767, 17098242],
    'Idioma': ['Ingl√©s', 'Chino', 'Hindi', 'Portugu√©s', 'Ruso']
}

paises_df = pd.DataFrame(data)
print(paises_df)

# Supongamos que paises_df ya est√° cargado con los datos de los pa√≠ses

# Filtrar pa√≠ses con poblaci√≥n > 100 millones y √°rea > 500,000 km^2
paises_grandes = paises_df[(paises_df['Poblacion'] > 100_000_000) & (paises_df['Area'] > 500_000)]
print(paises_grandes)
```

**Ejercicio 2: Clasificar pa√≠ses por densidad de poblaci√≥n**

Deseas calcular y clasificar los pa√≠ses en el DataFrame seg√∫n su densidad de poblaci√≥n (poblaci√≥n dividida por √°rea) en orden descendente.

```python
# Calcular densidad de poblaci√≥n y clasificar los pa√≠ses
paises_df['Densidad_Poblacion'] = paises_df['Poblacion'] / paises_df['Area']
paises_ordenados = paises_df.sort_values(by='Densidad_Poblacion', ascending=False)
print(paises_ordenados)
```

**Ejercicio 3: Filtrar pa√≠ses con idiomas espec√≠ficos**

Supongamos que tienes una lista de idiomas populares y quieres filtrar los pa√≠ses que tienen uno de esos idiomas como idioma oficial.

```python
idiomas_populares = ['Ingl√©s', 'Espa√±ol', 'Franc√©s']

# Filtrar pa√≠ses con idiomas populares como idioma oficial
paises_con_idiomas_populares = paises_df[paises_df['Idioma'].isin(idiomas_populares)]
print(paises_con_idiomas_populares)
```

**Ejercicio 4: Filtrar pa√≠ses por continente y caracter√≠sticas**

Si tienes un DataFrame con informaci√≥n detallada sobre los pa√≠ses, puedes realizar filtrados m√°s espec√≠ficos. Supongamos que deseas filtrar los pa√≠ses en el continente "Asia" con poblaci√≥n entre 50 y 100 millones y √°rea entre 200,000 y 500,000 kil√≥metros cuadrados.

```python
# Filtrar pa√≠ses en Asia con poblaci√≥n y √°rea espec√≠ficas
asia_filtros = (paises_df['Continente'] == 'Asia') & \
               (paises_df['Poblacion'].between(50_000_000, 100_000_000)) & \
               (paises_df['Area'].between(200_000, 500_000))

paises_asia_filtrados = paises_df[asia_filtros]
print(paises_asia_filtrados)
```

Recuerda que estos ejercicios son ejemplos simplificados para ilustrar el uso de condiciones booleanas en el filtrado y an√°lisis de datos en DataFrames de Pandas. En situaciones reales, los conjuntos de datos y los an√°lisis pueden ser m√°s complejos, pero estos ejercicios te brindan una base s√≥lida para abordar desaf√≠os m√°s avanzados.
