# **22. Las Mejores Métricas para Evaluar Modelos de Regresión con Scikit-Learn: R2, MSE, RMSE, MAE y otras**

## **Qué es un modelo de regresión**

Un modelo de regresión es una técnica en estadísticas y aprendizaje automático que se utiliza para predecir valores numéricos a partir de variables independientes. En otras palabras, un modelo de regresión busca establecer una relación matemática entre una variable dependiente (o de respuesta) y una o más variables independientes (o predictoras).

Aquí tienes una explicación detallada junto con un ejemplo y gráficas para comprender mejor:

**Tipos de Regresión:**
Existen varios tipos de modelos de regresión, como la regresión lineal, regresión polinómica, regresión logística, regresión ridge, regresión LASSO, etc. En esta explicación, nos centraremos en la regresión lineal simple, que es uno de los tipos más simples y ampliamente utilizados.

**Regresión Lineal Simple:**
La regresión lineal simple modela la relación entre una variable independiente (X) y una variable dependiente (Y) como una línea recta. La ecuación general de una regresión lineal simple es: Y = mX + b, donde "m" es la pendiente de la línea y "b" es la ordenada al origen.

**Ejemplo: Predicción de Puntuaje en un Examen:**
Supongamos que estamos tratando de predecir los puntajes en un examen basados en el número de horas de estudio. Tenemos los siguientes datos:

| Horas de Estudio (X) | Puntuaje en el Examen (Y) |
|----------------------|--------------------------|
| 2                    | 60                       |
| 3                    | 75                       |
| 4                    | 85                       |
| 5                    | 90                       |

**Gráfica de Dispersión:**
Primero, trazamos los puntos en un gráfico de dispersión para visualizar los datos:

```python
import matplotlib.pyplot as plt

horas_estudio = [2, 3, 4, 5]
puntajes_examen = [60, 75, 85, 90]

plt.scatter(horas_estudio, puntajes_examen)
plt.xlabel("Horas de Estudio")
plt.ylabel("Puntuaje en el Examen")
plt.title("Gráfico de Dispersión: Horas de Estudio vs. Puntuaje en el Examen")
plt.show()
```

**Ajuste de la Regresión Lineal:**

Luego, ajustamos una línea recta (modelo de regresión lineal) que mejor se ajuste a los puntos:

```python
from sklearn.linear_model import LinearRegression

X = [[2], [3], [4], [5]]
Y = [60, 75, 85, 90]

modelo = LinearRegression()
modelo.fit(X, Y)

pendiente = modelo.coef_[0]
ordenada_origen = modelo.intercept_
```

**Gráfica de la Regresión Lineal:**

Agregamos la línea de regresión al gráfico de dispersión:

```python
import numpy as np

plt.scatter(horas_estudio, puntajes_examen)
plt.plot(horas_estudio, modelo.predict(X), color='red')  # Línea de regresión
plt.xlabel("Horas de Estudio")
plt.ylabel("Puntuaje en el Examen")
plt.title("Regresión Lineal: Horas de Estudio vs. Puntuaje en el Examen")
plt.show()
```

**Predicción:**

Ahora podemos usar nuestro modelo para predecir el puntuaje en el examen para una cierta cantidad de horas de estudio:

```python
horas_nuevas = [[6]]
puntaje_predicho = modelo.predict(horas_nuevas)
print(f"Puntaje predicho para 6 horas de estudio: {puntaje_predicho[0]}")
```

En resumen, un modelo de regresión es una herramienta poderosa para predecir valores numéricos basados en relaciones matemáticas entre variables. En este ejemplo, utilizamos la regresión lineal simple para predecir puntuajes en un examen en función de las horas de estudio.

## **Error absoluto máximo**

El Error Absoluto Máximo, también conocido como MAE (Mean Absolute Error), es una métrica común utilizada para evaluar el rendimiento de un modelo de regresión al comparar las predicciones del modelo con los valores reales. Mide la diferencia absoluta entre las predicciones y los valores reales, sin considerar la dirección de la diferencia. En otras palabras, calcula el promedio de las diferencias absolutas entre las predicciones y los valores reales.

**Cálculo del Error Absoluto Máximo (MAE):**

La fórmula para calcular el MAE es la siguiente:

$MAE = \frac{1}{n} \sum_{i=1}^{n} |y_{i} - \hat{y}_{i}|$

Donde:

- ( $n$ ) es el número de observaciones (pares de valores reales y predichos).
- ( $y_{i}$ ) es el valor real de la observación ( $i$ ).
- ( $\hat{y}_{i}$ ) es el valor predicho por el modelo para la observación ( $i$ ).

**Ejemplo: Predicción de Precios de Casas:**

Supongamos que tenemos un modelo de regresión que predice los precios de casas en función de características como el tamaño, la ubicación, el número de habitaciones, etc. Tenemos los siguientes valores reales y predichos:

| Valor Real ( $y$ ) | Valor Predicho ( $\hat{y}$ ) |
|----------------------|-----------------------------|
| 200000               | 210000                      |
| 300000               | 280000                      |
| 150000               | 155000                      |
| 400000               | 390000                      |

**Cálculo del MAE:**

Calculamos la diferencia absoluta entre los valores reales y predichos, y luego tomamos el promedio de estas diferencias:

$MAE = \frac{|200000 - 210000| + |300000 - 280000| + |150000 - 155000| + |400000 - 390000|}{4} = \frac{10000 + 20000 + 5000 + 10000}{4} = 12500$

**Interpretación:**

Un MAE más bajo indica que el modelo tiene un mejor ajuste a los datos, ya que las diferencias entre las predicciones y los valores reales son más pequeñas.

**Gráfica del Error Absoluto Máximo:**

Podemos visualizar las diferencias absolutas entre las predicciones y los valores reales en un gráfico de barras:

```python
import matplotlib.pyplot as plt

valores_reales = [200000, 300000, 150000, 400000]
valores_predichos = [210000, 280000, 155000, 390000]

diferencias_absolutas = [abs(real - pred) for real, pred in zip(valores_reales, valores_predichos)]

plt.bar(range(len(diferencias_absolutas)), diferencias_absolutas)
plt.xlabel("Observaciones")
plt.ylabel("Diferencia Absoluta")
plt.title("Gráfico de Diferencias Absolutas entre Valores Reales y Predichos")
plt.show()
```

En resumen, el Error Absoluto Máximo (MAE) es una métrica que cuantifica la diferencia absoluta promedio entre las predicciones de un modelo de regresión y los valores reales. Es útil para evaluar el rendimiento del modelo y comprender cuánto varían las predicciones en comparación con los valores reales.

## **Error absoluto medio (MAE)**

El Error Absoluto Medio, también conocido como MAE (Mean Absolute Error), es una métrica fundamental en la evaluación de modelos de regresión. Mide el promedio de las diferencias absolutas entre las predicciones de un modelo y los valores reales. En otras palabras, representa cuánto en promedio se desvían las predicciones del modelo de los valores reales, sin considerar la dirección de la diferencia.

**Cálculo del Error Absoluto Medio (MAE):**

La fórmula para calcular el MAE es la siguiente:

$MAE = \frac{1}{n} \sum_{i=1}^{n} |y_{i} - \hat{y}_{i}|$

Donde:

- ( $n$ ) es el número de observaciones (pares de valores reales y predichos).
- ( $y_{i}$ ) es el valor real de la observación ( $i$ ).
- ( $\hat{y}_{i}$ ) es el valor predicho por el modelo para la observación ( $i$ ).

**Ejemplo: Predicción de Temperaturas:**

Supongamos que tenemos un modelo de regresión que predice las temperaturas diarias en función de características climáticas. Tenemos los siguientes valores reales y predichos:

| Valor Real ( $y$ ) | Valor Predicho ( $\hat{y}$ ) |
|----------------------|-----------------------------|
| 25.0                 | 24.5                        |
| 28.0                 | 28.2                        |
| 22.5                 | 22.0                        |
| 30.0                 | 29.8                        |

**Cálculo del MAE:**

Calculamos la diferencia absoluta entre los valores reales y predichos, y luego tomamos el promedio de estas diferencias:

$MAE = \frac{|25.0 - 24.5| + |28.0 - 28.2| + |22.5 - 22.0| + |30.0 - 29.8|}{4} = \frac{0.5 + 0.2 + 0.5 + 0.2}{4} = 0.35$

**Interpretación:**

Un MAE más bajo indica que el modelo tiene un mejor ajuste a los datos, ya que las diferencias entre las predicciones y los valores reales son más pequeñas.

**Gráfica del Error Absoluto Medio:**

Podemos visualizar las diferencias absolutas entre las predicciones y los valores reales en un gráfico de barras:

```python
import matplotlib.pyplot as plt

valores_reales = [25.0, 28.0, 22.5, 30.0]
valores_predichos = [24.5, 28.2, 22.0, 29.8]

diferencias_absolutas = [abs(real - pred) for real, pred in zip(valores_reales, valores_predichos)]

plt.bar(range(len(diferencias_absolutas)), diferencias_absolutas)
plt.xlabel("Observaciones")
plt.ylabel("Diferencia Absoluta")
plt.title("Gráfico de Diferencias Absolutas entre Valores Reales y Predichos")
plt.show()
```

En resumen, el Error Absoluto Medio (MAE) es una métrica que calcula el promedio de las diferencias absolutas entre las predicciones de un modelo de regresión y los valores reales. Es una medida de la precisión del modelo y permite evaluar cuánto se desvían en promedio las predicciones del valor real.

## **Error cuadrático medio (MSE)**

El Error Cuadrático Medio, también conocido como MSE (Mean Squared Error), es una métrica ampliamente utilizada en la evaluación de modelos de regresión. A diferencia del Error Absoluto Medio (MAE), el MSE considera las diferencias cuadráticas entre las predicciones y los valores reales. Esta métrica penaliza de manera más significativa los errores más grandes, lo que puede ser útil para resaltar la importancia de ajustar correctamente el modelo a los valores extremos.

**Cálculo del Error Cuadrático Medio (MSE):**

La fórmula para calcular el MSE es la siguiente:

$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_{i} - \hat{y}_{i})^2$

Donde:

- ( $n$ ) es el número de observaciones (pares de valores reales y predichos).
- ( $y_{i}$ ) es el valor real de la observación ( $i$ ).
- ( $\hat{y}_{i}$ ) es el valor predicho por el modelo para la observación ( $i$ ).

**Ejemplo: Predicción de Precios de Casas:**

Supongamos que tenemos un modelo de regresión que predice los precios de casas en función de características como el tamaño y la ubicación. Tenemos los siguientes valores reales y predichos:

| Valor Real ( $y$ ) | Valor Predicho ( $\hat{y}$ ) |
|----------------------|-----------------------------|
| 250,000              | 240,000                     |
| 320,000              | 310,000                     |
| 180,000              | 185,000                     |
| 400,000              | 390,000                     |

**Cálculo del MSE:**

Calculamos las diferencias al cuadrado entre los valores reales y predichos, y luego tomamos el promedio de estos valores:

$MSE = \frac{(250,000 - 240,000)^2 + (320,000 - 310,000)^2 + (180,000 - 185,000)^2 + (400,000 - 390,000)^2}{4}$

$MSE = \frac{10,000,000 + 10,000,000 + 25,000 + 100,000}{4} = \frac{20,135,000}{4} = 5,033,750$

**Interpretación:**

Un MSE más bajo indica que el modelo tiene un mejor ajuste a los datos, ya que las diferencias al cuadrado entre las predicciones y los valores reales son más pequeñas. Al elevar al cuadrado las diferencias, el MSE penaliza de manera más significativa los errores más grandes.

**Gráfica del Error Cuadrático Medio:**

Podemos visualizar las diferencias al cuadrado entre las predicciones y los valores reales en un gráfico de barras:

```python
import matplotlib.pyplot as plt

valores_reales = [250000, 320000, 180000, 400000]
valores_predichos = [240000, 310000, 185000, 390000]

diferencias_cuadraticas = [(real - pred)**2 for real, pred in zip(valores_reales, valores_predichos)]

plt.bar(range(len(diferencias_cuadraticas)), diferencias_cuadraticas)
plt.xlabel("Observaciones")
plt.ylabel("Diferencia Cuadrática")
plt.title("Gráfico de Diferencias Cuadráticas entre Valores Reales y Predichos")
plt.show()
```

En resumen, el Error Cuadrático Medio (MSE) es una métrica que calcula el promedio de las diferencias al cuadrado entre las predicciones de un modelo de regresión y los valores reales. Es una medida de la precisión del modelo y penaliza de manera más significativa los errores más grandes.

## **Suma de los cuadrados de los residuos (RSS)**

La Suma de los Cuadrados de los Residuos (RSS, por sus siglas en inglés) es una métrica utilizada en la evaluación de modelos de regresión para medir la cantidad total de variación no explicada por el modelo. En otras palabras, mide la diferencia entre los valores observados y las predicciones del modelo, al cuadrado, y luego los suma.

**Cálculo de la Suma de los Cuadrados de los Residuos (RSS):**

La fórmula para calcular el RSS es la siguiente:

$RSS = \sum_{i=1}^{n} (y_{i} - \hat{y}_{i})^2$

Donde:

- ( $n$ ) es el número de observaciones.
- ( $y_{i}$ ) es el valor real de la observación ( $i$ ).
- ( $\hat{y}_{i}$ ) es el valor predicho por el modelo para la observación ( $i$ ).

El RSS representa la cantidad total de error residual en el modelo, es decir, la suma de todas las diferencias al cuadrado entre los valores reales y las predicciones del modelo.

**Ejemplo: Predicción de Puntajes de Pruebas:**

Supongamos que tenemos un modelo de regresión que predice los puntajes de pruebas de estudiantes en función de las horas de estudio. Tenemos los siguientes valores reales y predichos:

| Valor Real ( $y$ ) | Valor Predicho ( $\hat{y}$ ) |
|----------------------|-----------------------------|
| 85                   | 80                          |
| 70                   | 75                          |
| 90                   | 88                          |
| 60                   | 62                          |

**Cálculo del RSS:**

Calculamos las diferencias al cuadrado entre los valores reales y predichos, y luego sumamos estos valores:

$RSS = (85 - 80)^2 + (70 - 75)^2 + (90 - 88)^2 + (60 - 62)^2$

$RSS = 25 + 25 + 4 + 4 = 58$

**Interpretación:**

El RSS representa la cantidad total de variación no explicada por el modelo. Un valor bajo de RSS indica que el modelo se ajusta bien a los datos, ya que las diferencias entre los valores reales y las predicciones son pequeñas en general.

**Gráfica del RSS:**

Podemos visualizar las diferencias al cuadrado entre los valores reales y predichos en un gráfico de barras:

```python
import matplotlib.pyplot as plt

valores_reales = [85, 70, 90, 60]
valores_predichos = [80, 75, 88, 62]

diferencias_cuadraticas = [(real - pred)**2 for real, pred in zip(valores_reales, valores_predichos)]

plt.bar(range(len(diferencias_cuadraticas)), diferencias_cuadraticas)
plt.xlabel("Observaciones")
plt.ylabel("Diferencia Cuadrática")
plt.title("Gráfico de Diferencias Cuadráticas entre Valores Reales y Predichos")
plt.show()
```

En resumen, la Suma de los Cuadrados de los Residuos (RSS) es una métrica que mide la cantidad total de error residual en un modelo de regresión. Representa la cantidad de variación no explicada por el modelo y se calcula sumando las diferencias al cuadrado entre los valores reales y las predicciones del modelo.

## **Raíz cuadrada del error cuadrático medio (RMSE)**

La Raíz Cuadrada del Error Cuadrático Medio (RMSE, por sus siglas en inglés) es una métrica ampliamente utilizada en la evaluación de modelos de regresión para medir la magnitud del error entre los valores reales y las predicciones del modelo. Es una forma de resumir cuán dispersos están los errores entre las predicciones y los valores reales, tomando en cuenta la magnitud de esos errores.

**Cálculo de la Raíz Cuadrada del Error Cuadrático Medio (RMSE):**

La fórmula para calcular el RMSE es la siguiente:

$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_{i} - \hat{y}_{i})^2}$

Donde:

- ( $n$ ) es el número de observaciones.
- ( $y_{i}$ ) es el valor real de la observación ( $i$ ).
- ( $\hat{y}_{i}$ ) es el valor predicho por el modelo para la observación ( $i$ ).
- ( $\sqrt{}$ ) representa la operación de raíz cuadrada.

El RMSE calcula la raíz cuadrada del promedio de los errores cuadráticos entre los valores reales y las predicciones del modelo. Al calcular la raíz cuadrada, se convierten las diferencias al cuadrado nuevamente a la misma escala que los valores originales, lo que facilita la interpretación.

**Ejemplo: Predicción de Puntajes de Pruebas:**

Continuando con el ejemplo anterior, supongamos que tenemos los siguientes valores reales y predichos:

| Valor Real ( $y$ ) | Valor Predicho ( $\hat{y}$ ) |
|----------------------|-----------------------------|
| 85                   | 80                          |
| 70                   | 75                          |
| 90                   | 88                          |
| 60                   | 62                          |

**Cálculo del RMSE:**

1. Calculamos las diferencias al cuadrado entre los valores reales y predichos, y luego sumamos estos valores:

$\sum_{i=1}^{n} (y_{i} - \hat{y}_{i})^2 = (85 - 80)^2 + (70 - 75)^2 + (90 - 88)^2 + (60 - 62)^2 = 25 + 25 + 4 + 4 = 58$

2. Dividimos la suma de los errores cuadráticos entre el número de observaciones:

$\frac{1}{n} \sum_{i=1}^{n} (y_{i} - \hat{y}_{i})^2 = \frac{58}{4} = 14.5$

3. Finalmente, calculamos la raíz cuadrada de este promedio:

$RMSE = \sqrt{14.5} \approx 3.81$

**Interpretación:**

Un valor bajo de RMSE indica que el modelo tiene un buen ajuste a los datos, ya que los errores entre las predicciones y los valores reales son pequeños en general.

**Gráfica del RMSE:**

Podemos visualizar los errores cuadráticos entre los valores reales y predichos en un gráfico de barras:

```python
import matplotlib.pyplot as plt
import numpy as np

valores_reales = np.array([85, 70, 90, 60])
valores_predichos = np.array([80, 75, 88, 62])

errores_cuadraticos = (valores_reales - valores_predichos) ** 2

plt.bar(range(len(errores_cuadraticos)), errores_cuadraticos)
plt.xlabel("Observaciones")
plt.ylabel("Error Cuadrático")
plt.title("Gráfico de Errores Cuadráticos entre Valores Reales y Predichos")
plt.show()
```

En resumen, la Raíz Cuadrada del Error Cuadrático Medio (RMSE) es una métrica que mide la magnitud del error entre los valores reales y las predicciones de un modelo de regresión. Calcula la raíz cuadrada del promedio de los errores cuadráticos y proporciona una medida de cuán dispersos están los errores en la misma escala que los valores originales.

## **Coeficiente de determinación o R2**

El Coeficiente de Determinación, también conocido como ( $R^2$ ) (R cuadrado), es una métrica utilizada para evaluar la calidad de un modelo de regresión. Representa la proporción de la variabilidad total en la variable dependiente que es explicada por las variables independientes del modelo. En otras palabras, ( $R^2$ ) mide qué tan bien las predicciones del modelo se ajustan a los valores reales de la variable dependiente.

**Cálculo del Coeficiente de Determinación ( $R^2$ ):**

La fórmula para calcular el coeficiente de determinación ( $R^2$ ) es la siguiente:

$R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$

Donde:

- ( $SS_{res}$ ) es la suma de los cuadrados de los residuos (suma de los errores al cuadrado entre las predicciones y los valores reales).
- ( $SS_{tot}$ ) es la suma total de los cuadrados (suma de las diferencias al cuadrado entre los valores reales y la media de los valores reales).

El valor de ( $R^2$ ) puede variar entre 0 y 1. Un ( $R^2$ ) más cercano a 1 indica que el modelo es capaz de explicar una gran parte de la variabilidad de la variable dependiente, mientras que un ( $R^2$ ) más cercano a 0 indica que el modelo no está explicando bien la variabilidad y sus predicciones son pobres.

**Ejemplo: Predicción de Puntajes de Pruebas:**

Supongamos que tenemos los siguientes valores reales y predichos:

| Valor Real ( $y$ ) | Valor Predicho ( $\hat{y}$ ) |
|----------------------|-----------------------------|
| 85                   | 80                          |
| 70                   | 75                          |
| 90                   | 88                          |
| 60                   | 62                          |

**Cálculo del ( $R^2$ ):**

1. Calculamos la suma total de los cuadrados ( $SS_{tot}$ ):

$SS_{tot} = (85 - \bar{y})^2 + (70 - \bar{y})^2 + (90 - \bar{y})^2 + (60 - \bar{y})^2$

Donde ( $\bar{y}$ ) es la media de los valores reales: ( $\bar{y} = \frac{85 + 70 + 90 + 60}{4} = 76.25$ ).

2. Calculamos la suma de los cuadrados de los residuos ( $SS_{res}$ ):

$SS_{res} = (85 - 80)^2 + (70 - 75)^2 + (90 - 88)^2 + (60 - 62)^2$

3. Aplicamos la fórmula del ( $R^2$ ):

$R^2 = 1 - \frac{SS_{res}}{SS_{tot}} = 1 - \frac{58}{147.5} \approx 0.608$

**Interpretación:**

En este caso, ( $R^2$ ) es aproximadamente 0.608. Esto significa que alrededor del 60.8% de la variabilidad en la variable dependiente (puntajes de pruebas) es explicada por las variables independientes del modelo. En otras palabras, el modelo puede hacer predicciones razonablemente buenas sobre los puntajes de pruebas.

**Gráfica del ( $R^2$ ):**

Podemos visualizar el concepto de ( $R^2$ ) en una gráfica de dispersión:

```python
import matplotlib.pyplot as plt
import numpy as np

valores_reales = np.array([85, 70, 90, 60])
valores_predichos = np.array([80, 75, 88, 62])
media_valores_reales = np.mean(valores_reales)

ss_tot = np.sum((valores_reales - media_valores_reales) ** 2)
ss_res = np.sum((valores_reales - valores_predichos) ** 2)
r_squared = 1 - (ss_res / ss_tot)

plt.scatter(valores_reales, valores_predichos)
plt.plot([min(valores_reales), max(valores_reales)], [min(valores_predichos), max(valores_predichos)], color='red', linestyle='--')
plt.xlabel("Valores Reales")
plt.ylabel("Valores Predichos")
plt.title(f"Gráfico de Dispersión y Ajuste del Modelo (R^2 = {r_squared:.3f})")
plt.show()
```

En resumen, el Coeficiente de Determinación ( $R^2$ ) es una métrica que evalúa la proporción de la variabilidad total

 en la variable dependiente que es explicada por las variables independientes del modelo. Un valor cercano a 1 indica un buen ajuste del modelo, mientras que un valor cercano a 0 indica un ajuste pobre. Se calcula mediante la proporción de los cuadrados de los residuos respecto a la variabilidad total.
