El Error Absoluto M谩ximo, tambi茅n conocido como MAE (Mean Absolute Error), es una m茅trica com煤n utilizada para evaluar el rendimiento de un modelo de regresi贸n al comparar las predicciones del modelo con los valores reales. **C谩lculo del Error Absoluto M谩ximo (MAE):**

La f贸rmula para calcular el MAE es la siguiente:

$MAE = \frac{1}{n} \sum_{i=1}^{n} |y_{i} - \hat{y}_{i}|$

Donde:

- ( $n$ ) es el n煤mero de observaciones (pares de valores reales y predichos).
- ( $y_{i}$ ) es el valor real de la observaci贸n ( $i$ ).
- ( $\hat{y}_{i}$ ) es el valor predicho por el modelo para la observaci贸n ( $i$ ).

**C谩lculo del MAE:**

**Interpretaci贸n:**

**Gr谩fica del Error Absoluto M谩ximo:**

En resumen, el Error Absoluto M谩ximo (MAE) es una m茅trica que cuantifica la diferencia absoluta promedio entre las predicciones de un modelo de regresi贸n y los valores reales. ## **Ra铆z cuadrada del error cuadr谩tico medio (RMSE)**

La Ra铆z Cuadrada del Error Cuadr谩tico Medio (RMSE, por sus siglas en ingl茅s) es una m茅trica ampliamente utilizada en la evaluaci贸n de modelos de regresi贸n para medir la magnitud del error entre los valores reales y las predicciones del modelo. **C谩lculo de la Ra铆z Cuadrada del Error Cuadr谩tico Medio (RMSE):**

La f贸rmula para calcular el RMSE es la siguiente:

$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_{i} - \hat{y}_{i})^2}$

Donde:

- ( $n$ ) es el n煤mero de observaciones.
- ( $y_{i}$ ) es el valor real de la observaci贸n ( $i$ ).
- ( $\hat{y}_{i}$ ) es el valor predicho por el modelo para la observaci贸n ( $i$ ).
- ( $\sqrt{}$ ) representa la operaci贸n de ra铆z cuadrada.

El RMSE calcula la ra铆z cuadrada del promedio de los errores cuadr谩ticos entre los valores reales y las predicciones del modelo. Al calcular la ra铆z cuadrada, se convierten las diferencias al cuadrado nuevamente a la misma escala que los valores originales, lo que facilita la interpretaci贸n.

**C谩lculo del RMSE:**

1. Calculamos las diferencias al cuadrado entre los valores reales y predichos, y luego sumamos estos valores:

$\sum_{i=1}^{n} (y_{i} - \hat{y}_{i})^2 = (85 - 80)^2 + (70 - 75)^2 + (90 - 88)^2 + (60 - 62)^2 = 25 + 25 + 4 + 4 = 58$

2. Dividimos la suma de los errores cuadr谩ticos entre el n煤mero de observaciones:

$\frac{1}{n} \sum_{i=1}^{n} (y_{i} - \hat{y}_{i})^2 = \frac{58}{4} = 14.5$

**Interpretaci贸n:**

En resumen, la Ra铆z Cuadrada del Error Cuadr谩tico Medio (RMSE) es una m茅trica que mide la magnitud del error entre los valores reales y las predicciones de un modelo de regresi贸n. ## **Coeficiente de determinaci贸n o R2**

El Coeficiente de Determinaci贸n, tambi茅n conocido como ( $R^2$ ) (R cuadrado), es una m茅trica utilizada para evaluar la calidad de un modelo de regresi贸n. Representa la proporci贸n de la variabilidad total en la variable dependiente que es explicada por las variables independientes del modelo. En otras palabras, ( $R^2$ ) mide qu茅 tan bien las predicciones del modelo se ajustan a los valores reales de la variable dependiente.

**C谩lculo del Coeficiente de Determinaci贸n ( $R^2$ ):**

La f贸rmula para calcular el coeficiente de determinaci贸n ( $R^2$ ) es la siguiente:

$R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$

Donde:

- ( $SS_{res}$ ) es la suma de los cuadrados de los residuos (suma de los errores al cuadrado entre las predicciones y los valores reales).
- ( $SS_{tot}$ ) es la suma total de los cuadrados (suma de las diferencias al cuadrado entre los valores reales y la media de los valores reales).

El valor de ( $R^2$ ) puede variar entre 0 y 1. Un ( $R^2$ ) m谩s cercano a 1 indica que el modelo es capaz de explicar una gran parte de la variabilidad de la variable dependiente, mientras que un ( $R^2$ ) m谩s cercano a 0 indica que el modelo no est谩 explicando bien la variabilidad y sus predicciones son pobres.

**Ejemplo: Predicci贸n de Puntajes de Pruebas:**

Supongamos que tenemos los siguientes valores reales y predichos:

| Valor Real ( $y$ ) | Valor Predicho ( $\hat{y}$ ) |
|----------------------|-----------------------------|
| 85                   | 80                          |
| 70                   | 75                          |
| 90                   | 88                          |
| 60                   | 62                          |

**C谩lculo del ( $R^2$ ):**

1. Calculamos la suma total de los cuadrados ( $SS_{tot}$ ):

$SS_{tot} = (85 - \bar{y})^2 + (70 - \bar{y})^2 + (90 - \bar{y})^2 + (60 - \bar{y})^2$

Donde ( $\bar{y}$ ) es la media de los valores reales: ( $\bar{y} = \frac{85 + 70 + 90 + 60}{4} = 76.25$ ).

2. Calculamos la suma de los cuadrados de los residuos ( $SS_{res}$ ):

$SS_{res} = (85 - 80)^2 + (70 - 75)^2 + (90 - 88)^2 + (60 - 62)^2$

3. Aplicamos la f贸rmula del ( $R^2$ ):

$R^2 = 1 - \frac{SS_{res}}{SS_{tot}} = 1 - \frac{58}{147.5} \approx 0.608$

**Interpretaci贸n:**

En este caso, ( $R^2$ ) es aproximadamente 0.608. Esto significa que alrededor del 60.8% de la variabilidad en la variable dependiente (puntajes de pruebas) es explicada por las variables independientes del modelo. En otras palabras, el modelo puede hacer predicciones razonablemente buenas sobre los puntajes de pruebas.

**Gr谩fica del ( $R^2$ ):**

Podemos visualizar el concepto de ( $R^2$ ) en una gr谩fica de dispersi贸n:

```python
import matplotlib.pyplot as plt
import numpy as np

valores_reales = np.array([85, 70, 90, 60])
valores_predichos = np.array([80, 75, 88, 62])
media_valores_reales = np.mean(valores_reales)

ss_tot = np.sum((valores_reales - media_valores_reales) ** 2)
ss_res = np.sum((valores_reales - valores_predichos) ** 2)
r_squared = 1 - (ss_res / ss_tot)

plt.scatter(valores_reales, valores_predichos)
plt.plot([min(valores_reales), max(valores_reales)], [min(valores_predichos), max(valores_predichos)], color='red', linestyle='--')
plt.xlabel("Valores Reales")
plt.ylabel("Valores Predichos")
plt.title(f"Gr谩fico de Dispersi贸n y Ajuste del Modelo (R^2 = {r_squared:.3f})")
plt.show()
```

En resumen, el Coeficiente de Determinaci贸n ( $R^2$ ) es una m茅trica que eval煤a la proporci贸n de la variabilidad total

 en la variable dependiente que es explicada por las variables independientes del modelo. Un valor cercano a 1 indica un buen ajuste del modelo, mientras que un valor cercano a 0 indica un ajuste pobre. Se calcula mediante la proporci贸n de los cuadrados de los residuos respecto a la variabilidad total.
