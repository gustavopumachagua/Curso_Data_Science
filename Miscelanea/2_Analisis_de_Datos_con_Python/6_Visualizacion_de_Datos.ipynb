# **6. Visualizaci√≥n de Datos usando Matplotlib de Python - Curso introductorio a la creaci√≥n de gr√°ficas**

## **Introducci√≥n a la Visualizaci√≥n de Datos**

¬°Por supuesto! La visualizaci√≥n de datos es una parte esencial del an√°lisis de datos, ya que te permite comprender patrones, tendencias y relaciones en los datos de manera m√°s intuitiva. En Python, una biblioteca com√∫nmente utilizada para la visualizaci√≥n de datos es Matplotlib. A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada junto con ejemplos y gr√°ficos correspondientes.

**Paso 1: Importar Matplotlib y Datos de Ejemplo**

Comencemos importando la biblioteca Matplotlib y creando datos de ejemplo que utilizaremos para crear visualizaciones.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
ventas = [1000, 1200, 800, 1500, 1100]
```

**Paso 2: Gr√°fico de L√≠nea**

Un gr√°fico de l√≠nea es √∫til para visualizar tendencias a lo largo del tiempo o de una variable a otra. Vamos a crear un gr√°fico de l√≠nea para visualizar las ventas mensuales.

```python
plt.plot(meses, ventas, marker='o')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Ventas Mensuales')
plt.grid(True)
plt.show()
```

**Paso 3: Gr√°fico de Barras**

Los gr√°ficos de barras son ideales para comparar diferentes categor√≠as o valores. Vamos a crear un gr√°fico de barras para comparar las ventas de diferentes meses.

```python

plt.bar(meses, ventas)
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Ventas Mensuales')
plt.show()
```

**Paso 4: Gr√°fico de Pastel**

Un gr√°fico de pastel es √∫til para mostrar la proporci√≥n de diferentes categor√≠as en un conjunto de datos. Vamos a crear un gr√°fico de pastel para visualizar la proporci√≥n de ventas en diferentes meses.

```python
plt.pie(ventas, labels=meses, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribuci√≥n de Ventas Mensuales')
plt.show()
```

**Paso 5: Gr√°fico de Dispersi√≥n**

Los gr√°ficos de dispersi√≥n se utilizan para mostrar la relaci√≥n entre dos variables. Crearemos un gr√°fico de dispersi√≥n para visualizar la relaci√≥n entre los meses y las ventas.

```python

plt.scatter(meses, ventas, color='blue')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Relaci√≥n entre Meses y Ventas')
plt.show()
```

Estos son solo ejemplos b√°sicos de c√≥mo utilizar Matplotlib para crear diferentes tipos de gr√°ficos y visualizar datos. Puedes explorar m√°s tipos de gr√°ficos, personalizar colores, estilos y agregar m√°s detalles seg√∫n tus necesidades. La visualizaci√≥n de datos te permite comunicar informaci√≥n de manera efectiva y descubrir insights importantes en tus conjuntos de datos.

## **Anatom√≠a de la Figura de Matplotlib**

¬°Por supuesto! La anatom√≠a de una figura en Matplotlib se refiere a las diferentes partes que componen un gr√°fico creado utilizando esta biblioteca. Aqu√≠ te proporcionar√© una explicaci√≥n detallada de las principales componentes de una figura de Matplotlib, acompa√±ada de un ejemplo para ilustrar cada parte.

**Anatom√≠a de la Figura de Matplotlib:**

1. **Figura (`figure`):** La figura es el lienzo en el que se crean los gr√°ficos. Puede contener uno o varios gr√°ficos. Se puede acceder a la figura utilizando el m√©todo `plt.figure()`.

2. **Ejes (`axes`):** Los ejes son los espacios dentro de la figura donde se dibujan los gr√°ficos. Puedes pensar en ellos como los "subgr√°ficos" individuales. Se pueden crear con `fig.add_subplot()` o `plt.subplots()`.

3. **T√≠tulo (`title`):** El t√≠tulo proporciona una descripci√≥n general del gr√°fico y se coloca en la parte superior del gr√°fico.

4. **Etiquetas de Ejes (`xlabel` y `ylabel`):** Las etiquetas de ejes describen las dimensiones del gr√°fico, como las etiquetas para el eje x (horizontal) y el eje y (vertical).

5. **Leyenda (`legend`):** La leyenda muestra las etiquetas para los diferentes elementos en el gr√°fico. Se puede crear utilizando el par√°metro `label` en las funciones de trazado y luego activarla con `plt.legend()`.

6. **Gr√°ficos (`plot`):** Los gr√°ficos son los datos visualizados en la figura. Pueden ser l√≠neas, barras, puntos, entre otros.

7. **Cuadr√≠cula (`grid`):** La cuadr√≠cula a√±ade l√≠neas de referencia en el gr√°fico para facilitar la lectura de los datos.

8. **Texto (`text`):** Se pueden agregar anotaciones de texto en diferentes ubicaciones del gr√°fico utilizando la funci√≥n `plt.text()`.

9. **Tama√±o de la Figura (`figsize`):** El tama√±o de la figura se define mediante el par√°metro `figsize` al crear la figura. Especifica el ancho y alto en pulgadas.

**Ejemplo:**

Aqu√≠ tienes un ejemplo que ilustra la anatom√≠a de una figura de Matplotlib utilizando un gr√°fico de l√≠nea:

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
ventas = [1000, 1200, 800, 1500, 1100]

# Crear una figura y ejes
fig, ax = plt.subplots()

# Trazar el gr√°fico de l√≠nea
ax.plot(meses, ventas, marker='o')

# Etiquetas y t√≠tulo
ax.set_xlabel('Mes')
ax.set_ylabel('Ventas')
ax.set_title('Ventas Mensuales')

# Leyenda
ax.legend(['Ventas'])

# Mostrar cuadr√≠cula
ax.grid(True)

# Mostrar la figura
plt.show()
```

Este ejemplo demuestra c√≥mo cada una de las partes mencionadas se integra en un gr√°fico de l√≠nea creado con Matplotlib. Recuerda que puedes personalizar a√∫n m√°s tu figura ajustando colores, estilos y otros detalles seg√∫n tus necesidades espec√≠ficas.

## **Personalizaci√≥n de Figuras**

¬°Por supuesto! La personalizaci√≥n de figuras en Matplotlib te permite ajustar colores, estilos, t√≠tulos, etiquetas y muchos otros aspectos para crear visualizaciones atractivas y efectivas. A continuaci√≥n, te mostrar√© c√≥mo personalizar diferentes aspectos de una figura utilizando ejemplos.

**Personalizaci√≥n de Colores y Estilos:**

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y1 = [10, 25, 15, 30, 20]
y2 = [5, 18, 12, 22, 8]

# Crear una figura y ejes
fig, ax = plt.subplots()

# Trazar los gr√°ficos de l√≠nea con colores y estilos diferentes
ax.plot(x, y1, color='blue', linestyle='-', marker='o', label='Serie 1')
ax.plot(x, y2, color='green', linestyle='--', marker='x', label='Serie 2')

# Etiquetas y t√≠tulo
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Gr√°fico Personalizado')

# Leyenda
ax.legend()

# Mostrar la figura
plt.show()
```

**Personalizaci√≥n de T√≠tulos y Etiquetas:**

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D']
valores = [20, 35, 25, 40]

# Crear una figura y ejes
plt.figure()
plt.bar(categorias, valores)

# Etiquetas y t√≠tulo
plt.xlabel('Categor√≠as')
plt.ylabel('Valores')
plt.title('Gr√°fico de Barras Personalizado')

# Mostrar la figura
plt.show()
```

**Personalizaci√≥n de Ejes y Cuadr√≠cula:**

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [10, 25, 15, 30, 20]

# Crear una figura y ejes
plt.figure()
plt.scatter(x, y, color='red')

# Etiquetas y t√≠tulo
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gr√°fico de Dispersi√≥n Personalizado')

# Personalizaci√≥n de ejes y cuadr√≠cula
plt.xticks([1, 2, 3, 4, 5])
plt.yticks([0, 10, 20, 30, 40])
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar la figura
plt.show()
```

Estos ejemplos te muestran c√≥mo personalizar diferentes aspectos de las figuras en Matplotlib, incluyendo colores, estilos, t√≠tulos, etiquetas, ejes y cuadr√≠cula. Puedes explorar a√∫n m√°s opciones de personalizaci√≥n en la documentaci√≥n de Matplotlib para adaptar tus visualizaciones a tus necesidades y preferencias espec√≠ficas.

## **Gr√°ficas de Barras Verticales**

Las gr√°ficas de barras verticales son un tipo de visualizaci√≥n que se utiliza para representar datos en forma de barras rectangulares dispuestas verticalmente. Cada barra representa una categor√≠a o valor, y la altura de la barra est√° relacionada con la magnitud del valor que representa. Estas gr√°ficas son √∫tiles para comparar datos entre diferentes categor√≠as y destacar diferencias y patrones. A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada junto con un ejemplo y su respectiva gr√°fica de barras verticales.

**Ejemplo: Ventas Mensuales de Productos**

Supongamos que tienes datos sobre las ventas mensuales de tres productos (A, B, C) en un a√±o. Quieres visualizar estas ventas utilizando una gr√°fica de barras verticales.

**Paso 1: Importar Matplotlib y Crear Datos de Ejemplo**

Comenzamos importando la biblioteca Matplotlib y creando los datos de ejemplo para las ventas mensuales de tres productos.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
ventas_producto_a = [100, 120, 80, 150, 110]
ventas_producto_b = [80, 110, 90, 130, 100]
ventas_producto_c = [120, 100, 110, 90, 130]
```

**Paso 2: Crear la Gr√°fica de Barras Verticales**

Ahora utilizamos Matplotlib para crear la gr√°fica de barras verticales que visualizar√° las ventas mensuales de los tres productos.

```python
# Crear una figura y ejes
plt.figure()

# Crear las barras verticales
plt.bar(meses, ventas_producto_a, label='Producto A')
plt.bar(meses, ventas_producto_b, label='Producto B')
plt.bar(meses, ventas_producto_c, label='Producto C')

# Etiquetas y t√≠tulo
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.title('Ventas Mensuales de Productos')
plt.legend()

# Mostrar la figura
plt.show()
```

**Resultado:**

La gr√°fica de barras verticales resultante tendr√° tres conjuntos de barras correspondientes a los productos A, B y C. Cada barra mostrar√° las ventas mensuales para el producto correspondiente en cada mes.

Recuerda que esta es una representaci√≥n simple de una gr√°fica de barras verticales. Puedes personalizarla a√∫n m√°s ajustando colores, estilos, t√≠tulos y otros detalles seg√∫n tus preferencias y necesidades espec√≠ficas. Las gr√°ficas de barras verticales son especialmente √∫tiles cuando deseas comparar magnitudes en diferentes categor√≠as.

Cada barra representa una categor√≠a o valor, y la longitud de la barra est√° relacionada con la magnitud del valor que representa. Estas gr√°ficas son excelentes para comparar datos entre diferentes categor√≠as y destacar diferencias y patrones, especialmente cuando se desea enfocar en una comparaci√≥n de magnitudes. A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada junto con un ejemplo y su respectiva gr√°fica de barras horizontales.

**Ejemplo: Poblaci√≥n de Pa√≠ses**

Supongamos que tienes datos sobre la poblaci√≥n de cinco pa√≠ses en millones. Quieres visualizar estas poblaciones utilizando una gr√°fica de barras horizontales.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
paises = ['China', 'India', 'Estados Unidos', 'Indonesia', 'Pakist√°n']
poblacion = [1439323776, 1380004385, 331002651, 273523615, 220892340]
```

Recuerda que esta es una representaci√≥n simple de una gr√°fica de barras horizontales. Puedes personalizarla a√∫n m√°s ajustando colores, estilos, t√≠tulos y otros detalles seg√∫n tus preferencias y necesidades espec√≠ficas. Las gr√°ficas de barras horizontales son especialmente √∫tiles cuando deseas comparar magnitudes entre diferentes categor√≠as, y resultan particularmente efectivas cuando tienes etiquetas de texto largas, como nombres de pa√≠ses.

## **Gr√°ficas de Barras Apiladas**

Las gr√°ficas de barras apiladas son una forma de visualizaci√≥n que se utiliza para representar datos en forma de barras rectangulares apiladas unas sobre otras. Cada barra representa una categor√≠a o valor, y las barras se apilan para mostrar la contribuci√≥n individual de cada subvalor a la categor√≠a total. Estas gr√°ficas son √∫tiles para comparar la relaci√≥n entre los subvalores dentro de cada categor√≠a y ver c√≥mo cambia esta relaci√≥n entre diferentes categor√≠as. A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada junto con un ejemplo y su respectiva gr√°fica de barras apiladas.

**Ejemplo: Ingresos de Departamentos**

Supongamos que tienes datos sobre los ingresos mensuales de tres departamentos (Ventas, Marketing, Finanzas) en diferentes trimestres. Quieres visualizar estos ingresos utilizando una gr√°fica de barras apiladas.

**Paso 1: Importar Matplotlib y Crear Datos de Ejemplo**

Comenzamos importando la biblioteca Matplotlib y creando los datos de ejemplo para los ingresos de los tres departamentos en diferentes trimestres.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
trimestres = ['Q1', 'Q2', 'Q3', 'Q4']
ingresos_ventas = [200, 250, 220, 270]
ingresos_marketing = [150, 180, 160, 190]
ingresos_finanzas = [120, 130, 110, 140]
```

**Paso 2: Crear la Gr√°fica de Barras Apiladas**

Ahora utilizamos Matplotlib para crear la gr√°fica de barras apiladas que visualizar√° los ingresos de los tres departamentos en diferentes trimestres.

```python
# Crear una figura y ejes
plt.figure()

# Crear las barras apiladas
plt.bar(trimestres, ingresos_ventas, label='Ventas')
plt.bar(trimestres, ingresos_marketing, bottom=ingresos_ventas, label='Marketing')
plt.bar(trimestres, ingresos_finanzas, bottom=[i+j for i,j in zip(ingresos_ventas, ingresos_marketing)], label='Finanzas')

# Etiquetas y t√≠tulo
plt.xlabel('Trimestres')
plt.ylabel('Ingresos')
plt.title('Ingresos de Departamentos por Trimestre')
plt.legend()

# Mostrar la figura
plt.show()
```

**Resultado:**

La gr√°fica de barras apiladas resultante mostrar√° cuatro barras apiladas correspondientes a cada trimestre. Cada barra estar√° compuesta por segmentos que representan los ingresos de los tres departamentos. La altura total de la barra representar√° el ingreso total en ese trimestre.

Recuerda que esta es una representaci√≥n simple de una gr√°fica de barras apiladas. Puedes personalizarla a√∫n m√°s ajustando colores, estilos, t√≠tulos y otros detalles seg√∫n tus preferencias y necesidades espec√≠ficas. Las gr√°ficas de barras apiladas son √∫tiles para mostrar c√≥mo contribuyen diferentes subvalores a una categor√≠a total.

## **Gr√°ficas de Dispersi√≥n**

Las gr√°ficas de dispersi√≥n son una forma de visualizaci√≥n que se utiliza para mostrar la relaci√≥n entre dos variables. Cada punto en la gr√°fica representa un par de valores correspondientes a las dos variables. Estas gr√°ficas son ideales para identificar patrones, tendencias, correlaciones y valores at√≠picos en los datos. A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada junto con un ejemplo y su respectiva gr√°fica de dispersi√≥n.

**Ejemplo: Relaci√≥n entre Horas de Estudio y Calificaci√≥n de Examen**

Supongamos que tienes datos sobre las horas de estudio y las calificaciones obtenidas por varios estudiantes en un examen. Quieres visualizar la relaci√≥n entre estas dos variables utilizando una gr√°fica de dispersi√≥n.

**Paso 1: Importar Matplotlib y Crear Datos de Ejemplo**

Comenzamos importando la biblioteca Matplotlib y creando los datos de ejemplo para las horas de estudio y las calificaciones de los estudiantes.

## **Gr√°ficas de Pie**

Las gr√°ficas de pastel, tambi√©n conocidas como gr√°ficas circulares, son una forma de visualizaci√≥n que se utiliza para mostrar la proporci√≥n de diferentes partes con respecto a un todo. Estas gr√°ficas representan cada parte como una porci√≥n de un c√≠rculo, donde el tama√±o del sector es proporcional al valor que representa. Son ideales para mostrar la composici√≥n relativa de categor√≠as en un conjunto de datos. A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada junto con un ejemplo y su respectiva gr√°fica de pastel.

**Ejemplo: Composici√≥n de Gastos Mensuales**

Supongamos que tienes datos sobre los gastos mensuales en diferentes categor√≠as (Alimentaci√≥n, Vivienda, Transporte, Entretenimiento). Quieres visualizar la composici√≥n de gastos utilizando una gr√°fica de pastel.

**Paso 1: Importar Matplotlib y Crear Datos de Ejemplo**

Comenzamos importando la biblioteca Matplotlib y creando los datos de ejemplo para los gastos mensuales en diferentes categor√≠as.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['Alimentaci√≥n', 'Vivienda', 'Transporte', 'Entretenimiento']
gastos = [600, 1000, 400, 300]
```

**Paso 2: Crear la Gr√°fica de Pastel**

Ahora utilizamos Matplotlib para crear la gr√°fica de pastel que visualizar√° la composici√≥n de gastos en diferentes categor√≠as.

```python
# Crear una figura y ejes
plt.figure()

# Crear la gr√°fica de pastel
plt.pie(gastos, labels=categorias, autopct='%1.1f%%', startangle=140)

# Propiedades adicionales
plt.axis('equal')  # Para que el c√≠rculo de la gr√°fica se vea circular
plt.title('Composici√≥n de Gastos Mensuales')

# Mostrar la figura
plt.show()
```

**Resultado:**

La gr√°fica de pastel resultante mostrar√° sectores circulares, donde cada sector representa una categor√≠a de gastos. El tama√±o de cada sector es proporcional al gasto en esa categor√≠a.

Recuerda que esta es una representaci√≥n simple de una gr√°fica de pastel. Puedes personalizarla a√∫n m√°s ajustando colores, porcentaje de formato, t√≠tulos y otros detalles seg√∫n tus preferencias y necesidades espec√≠ficas. Las gr√°ficas de pastel son √∫tiles para mostrar la distribuci√≥n proporcional de partes dentro de un todo.

## **Combinar Gr√°ficas**

La combinaci√≥n de gr√°ficas se refiere a la pr√°ctica de superponer o mostrar m√∫ltiples tipos de gr√°ficas en una misma figura para comparar y visualizar datos desde diferentes perspectivas. Puedes combinar gr√°ficas para presentar informaci√≥n m√°s completa y hacer que tus visualizaciones sean m√°s efectivas. Aqu√≠ tienes una explicaci√≥n detallada junto con un ejemplo y sus respectivas gr√°ficas de combinaci√≥n.

**Ejemplo: Combinaci√≥n de Gr√°fica de L√≠nea y Gr√°fica de Barras**

Supongamos que tienes datos sobre los ingresos y los gastos mensuales de un negocio durante un a√±o. Quieres visualizar tanto los ingresos como los gastos en la misma figura para comparar su tendencia a lo largo del tiempo.

**Paso 1: Importar Matplotlib y Crear Datos de Ejemplo**

Comenzamos importando la biblioteca Matplotlib y creando los datos de ejemplo para los ingresos y los gastos mensuales.

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
ingresos = [5000, 5500, 6000, 6200, 7000, 7500, 8000, 7800, 8500, 9000, 9500, 10000]
gastos = [4000, 4200, 4500, 4800, 5000, 5500, 6000, 5800, 6200, 6500, 7000, 7500]
```

**Paso 2: Crear la Combinaci√≥n de Gr√°ficas**

Ahora utilizamos Matplotlib para crear una figura con una gr√°fica de l√≠nea para los ingresos y una gr√°fica de barras para los gastos.

```python
# Crear una figura y ejes
fig, ax1 = plt.subplots()

# Trazar la gr√°fica de l√≠nea para los ingresos
ax1.plot(meses, ingresos, color='blue', marker='o', label='Ingresos')
ax1.set_xlabel('Meses')
ax1.set_ylabel('Ingresos', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.legend(loc='upper left')

# Crear un segundo eje y para la gr√°fica de barras de gastos
ax2 = ax1.twinx()
ax2.bar(meses, gastos, color='red', alpha=0.3, label='Gastos')
ax2.set_ylabel('Gastos', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.legend(loc='upper right')

# T√≠tulo
plt.title('Combinaci√≥n de Gr√°fica de L√≠nea e Gr√°fica de Barras')

# Mostrar la figura
plt.show()
```

**Resultado:**

La figura resultante mostrar√° una gr√°fica de l√≠nea para los ingresos (con eje y izquierdo) y una gr√°fica de barras para los gastos (con eje y derecho). Las dos gr√°ficas est√°n superpuestas en el mismo espacio y permiten comparar f√°cilmente las tendencias de ingresos y gastos a lo largo del tiempo.

Recuerda que esta es una representaci√≥n simple de c√≥mo combinar gr√°ficas. Puedes ajustar colores, estilos, etiquetas y otros detalles para personalizar la visualizaci√≥n seg√∫n tus preferencias y necesidades espec√≠ficas. La combinaci√≥n de gr√°ficas es una t√©cnica poderosa para presentar informaci√≥n completa en una sola figura.
