## **Intervalo de confianza**

El intervalo de confianza es un concepto estad铆stico que se utiliza para estimar un rango de valores en el cual es probable que se encuentre un par谩metro poblacional desconocido, bas谩ndose en una muestra de datos. Se expresa como un rango de valores junto con un nivel de confianza que indica qu茅 tan seguro estamos de que el par谩metro real est谩 dentro de ese rango.

Por ejemplo, supongamos que est谩s realizando una encuesta para estimar la estatura promedio de una poblaci贸n. Despu茅s de recolectar datos de una muestra, puedes calcular un intervalo de confianza para la estatura promedio de la poblaci贸n. Un intervalo de confianza t铆pico se expresa como "X% de confianza", donde X es un porcentaje (por ejemplo, 95% o 99%). Esto significa que si repetimos el proceso de muestreo y c谩lculo del intervalo muchas veces, X% de los intervalos generados contendr谩n el verdadero valor del par谩metro.

La f贸rmula general para calcular un intervalo de confianza para la media poblacional (cuando se cumple el teorema del l铆mite central) es:

$\text{Intervalo de confianza} = \text{Media muestral} \pm \text{Margen de error}$

El margen de error se calcula utilizando el valor cr铆tico z correspondiente al nivel de confianza y la desviaci贸n est谩ndar muestral.

Aqu铆 tienes un ejemplo de c贸mo calcular el intervalo de confianza utilizando Python y NumPy:

````python
import numpy as np
from scipy import stats

# Datos de la muestra
data = np.array([165, 170, 160, 175, 168, 180, 172, 167, 169, 171])

# Nivel de confianza (por ejemplo, 95%)
confidence_level = 0.95

# Calcular media y desviaci贸n est谩ndar muestrales
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Usar ddof=1 para c谩lculo no sesgado

# Calcular el valor cr铆tico z para el nivel de confianza
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

# Calcular el margen de error
margin_of_error = z_critical * (sample_std / np.sqrt(len(data)))

# Calcular el intervalo de confianza
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("Intervalo de confianza:", confidence_interval)
````

Recuerda que estos ejemplos son solo para ilustrar c贸mo se calcula un intervalo de confianza. En aplicaciones reales, debes ajustar los valores seg煤n tus datos y necesidades.

## **Combinar gr谩ficas**

Combinar gr谩ficas significa superponer o colocar m煤ltiples tipos de gr谩ficos en una misma figura para visualizar y comparar diferentes conjuntos de datos. Seaborn y Matplotlib ofrecen formas flexibles de combinar gr谩ficas para crear visualizaciones m谩s ricas y significativas. Aqu铆 tienes ejemplos de c贸mo combinar diferentes tipos de gr谩ficos utilizando Seaborn y Matplotlib:

**Ejemplo de Combinaci贸n de Gr谩ficas:**

Supongamos que tienes datos de ventas mensuales y deseas comparar la distribuci贸n de ventas y las ventas promedio por mes utilizando un diagrama de viol铆n y un gr谩fico de l铆neas.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear datos simulados
data = {
    "mes": ["Ene", "Feb", "Mar", "Abr", "May", "Jun"],
    "ventas": [1500, 1800, 2200, 2500, 2300, 2800]
}
df = pd.DataFrame(data)

# Crear un gr谩fico combinado de viol铆n y l铆neas
plt.figure(figsize=(10, 6))

# Subplot 1: Diagrama de Viol铆n
plt.subplot(2, 1, 1)
sns.violinplot(x="mes", y="ventas", data=df)
plt.title("Diagrama de Viol铆n: Distribuci贸n de Ventas por Mes")

# Subplot 2: Gr谩fico de L铆neas
plt.subplot(2, 1, 2)
sns.lineplot(x="mes", y="ventas", data=df, marker="o")
plt.title("Gr谩fico de L铆neas: Ventas Mensuales")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````

En este ejemplo, utilizamos `plt.subplot()` para crear dos subtramas en una figura. En la primera subtrama, creamos un diagrama de viol铆n que muestra la distribuci贸n de las ventas por mes. En la segunda subtrama, creamos un gr谩fico de l铆neas que muestra las ventas mensuales con puntos de datos resaltados. Luego, usamos `plt.tight_layout()` para ajustar el dise帽o de la figura y mostrar las gr谩ficas de manera ordenada.

Este es solo un ejemplo de c贸mo combinar diferentes tipos de gr谩ficos. Puedes experimentar con otras combinaciones y ajustar los par谩metros para obtener la visualizaci贸n que mejor se adapte a tus datos y objetivos.
