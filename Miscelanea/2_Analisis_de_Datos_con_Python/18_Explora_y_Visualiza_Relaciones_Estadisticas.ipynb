## **Introducción a la visualización de datos**

La visualización de datos es una herramienta fundamental en el análisis de datos y la comunicación de resultados. Permite representar información de manera gráfica para identificar patrones, tendencias y relaciones en los datos. En esta explicación, te mostraré los conceptos básicos de la visualización de datos utilizando la biblioteca Matplotlib en Python.

**Ejemplo de Visualización de Datos:**

Supongamos que tenemos datos de ventas mensuales de dos productos diferentes: "Producto A" y "Producto B". Queremos visualizar cómo han evolucionado las ventas a lo largo de los meses.

````python
import matplotlib.pyplot as plt

# Datos de ejemplo
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
ventas_producto_a = [1500, 1800, 2200, 2500, 2300, 2800]
ventas_producto_b = [1200, 1600, 1800, 2100, 2400, 2700]

# Crear un gráfico de líneas
plt.figure(figsize=(8, 5))
plt.plot(meses, ventas_producto_a, marker='o', label='Producto A')
plt.plot(meses, ventas_producto_b, marker='o', label='Producto B')

# Agregar etiquetas y título
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Ventas Mensuales de Productos')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
````

**Explicación:**

1. Importamos la biblioteca Matplotlib como `plt`.

2. Definimos los datos de ejemplo para los meses y las ventas de los productos A y B.

3. Creamos un gráfico de líneas utilizando `plt.plot()`, donde especificamos los meses en el eje x y las ventas en el eje y. Usamos el argumento `marker='o'` para mostrar puntos en los datos.

4. Agregamos etiquetas y un título al gráfico utilizando `plt.xlabel()`, `plt.ylabel()` y `plt.title()`.

5. Usamos `plt.legend()` para mostrar una leyenda que identifica los productos.

6. Finalmente, usamos `plt.grid(True)` para agregar una cuadrícula y `plt.show()` para mostrar el gráfico.

Este ejemplo demuestra cómo crear un gráfico de líneas simple para visualizar las ventas mensuales de dos productos. Matplotlib ofrece muchas más opciones de personalización para adaptarse a tus necesidades de visualización de datos, incluyendo gráficos de dispersión, histogramas, gráficos de barras y más. La visualización de datos es una herramienta poderosa para explorar y comunicar información de manera efectiva.

## **Tipos de gráficas en Seaborn**

Seaborn es una biblioteca de visualización de datos en Python que se basa en Matplotlib y proporciona una interfaz más sencilla y estilizada para crear gráficos atractivos y efectivos. Aquí tienes algunos ejemplos de los tipos de gráficos que puedes crear utilizando Seaborn:

7. **Mapa de Calor (Heatmap):**

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Crear datos simulados
data = {
    "edad": [25, 30, 22, 35, 28, 40, 32, 26, 29, 31],
    "altura": [165, 170, 160, 175, 168, 180, 172, 167, 169, 171],
    "ventas": [1500, 1800, 2200, 2500, 2300, 2800, 2000, 2100, 2400, 1900]
}
df = pd.DataFrame(data)

# Calcular la matriz de correlación
correlation_matrix = df.corr()

# Crear un mapa de calor
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Mapa de Calor: Matriz de Correlación")
plt.show()
````

## **Intervalo de confianza**

El intervalo de confianza es un concepto estadístico que se utiliza para estimar un rango de valores en el cual es probable que se encuentre un parámetro poblacional desconocido, basándose en una muestra de datos. Se expresa como un rango de valores junto con un nivel de confianza que indica qué tan seguro estamos de que el parámetro real está dentro de ese rango.

Por ejemplo, supongamos que estás realizando una encuesta para estimar la estatura promedio de una población. Después de recolectar datos de una muestra, puedes calcular un intervalo de confianza para la estatura promedio de la población. Un intervalo de confianza típico se expresa como "X% de confianza", donde X es un porcentaje (por ejemplo, 95% o 99%). Esto significa que si repetimos el proceso de muestreo y cálculo del intervalo muchas veces, X% de los intervalos generados contendrán el verdadero valor del parámetro.

La fórmula general para calcular un intervalo de confianza para la media poblacional (cuando se cumple el teorema del límite central) es:

$\text{Intervalo de confianza} = \text{Media muestral} \pm \text{Margen de error}$

El margen de error se calcula utilizando el valor crítico z correspondiente al nivel de confianza y la desviación estándar muestral.

Aquí tienes un ejemplo de cómo calcular el intervalo de confianza utilizando Python y NumPy:

````python
import numpy as np
from scipy import stats

# Datos de la muestra
data = np.array([165, 170, 160, 175, 168, 180, 172, 167, 169, 171])

# Nivel de confianza (por ejemplo, 95%)
confidence_level = 0.95

# Calcular media y desviación estándar muestrales
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Usar ddof=1 para cálculo no sesgado

# Calcular el valor crítico z para el nivel de confianza
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

# Calcular el margen de error
margin_of_error = z_critical * (sample_std / np.sqrt(len(data)))

# Calcular el intervalo de confianza
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("Intervalo de confianza:", confidence_interval)
````

Recuerda que estos ejemplos son solo para ilustrar cómo se calcula un intervalo de confianza. En aplicaciones reales, debes ajustar los valores según tus datos y necesidades.

## **Combinar gráficas**

Combinar gráficas significa superponer o colocar múltiples tipos de gráficos en una misma figura para visualizar y comparar diferentes conjuntos de datos. Seaborn y Matplotlib ofrecen formas flexibles de combinar gráficas para crear visualizaciones más ricas y significativas. Aquí tienes ejemplos de cómo combinar diferentes tipos de gráficos utilizando Seaborn y Matplotlib:

**Ejemplo de Combinación de Gráficas:**

Supongamos que tienes datos de ventas mensuales y deseas comparar la distribución de ventas y las ventas promedio por mes utilizando un diagrama de violín y un gráfico de líneas.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear datos simulados
data = {
    "mes": ["Ene", "Feb", "Mar", "Abr", "May", "Jun"],
    "ventas": [1500, 1800, 2200, 2500, 2300, 2800]
}
df = pd.DataFrame(data)

# Crear un gráfico combinado de violín y líneas
plt.figure(figsize=(10, 6))

# Subplot 1: Diagrama de Violín
plt.subplot(2, 1, 1)
sns.violinplot(x="mes", y="ventas", data=df)
plt.title("Diagrama de Violín: Distribución de Ventas por Mes")

# Subplot 2: Gráfico de Líneas
plt.subplot(2, 1, 2)
sns.lineplot(x="mes", y="ventas", data=df, marker="o")
plt.title("Gráfico de Líneas: Ventas Mensuales")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````

En este ejemplo, utilizamos `plt.subplot()` para crear dos subtramas en una figura. En la primera subtrama, creamos un diagrama de violín que muestra la distribución de las ventas por mes. En la segunda subtrama, creamos un gráfico de líneas que muestra las ventas mensuales con puntos de datos resaltados. Luego, usamos `plt.tight_layout()` para ajustar el diseño de la figura y mostrar las gráficas de manera ordenada.

Este es solo un ejemplo de cómo combinar diferentes tipos de gráficos. Puedes experimentar con otras combinaciones y ajustar los parámetros para obtener la visualización que mejor se adapte a tus datos y objetivos.
