# **24. Las 5 formas m√°s Simples de Predecir el FUTURO en Series de Tiempo con Python: Media M√≥vil y ‚úö**

## **Introducci√≥n**

Predecir el futuro en series de tiempo es un objetivo clave en el an√°lisis de datos. Una de las formas m√°s simples de predecir valores futuros es utilizando t√©cnicas de promedio m√≥vil. Aqu√≠ te explicar√© dos de estas t√©cnicas: la media m√≥vil y la media m√≥vil ponderada, junto con ejemplos y gr√°ficas correspondientes utilizando Python y las bibliotecas Pandas y Matplotlib.

Aseg√∫rate de tener las bibliotecas instaladas utilizando el siguiente c√≥digo:

**1. Media M√≥vil:**

La media m√≥vil es una t√©cnica que suaviza una serie de tiempo calculando el promedio de los valores en un intervalo de tiempo espec√≠fico. Esto ayuda a eliminar el ruido y revelar tendencias subyacentes.

**Ejemplo de Media M√≥vil:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de ejemplo
np.random.seed(0)
dates = pd.date_range(start='2021-01-01', periods=50, freq='D')
values = np.sin(np.linspace(0, 4 * np.pi, 50)) + np.random.normal(0, 0.2, 50)

# Crear un DataFrame de Pandas
data = pd.DataFrame({'Date': dates, 'Values': values})
data.set_index('Date', inplace=True)

# Calcular la media m√≥vil con ventana de tama√±o 7
rolling_mean = data['Values'].rolling(window=7).mean()

# Graficar la serie original y la media m√≥vil
plt.figure(figsize=(10, 6))
plt.plot(data['Values'], label='Serie Original')
plt.plot(rolling_mean, label='Media M√≥vil (Ventana 7)')
plt.legend()
plt.title('Media M√≥vil en Series de Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```

En este ejemplo, hemos creado una serie de tiempo que es la suma de una funci√≥n sinusoidal y ruido. Calculamos la media m√≥vil utilizando una ventana de tama√±o 7 para suavizar la serie. La gr√°fica muestra la serie original y la media m√≥vil.

**2. Media M√≥vil Ponderada:**

La media m√≥vil ponderada es similar a la media m√≥vil, pero en lugar de asignar un peso uniforme a todos los valores en la ventana, se asignan pesos diferentes seg√∫n su posici√≥n en la ventana.

**Ejemplo de Media M√≥vil Ponderada:**

```python

# Calcular la media m√≥vil ponderada con ventana de tama√±o 5 y pesos [0.1, 0.2, 0.3, 0.2, 0.1]
weights = [0.1, 0.2, 0.3, 0.2, 0.1]
weighted_rolling_mean = data['Values'].rolling(window=5).apply(lambda x: np.dot(x, weights), raw=True)

# Graficar la serie original y la media m√≥vil ponderada
plt.figure(figsize=(10, 6))
plt.plot(data['Values'], label='Serie Original')
plt.plot(weighted_rolling_mean, label='Media M√≥vil Ponderada (Ventana 5)')
plt.legend()
plt.title('Media M√≥vil Ponderada en Series de Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```

En este ejemplo, hemos calculado la media m√≥vil ponderada utilizando una ventana de tama√±o 5 y pesos predefinidos. La gr√°fica muestra la serie original y la media m√≥vil ponderada.

En resumen, las t√©cnicas de promedio m√≥vil, como la media m√≥vil y la media m√≥vil ponderada, son m√©todos simples pero efectivos para predecir valores futuros en series de tiempo. Estas t√©cnicas suavizan los datos y pueden ayudar a identificar tendencias subyacentes al eliminar el ruido.

## **M√©todos de Pron√≥stico Simples**

Los m√©todos de pron√≥stico simples son enfoques b√°sicos utilizados para hacer predicciones en series de tiempo. Estos m√©todos no tienen en cuenta patrones complicados ni consideran relaciones complejas entre las observaciones pasadas y futuras. A continuaci√≥n, te explicar√© tres de los m√©todos de pron√≥stico m√°s simples: Na√Øve, Media y Aleatorio, junto con ejemplos y gr√°ficas correspondientes utilizando Python y las bibliotecas Pandas y Matplotlib.

**1. M√©todo Na√Øve (Inocente):**

El m√©todo Na√Øve es uno de los m√©todos m√°s simples de pron√≥stico. Consiste en asumir que el pr√≥ximo valor ser√° igual al valor m√°s reciente en la serie de tiempo.

**Ejemplo de M√©todo Na√Øve:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de ejemplo
np.random.seed(0)
dates = pd.date_range(start='2021-01-01', periods=50, freq='D')
values = np.sin(np.linspace(0, 4 * np.pi, 50)) + np.random.normal(0, 0.2, 50)

# Crear un DataFrame de Pandas
data = pd.DataFrame({'Date': dates, 'Values': values})
data.set_index('Date', inplace=True)

# Calcular el pron√≥stico Na√Øve
naive_forecast = data['Values'].shift(1)

# Graficar la serie original y el pron√≥stico Na√Øve
plt.figure(figsize=(10, 6))
plt.plot(data['Values'], label='Serie Original')
plt.plot(naive_forecast, label='Pron√≥stico Na√Øve')
plt.legend()
plt.title('Pron√≥stico Na√Øve en Series de Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```

En este ejemplo, hemos creado una serie de tiempo con una funci√≥n sinusoidal y ruido. El pron√≥stico Na√Øve simplemente es el valor m√°s reciente de la serie de tiempo, desplazado hacia adelante. La gr√°fica muestra la serie original y el pron√≥stico Na√Øve.

**2. M√©todo de la Media:**

El m√©todo de la Media pronostica los valores futuros utilizando el promedio de los valores pasados en la serie de tiempo.

**Ejemplo de M√©todo de la Media:**

```python
# Calcular el pron√≥stico de la Media
mean_forecast = data['Values'].rolling(window=5).mean()

# Graficar la serie original y el pron√≥stico de la Media
plt.figure(figsize=(10, 6))
plt.plot(data['Values'], label='Serie Original')
plt.plot(mean_forecast, label='Pron√≥stico de la Media (Ventana 5)')
plt.legend()
plt.title('Pron√≥stico de la Media en Series de Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```

En este ejemplo, hemos calculado el pron√≥stico de la Media utilizando el promedio m√≥vil con una ventana de tama√±o 5. La gr√°fica muestra la serie original y el pron√≥stico de la Media.

**3. M√©todo Aleatorio:**

El m√©todo Aleatorio pronostica valores futuros seleccionando aleatoriamente valores anteriores de la serie de tiempo.

**Ejemplo de M√©todo Aleatorio:**

```python

# Calcular el pron√≥stico Aleatorio
random_forecast = np.random.choice(data['Values'], size=len(data))

# Graficar la serie original y el pron√≥stico Aleatorio
plt.figure(figsize=(10, 6))
plt.plot(data['Values'], label='Serie Original')
plt.plot(data.index, random_forecast, label='Pron√≥stico Aleatorio')
plt.legend()
plt.title('Pron√≥stico Aleatorio en Series de Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```

En este ejemplo, hemos calculado el pron√≥stico Aleatorio seleccionando valores anteriores de la serie de tiempo de manera aleatoria. La gr√°fica muestra la serie original y el pron√≥stico Aleatorio.

En resumen, los m√©todos de pron√≥stico simples, como Na√Øve, Media y Aleatorio, son enfoques b√°sicos para predecir valores futuros en series de tiempo. Estos m√©todos no tienen en cuenta patrones complejos y se utilizan generalmente como puntos de referencia para comparar con m√©todos m√°s avanzados. Las gr√°ficas ilustran c√≥mo se comparan los pron√≥sticos con la serie original.

## **M√©todo Ingenuo**

El m√©todo ingenuo, tambi√©n conocido como "Na√Øve" o "Inocente", es uno de los m√©todos de pron√≥stico m√°s simples y directos en el an√°lisis de series de tiempo. Consiste en asumir que el pr√≥ximo valor en la serie de tiempo ser√° igual al valor m√°s reciente observado. Aunque este enfoque puede parecer demasiado simplista, es √∫til como punto de referencia inicial y para evaluar la eficacia de otros m√©todos m√°s complejos.

A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada, un ejemplo y las respectivas gr√°ficas para ilustrar el m√©todo ingenuo utilizando Python y las bibliotecas Pandas y Matplotlib.


**Ejemplo de M√©todo Ingenuo:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de ejemplo
np.random.seed(0)
dates = pd.date_range(start='2021-01-01', periods=50, freq='D')
values = np.sin(np.linspace(0, 4 * np.pi, 50)) + np.random.normal(0, 0.2, 50)

# Crear un DataFrame de Pandas
data = pd.DataFrame({'Date': dates, 'Values': values})
data.set_index('Date', inplace=True)

# Calcular el pron√≥stico ingenuo (Na√Øve)
naive_forecast = data['Values'].shift(1)

# Graficar la serie original y el pron√≥stico ingenuo (Na√Øve)
plt.figure(figsize=(10, 6))
plt.plot(data['Values'], label='Serie Original')
plt.plot(naive_forecast, label='Pron√≥stico Ingenuo (Na√Øve)')
plt.legend()
plt.title('Pron√≥stico Ingenuo en Series de Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```

En este ejemplo, hemos creado una serie de tiempo que es la suma de una funci√≥n sinusoidal y ruido. El pron√≥stico ingenuo simplemente toma el valor m√°s reciente de la serie de tiempo y lo desplaza hacia adelante. La gr√°fica muestra la serie original y el pron√≥stico ingenuo.

**Interpretaci√≥n:**

- La l√≠nea azul representa la serie de tiempo original.
- La l√≠nea naranja representa el pron√≥stico ingenuo, donde cada valor es el mismo que el valor en el tiempo anterior.

Aunque el m√©todo ingenuo es muy b√°sico y no considera patrones o tendencias en los datos, puede ser √∫til como un punto de partida para comprender la naturaleza de los datos y para evaluar la eficacia de otros m√©todos de pron√≥stico m√°s complejos. En ocasiones, m√©todos m√°s sofisticados no superan al m√©todo ingenuo si los datos no presentan patrones claros.

## **M√©todo de la Media**

El m√©todo de la media es uno de los m√©todos de pron√≥stico m√°s simples en el an√°lisis de series de tiempo. Consiste en calcular el promedio de los valores pasados en la serie y utilizar ese promedio como pron√≥stico para los valores futuros. Aunque este m√©todo no considera patrones complejos ni tendencias en los datos, puede ser √∫til para suavizar ruido y obtener una idea general de la direcci√≥n en la que se mueven los datos.

A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada, un ejemplo y las respectivas gr√°ficas para ilustrar el m√©todo de la media utilizando Python y las bibliotecas Pandas y Matplotlib.

En este ejemplo, hemos creado una serie de tiempo que es la suma de una funci√≥n sinusoidal y ruido. El pron√≥stico de la media calcula el promedio m√≥vil utilizando una ventana de tama√±o 5 para suavizar la serie. La gr√°fica muestra la serie original y el pron√≥stico de la media.

**Interpretaci√≥n:**

- La l√≠nea azul representa la serie de tiempo original.
- La l√≠nea naranja representa el pron√≥stico de la media, donde cada valor es el promedio de los valores en una ventana m√≥vil de tama√±o 5.

Aunque el m√©todo de la media es simple y no considera patrones complejos, puede ser √∫til para identificar tendencias generales en los datos al eliminar el ruido. Sin embargo, este m√©todo puede ser sensible a cambios abruptos en los datos debido a la naturaleza de suavizaci√≥n de la media m√≥vil.

## **M√©todo de la Deriva**

El M√©todo de la Deriva, tambi√©n conocido como m√©todo de tendencia lineal, es un enfoque simple de pron√≥stico en el an√°lisis de series de tiempo que se basa en la extrapolaci√≥n de una tendencia lineal en los datos pasados para predecir valores futuros. Este m√©todo asume que la serie de tiempo sigue una trayectoria lineal y utiliza la ecuaci√≥n de una l√≠nea para hacer pron√≥sticos.

A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada, un ejemplo y las respectivas gr√°ficas para ilustrar el M√©todo de la Deriva utilizando Python y las bibliotecas Pandas y Matplotlib.

**Ejemplo de M√©todo de la Deriva:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de ejemplo
np.random.seed(0)
dates = pd.date_range(start='2021-01-01', periods=50, freq='D')
trend = 0.1 * np.arange(50)
values = trend + np.random.normal(0, 0.2, 50)

# Crear un DataFrame de Pandas
data = pd.DataFrame({'Date': dates, 'Values': values})
data.set_index('Date', inplace=True)

# Calcular la tendencia lineal utilizando regresi√≥n
coefficients = np.polyfit(np.arange(len(data)), data['Values'], 1)
trend_line = np.polyval(coefficients, np.arange(len(data)))

# Graficar la serie original y la tendencia lineal
plt.figure(figsize=(10, 6))
plt.plot(data['Values'], label='Serie Original')
plt.plot(trend_line, label='Tendencia Lineal')
plt.legend()
plt.title('M√©todo de la Deriva en Series de Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```

En este ejemplo, hemos creado una serie de tiempo que sigue una tendencia lineal (la variable "trend") junto con ruido aleatorio. Utilizamos regresi√≥n lineal para calcular la ecuaci√≥n de la l√≠nea de tendencia. La gr√°fica muestra la serie original y la tendencia lineal.

**Interpretaci√≥n:**

- La l√≠nea azul representa la serie de tiempo original con tendencia y ruido.
- La l√≠nea naranja representa la tendencia lineal extrapolada utilizando el m√©todo de la deriva.

El m√©todo de la deriva es √∫til cuando existe una tendencia lineal clara en los datos y se desea proyectar esa tendencia en el futuro. Sin embargo, es importante tener en cuenta que este m√©todo no captura patrones m√°s complejos y no es adecuado para series de tiempo con fluctuaciones irregulares o no lineales. En situaciones donde la tendencia es m√°s compleja, es posible que sea necesario utilizar m√©todos m√°s avanzados.

## **M√©todo Ingenuo Estacional**

El M√©todo Ingenuo Estacional, tambi√©n conocido como m√©todo de descomposici√≥n aditiva estacional, es una t√©cnica simple de pron√≥stico en el an√°lisis de series de tiempo que tiene en cuenta las estacionalidades en los datos. Este m√©todo asume que el pr√≥ximo valor en la serie de tiempo ser√° igual al valor observado en el mismo per√≠odo en el ciclo estacional anterior. Es especialmente √∫til cuando los datos tienen patrones estacionales consistentes.

A continuaci√≥n, te proporcionar√© una explicaci√≥n detallada, un ejemplo y las respectivas gr√°ficas para ilustrar el M√©todo Ingenuo Estacional utilizando Python y las bibliotecas Pandas y Matplotlib.


**Ejemplo de M√©todo Ingenuo Estacional:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de ejemplo con estacionalidad
np.random.seed(0)
dates = pd.date_range(start='2021-01-01', periods=100, freq='D')
seasonality = np.sin(np.linspace(0, 4 * np.pi, 100))
values = 3 * seasonality + np.random.normal(0, 0.5, 100)

# Crear un DataFrame de Pandas
data = pd.DataFrame({'Date': dates, 'Values': values})
data.set_index('Date', inplace=True)

# Calcular el pron√≥stico ingenuo estacional
seasonal_forecast = data['Values'].shift(1)

# Graficar la serie original y el pron√≥stico ingenuo estacional
plt.figure(figsize=(10, 6))
plt.plot(data['Values'], label='Serie Original')
plt.plot(seasonal_forecast, label='Pron√≥stico Ingenuo Estacional')
plt.legend()
plt.title('M√©todo Ingenuo Estacional en Series de Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```

En este ejemplo, hemos creado una serie de tiempo que incluye estacionalidad sinusoidal y ruido. El pron√≥stico ingenuo estacional asume que el pr√≥ximo valor ser√° igual al valor en el mismo per√≠odo en el ciclo estacional anterior. La gr√°fica muestra la serie original y el pron√≥stico ingenuo estacional.

**Interpretaci√≥n:**

- La l√≠nea azul representa la serie de tiempo original con estacionalidad y ruido.
- La l√≠nea naranja representa el pron√≥stico ingenuo estacional, donde cada valor es el mismo que el valor en el mismo per√≠odo del ciclo estacional anterior.

El m√©todo ingenuo estacional es especialmente √∫til cuando los datos tienen patrones estacionales consistentes, como en este ejemplo. Sin embargo, este m√©todo no captura patrones de tendencia o cambios en el ciclo estacional a lo largo del tiempo. Para datos m√°s complejos, es posible que necesites utilizar m√©todos m√°s avanzados que consideren tendencias y estacionalidades cambiantes.

**2. Pron√≥stico de la Media M√≥vil:**

El pron√≥stico de la Media M√≥vil calcula el promedio de los valores pasados en un intervalo de tiempo definido y lo utiliza como pron√≥stico para valores futuros.

**Ejemplo de Pron√≥stico de la Media M√≥vil:**

```python
# Calcular el pron√≥stico de la Media M√≥vil con ventana de tama√±o 5
rolling_mean_forecast = data['Values'].rolling(window=5).mean()

# Graficar la serie original y el pron√≥stico de la Media M√≥vil
plt.figure(figsize=(10, 6))
plt.plot(data['Values'], label='Serie Original')
plt.plot(rolling_mean_forecast, label='Pron√≥stico de la Media M√≥vil (Ventana 5)')
plt.legend()
plt.title('Pron√≥stico de la Media M√≥vil en Series de Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```

**3. Pron√≥stico de la Deriva (Tendencia Lineal):**

El pron√≥stico de la Deriva asume una tendencia lineal en los datos y proyecta esta tendencia hacia el futuro.

**Ejemplo de Pron√≥stico de la Deriva:**

```python
# Calcular la tendencia lineal utilizando regresi√≥n
coefficients = np.polyfit(np.arange(len(data)), data['Values'], 1)
trend_line_forecast = np.polyval(coefficients, np.arange(len(data)))

# Graficar la serie original y el pron√≥stico de la Deriva
plt.figure(figsize=(10, 6))
plt.plot(data['Values'], label='Serie Original')
plt.plot(trend_line_forecast, label='Pron√≥stico de la Deriva (Tendencia Lineal)')
plt.legend()
plt.title('Pron√≥stico de la Deriva en Series de Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```
