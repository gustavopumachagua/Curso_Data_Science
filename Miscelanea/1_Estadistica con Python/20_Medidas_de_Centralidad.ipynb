# **20. Mejora tu Análisis de Datos con Medidas de Centralidad para Datos Categóricos usando Python**

## **Introducción**

Para mejorar el análisis de datos categóricos, es fundamental utilizar medidas de centralidad específicas para este tipo de datos. A diferencia de los datos numéricos, los datos categóricos no se pueden analizar directamente con medidas como la media o la mediana. En su lugar, se utilizan medidas como la moda y la frecuencia relativa para comprender mejor la distribución de categorías en los datos. A continuación, te proporcionaré una explicación detallada junto con ejemplos y gráficos para ilustrar cómo realizar un análisis de datos categóricos utilizando Python.

**Moda para Datos Categóricos:**

La moda es la categoría que aparece con mayor frecuencia en un conjunto de datos categóricos. En otras palabras, es la categoría más común entre los datos. Calcular la moda te permite identificar la categoría dominante en tus datos categóricos.

**Ejemplo:**

Supongamos que tienes un conjunto de datos que registra las preferencias de frutas de un grupo de personas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame de ejemplo
data = {
    'Nombre': ['Ana', 'Juan', 'María', 'Carlos', 'Luis', 'Laura', 'David', 'Sofía', 'Pedro', 'Elena'],
    'Fruta_Favorita': ['Manzana', 'Plátano', 'Manzana', 'Uva', 'Plátano', 'Manzana', 'Uva', 'Manzana', 'Plátano', 'Uva']
}

df = pd.DataFrame(data)

# Calcular la moda de las frutas favoritas
moda = df['Fruta_Favorita'].mode()[0]

print("Moda:", moda)
```

**Creación de Gráficos:**

Podemos visualizar la distribución de las frutas favoritas utilizando un gráfico de barras.

```python

# Calcular la frecuencia de las frutas
fruta_counts = df['Fruta_Favorita'].value_counts()

# Crear un gráfico de barras para las frutas favoritas
plt.bar(fruta_counts.index, fruta_counts.values, color='purple', alpha=0.7)
plt.xlabel('Fruta Favorita')
plt.ylabel('Frecuencia')
plt.title('Distribución de Frutas Favoritas')
plt.show()
```

**Interpretación del Gráfico:**

En este gráfico de barras, se muestra la distribución de frutas favoritas entre las personas en el conjunto de datos. Puedes identificar visualmente la fruta más común.

**Frecuencia Relativa para Datos Categóricos:**

La frecuencia relativa es la proporción de veces que una categoría específica aparece en relación con el total de observaciones. Te ayuda a comprender la proporción de cada categoría en tus datos.

**Ejemplo:**

Utilizando el mismo conjunto de datos de frutas favoritas:

```python
# Calcular la frecuencia relativa de las frutas favoritas
frecuencia_relativa = df['Fruta_Favorita'].value_counts(normalize=True)

print("Frecuencia Relativa:\n", frecuencia_relativa)
```

**Creación de Gráficos:**

Podemos visualizar la frecuencia relativa de las frutas favoritas utilizando un gráfico de pastel.

```python

# Crear un gráfico de pastel para la frecuencia relativa de las frutas favoritas
plt.pie(frecuencia_relativa, labels=frecuencia_relativa.index, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title('Frecuencia Relativa de Frutas Favoritas')
plt.axis('equal')  # Para que el gráfico de pastel sea un círculo
plt.show()
```

**Interpretación del Gráfico:**

En este gráfico de pastel, se muestra la proporción de cada fruta favorita en relación con el total de observaciones. Cada porción del pastel representa la frecuencia relativa de una fruta.

**Usos y Consideraciones:**

- La moda es útil para identificar la categoría más común en datos categóricos.
- La frecuencia relativa te ayuda a comprender la distribución proporcional de categorías en tus datos.
- Los gráficos de barras y pastel son excelentes para visualizar la distribución de categorías y frecuencias relativas.

En resumen, al analizar datos categóricos, es esencial utilizar medidas de centralidad adecuadas como la moda y la frecuencia relativa. Estas medidas te ayudan a comprender mejor la distribución de categorías en tus datos y a tomar decisiones informadas basadas en la información categórica.

## **Categóricos Nominales y Ordinales**

Los datos categóricos se refieren a variables que representan diferentes categorías o grupos, pero que no tienen un valor numérico que pueda ser interpretado como una cantidad. Los datos categóricos se dividen en dos tipos principales: categóricos nominales y categóricos ordinales. Estos dos tipos de datos tienen características y propiedades distintas, y se utilizan en diferentes contextos según la naturaleza de la variable que representan.

**Datos Categóricos Nominales:**

Los datos categóricos nominales son categorías que no tienen un orden inherente entre ellas. En otras palabras, estas categorías son mutuamente exclusivas y no se pueden clasificar en un orden específico. Los ejemplos de datos categóricos nominales pueden ser colores, géneros, tipos de frutas, estados civiles, etc.

**Ejemplo:**

Supongamos que tenemos datos sobre los colores favoritos de un grupo de personas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame de ejemplo
data = {
    'Nombre': ['Ana', 'Juan', 'María', 'Carlos', 'Luis', 'Laura'],
    'Color_Favorito': ['Azul', 'Rojo', 'Verde', 'Azul', 'Rojo', 'Verde']
}

df = pd.DataFrame(data)

# Calcular la frecuencia de colores favoritos
color_counts = df['Color_Favorito'].value_counts()

# Crear un gráfico de barras para los colores favoritos
plt.bar(color_counts.index, color_counts.values, color='blue')
plt.xlabel('Color Favorito')
plt.ylabel('Frecuencia')
plt.title('Distribución de Colores Favoritos')
plt.show()
```

En este ejemplo, los colores favoritos son datos categóricos nominales. No existe un orden específico entre los colores; simplemente son categorías distintas.

**Datos Categóricos Ordinales:**

Los datos categóricos ordinales también representan categorías, pero en este caso, hay un orden significativo entre las categorías. Las categorías ordinales pueden ser clasificadas en un orden específico, pero la distancia entre las categorías no es necesariamente uniforme ni cuantificable. Los ejemplos de datos categóricos ordinales pueden ser niveles de educación, clasificaciones de satisfacción (como "bajo", "medio" y "alto"), puntuaciones de clasificación, etc.

**Ejemplo:**

Supongamos que tenemos datos sobre niveles de educación de un grupo de personas.

```python
# Crear un DataFrame de ejemplo
data = {
    'Nombre': ['Ana', 'Juan', 'María', 'Carlos', 'Luis', 'Laura'],
    'Educacion': ['Licenciatura', 'Maestría', 'Doctorado', 'Licenciatura', 'Bachillerato', 'Maestría']
}

df = pd.DataFrame(data)

# Orden de niveles de educación
orden_educacion = ['Bachillerato', 'Licenciatura', 'Maestría', 'Doctorado']

# Calcular la frecuencia de niveles de educación
educacion_counts = df['Educacion'].value_counts()

# Crear un gráfico de barras para los niveles de educación
plt.bar(educacion_counts.index, educacion_counts.values, color='green')
plt.xlabel('Nivel de Educación')
plt.ylabel('Frecuencia')
plt.title('Distribución de Niveles de Educación')
plt.xticks(orden_educacion)  # Mostrar las categorías en el orden correcto en el eje x
plt.show()
```

En este ejemplo, los niveles de educación son datos categóricos ordinales. Hay un orden específico entre los niveles de educación, lo que permite clasificarlos de menor a mayor.

**Usos y Consideraciones:**

- Los datos categóricos nominales se utilizan para representar categorías sin un orden específico, mientras que los datos categóricos ordinales se utilizan cuando existe un orden significativo entre las categorías.
- Las medidas de centralidad y visualizaciones para datos categóricos varían según si son nominales u ordinales. Por ejemplo, la moda y la frecuencia relativa son especialmente útiles para datos categóricos nominales, mientras que las gráficas de barras y de puntos son relevantes para ambos tipos de datos.
- Es importante tener en cuenta la naturaleza de los datos categóricos al realizar análisis y tomar decisiones basadas en ellos.

En resumen, los datos categóricos nominales y ordinales son tipos de variables utilizadas para representar diferentes categorías o grupos. Comprender la diferencia entre estos dos tipos de datos es fundamental para realizar análisis precisos y tomar decisiones informadas.

## **Medidas para Nominales**

Las medidas para datos categóricos nominales son herramientas estadísticas que nos permiten resumir y analizar la distribución de categorías en un conjunto de datos. Dado que los datos categóricos nominales representan categorías sin un orden específico, las medidas utilizadas en este contexto son diferentes de las que se aplican a datos numéricos. A continuación, te proporcionaré una explicación detallada junto con ejemplos y gráficos para ilustrar las medidas comunes utilizadas para datos categóricos nominales.

**Moda para Datos Categóricos Nominales:**

La moda es la categoría que aparece con mayor frecuencia en un conjunto de datos categóricos nominales. Calculamos la moda para identificar la categoría más común entre las observaciones.

**Ejemplo:**

Supongamos que tenemos un conjunto de datos que registra las preferencias de colores de un grupo de personas.

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'Nombre': ['Ana', 'Juan', 'María', 'Carlos', 'Luis', 'Laura', 'David', 'Sofía', 'Pedro', 'Elena'],
    'Color_Favorito': ['Azul', 'Rojo', 'Verde', 'Azul', 'Rojo', 'Verde', 'Azul', 'Verde', 'Rojo', 'Verde']
}

df = pd.DataFrame(data)

# Calcular la moda de los colores favoritos
moda = df['Color_Favorito'].mode()[0]

print("Moda:", moda)
```

**Creación de Gráficos:**

Para visualizar la distribución de colores favoritos, podemos crear un gráfico de barras.

```python
import matplotlib.pyplot as plt

# Calcular la frecuencia de colores favoritos
color_counts = df['Color_Favorito'].value_counts()

# Crear un gráfico de barras para los colores favoritos
plt.bar(color_counts.index, color_counts.values, color='purple', alpha=0.7)
plt.xlabel('Color Favorito')
plt.ylabel('Frecuencia')
plt.title('Distribución de Colores Favoritos')
plt.xticks(rotation=45)
plt.show()
```

**Interpretación del Gráfico:**

En el gráfico de barras, se muestra la distribución de colores favoritos entre las personas del conjunto de datos. La categoría "Verde" es la moda, ya que es la más común entre las preferencias de color.

**Frecuencia Relativa para Datos Categóricos Nominales:**

La frecuencia relativa es la proporción de veces que una categoría específica aparece en relación con el total de observaciones. Calcular la frecuencia relativa nos ayuda a comprender la distribución proporcional de las categorías en los datos categóricos nominales.

**Ejemplo:**

Utilizando el mismo conjunto de datos de colores favoritos:

```python
# Calcular la frecuencia relativa de colores favoritos
frecuencia_relativa = df['Color_Favorito'].value_counts(normalize=True)

print("Frecuencia Relativa:\n", frecuencia_relativa)
```

**Creación de Gráficos:**

Podemos visualizar la frecuencia relativa de colores favoritos utilizando un gráfico de pastel.

```python
# Crear un gráfico de pastel para la frecuencia relativa de colores favoritos
plt.pie(frecuencia_relativa, labels=frecuencia_relativa.index, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title('Frecuencia Relativa de Colores Favoritos')
plt.axis('equal')
plt.show()
```

**Interpretación del Gráfico:**

En el gráfico de pastel, se muestra la proporción de cada color favorito en relación con el total de observaciones. Cada porción del pastel representa la frecuencia relativa de un color.

**Usos y Consideraciones:**

- La moda y la frecuencia relativa son medidas útiles para resumir la distribución de categorías en datos categóricos nominales.
- Los gráficos de barras y de pastel son herramientas visuales efectivas para mostrar la distribución de categorías y frecuencias relativas.
- Estas medidas son especialmente relevantes cuando se trabaja con características categóricas en análisis de datos y toma de decisiones.

En resumen, las medidas para datos categóricos nominales, como la moda y la frecuencia relativa, nos ayudan a comprender la distribución de categorías en un conjunto de datos y a tomar decisiones informadas basadas en la información categórica.

## **Medidas para Ordinales**

Las medidas para datos categóricos ordinales son herramientas estadísticas que nos permiten resumir y analizar la distribución de categorías que tienen un orden específico entre ellas. A diferencia de los datos categóricos nominales, en los datos categóricos ordinales, las categorías pueden ser clasificadas en un orden significativo, pero no necesariamente se puede cuantificar la distancia entre las categorías. A continuación, te proporcionaré una explicación detallada junto con ejemplos y gráficos para ilustrar las medidas comunes utilizadas para datos categóricos ordinales.

**Ordenamiento y Etiquetas Ordinales:**

Antes de analizar los datos ordinales, es importante establecer un orden claro entre las categorías. Las etiquetas ordinales deben ser significativas y reflejar la jerarquía de las categorías. Por ejemplo, en una escala de satisfacción, podríamos tener etiquetas "Bajo", "Medio" y "Alto" para representar diferentes niveles de satisfacción.

**Frecuencia Ordinal:**

La frecuencia ordinal es el recuento de observaciones en cada categoría ordinal. Esta medida nos proporciona una idea de la distribución de las categorías en los datos.

**Ejemplo:**

Supongamos que tenemos un conjunto de datos que registra el nivel de satisfacción de los clientes en una escala ordinal.

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'Cliente': ['Ana', 'Juan', 'María', 'Carlos', 'Luis', 'Laura'],
    'Satisfaccion': ['Alto', 'Medio', 'Bajo', 'Medio', 'Alto', 'Medio']
}

df = pd.DataFrame(data)

# Calcular la frecuencia ordinal de satisfacción
frecuencia_ordinal = df['Satisfaccion'].value_counts()

print("Frecuencia Ordinal:\n", frecuencia_ordinal)
```

**Creación de Gráficos:**

Podemos visualizar la frecuencia ordinal de satisfacción utilizando un gráfico de barras.

```python
import matplotlib.pyplot as plt

# Crear un gráfico de barras para la frecuencia ordinal de satisfacción
plt.bar(frecuencia_ordinal.index, frecuencia_ordinal.values, color='green')
plt.xlabel('Nivel de Satisfacción')
plt.ylabel('Frecuencia')
plt.title('Distribución de Niveles de Satisfacción')
plt.show()
```

**Interpretación del Gráfico:**

En el gráfico de barras, se muestra la distribución de los niveles de satisfacción entre los clientes. Puedes identificar visualmente el nivel de satisfacción más común.

**Orden Mediano (Mediana Ordinal):**

La mediana ordinal es el valor que divide el conjunto de datos ordinales en dos partes iguales. Es el valor que se encuentra en medio de las observaciones cuando están ordenadas. Calcular la mediana ordinal es útil para entender la posición central de los datos ordinales.

**Ejemplo:**

Utilizando el mismo conjunto de datos de satisfacción:

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'Persona': ['Ana', 'Juan', 'María', 'Carlos', 'Luis', 'Laura'],
    'Nivel_Educacion': ['Licenciatura', 'Maestría', 'Bachillerato', 'Licenciatura', 'Doctorado', 'Maestría']
}

df = pd.DataFrame(data)

# Calcular la moda del nivel de educación
moda_educacion = df['Nivel_Educacion'].mode()[0]

print("Moda del Nivel de Educación:", moda_educacion)
```

```python
import matplotlib.pyplot as plt

# Calcular la frecuencia de niveles de educación
educacion_counts = df['Nivel_Educacion'].value_counts()

# Crear un gráfico de barras para los niveles de educación
plt.bar(educacion_counts.index, educacion_counts.values, color='blue', alpha=0.7)
plt.xlabel('Nivel de Educación')
plt.ylabel('Frecuencia')
plt.title('Distribución de Niveles de Educación')
plt.xticks(rotation=45)
plt.show()
```

**Usos y Consideraciones:**

- Las medidas para datos categóricos ordinales, como la frecuencia ordinal y la mediana ordinal, nos permiten resumir la distribución y la posición central de las categorías en un conjunto de datos.
- Los gráficos de barras son útiles para visualizar la distribución de categorías ordinales y la frecuencia de cada categoría.
- Estas medidas son relevantes cuando trabajamos con características ordinales en análisis de datos y toma de decisiones.

En resumen, las medidas para datos categóricos ordinales, como la frecuencia ordinal y la mediana ordinal, nos ayudan a comprender la distribución y la posición central de las categorías en un conjunto de datos. Estas medidas son esenciales para analizar y tomar decisiones basadas en datos con ordenamiento jerárquico.

## **Centralidad en Categóricos con Python**

La centralidad en datos categóricos se refiere a las medidas que resumen la ubicación central o típica de las categorías en un conjunto de datos categóricos. A diferencia de los datos numéricos, donde las medidas de centralidad como la media y la mediana tienen sentido, en los datos categóricos se utilizan otras medidas para resumir la ubicación central. A continuación, te proporcionaré una explicación detallada junto con ejemplos y gráficos para ilustrar las medidas de centralidad en datos categóricos utilizando Python.

**Moda en Datos Categóricos:**

La moda es la categoría que aparece con mayor frecuencia en un conjunto de datos categóricos. Es una medida de centralidad que identifica la categoría más común entre las observaciones.

**Ejemplo:**

Supongamos que tenemos un conjunto de datos que registra la preferencia de género musical de un grupo de personas.

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'Persona': ['Ana', 'Juan', 'María', 'Carlos', 'Luis', 'Laura'],
    'Genero_Musical': ['Pop', 'Rock', 'Pop', 'Reggaeton', 'Rock', 'Pop']
}

df = pd.DataFrame(data)

# Calcular la moda del género musical
moda_genero = df['Genero_Musical'].mode()[0]

print("Moda del Género Musical:", moda_genero)
```

**Creación de Gráficos:**

Podemos visualizar la frecuencia de los géneros musicales utilizando un gráfico de barras.

```python
import matplotlib.pyplot as plt

# Calcular la frecuencia de géneros musicales
genero_counts = df['Genero_Musical'].value_counts()

# Crear un gráfico de barras para los géneros musicales
plt.bar(genero_counts.index, genero_counts.values, color='green', alpha=0.7)
plt.xlabel('Género Musical')
plt.ylabel('Frecuencia')
plt.title('Distribución de Géneros Musicales')
plt.xticks(rotation=45)
plt.show()
```

**Interpretación del Gráfico:**

En el gráfico de barras, se muestra la distribución de los géneros musicales entre las personas del conjunto de datos. La categoría con la barra más alta es la moda y representa el género musical más común.

**Consideraciones:**

- La moda es una medida común de centralidad para datos categóricos, ya que ayuda a identificar la categoría más representativa.
- Los gráficos de barras son herramientas visuales útiles para resumir y comparar la frecuencia de diferentes categorías.

**Conclusiones:**

En resumen, la moda es una medida de centralidad útil para resumir datos categóricos, y los gráficos de barras nos permiten visualizar la distribución de las categorías en el conjunto de datos. Estas herramientas son esenciales para analizar y comunicar la ubicación central de las categorías en un contexto categórico.
