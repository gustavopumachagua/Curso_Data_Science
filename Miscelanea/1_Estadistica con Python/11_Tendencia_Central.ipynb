# **11. Las 3 Medidas de Tendencia Central m√°s Utilizadas: Promedio (o Media), Mediana y Moda con Python**

## **Introducci√≥n a la medidas de tendencia central**

Las tres medidas de tendencia central m√°s utilizadas son el promedio (o media), la mediana y la moda. Estas medidas nos ayudan a resumir y comprender la distribuci√≥n de datos en un conjunto de observaciones. A continuaci√≥n, explicar√© cada una de estas medidas con ejemplos y sus respectivas gr√°ficas en Python.

1. **Promedio (o Media):**

El promedio es una medida de tendencia central que se calcula sumando todos los valores en el conjunto de datos y dividi√©ndolos entre el n√∫mero total de observaciones. Es la suma de los valores dividida por el n√∫mero de elementos en el conjunto de datos.

**Ejemplo de Promedio con Python:**

Supongamos que tenemos el siguiente conjunto de datos que representa las edades de un grupo de personas:

````python
edades = [25, 30, 22, 28, 35, 27, 31, 29, 26, 24]
````

Para calcular el promedio en Python, podemos utilizar la funci√≥n `mean()` de la librer√≠a NumPy:

````python
import numpy as np

promedio = np.mean(edades)

print("Promedio:", promedio)
````

**Gr√°fica:**

Podemos visualizar la distribuci√≥n de edades con un histograma para tener una idea de c√≥mo se agrupan los datos alrededor del promedio:

````python
import matplotlib.pyplot as plt

plt.hist(edades, bins=5, edgecolor='black')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades')
plt.axvline(promedio, color='red', linestyle='dashed', linewidth=2, label='Promedio')
plt.legend()
plt.show()
````

**Resultado:**

La gr√°fica mostrar√° un histograma de las edades con una l√≠nea vertical roja punteada en el valor del promedio (28.7). Esta l√≠nea representa el promedio y nos indica c√≥mo se agrupan los datos alrededor de este valor.

2. **Mediana:**

La mediana es la medida de tendencia central que se encuentra en el centro de un conjunto de datos ordenados. Para calcular la mediana, es necesario ordenar los datos de manera ascendente o descendente y luego encontrar el valor que se encuentra en la posici√≥n central. Si hay un n√∫mero impar de observaciones, la mediana ser√° el valor en la posici√≥n central. Si hay un n√∫mero par de observaciones, la mediana ser√° el promedio de los dos valores centrales.

**Ejemplo de Mediana con Python:**

Continuando con el ejemplo anterior, vamos a calcular la mediana de las edades utilizando la funci√≥n `median()` de NumPy:

````python
mediana = np.median(edades)

print("Mediana:", mediana)
````

**Gr√°fica:**

Podemos visualizar la mediana en el histograma anterior agregando una l√≠nea vertical verde punteada:

````python
plt.hist(edades, bins=5, edgecolor='black')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades')
plt.axvline(promedio, color='red', linestyle='dashed', linewidth=2, label='Promedio')
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label='Mediana')
plt.legend()
plt.show()
````

**Resultado:**

La gr√°fica mostrar√° el histograma de las edades con dos l√≠neas verticales punteadas: una roja para el promedio y una verde para la mediana. Esto nos permite comparar la ubicaci√≥n del promedio y la mediana en la distribuci√≥n de datos.

3. **Moda:**

La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. Es posible que un conjunto de datos tenga una moda (unimodal), dos modas (bimodal) o m√°s modas (multimodal). Tambi√©n es posible que un conjunto de datos no tenga una moda, si todos los valores son diferentes y no hay repetici√≥n.

**Ejemplo de Moda con Python:**

Continuando con el mismo ejemplo, vamos a calcular la moda de las edades utilizando la funci√≥n `mode()` de la librer√≠a SciPy:

````python
from collections import Counter

edades = [25, 30, 22, 28, 35, 27, 31, 29, 26, 24]

# Calcular la moda
moda_counter = Counter(edades)
moda = moda_counter.most_common(1)[0][0]
frecuencia_moda = moda_counter.most_common(1)[0][1]

print("Moda:", moda)
print("Frecuencia de la Moda:", frecuencia_moda)
````

**Gr√°fica:**

Podemos visualizar la moda en el histograma anterior agregando una l√≠nea vertical azul punteada:

````python
plt.hist(edades, bins=5, edgecolor='black')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades')
plt.axvline(promedio, color='red', linestyle='dashed', linewidth=2, label='Promedio')
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label='Mediana')
plt.axvline(moda, color='blue', linestyle='dashed', linewidth=2, label='Moda')
plt.legend()
plt.show()
````

**Resultado:**

La gr√°fica mostrar√° el histograma de las edades con tres l√≠neas verticales punteadas: una roja para el promedio, una verde para la mediana y una azul para la moda. Esto nos permite identificar visualmente el valor que aparece con mayor frecuencia en la distribuci√≥n de datos.

En resumen, el promedio, la mediana y la moda son tres medidas de tendencia central utilizadas para resumir y comprender la distribuci√≥n de datos en un conjunto de observaciones. Cada una de estas medidas tiene sus propias ventajas y limitaciones, y es importante utilizarlas en funci√≥n del tipo de datos y el objetivo del an√°lisis.

## **Promedio o media aritm√©tica**

La media aritm√©tica, com√∫nmente conocida como promedio, es una medida estad√≠stica que se utiliza para representar el valor t√≠pico o central de un conjunto de datos num√©ricos. Se calcula sumando todos los valores en el conjunto y luego dividiendo esa suma entre el n√∫mero total de elementos en el conjunto.

La f√≥rmula para calcular el promedio de un conjunto de datos de tama√±o $n$ es:

$\text{Promedio} = \frac{X_1 + X_2 + \ldots + X_n}{n}$

Donde $X_1$, $X_2$, $\ldots$, $X_n$ son los elementos del conjunto de datos y $n$ es el n√∫mero total de elementos.

Veamos un ejemplo para calcular el promedio de las siguientes notas: 80, 85, 90, 70, 95.

Paso 1: Sumar todas las notas.

$\text{Suma} = 80 + 85 + 90 + 70 + 95 = 420$

Paso 2: Calcular el promedio.

$\text{Promedio} = \frac{420}{5} = 84$

Por lo tanto, el promedio de estas notas es 84.

Ahora, vamos a visualizar esta informaci√≥n en una gr√°fica de barras usando Python. Utilizaremos la biblioteca `matplotlib` para graficar los datos.

````python
import matplotlib.pyplot as plt

# Datos
notas = [80, 85, 90, 70, 95]
n = len(notas)

# C√°lculo del promedio
suma = sum(notas)
promedio = suma / n

# Gr√°fica de barras
plt.bar(range(n), notas, tick_label=["Nota 1", "Nota 2", "Nota 3", "Nota 4", "Nota 5"])
plt.axhline(y=promedio, color='r', linestyle='dashed', label='Promedio')
plt.xlabel('Notas')
plt.ylabel('Valor')
plt.title('Notas y su Promedio')
plt.legend()
plt.show()
````

En esta gr√°fica, las barras representan las notas individuales, y la l√≠nea roja punteada representa el promedio (84) de las notas. La l√≠nea roja nos muestra el valor t√≠pico o central del conjunto de datos.

Recuerda que el promedio es solo una medida estad√≠stica b√°sica y puede no reflejar la distribuci√≥n completa de los datos.

 Por ejemplo, si tenemos un conjunto de datos {1, 2, 3, 1000}, el promedio ser√≠a (1 + 2 + 3 + 1000) / 4 = 251.5, que no representa bien la mayor√≠a de los datos. En tales casos, otras medidas como la mediana o la desviaci√≥n est√°ndar pueden proporcionar una mejor comprensi√≥n de la distribuci√≥n.

## **Mediana**

La mediana es otra medida estad√≠stica utilizada para representar el valor central de un conjunto de datos num√©ricos. A diferencia del promedio, la mediana no se ve afectada por valores extremos o at√≠picos en los datos y, por lo tanto, proporciona una mejor representaci√≥n del valor t√≠pico cuando hay valores at√≠picos presentes.

Para calcular la mediana, primero se deben ordenar los datos en orden ascendente o descendente, y luego se selecciona el valor que se encuentra en la posici√≥n central del conjunto de datos. Si el conjunto de datos tiene un n√∫mero impar de elementos, la mediana ser√° el valor que est√° justo en el medio. Si el conjunto de datos tiene un n√∫mero par de elementos, la mediana se calcula promediando los dos valores centrales.

Veamos un ejemplo para calcular la mediana del siguiente conjunto de datos: 10, 20, 30, 40, 50.

**Paso 1:** Ordenar los datos en orden ascendente: 10, 20, 30, 40, 50.

**Paso 2:** Identificar el valor central, que en este caso es el tercer valor, que es 30.

Por lo tanto, la mediana de este conjunto de datos es 30.

Ahora, vamos a visualizar esta informaci√≥n en una gr√°fica de barras usando Python. Utilizaremos la biblioteca matplotlib para graficar los datos.

````python
import matplotlib.pyplot as plt

# Datos
datos = [10, 20, 30, 40, 50]

# C√°lculo de la mediana
n = len(datos)
datos_ordenados = sorted(datos)
mediana = datos_ordenados[n // 2] if n % 2 != 0 else (datos_ordenados[n // 2 - 1] + datos_ordenados[n // 2]) / 2

# Gr√°fica de barras
plt.bar(range(n), datos, tick_label=["Dato 1", "Dato 2", "Dato 3", "Dato 4", "Dato 5"])
plt.axhline(y=mediana, color='r', linestyle='dashed', label='Mediana')
plt.xlabel('Datos')
plt.ylabel('Valor')
plt.title('Datos y su Mediana')
plt.legend()
plt.show()
````

En esta gr√°fica, las barras representan los datos individuales, y la l√≠nea roja punteada representa la mediana (30) del conjunto de datos. La mediana nos proporciona una mejor idea del valor central de los datos, ya que no se ve afectada por los valores extremos, como el valor 10 que es m√°s peque√±o y el valor 50 que es m√°s grande.

La mediana es especialmente √∫til cuando tenemos datos sesgados o con valores at√≠picos, ya que no se ve influenciada por esos valores y ofrece una perspectiva m√°s robusta del valor central del conjunto de datos.

## **Moda**

La moda es otra medida estad√≠stica utilizada para representar el valor m√°s frecuente o repetido en un conjunto de datos num√©ricos o categ√≥ricos. En otras palabras, la moda es el valor que aparece con mayor frecuencia en el conjunto de datos.

En un conjunto de datos, puede haber uno o m√°s valores que se repiten con mayor frecuencia. Si un conjunto de datos tiene un solo valor que aparece con mayor frecuencia, se dice que tiene una "moda unimodal". Si hay dos o m√°s valores que tienen la misma frecuencia m√°xima y son m√°s frecuentes que los dem√°s valores, se dice que tiene una "moda bimodal" o "multimodal".

Veamos un ejemplo para calcular la moda del siguiente conjunto de datos: 10, 20, 30, 40, 30, 50, 30.

**Paso 1:** Identificar los valores que se repiten y contar su frecuencia:

- 10 aparece 1 vez.
- 20 aparece 1 vez.
- 30 aparece 3 veces.
- 40 aparece 1 vez.
- 50 aparece 1 vez.

**Paso 2:** Identificar el valor con la mayor frecuencia, que es 30. Por lo tanto, la moda de este conjunto de datos es 30.

Ahora, vamos a visualizar esta informaci√≥n en una gr√°fica de barras usando Python. Utilizaremos la biblioteca `matplotlib` para graficar los datos.

````python
import matplotlib.pyplot as plt
from collections import Counter

# Datos
datos = [10, 20, 30, 40, 30, 50, 30]

# C√°lculo de la moda
frecuencias = Counter(datos)
moda = max(frecuencias, key=frecuencias.get)

# Gr√°fica de barras
plt.bar(frecuencias.keys(), frecuencias.values())
plt.axhline(y=frecuencias[moda], color='r', linestyle='dashed', label='Moda')
plt.xlabel('Datos')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Datos y su Moda')
plt.legend()
plt.show()
````

En esta gr√°fica, las barras representan los datos individuales, y la l√≠nea roja punteada representa la moda (30) del conjunto de datos. Podemos observar que el valor 30 es el que tiene la frecuencia m√°s alta y, por lo tanto, es la moda del conjunto de datos.

La moda es √∫til para identificar cu√°l es el valor m√°s representativo o frecuente en un conjunto de datos. Puede ser especialmente relevante en datos categ√≥ricos, como tipos de frutas o colores, donde la moda nos dir√° cu√°l es la opci√≥n m√°s popular. Adem√°s, tambi√©n puede ser √∫til en datos num√©ricos para identificar los valores m√°s comunes y frecuentes en una distribuci√≥n.

## **Medidas de tendencia central con Python**

Las medidas de tendencia central son estad√≠sticas descriptivas que nos permiten resumir y representar un conjunto de datos en un solo valor que representa su "centro". Las tres medidas de tendencia central m√°s comunes son el promedio (o media aritm√©tica), la mediana y la moda, que ya hemos explicado individualmente en preguntas anteriores. Ahora, vamos a proporcionar una explicaci√≥n detallada de cada una junto con ejemplos y gr√°ficas en Python para cada medida.

**1. Promedio (Media Aritm√©tica):**

Ya hemos explicado el concepto y el c√°lculo del promedio en la primera pregunta. Repasemos r√°pidamente:

El promedio es la suma de todos los valores en un conjunto de datos dividida por el n√∫mero total de elementos.

**Ejemplo:**

Consideremos el siguiente conjunto de datos: 5, 10, 15, 20, 25.

Paso 1: Sumar todos los valores: 5 + 10 + 15 + 20 + 25 = 75.

Paso 2: Calcular el promedio: 75 / 5 = 15.

````python
import matplotlib.pyplot as plt

# Datos
datos = [5, 10, 15, 20, 25]

# C√°lculo del promedio
promedio = sum(datos) / len(datos)

# Gr√°fica de barras
plt.bar(range(len(datos)), datos, tick_label=["Dato 1", "Dato 2", "Dato 3", "Dato 4", "Dato 5"])
plt.axhline(y=promedio, color='r', linestyle='dashed', label='Promedio')
plt.xlabel('Datos')
plt.ylabel('Valor')
plt.title('Datos y su Promedio')
plt.legend()
plt.show()
````

**2. Mediana:**
Ya hemos explicado el concepto y el c√°lculo de la mediana en la segunda pregunta. Repasemos r√°pidamente:

La mediana es el valor central de un conjunto de datos una vez que los datos se han ordenado.

**Ejemplo:**

Consideremos el siguiente conjunto de datos: 10, 20, 30, 40, 50.

Paso 1: Ordenar los datos en orden ascendente: 10, 20, 30, 40, 50.

Paso 2: Identificar el valor central, que en este caso es el tercer valor, que es 30.

````python
import matplotlib.pyplot as plt

# Datos
datos = [10, 20, 30, 40, 50]

# C√°lculo de la mediana
n = len(datos)
datos_ordenados = sorted(datos)
mediana = datos_ordenados[n // 2] if n % 2 != 0 else (datos_ordenados[n // 2 - 1] + datos_ordenados[n // 2]) / 2

# Gr√°fica de barras
plt.bar(range(n), datos, tick_label=["Dato 1", "Dato 2", "Dato 3", "Dato 4", "Dato 5"])
plt.axhline(y=mediana, color='r', linestyle='dashed', label='Mediana')
plt.xlabel('Datos')
plt.ylabel('Valor')
plt.title('Datos y su Mediana')
plt.legend()
plt.show()
````

**3. Moda:**

Ya hemos explicado el concepto y el c√°lculo de la moda en la tercera pregunta. Repasemos r√°pidamente:

La moda es el valor m√°s frecuente en un conjunto de datos.

**Ejemplo:**

Consideremos el siguiente conjunto de datos: 10, 20, 30, 40, 30, 50, 30.

Paso 1: Identificar los valores que se repiten y contar su frecuencia:

- 10 aparece 1 vez.
- 20 aparece 1 vez.
- 30 aparece 3 veces.
- 40 aparece 1 vez.
- 50 aparece 1 vez.

Paso 2: Identificar el valor con la mayor frecuencia, que es 30.

````python
import matplotlib.pyplot as plt
from collections import Counter

# Datos
datos = [10, 20, 30, 40, 30, 50, 30]

# C√°lculo de la moda
frecuencias = Counter(datos)
moda = max(frecuencias, key=frecuencias.get)

# Gr√°fica de barras
plt.bar(frecuencias.keys(), frecuencias.values())
plt.axhline(y=frecuencias[moda], color='r', linestyle='dashed', label='Moda')
plt.xlabel('Datos')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Datos y su Moda')
plt.legend()
plt.show()
````

Estas tres medidas de tendencia central (promedio, mediana y moda) son herramientas valiosas para resumir y entender conjuntos de datos. Es importante considerar cada una de ellas seg√∫n el tipo de datos que se est√© analizando y la naturaleza de la distribuci√≥n para obtener una visi√≥n completa de los datos.

## **Distribuci√≥n de datos y Medidas de tendencia central**

**Distribuci√≥n de datos:**

La distribuci√≥n de datos es una descripci√≥n de c√≥mo se encuentran distribuidos los valores en un conjunto de datos. En otras palabras, muestra c√≥mo se agrupan o dispersan los datos alrededor de diferentes valores en el conjunto. Entender la distribuci√≥n de los datos es fundamental para analizar y tomar decisiones basadas en los datos. Una distribuci√≥n puede ser sim√©trica o asim√©trica, y puede tener diferentes formas como la distribuci√≥n normal, la distribuci√≥n sesgada a la izquierda (negativa), la distribuci√≥n sesgada a la derecha (positiva), entre otras.

Algunas caracter√≠sticas importantes de una distribuci√≥n incluyen la ubicaci√≥n central (representada por las medidas de tendencia central), la dispersi√≥n (representada por las medidas de dispersi√≥n como la desviaci√≥n est√°ndar), y la forma general.

**Medidas de tendencia central:**

Las medidas de tendencia central son estad√≠sticas descriptivas que nos permiten resumir y representar un conjunto de datos en un solo valor que representa su "centro". Las tres medidas de tendencia central m√°s comunes son:

1. **Promedio (o media aritm√©tica):**

 Es la suma de todos los valores en un conjunto de datos dividida por el n√∫mero total de elementos. Representa el valor t√≠pico de los datos y puede ser afectado por valores extremos.

2. **Mediana:**

 Es el valor central de un conjunto de datos una vez que los datos se han ordenado. Es √∫til cuando hay valores extremos o at√≠picos, ya que no se ve afectada por ellos.

3. **Moda:**

 Es el valor m√°s frecuente en un conjunto de datos. Es √∫til para identificar el valor m√°s representativo o frecuente en un conjunto de datos.

**Ejemplo:**

Consideremos el siguiente conjunto de datos de calificaciones de estudiantes en un examen:

```
calificaciones = [70, 75, 80, 85, 90, 90, 95, 100, 100, 100]
```

**1. Promedio:**

Para calcular el promedio, sumamos todas las calificaciones y lo dividimos por el n√∫mero total de calificaciones.

```
promedio = (70 + 75 + 80 + 85 + 90 + 90 + 95 + 100 + 100 + 100) / 10
promedio = 90
```

**2. Mediana:**

Primero, ordenamos las calificaciones de forma ascendente:

```
calificaciones_ordenadas = [70, 75, 80, 85, 90, 90, 95, 100, 100, 100]
```

Como hay un n√∫mero par de calificaciones, la mediana ser√° el promedio de las dos calificaciones centrales.

```
mediana = (90 + 90) / 2
mediana = 90
```

**3. Moda:**

La moda es el valor m√°s frecuente, en este caso, la calificaci√≥n que aparece con mayor frecuencia es 100.

```
moda = 100
```

**Gr√°ficas:**

Veamos c√≥mo se ven estas medidas de tendencia central en gr√°ficas usando Python y la biblioteca `matplotlib`.

````python
import matplotlib.pyplot as plt

calificaciones = [70, 75, 80, 85, 90, 90, 95, 100, 100, 100]

# C√°lculo del promedio
promedio = sum(calificaciones) / len(calificaciones)

# C√°lculo de la mediana
n = len(calificaciones)
calificaciones_ordenadas = sorted(calificaciones)
mediana = calificaciones_ordenadas[n // 2] if n % 2 != 0 else (calificaciones_ordenadas[n // 2 - 1] + calificaciones_ordenadas[n // 2]) / 2

# C√°lculo de la moda
from collections import Counter
frecuencias = Counter(calificaciones)
moda = max(frecuencias, key=frecuencias.get)

# Gr√°fica de barras
plt.bar(range(n), calificaciones, tick_label=[f"{i+1}" for i in range(n)])
plt.axhline(y=promedio, color='r', linestyle='dashed', label='Promedio')
plt.axhline(y=mediana, color='g', linestyle='dashed', label='Mediana')
plt.axhline(y=moda, color='b', linestyle='dashed', label='Moda')
plt.xlabel('Estudiantes')
plt.ylabel('Calificaciones')
plt.title('Calificaciones y sus Medidas de Tendencia Central')
plt.legend()
plt.show()
````

En esta gr√°fica, las barras representan las calificaciones de cada estudiante, y las l√≠neas punteadas representan el promedio (en rojo), la mediana (en verde) y la moda (en azul). Podemos ver c√≥mo estas medidas de tendencia central resumen y representan diferentes aspectos de la distribuci√≥n de datos. En este caso, las tres medidas coinciden en 90, lo que indica que la mayor√≠a de los estudiantes obtuvieron una calificaci√≥n alta de 90.
