# **1. Qué es la Estadística Inferencial: Fundamentos de la Ciencia de Datos con Python**

## **Introducción**

La Estadística Inferencial es una rama de la estadística que se enfoca en tomar muestras de datos y utilizarlas para hacer inferencias o predicciones sobre una población más amplia. Su objetivo principal es obtener conclusiones y generalizaciones a partir de la información recopilada a través de muestras, permitiendo tomar decisiones informadas y realizar pronósticos basados en evidencia estadística.

Para entender mejor la Estadística Inferencial, es fundamental comprender los siguientes conceptos clave:

1. **Población y muestra:**

   - La población es el conjunto completo de individuos, objetos o eventos que queremos estudiar y sobre los cuales queremos hacer inferencias.
   - La muestra es una parte representativa de la población, seleccionada para que sus características reflejen las de la población original.

2. **Parámetros y estadísticos:**

   - Los parámetros son medidas numéricas que describen características de la población completa.
   - Los estadísticos son medidas numéricas calculadas a partir de la muestra y utilizadas para hacer inferencias sobre los parámetros poblacionales.

3. **Distribución de probabilidad:**

   - La distribución de probabilidad describe cómo están distribuidos los valores en una población o muestra particular.
   - Es esencial conocer la distribución subyacente para realizar inferencias precisas.

4. **Principales técnicas de Estadística Inferencial:**

   - Estimación de parámetros: Utilizando muestras para calcular estimaciones puntuales o intervalos de confianza para los parámetros de la población.
   - Pruebas de hipótesis: Evaluando afirmaciones sobre los parámetros poblacionales y determinando si las diferencias observadas son estadísticamente significativas o simplemente resultado del azar.

A continuación, proporcionaré algunos ejemplos de cómo se aplican los conceptos de Estadística Inferencial utilizando Python:

**Ejemplo 1: Estimación de un parámetro**

Supongamos que tenemos una población de 1000 estudiantes y queremos estimar la altura promedio de todos los estudiantes en la universidad. En lugar de medir a todos los estudiantes, tomamos una muestra aleatoria de 100 estudiantes y medimos sus alturas.

```python
import numpy as np

# Alturas de los 100 estudiantes seleccionados en la muestra
altura_muestra = np.random.normal(loc=170, scale=5, size=100)

# Estimación puntual de la altura promedio en la población
altura_promedio_estimada = altura_muestra.mean()
print("Estimación puntual de la altura promedio:", altura_promedio_estimada)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Alturas de los 100 estudiantes seleccionados en la muestra
altura_muestra = np.random.normal(loc=170, scale=5, size=100)

# Estimación puntual de la altura promedio en la población
altura_promedio_estimada = altura_muestra.mean()

# Crear histograma para visualizar la distribución de alturas en la muestra
plt.hist(altura_muestra, bins=15, edgecolor='black')
plt.xlabel('Altura')
plt.ylabel('Frecuencia')
plt.title('Distribución de Alturas en la Muestra')
plt.show()
```

**Ejemplo 2: Intervalo de confianza**

Continuando con el ejemplo anterior, además de la estimación puntual, queremos calcular un intervalo de confianza del 95% para la altura promedio de todos los estudiantes en la universidad.

```python
from scipy import stats

# Calculamos el intervalo de confianza al 95% para la altura promedio
intervalo_confianza = stats.norm.interval(0.95, loc=altura_promedio_estimada, scale=altura_muestra.std()/np.sqrt(len(altura_muestra)))

print("Intervalo de confianza al 95%:", intervalo_confianza)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Alturas de los 100 estudiantes seleccionados en la muestra
altura_muestra = np.random.normal(loc=170, scale=5, size=100)

# Estimación puntual de la altura promedio en la población
altura_promedio_estimada = altura_muestra.mean()

# Calculamos el intervalo de confianza al 95% para la altura promedio
intervalo_confianza = stats.norm.interval(0.95, loc=altura_promedio_estimada, scale=altura_muestra.std()/np.sqrt(len(altura_muestra)))

# Crear histograma para visualizar la distribución de alturas en la muestra
plt.hist(altura_muestra, bins=15, edgecolor='black')
plt.xlabel('Altura')
plt.ylabel('Frecuencia')
plt.title('Distribución de Alturas en la Muestra')

# Mostrar el intervalo de confianza en la gráfica
plt.axvline(intervalo_confianza[0], color='red', linestyle='--', label='Intervalo de Confianza (95%)')
plt.axvline(intervalo_confianza[1], color='red', linestyle='--')

plt.legend()
plt.show()
```


**Ejemplo 3: Prueba de hipótesis**

Supongamos que tenemos dos grupos de estudiantes, uno que tomó un curso de preparación para un examen y otro que no. Queremos determinar si el curso de preparación tuvo un impacto significativo en las calificaciones del examen.

```python
# Calificaciones de los dos grupos de estudiantes
grupo_control = np.random.normal(loc=75, scale=10, size=50)
grupo_preparacion = np.random.normal(loc=80, scale=12, size=50)

# Realizamos una prueba t de Student para comparar las medias de ambos grupos
resultado_prueba = stats.ttest_ind(grupo_control, grupo_preparacion)

print("Resultado de la prueba de hipótesis:", resultado_prueba)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Calificaciones de los dos grupos de estudiantes
grupo_control = np.random.normal(loc=75, scale=10, size=50)
grupo_preparacion = np.random.normal(loc=80, scale=12, size=50)

# Realizamos una prueba t de Student para comparar las medias de ambos grupos
resultado_prueba = stats.ttest_ind(grupo_control, grupo_preparacion)

# Crear histograma para visualizar las calificaciones en ambos grupos
plt.hist(grupo_control, bins=15, alpha=0.5, label='Grupo Control')
plt.hist(grupo_preparacion, bins=15, alpha=0.5, label='Grupo Preparación')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Distribución de Calificaciones en Ambos Grupos')
plt.legend()

# Mostrar el resultado de la prueba de hipótesis en la gráfica
if resultado_prueba.pvalue < 0.05:
    plt.text(70, 12, 'Diferencia significativa', color='red')
else:
    plt.text(70, 12, 'No hay diferencia significativa', color='green')

plt.show()
```

En este ejemplo, la función `ttest_ind` realiza una prueba t de Student para determinar si hay una diferencia significativa entre las medias de los dos grupos.

Estos son solo ejemplos básicos para ilustrar los fundamentos de la Estadística Inferencial con Python. La Estadística Inferencial es una herramienta poderosa para tomar decisiones basadas en datos y extraer información valiosa de muestras para hacer inferencias sobre poblaciones más grandes.

## **Estadística descriptiva e inferencial**

**Estadística Descriptiva:**

La Estadística Descriptiva es una rama de la estadística que se centra en resumir y describir datos de una manera concisa y significativa. Su objetivo principal es organizar y presentar la información de manera que sea más fácil de entender y analizar, sin hacer inferencias sobre una población más amplia. Algunas de las técnicas comunes utilizadas en la Estadística Descriptiva incluyen medidas de tendencia central, medidas de dispersión y gráficos.

**Ejemplo:** Puntuaciones de un examen

Supongamos que tenemos las siguientes puntuaciones (en escala de 100) de los estudiantes en un examen:

85, 90, 78, 92, 80, 88, 95, 87, 82, 89

Para realizar un análisis descriptivo, podemos calcular medidas como la media, la mediana y la desviación estándar:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de las puntuaciones de los estudiantes
puntuaciones = [85, 90, 78, 92, 80, 88, 95, 87, 82, 89]

# Calculamos la media y la mediana
media = np.mean(puntuaciones)
mediana = np.median(puntuaciones)

# Calculamos la desviación estándar
desviacion_estandar = np.std(puntuaciones)

# Mostramos los resultados
print("Media:", media)
print("Mediana:", mediana)
print("Desviación Estándar:", desviacion_estandar)

# Crear un histograma para visualizar las puntuaciones
plt.hist(puntuaciones, bins=5, edgecolor='black')
plt.xlabel('Puntuaciones')
plt.ylabel('Frecuencia')
plt.title('Distribución de Puntuaciones en el Examen')
plt.axvline(media, color='red', linestyle='--', label='Media')
plt.axvline(mediana, color='green', linestyle='--', label='Mediana')
plt.legend()
plt.show()
```

En este ejemplo, calculamos la media (86.6), la mediana (87), y la desviación estándar (5.78) para resumir la distribución de las puntuaciones. Además, hemos creado un histograma que muestra la frecuencia de las puntuaciones, junto con las líneas verticales para representar la media y la mediana.

**Estadística Inferencial:**

La Estadística Inferencial es una rama de la estadística que se enfoca en tomar muestras de datos y utilizarlas para hacer inferencias o predicciones sobre una población más amplia. Su objetivo principal es generalizar los resultados de la muestra para hacer afirmaciones sobre una población completa, lo que permite tomar decisiones informadas y realizar pronósticos basados en evidencia estadística. Algunas de las técnicas comunes utilizadas en la Estadística Inferencial incluyen estimaciones de parámetros y pruebas de hipótesis.

**Ejemplo:** Estimación de la proporción de estudiantes que aprobarán el examen

Supongamos que queremos estimar la proporción de estudiantes en toda la escuela que aprobarán el examen (puntuación igual o superior a 60). Para hacerlo, tomamos una muestra aleatoria de 50 estudiantes y registramos si aprobaron o no:

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de la muestra: 1 representa aprobado, 0 representa no aprobado
muestra = np.random.choice([0, 1], size=50, p=[0.4, 0.6])

# Calculamos la proporción de estudiantes que aprobaron en la muestra
proporcion_aprobados_muestra = np.mean(muestra)

# Mostramos el resultado
print("Proporción de estudiantes que aprobaron en la muestra:", proporcion_aprobados_muestra)

# Crear un gráfico de barras para visualizar la proporción de aprobados en la muestra
plt.bar(["Aprobados", "No Aprobados"], [proporcion_aprobados_muestra, 1 - proporcion_aprobados_muestra])
plt.ylabel('Proporción')
plt.title('Proporción de Estudiantes que Aprobaron en la Muestra')
plt.show()
```

En este ejemplo, hemos estimado la proporción de estudiantes que aprobaron en nuestra muestra (la cual puede variar en diferentes ejecuciones debido al muestreo aleatorio). Para hacer una inferencia sobre la población completa, necesitaríamos realizar una estimación puntual o un intervalo de confianza. Pero en esta instancia, solo hemos mostrado la proporción en la muestra mediante un gráfico de barras.

Espero que estas explicaciones, ejemplos y visualizaciones te ayuden a comprender mejor las diferencias entre la Estadística Descriptiva y la Estadística Inferencial, y cómo ambas son fundamentales para el análisis de datos.

## **Librerías de Python para estadística inferencial**

En Python, hay varias bibliotecas populares que facilitan la realización de análisis estadísticos e inferenciales. Algunas de las principales bibliotecas son:

1. **NumPy:**

 NumPy es una biblioteca fundamental para la computación científica en Python. Proporciona soporte para arrays multidimensionales y operaciones matemáticas eficientes. Es ampliamente utilizado en cálculos estadísticos y en la generación de muestras aleatorias para realizar pruebas de hipótesis y estimaciones.

2. **SciPy:**

 SciPy es una biblioteca que se basa en NumPy y agrega funciones adicionales para la optimización, la interpolación, las pruebas estadísticas y otras operaciones científicas. Contiene herramientas útiles para realizar pruebas de hipótesis y estimaciones de parámetros.

3. **StatsModels:**

 StatsModels es una biblioteca dedicada a la estimación de modelos estadísticos y realización de pruebas de hipótesis. Es especialmente útil para ajustar modelos lineales, realizar regresiones y análisis de varianza.

4. **pandas:**

 pandas es una biblioteca popular para el análisis de datos. Ofrece estructuras de datos flexibles y herramientas para manipular y analizar datos tabulares. Se utiliza ampliamente en la preparación de datos para análisis inferenciales.

5. **matplotlib:**

 matplotlib es una biblioteca de trazado que permite crear gráficos y visualizaciones de alta calidad en Python. Es útil para representar histogramas, diagramas de dispersión y otros gráficos que ayudan a comprender y comunicar resultados estadísticos.

6. **seaborn:**

 seaborn es una biblioteca de visualización basada en matplotlib que ofrece una interfaz más sencilla para crear gráficos estadísticos atractivos. Es especialmente útil para crear gráficos de distribución, gráficos de caja y bigotes, y matrices de correlación.

A continuación, presentaré ejemplos de cómo utilizar estas bibliotecas para realizar algunas técnicas de Estadística Inferencial:

**Ejemplo 1:**

 **Prueba t de Student con SciPy**

Supongamos que tenemos dos grupos de estudiantes, uno que tomó un curso de preparación para un examen y otro que no. Queremos determinar si hay una diferencia significativa en las calificaciones del examen entre los dos grupos.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Calificaciones de los dos grupos de estudiantes
grupo_control = np.random.normal(loc=75, scale=10, size=50)
grupo_preparacion = np.random.normal(loc=80, scale=12, size=50)

# Realizamos una prueba t de Student para comparar las medias de ambos grupos
resultado_prueba = stats.ttest_ind(grupo_control, grupo_preparacion)

# Crear histogramas para visualizar las calificaciones en ambos grupos
plt.hist(grupo_control, bins=15, alpha=0.5, label='Grupo Control')
plt.hist(grupo_preparacion, bins=15, alpha=0.5, label='Grupo Preparación')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Distribución de Calificaciones en Ambos Grupos')
plt.legend()

# Mostrar el resultado de la prueba de hipótesis en la gráfica
if resultado_prueba.pvalue < 0.05:
    plt.text(70, 12, 'Diferencia significativa', color='red')
else:
    plt.text(70, 12, 'No hay diferencia significativa', color='green')

plt.show()
```

En este ejemplo, utilizamos la función `ttest_ind`` de SciPy para realizar una prueba t de Student y determinar si hay una diferencia significativa entre las medias de las calificaciones de los dos grupos. Luego, utilizamos matplotlib para crear histogramas que representan las distribuciones de calificaciones en ambos grupos, y se muestra en la gráfica el resultado de la prueba de hipótesis.

**Ejemplo 2:**

**Intervalo de Confianza con pandas y seaborn**

Supongamos que queremos calcular un intervalo de confianza del 95% para la altura promedio de una muestra de estudiantes.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Alturas de los estudiantes en una muestra
altura_muestra = np.random.normal(loc=170, scale=5, size=100)

# Calculamos el intervalo de confianza al 95% para la altura promedio
media_estimada = altura_muestra.mean()
desviacion_estandar = altura_muestra.std()
tamaño_muestra = len(altura_muestra)
intervalo_confianza = stats.norm.interval(0.95, loc=media_estimada, scale=desviacion_estandar/np.sqrt(tamaño_muestra))

# Crear un diagrama de dispersión para visualizar las alturas de los estudiantes
sns.scatterplot(x=range(1, tamaño_muestra + 1), y=altura_muestra)
plt.axhline(media_estimada, color='red', linestyle='--', label='Media Estimada')
plt.axhline(intervalo_confianza[0], color='orange', linestyle='--', label='Intervalo Confianza (95%)')
plt.axhline(intervalo_confianza[1], color='orange', linestyle='--')
plt.xlabel('Estudiante')
plt.ylabel('Altura')
plt.title('Altura de los Estudiantes en la Muestra')
plt.legend()
plt.show()
```

En este ejemplo, utilizamos NumPy para generar una muestra de alturas de estudiantes. Luego, usamos pandas y seaborn para crear un diagrama de dispersión que muestra las alturas de los estudiantes en la muestra, junto con la media estimada y el intervalo de confianza del 95% para la altura promedio.
