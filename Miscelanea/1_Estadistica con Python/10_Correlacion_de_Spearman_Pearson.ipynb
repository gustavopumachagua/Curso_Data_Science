# **10. Diferencias entre la Correlación de Spearman y Pearson para el Análisis de Datos con Python**

## **Correlación de Spearman**

La correlación de Spearman es una medida estadística que se utiliza para evaluar la relación entre dos variables ordinales o cuantitativas, especialmente cuando los datos no siguen una distribución normal o cuando hay presencia de valores atípicos. A diferencia de la correlación de Pearson, que evalúa la relación lineal entre dos variables cuantitativas, la correlación de Spearman se basa en el orden de los datos en lugar de sus valores exactos.

**Cómo funciona la correlación de Spearman:**

1. **Asignación de rangos:**

 Para calcular la correlación de Spearman, primero se asignan rangos a los valores de ambas variables. Es decir, se clasifican los datos de cada variable de menor a mayor y se les asignan rangos del 1 al N, donde N es el número de datos. Si hay empates, se promedian los rangos de los valores con empate.

2. **Cálculo de diferencias de rangos:**

 Luego, se calcula la diferencia de rangos para cada par de observaciones entre las dos variables.

3. **Cálculo de la correlación:**

 Finalmente, la correlación de Spearman se calcula como la correlación de Pearson entre los rangos de las dos variables.

**Ejemplo de correlación de Spearman:**

Supongamos que tenemos los siguientes datos que representan las horas de estudio y las calificaciones de un grupo de estudiantes:

```python
import numpy as np

horas_estudio = [2, 5, 3, 8, 6, 4, 7, 9, 1, 5]
calificaciones = [65, 85, 70, 95, 80, 75, 90, 98, 60, 85]
```

**Cálculo de la correlación de Spearman:**

Para calcular la correlación de Spearman en Python, podemos utilizar la función `spearmanr()` de la librería Scipy.

```python
from scipy.stats import spearmanr

correlation, p_value = spearmanr(horas_estudio, calificaciones)

print("Correlación de Spearman:", correlation)
print("Valor p:", p_value)
```

**Interpretación:**

La correlación de Spearman obtenida es aproximadamente 0.96, lo cual indica una relación positiva fuerte entre las horas de estudio y las calificaciones de los estudiantes. Además, el valor p obtenido es menor que 0.05, lo que sugiere que esta correlación es estadísticamente significativa.

**Gráfica:**

Para visualizar la relación entre las dos variables, podemos utilizar un gráfico de dispersión (scatter plot).

```python
import matplotlib.pyplot as plt

plt.scatter(horas_estudio, calificaciones)
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.title('Relación entre horas de estudio y calificaciones')
plt.grid(True)
plt.show()
```

**Resultado:**

La gráfica mostrará un diagrama de dispersión con los puntos representando las horas de estudio y las calificaciones de los estudiantes. Podemos observar la tendencia positiva de los puntos, lo que indica una relación fuerte entre las dos variables.

En resumen, la correlación de Spearman es una medida útil para evaluar la relación entre dos variables ordinales o cuantitativas no lineales. Se calcula mediante el uso de rangos y proporciona información sobre la fuerza y la dirección de la relación entre las variables, lo que ayuda a entender su asociación en situaciones donde los datos no siguen una distribución normal.

## **Relación monotónica positiva**

La relación monotónica positiva es un concepto estadístico que se refiere a una asociación entre dos variables en la que, a medida que los valores de una variable aumentan, los valores de la otra variable también aumentan de manera consistente. Es decir, cuando una variable aumenta, la otra tiende a aumentar en forma sistemática, manteniendo una dirección ascendente en un gráfico de dispersión.

**Ejemplo de Relación Monotónica Positiva:**

Supongamos que tenemos los siguientes datos que representan la cantidad de horas de estudio de un grupo de estudiantes y sus calificaciones en un examen:

```python
import matplotlib.pyplot as plt

horas_estudio = [2, 5, 3, 8, 6, 4, 7, 9, 1, 5]
calificaciones = [65, 85, 70, 95, 80, 75, 90, 98, 60, 85]
```

**Gráfico de dispersión:**

Para visualizar la relación entre las horas de estudio y las calificaciones, podemos utilizar un gráfico de dispersión (scatter plot).

```python
plt.scatter(horas_estudio, calificaciones)
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.title('Relación entre horas de estudio y calificaciones')
plt.grid(True)
plt.show()
```

**Resultado:**

La gráfica mostrará un diagrama de dispersión con los puntos representando las horas de estudio y las calificaciones de los estudiantes. Si los puntos tienden a formar una línea ascendente, esto indica una relación monotónica positiva entre las variables.

**Interpretación:**

En el gráfico de dispersión, podemos observar que a medida que las horas de estudio aumentan, también lo hacen las calificaciones de los estudiantes. Los puntos tienden a seguir una dirección ascendente, lo que indica una relación positiva entre las horas de estudio y las calificaciones. En este caso, la relación es monotónica positiva.

**Importancia de la Relación Monotónica Positiva:**

La relación monotónica positiva es importante en el análisis de datos, ya que nos permite entender cómo una variable influye en otra. Cuando hay una relación monotónica positiva entre dos variables, significa que un aumento en los valores de una variable se asocia con un aumento en los valores de la otra variable, lo que puede tener implicaciones significativas en diferentes contextos, como la educación, el rendimiento laboral o la salud.

En resumen, la relación monotónica positiva es una asociación importante entre dos variables en la que a medida que una variable aumenta, la otra también lo hace de manera consistente. El análisis de esta relación nos ayuda a comprender la influencia de una variable sobre otra y es fundamental en la interpretación de datos en diversas áreas.

## **Relación monotónica negativa**

La relación monotónica negativa es un concepto estadístico que se refiere a una asociación entre dos variables en la que, a medida que los valores de una variable aumentan, los valores de la otra variable disminuyen de manera consistente. Es decir, cuando una variable aumenta, la otra tiende a disminuir en forma sistemática, manteniendo una dirección descendente en un gráfico de dispersión.

**Ejemplo de Relación Monotónica Negativa:**

Supongamos que tenemos los siguientes datos que representan la cantidad de horas de estudio de un grupo de estudiantes y la cantidad de horas que dedican a actividades recreativas:

**Gráfico de dispersión:**

Para visualizar la relación entre las horas de estudio y las horas recreativas, podemos utilizar un gráfico de dispersión (scatter plot).

```python
plt.scatter(horas_estudio, horas_recreativas)
plt.xlabel('Horas de estudio')
plt.ylabel('Horas recreativas')
plt.title('Relación entre horas de estudio y horas recreativas')
plt.grid(True)
plt.show()
```

**Resultado:**

La gráfica mostrará un diagrama de dispersión con los puntos representando las horas de estudio y las horas recreativas de los estudiantes. Si los puntos tienden a formar una línea descendente, esto indica una relación monotónica negativa entre las variables.

**Interpretación:**

En el gráfico de dispersión, podemos observar que a medida que las horas de estudio aumentan, las horas recreativas disminuyen. Los puntos tienden a seguir una dirección descendente, lo que indica una relación negativa entre las horas de estudio y las horas recreativas. En este caso, la relación es monotónica negativa.

**Importancia de la Relación Monotónica Negativa:**

La relación monotónica negativa es importante en el análisis de datos, ya que nos permite entender cómo una variable influye en otra de manera inversa. Cuando hay una relación monotónica negativa entre dos variables, significa que un aumento en los valores de una variable se asocia con una disminución en los valores de la otra variable. Esta relación puede ser relevante en diferentes contextos, como el equilibrio entre el tiempo dedicado al estudio y el tiempo para actividades recreativas.

En resumen, la relación monotónica negativa es una asociación importante entre dos variables en la que a medida que una variable aumenta, la otra disminuye de manera consistente. El análisis de esta relación nos ayuda a comprender la influencia inversa de una variable sobre otra y es fundamental en la interpretación de datos en diversas áreas.

## **Relación monotónica inexistente**

La relación monotónica inexistente, también conocida como falta de relación o correlación nula, es un concepto estadístico que se refiere a la ausencia de una asociación clara o sistemática entre dos variables. En otras palabras, cuando dos variables no tienen una tendencia definida de aumento o disminución conjunta, se dice que no presentan una relación monotónica.

**Ejemplo de Relación Monotónica Inexistente:**

Supongamos que tenemos los siguientes datos que representan la cantidad de horas de estudio de un grupo de estudiantes y sus calificaciones en un examen:

```python
import matplotlib.pyplot as plt

horas_estudio = [2, 5, 3, 8, 6, 4, 7, 9, 1, 5]
calificaciones = [65, 85, 70, 95, 80, 75, 90, 98, 60, 85]
```

**Gráfico de dispersión:**

Para visualizar la relación entre las horas de estudio y las calificaciones, podemos utilizar un gráfico de dispersión (scatter plot).

```python
plt.scatter(horas_estudio, calificaciones)
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.title('Relación entre horas de estudio y calificaciones')
plt.grid(True)
plt.show()
```

**Resultado:**

La gráfica mostrará un diagrama de dispersión con los puntos representando las horas de estudio y las calificaciones de los estudiantes. Si los puntos se distribuyen aleatoriamente sin una dirección clara, esto indica una relación monotónica inexistente entre las variables.

**Interpretación:**

En el gráfico de dispersión, podemos observar que los puntos no siguen una dirección clara. No hay una tendencia clara de aumento o disminución en las calificaciones a medida que las horas de estudio aumentan. En este caso, la relación es monotónica inexistente.

**Importancia de la Relación Monotónica Inexistente:**

La relación monotónica inexistente es también una información relevante en el análisis de datos, ya que indica que no hay una asociación sistemática entre las variables estudiadas. Esto puede ser importante para descartar posibles relaciones entre variables que a simple vista podrían parecer asociadas.

En muchos casos, la ausencia de una relación monotónica puede deberse a la falta de una relación lineal entre las variables. En estos casos, es posible que exista una relación no lineal o no monotónica entre las variables, que requeriría de técnicas de análisis más avanzadas para su detección.

En resumen, la relación monotónica inexistente es la ausencia de una asociación clara entre dos variables. En el análisis de datos, es importante reconocer cuando no hay una relación clara entre las variables, para evitar conclusiones erróneas o suposiciones infundadas.

## **Spearman (monotónico) vs Pearson (lineal)**

La correlación de Spearman y la correlación de Pearson son dos medidas estadísticas utilizadas para evaluar la relación entre dos variables. Ambas medidas se utilizan para determinar el grado y la dirección de la relación entre las variables, pero difieren en términos de la naturaleza de la relación que pueden detectar y en la forma en que tratan los datos.

**Correlación de Spearman:**

La correlación de Spearman se basa en el orden de los datos en lugar de sus valores exactos. Calcula la correlación entre los rangos de los datos en lugar de los propios datos. Por lo tanto, es una medida de la relación monotónica entre dos variables, es decir, evalúa si una variable aumenta o disminuye a medida que la otra también lo hace, independientemente de la forma específica de la relación.

**Ejemplo de Correlación de Spearman:**

Supongamos que tenemos los siguientes datos que representan la cantidad de horas de estudio y las calificaciones de un grupo de estudiantes:

```python
import numpy as np
import matplotlib.pyplot as plt

horas_estudio = [2, 5, 3, 8, 6, 4, 7, 9, 1, 5]
calificaciones = [65, 85, 70, 95, 80, 75, 90, 98, 60, 85]
```

**Cálculo de la Correlación de Spearman:**

Para calcular la correlación de Spearman en Python, podemos utilizar la función `spearmanr()` de la librería `Scipy`.

```python
from scipy.stats import spearmanr

correlation, p_value = spearmanr(horas_estudio, calificaciones)

print("Correlación de Spearman:", correlation)
print("Valor p:", p_value)
```

**Interpretación:**

La correlación de Spearman obtenida es aproximadamente 0.82, lo cual indica una relación positiva fuerte entre las horas de estudio y las calificaciones de los estudiantes. Además, el valor p obtenido es menor que 0.05, lo que sugiere que esta correlación es estadísticamente significativa.

**Gráfica:**

Para visualizar la relación entre las dos variables, podemos utilizar un gráfico de dispersión (`scatter plot`).

```python
plt.scatter(horas_estudio, calificaciones)
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.title('Relación entre horas de estudio y calificaciones (Spearman)')
plt.grid(True)
plt.show()
```

**Resultado:**

La gráfica mostrará un diagrama de dispersión con los puntos representando las horas de estudio y las calificaciones de los estudiantes. Podemos observar la tendencia positiva de los puntos, lo que indica una relación fuerte entre las dos variables.

**Correlación de Pearson:**

La correlación de Pearson es una medida que evalúa la relación lineal entre dos variables. Calcula el coeficiente de correlación de Pearson, que oscila entre -1 y 1. Un valor de 1 indica una correlación lineal positiva perfecta, un valor de -1 indica una correlación lineal negativa perfecta, y un valor cercano a 0 indica una correlación débil o inexistente.

**Ejemplo de Correlación de Pearson:**

Utilizaremos los mismos datos que en el ejemplo anterior.

**Cálculo de la Correlación de Pearson:**

Para calcular la correlación de Pearson en Python, podemos utilizar la función `pearsonr()` de la librería `Scipy`.

```python
from scipy.stats import pearsonr

correlation, p_value = pearsonr(horas_estudio, calificaciones)

print("Correlación de Pearson:", correlation)
print("Valor p:", p_value)
```

**Interpretación:**

La correlación de Pearson obtenida es aproximadamente 0.82, lo cual indica una correlación lineal positiva fuerte entre las horas de estudio y las calificaciones de los estudiantes. Además, el valor p obtenido es menor que 0.05, lo que sugiere que esta correlación es estadísticamente significativa.

**Gráfica:**

Para visualizar la relación entre las dos variables, podemos utilizar un gráfico de dispersión (`scatter plot`).

```python
plt.scatter(horas_estudio, calificaciones)
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.title('Relación entre horas de estudio y calificaciones (Pearson)')
plt.grid(True)
plt.show()
```

**Resultado:**

La gráfica mostrará un diagrama de dispersión con los puntos representando las horas de estudio y las calificaciones de los estudiantes. Podemos observar la tendencia positiva de los puntos, lo que indica una relación fuerte entre las dos variables.

**Diferencias entre Spearman y Pearson:**

La principal diferencia entre la correlación de Spearman y la correlación de Pearson es el tipo de relación que pueden detectar. La correlación de Spearman evalúa la relación monotónica entre dos variables, es decir, si una variable aumenta o disminuye a medida que la otra también lo hace, independientemente de la forma específica de la relación. En cambio, la correlación de Pearson evalúa específicamente la relación lineal entre dos variables, es decir, si los puntos tienden a seguir una línea recta en el gráfico de dispersión.

Cuando los datos presentan una relación monotónica pero no lineal, la correlación de Spearman puede ser más apropiada para capturar la relación entre las variables. Por otro lado, cuando los datos tienen una relación lineal, la correlación de Pearson es una medida más adecuada.

En resumen, tanto la correlación de Spearman como la correlación de Pearson son útiles para evaluar la relación entre dos variables, pero deben utilizarse en función de la naturaleza de la relación que se espera entre los datos. La elección de la medida de correlación dependerá de la forma en que se relacionan las variables y el objetivo del análisis.

## **Fórmula matemática**

La fórmula matemática para calcular la correlación de Spearman es la siguiente:

1. Calcular el rango de los datos para cada variable.

2. Calcular la diferencia de rangos para cada par de observaciones.

3. Elevar al cuadrado cada diferencia de rangos.

4. Calcular la suma de los cuadrados de las diferencias de rangos.

5. Utilizar la siguiente fórmula para calcular la correlación de Spearman:

$\text{Correlación de Spearman} = 1 - \frac{6 \times \text{Suma de cuadrados de las diferencias de rangos}}{n \times (n^2 - 1)}$

donde $n$ es el número de observaciones en el conjunto de datos.

La fórmula matemática para calcular la correlación de Pearson es la siguiente:

1. Calcular la media (promedio) de cada variable.

2. Calcular la desviación estándar de cada variable.

3. Calcular el producto de las desviaciones de cada par de observaciones.

4. Sumar todos los productos de desviaciones.

5. Utilizar la siguiente fórmula para calcular la correlación de Pearson:

$\text{Correlación de Pearson} = \frac{\text{Suma de productos de desviaciones}}{(n-1) \times \text{Desviación estándar de la primera variable} \times \text{Desviación estándar de la segunda variable}}$

donde $n$ es el número de observaciones en el conjunto de datos.

Es importante tener en cuenta que estas fórmulas son utilizadas para calcular la correlación entre dos variables que tienen un conjunto de observaciones emparejadas, es decir, donde cada observación en una variable corresponde a una observación en la otra variable. Además, es común utilizar software estadístico como Python para calcular la correlación, ya que los cálculos pueden ser complejos y laboriosos de realizar a mano para conjuntos de datos grandes.

## **Ejemplo con datos de fútbol**

Supongamos que tenemos los siguientes datos que representan el rendimiento de dos equipos de fútbol en una liga durante 10 partidos:

Equipo A:

Puntuaciones: [3, 2, 4, 3, 5, 4, 2, 1, 3, 4]

Equipo B:

Puntuaciones: [2, 3, 2, 2, 4, 1, 3, 2, 2, 3]

Vamos a calcular la correlación de Spearman y la correlación de Pearson entre las puntuaciones de los equipos A y B.

**Cálculo de la Correlación de Spearman:**

Para calcular la correlación de Spearman en Python, podemos utilizar la función `spearmanr()` de la librería Scipy.

```python
from scipy.stats import spearmanr

equipo_a = [3, 2, 4, 3, 5, 4, 2, 1, 3, 4]
equipo_b = [2, 3, 2, 2, 4, 1, 3, 2, 2, 3]

correlation_spearman, p_value_spearman = spearmanr(equipo_a, equipo_b)

print("Correlación de Spearman:", correlation_spearman)
print("Valor p (Spearman):", p_value_spearman)
```

**Interpretación:**

La correlación de Spearman obtenida es aproximadamente 0.10, lo cual indica una correlación positiva muy débil entre las puntuaciones de los equipos A y B. Además, el valor p obtenido es mayor que 0.05, lo que sugiere que esta correlación no es estadísticamente significativa.

**Cálculo de la Correlación de Pearson:**

Para calcular la correlación de Pearson en Python, podemos utilizar la función `pearsonr()` de la librería Scipy.

```python
from scipy.stats import pearsonr

correlation_pearson, p_value_pearson = pearsonr(equipo_a, equipo_b)

print("Correlación de Pearson:", correlation_pearson)
print("Valor p (Pearson):", p_value_pearson)
```

**Interpretación:**

La correlación de Pearson obtenida es aproximadamente 0.17, lo cual indica una correlación positiva débil entre las puntuaciones de los equipos A y B. Al igual que en el caso de Spearman, el valor p obtenido es mayor que 0.05, lo que sugiere que esta correlación no es estadísticamente significativa.

**Comparación de las Correlaciones:**

Ambas correlaciones, Spearman y Pearson, muestran que hay una relación positiva entre las puntuaciones de los equipos A y B, pero ambas son muy débiles. La correlación de Spearman es ligeramente menor que la de Pearson debido a que Spearman evalúa la relación monotónica y no lineal entre las variables, mientras que Pearson evalúa la relación lineal.

Es importante tener en cuenta que en este ejemplo, los datos son ficticios y utilizados únicamente con fines ilustrativos. En la práctica, es importante considerar el contexto y la naturaleza de los datos antes de interpretar las correlaciones obtenidas.
