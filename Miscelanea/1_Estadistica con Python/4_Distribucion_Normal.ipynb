# **4. Distribución Normal (o Gaussiana) y la Regla Empírica 68-95-99.7 con Python**

## **Introducción a la distribución normal**

La Distribución Normal, también conocida como Distribución Gaussiana, es una de las distribuciones más importantes en estadística. Es una distribución continua que se caracteriza por su forma de campana y se utiliza ampliamente en diversos campos debido a su comportamiento bien conocido y a que muchos fenómenos naturales siguen aproximadamente esta distribución.

La Regla Empírica 68-95-99.7 es una propiedad clave de la Distribución Normal que proporciona información sobre el porcentaje de datos que caen dentro de ciertos intervalos de desviación estándar alrededor de la media. Esta regla es útil para interpretar y visualizar la distribución de los datos.

**Distribución Normal:**

La función de densidad de probabilidad de la Distribución Normal está dada por la siguiente fórmula:

![Función de densidad de probabilidad de la Distribución Normal](https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/Normal_Distribution_PDF.svg/450px-Normal_Distribution_PDF.svg.png)

Donde:

- μ (mu) es la media de la distribución, que representa el valor central alrededor del cual se agrupan los datos.

- σ (sigma) es la desviación estándar, que mide la dispersión o variabilidad de los datos.

- π (pi) es una constante matemática aproximadamente igual a 3.14159.

La curva de la Distribución Normal es simétrica respecto a la media y alcanza su máximo en el valor de la media μ. A medida que nos alejamos de la media, la probabilidad de obtener valores disminuye, pero nunca llega a cero, ya que la distribución es continua.

**Regla Empírica 68-95-99.7:**

La Regla Empírica establece que, en una Distribución Normal:

- Aproximadamente el 68% de los datos se encuentran dentro de una desviación estándar (σ) de la media (μ).

- Aproximadamente el 95% de los datos se encuentran dentro de dos desviaciones estándar (2σ) de la media (μ).

- Aproximadamente el 99.7% de los datos se encuentran dentro de tres desviaciones estándar (3σ) de la media (μ).

**Ejemplo con Python:**

Vamos a generar una muestra de datos que sigue una Distribución Normal, calcular la media y la desviación estándar de la muestra, y luego utilizar la Regla Empírica para visualizar la proporción de datos dentro de los intervalos 1σ, 2σ y 3σ de la media.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de datos con Distribución Normal
np.random.seed(42)  # Para reproducibilidad
media = 50
desviacion_estandar = 10
muestra_datos = np.random.normal(loc=media, scale=desviacion_estandar, size=1000)

# Calcular la media y la desviación estándar de la muestra
media_muestra = np.mean(muestra_datos)
desviacion_estandar_muestra = np.std(muestra_datos)

# Intervalos de 1σ, 2σ y 3σ alrededor de la media
intervalo_1_sigma = (media_muestra - desviacion_estandar_muestra, media_muestra + desviacion_estandar_muestra)
intervalo_2_sigma = (media_muestra - 2 * desviacion_estandar_muestra, media_muestra + 2 * desviacion_estandar_muestra)
intervalo_3_sigma = (media_muestra - 3 * desviacion_estandar_muestra, media_muestra + 3 * desviacion_estandar_muestra)

# Graficar la Distribución Normal
plt.hist(muestra_datos, bins=30, edgecolor='black', alpha=0.7, density=True)
plt.axvline(media_muestra, color='red', linestyle='--', label='Media')
plt.axvline(intervalo_1_sigma[0], color='green', linestyle='--', label='1σ')
plt.axvline(intervalo_1_sigma[1], color='green', linestyle='--')
plt.axvline(intervalo_2_sigma[0], color='orange', linestyle='--', label='2σ')
plt.axvline(intervalo_2_sigma[1], color='orange', linestyle='--')
plt.axvline(intervalo_3_sigma[0], color='purple', linestyle='--', label='3σ')
plt.axvline(intervalo_3_sigma[1], color='purple', linestyle='--')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Normal')
plt.legend()
plt.show()

# Imprimir la proporción de datos dentro de los intervalos de 1σ, 2σ y 3σ
print("Proporción de datos en el intervalo 1σ:", np.mean((muestra_datos >= intervalo_1_sigma[0]) & (muestra_datos <= intervalo_1_sigma[1])))
print("Proporción de datos en el intervalo 2σ:", np.mean((muestra_datos >= intervalo_2_sigma[0]) & (muestra_datos <= intervalo_2_sigma[1])))
print("Proporción de datos en el intervalo 3σ:", np.mean((muestra_datos >= intervalo_3_sigma[0]) & (muestra_datos <= intervalo_3_sigma[1])))
```

**Explicación del ejemplo:**

1. Importamos las bibliotecas NumPy y Matplotlib.

2. Generamos una muestra de datos de tamaño 1000 que sigue una Distribución Normal con media 50 y desviación estándar 10.

3. Calculamos la media y la desviación estándar de la muestra.

4. Calculamos los intervalos de 1σ, 2σ y 3σ alrededor de la media.

5. Graficamos el histograma de la muestra de datos y superponemos líneas verticales rojas para representar la media y líneas punteadas para los intervalos 1σ, 2σ y 3σ.

6. Calculamos la proporción de datos dentro de cada intervalo utilizando la función `np.mean()` y la operación lógica `&` (AND).

En la gráfica, podemos observar que la mayor densidad de datos se encuentra cerca de la media (línea roja) y disminuye a medida que nos alejamos de ella. Además, la Regla Empírica se cumple, ya que aproximadamente el 68%, el 95% y el 99.7% de los datos caen dentro de 1σ, 2σ y 3σ de la media, respectivamente.

 El peso de una población también suele tener una Distribución Normal. La mayoría de las personas tienen un peso cercano a la media, con pocas personas con pesos extremadamente altos o bajos.

 Las puntuaciones en pruebas estandarizadas, como el SAT o el GRE, tienden a distribuirse normalmente. La mayoría de los estudiantes obtienen puntuaciones cercanas a la media, con pocos estudiantes obteniendo puntuaciones muy altas o muy bajas.

4. **Tiempo de Respuesta de Sistemas Computacionales:**

 Muchas mediciones de laboratorio, como la concentración de una sustancia en una muestra, pueden tener una distribución aproximadamente normal.

## **Explicación Distribución Normal**

La Distribución Normal, también conocida como Distribución Gaussiana, es una de las distribuciones más importantes y ampliamente utilizadas en estadística y ciencias naturales. Se caracteriza por su forma de campana simétrica y está completamente determinada por dos parámetros: la media (μ) y la desviación estándar (σ).

**Forma de la Distribución Normal:**

La función de densidad de probabilidad de la Distribución Normal se define matemáticamente como:

![Función de densidad de probabilidad de la Distribución Normal](https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/Normal_Distribution_PDF.svg/450px-Normal_Distribution_PDF.svg.png)

Donde:

- x: Variable aleatoria (el valor que se quiere medir).

- μ (mu): Media de la distribución, que representa el valor central o el pico de la campana.

- σ (sigma): Desviación estándar, que controla la dispersión o la extensión de la campana.

La curva es simétrica respecto a la media (μ) y alcanza su valor máximo en μ. A medida que nos alejamos de la media, la probabilidad de obtener valores disminuye, pero nunca llega a cero, ya que la distribución es continua y se extiende infinitamente en ambos sentidos.

**Propiedades de la Distribución Normal:**

1. La suma de una cantidad suficientemente grande de variables aleatorias independientes que siguen una distribución normal también sigue una distribución normal.

2. La distribución normal es completamente determinada por su media (μ) y su desviación estándar (σ). Una vez que se conocen estos parámetros, la forma de la campana es siempre la misma.

3. La media, la mediana y la moda de la distribución normal son iguales, lo que significa que la distribución es simétrica respecto a su media.

4. La distribución normal está acotada, pero tiende hacia el infinito en ambos sentidos. Sin embargo, a medida que nos alejamos de la media, la probabilidad de obtener valores muy grandes o muy pequeños disminuye rápidamente.

**Aplicaciones de la Distribución Normal:**

La Distribución Normal es ampliamente utilizada en el análisis de datos, la inferencia estadística y la modelización en diversas áreas, incluyendo:

- Estimación de parámetros y construcción de intervalos de confianza.

- Pruebas de hipótesis.

- Análisis de regresión y ajuste de modelos lineales.

- Teoría de probabilidad y estadística matemática.

- Modelado de fenómenos naturales y científicos.

En resumen, la Distribución Normal es una herramienta esencial en estadística debido a su comportamiento bien conocido y sus propiedades matemáticas. Es utilizada ampliamente en la modelización de datos y proporciona una base sólida para realizar inferencias y tomar decisiones basadas en datos observados. Su forma de campana y la Regla Empírica 68-95-99.7 hacen que sea una distribución muy útil y aplicable en una amplia variedad de situaciones.

## **Regla Empírica 68-95-99.7**

La Regla Empírica 68-95-99.7, también conocida como la Regla 68-95-99.7 o la Regla de los Tres Sigmas, es una propiedad clave de la Distribución Normal que proporciona información sobre el porcentaje de datos que caen dentro de ciertos intervalos de desviación estándar alrededor de la media. Esta regla es una aproximación útil para interpretar y visualizar la distribución de los datos cuando se asume que siguen una Distribución Normal.

**La Regla Empírica establece lo siguiente:**

1. Aproximadamente el 68% de los datos se encuentran dentro de una desviación estándar (σ) de la media (μ).

2. Aproximadamente el 95% de los datos se encuentran dentro de dos desviaciones estándar (2σ) de la media (μ).

3. Aproximadamente el 99.7% de los datos se encuentran dentro de tres desviaciones estándar (3σ) de la media (μ).

**Visualización de la Regla Empírica:**

Para visualizar la Regla Empírica, podemos representar una Distribución Normal y trazar líneas verticales en la media y en ±σ, ±2σ y ±3σ alrededor de la media. La proporción de datos dentro de cada intervalo corresponde a los porcentajes mencionados anteriormente.

- El área sombreada en azul representa aproximadamente el 68% de los datos dentro de 1σ de la media.

- El área sombreada en verde representa aproximadamente el 95% de los datos dentro de 2σ de la media.

- El área sombreada en amarillo representa aproximadamente el 99.7% de los datos dentro de 3σ de la media.

**Aplicación de la Regla Empírica:**

La Regla Empírica es útil para evaluar rápidamente la distribución de los datos y comprender cuántos valores se encuentran dentro de ciertos intervalos alrededor de la media. Cuando se cumplen ciertas condiciones, la Regla Empírica puede utilizarse para realizar estimaciones aproximadas y tomar decisiones en análisis estadísticos sin la necesidad de calcular directamente probabilidades específicas.

Es importante tener en cuenta que la Regla Empírica se basa en la suposición de que los datos siguen una Distribución Normal. Si los datos no se ajustan a una distribución normal, la regla puede no aplicarse con precisión. Por lo tanto, es importante verificar si los datos se ajustan a una Distribución Normal antes de aplicar la Regla Empírica.

**Paso 1: Importar las bibliotecas necesarias**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Paso 2: Generar una muestra de datos con Distribución Normal**

Vamos a generar una muestra de 1000 datos que sigan una Distribución Normal con media 50 y desviación estándar 10.

```python
np.random.seed(42)  # Para reproducibilidad
media = 50
desviacion_estandar = 10
muestra_datos = np.random.normal(loc=media, scale=desviacion_estandar, size=1000)
```

**Paso 3: Calcular la media y la desviación estándar de la muestra**

```python
media_muestra = np.mean(muestra_datos)
desviacion_estandar_muestra = np.std(muestra_datos)
```

**Paso 4: Graficar la Distribución Normal y aplicar la Regla Empírica**

```python
# Crear histograma de la muestra de datos
plt.hist(muestra_datos, bins=30, edgecolor='black', alpha=0.7, density=True)

# Líneas verticales para representar la media y las desviaciones estándar
plt.axvline(media_muestra, color='red', linestyle='--', label='Media')
plt.axvline(media_muestra + desviacion_estandar_muestra, color='green', linestyle='--', label='1σ')
plt.axvline(media_muestra - desviacion_estandar_muestra, color='green', linestyle='--')
plt.axvline(media_muestra + 2 * desviacion_estandar_muestra, color='orange', linestyle='--', label='2σ')
plt.axvline(media_muestra - 2 * desviacion_estandar_muestra, color='orange', linestyle='--')
plt.axvline(media_muestra + 3 * desviacion_estandar_muestra, color='purple', linestyle='--', label='3σ')
plt.axvline(media_muestra - 3 * desviacion_estandar_muestra, color='purple', linestyle='--')

plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Normal')
plt.legend()
plt.show()
```

## **Ejemplificación de la Regla Empírica**

Vamos a generar una muestra de datos que sigue una Distribución Normal y a utilizar la Regla Empírica 68-95-99.7 para verificar si los porcentajes de datos dentro de 1σ, 2σ y 3σ de la media se acercan a los valores esperados. Utilizaremos Python y las bibliotecas NumPy y Matplotlib.

**Paso 1: Importar las bibliotecas necesarias**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Paso 2: Generar una muestra de datos con Distribución Normal**

Vamos a generar una muestra de 1000 datos que sigan una Distribución Normal con media 50 y desviación estándar 10.

```python
np.random.seed(42)  # Para reproducibilidad
media = 50
desviacion_estandar = 10
muestra_datos = np.random.normal(loc=media, scale=desviacion_estandar, size=1000)
```

**Paso 3: Calcular la media y la desviación estándar de la muestra**

```python
media_muestra = np.mean(muestra_datos)
desviacion_estandar_muestra = np.std(muestra_datos)
```

**Paso 4: Calcular la proporción de datos dentro de 1σ, 2σ y 3σ de la media**

```python
# Número de datos dentro de 1σ de la media
num_datos_1_sigma = np.sum(np.logical_and(muestra_datos >= media_muestra - desviacion_estandar_muestra,
                                          muestra_datos <= media_muestra + desviacion_estandar_muestra))

# Número de datos dentro de 2σ de la media
num_datos_2_sigma = np.sum(np.logical_and(muestra_datos >= media_muestra - 2 * desviacion_estandar_muestra,
                                          muestra_datos <= media_muestra + 2 * desviacion_estandar_muestra))

# Número de datos dentro de 3σ de la media
num_datos_3_sigma = np.sum(np.logical_and(muestra_datos >= media_muestra - 3 * desviacion_estandar_muestra,
                                          muestra_datos <= media_muestra + 3 * desviacion_estandar_muestra))

# Proporción de datos dentro de 1σ, 2σ y 3σ
prop_1_sigma = num_datos_1_sigma / len(muestra_datos)
prop_2_sigma = num_datos_2_sigma / len(muestra_datos)
prop_3_sigma = num_datos_3_sigma / len(muestra_datos)

# Valores esperados según la Regla Empírica
esperado_1_sigma = 0.68
esperado_2_sigma = 0.95
esperado_3_sigma = 0.997
```

**Paso 5: Mostrar resultados**

```python
print("Proporción de datos dentro de 1σ:", prop_1_sigma, "(Esperado:", esperado_1_sigma, ")")
print("Proporción de datos dentro de 2σ:", prop_2_sigma, "(Esperado:", esperado_2_sigma, ")")
print("Proporción de datos dentro de 3σ:", prop_3_sigma, "(Esperado:", esperado_3_sigma, ")")
```
