```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de datos con Distribuci贸n Normal
np.random.seed(42)  # Para reproducibilidad
media = 50
desviacion_estandar = 10
muestra_datos = np.random.normal(loc=media, scale=desviacion_estandar, size=1000)

# Calcular la media y la desviaci贸n est谩ndar de la muestra
media_muestra = np.mean(muestra_datos)
desviacion_estandar_muestra = np.std(muestra_datos)

# Intervalos de 1, 2 y 3 alrededor de la media
intervalo_1_sigma = (media_muestra - desviacion_estandar_muestra, media_muestra + desviacion_estandar_muestra)
intervalo_2_sigma = (media_muestra - 2 * desviacion_estandar_muestra, media_muestra + 2 * desviacion_estandar_muestra)
intervalo_3_sigma = (media_muestra - 3 * desviacion_estandar_muestra, media_muestra + 3 * desviacion_estandar_muestra)

# Graficar la Distribuci贸n Normal
plt.hist(muestra_datos, bins=30, edgecolor='black', alpha=0.7, density=True)
plt.axvline(media_muestra, color='red', linestyle='--', label='Media')
plt.axvline(intervalo_1_sigma[0], color='green', linestyle='--', label='1')
plt.axvline(intervalo_1_sigma[1], color='green', linestyle='--')
plt.axvline(intervalo_2_sigma[0], color='orange', linestyle='--', label='2')
plt.axvline(intervalo_2_sigma[1], color='orange', linestyle='--')
plt.axvline(intervalo_3_sigma[0], color='purple', linestyle='--', label='3')
plt.axvline(intervalo_3_sigma[1], color='purple', linestyle='--')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribuci贸n Normal')
plt.legend()
plt.show()

# Imprimir la proporci贸n de datos dentro de los intervalos de 1, 2 y 3
print("Proporci贸n de datos en el intervalo 1:", np.mean((muestra_datos >= intervalo_1_sigma[0]) & (muestra_datos <= intervalo_1_sigma[1])))
print("Proporci贸n de datos en el intervalo 2:", np.mean((muestra_datos >= intervalo_2_sigma[0]) & (muestra_datos <= intervalo_2_sigma[1])))
print("Proporci贸n de datos en el intervalo 3:", np.mean((muestra_datos >= intervalo_3_sigma[0]) & (muestra_datos <= intervalo_3_sigma[1])))
```
