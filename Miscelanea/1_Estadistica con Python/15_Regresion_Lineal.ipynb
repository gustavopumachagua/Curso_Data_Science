**Ejemplo: Regresi贸n Lineal Simple**

Supongamos que tenemos un conjunto de datos que contiene la cantidad de horas de estudio $x$ y las calificaciones de los estudiantes $y$ en un examen. Queremos construir un modelo de regresi贸n lineal simple para predecir las calificaciones en funci贸n de las horas de estudio.
