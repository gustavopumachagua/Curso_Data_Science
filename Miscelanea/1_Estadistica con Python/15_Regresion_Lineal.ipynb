# **15. Regresión Lineal con Python**

## **Qué es un modelo de regresión lineal**

Un modelo de regresión lineal es un enfoque estadístico que se utiliza para establecer una relación lineal entre una variable independiente (también conocida como característica o variable predictora) y una variable dependiente (también conocida como variable objetivo). El objetivo es encontrar la mejor línea recta que se ajuste a los datos para predecir o modelar el comportamiento de la variable dependiente en función de la variable independiente. A continuación, te proporcionaré una explicación detallada, ejemplos y gráficas para ilustrar el concepto de regresión lineal.

**Explicación:**

La regresión lineal asume que existe una relación lineal entre las variables. En su forma más simple, un modelo de regresión lineal se define por la ecuación:

$y = mx + b$

Donde:

- $y$ es la variable dependiente (variable objetivo).

- $x$ es la variable independiente (característica o variable predictora).

- $m$ es la pendiente de la línea, que representa el cambio en $y$ por un cambio unitario en $x$.

- $b$ es la ordenada al origen, que representa el valor de $y$ cuando $x$ es 0.

**Ejemplo: Regresión Lineal Simple**

Supongamos que tenemos un conjunto de datos que contiene la cantidad de horas de estudio $x$ y las calificaciones de los estudiantes $y$ en un examen. Queremos construir un modelo de regresión lineal simple para predecir las calificaciones en función de las horas de estudio.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
hours = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
scores = np.array([65, 75, 80, 85, 90, 92, 95, 98, 100])

# Crear una instancia de Regresión Lineal
model = LinearRegression()

# Entrenar el modelo con los datos
model.fit(hours.reshape(-1, 1), scores)

# Predicciones con el modelo
predicted_scores = model.predict(hours.reshape(-1, 1))

# Graficar los datos y la línea de regresión
plt.scatter(hours, scores, color='blue', label='Datos')
plt.plot(hours, predicted_scores, color='red', linewidth=2, label='Regresión Lineal')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.show()
```

En esta gráfica, los puntos azules representan los datos reales de horas de estudio y calificaciones, mientras que la línea roja es la línea de regresión que se ajusta mejor a los datos. El modelo de regresión lineal busca minimizar la distancia entre la línea y los puntos.

**Ventajas de la Regresión Lineal:**

- **Interpretación sencilla:** La relación lineal es fácil de interpretar y entender.
- **Buen punto de partida:** La regresión lineal es un buen punto de partida para la modelización de datos antes de explorar modelos más complejos.

**Nota:** La regresión lineal simple es adecuada cuando se asume que la relación entre las variables es lineal. En casos donde la relación no es lineal, se pueden explorar modelos de regresión más avanzados, como la regresión polinómica o la regresión lineal múltiple.

## **Fundamentos**

**Fundamentos de la Regresión Lineal:**

La regresión lineal es una técnica estadística que busca modelar la relación entre una variable independiente (predictora) y una variable dependiente (objetivo) mediante una línea recta. Su objetivo es encontrar la mejor línea que minimice la distancia entre los puntos de datos y la línea, lo que permite hacer predicciones y comprender cómo la variable independiente afecta a la variable dependiente. A continuación, te proporcionaré una explicación detallada, ejemplos y gráficas para entender los fundamentos de la regresión lineal.

**Explicación:**

En la regresión lineal, se asume que existe una relación lineal entre las variables independiente $x$ y dependiente $y$. La forma general de la ecuación de una línea recta es:

$y = mx + b$

Donde:

- $y$ es la variable dependiente (objetivo).

- $x$ es la variable independiente (predictora).

- $m$ es la pendiente de la línea, que indica cuánto cambia $y$ por un cambio unitario en $x$.

- $b$ es el término de intersección, que representa el valor de $y$ cuando $x$ es 0.

El objetivo de la regresión lineal es encontrar los valores de $m$ y $b$ que mejor se ajusten a los datos observados, minimizando la suma de los cuadrados de las diferencias entre los valores reales y los predichos $y - mx - b$.

**Ejemplo: Regresión Lineal Simple**

Supongamos que tenemos datos que muestran la relación entre la cantidad de horas de estudio $x$ y las calificaciones de los estudiantes $y$ en un examen. Queremos construir un modelo de regresión lineal simple para predecir las calificaciones en función de las horas de estudio.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
hours = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
scores = np.array([65, 75, 80, 85, 90, 92, 95, 98, 100])

# Crear una instancia de Regresión Lineal
model = LinearRegression()

# Entrenar el modelo con los datos
model.fit(hours.reshape(-1, 1), scores)

# Parámetros aprendidos
m = model.coef_[0]
b = model.intercept_

# Predicciones con el modelo
predicted_scores = model.predict(hours.reshape(-1, 1))

# Graficar los datos y la línea de regresión
plt.scatter(hours, scores, color='blue', label='Datos')
plt.plot(hours, predicted_scores, color='red', linewidth=2, label='Regresión Lineal')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.show()
```

En la gráfica, los puntos azules representan los datos reales, la línea roja es la línea de regresión ajustada y las distancias verticales entre los puntos y la línea representan los residuos.

**Ventajas y Limitaciones:**

- **Ventajas:** La regresión lineal es fácil de interpretar, computacionalmente eficiente y puede ser un buen punto de partida para el análisis. Ayuda a identificar relaciones y hacer predicciones simples.
- **Limitaciones:** La regresión lineal asume una relación lineal entre las variables, lo que puede no ser válido en todos los casos. No maneja bien relaciones no lineales ni datos atípicos.

**Nota:** La regresión lineal simple es una versión básica de la regresión lineal. Para problemas más complejos, como la regresión lineal múltiple, donde hay múltiples variables predictoras, se pueden aplicar técnicas más avanzadas.

## **Suposiciones**

En la regresión lineal, existen varias suposiciones fundamentales que deben cumplirse para que los resultados y las interpretaciones del modelo sean válidos y confiables. Estas suposiciones son esenciales para garantizar que el modelo de regresión lineal sea una representación precisa de los datos y las relaciones entre las variables. A continuación, se detallan las suposiciones clave en la regresión lineal:

**1. Linealidad:** La relación entre la variable independiente $x$ y la variable dependiente $y$ debe ser lineal. Esto significa que los cambios en $y$ son proporcionales a los cambios en $x$ a lo largo de toda la gama de datos.

**2. Homocedasticidad:** Los errores de predicción, también conocidos como residuos, deben tener una varianza constante a lo largo de todo el rango de valores de $x$. En otras palabras, la dispersión de los residuos no debe cambiar a medida que cambian los valores de $x$.

**3. Independencia de Residuos:** Los residuos deben ser independientes entre sí, lo que significa que el error en la predicción de un punto no debe estar correlacionado con el error en la predicción de otro punto.

**4. Normalidad de Residuos:** Los residuos deben seguir una distribución normal. Esto es importante para realizar inferencias estadísticas y pruebas de hipótesis sobre los coeficientes del modelo.

**5. Ausencia de Multicolinealidad:** En la regresión lineal múltiple (cuando hay más de una variable independiente), las variables independientes no deben estar altamente correlacionadas entre sí. La multicolinealidad puede dificultar la interpretación de los coeficientes y aumentar la varianza de los estimadores.

**Cómo Evaluar las Suposiciones:**

Para evaluar si las suposiciones de la regresión lineal se cumplen, se pueden realizar diversas pruebas y análisis:

- **Gráficos de Residuos:** Visualizar los residuos en gráficos de dispersión o histogramas para evaluar la homocedasticidad y la normalidad.

- **Gráficos de P-P y Q-Q:** Estos gráficos comparan la distribución de los residuos con una distribución normal para evaluar la normalidad de los residuos.

- **Análisis de Influencia:** Identificar puntos influyentes y atípicos que puedan afectar el modelo.

- **Pruebas de Hipótesis:** Realizar pruebas de normalidad en los residuos y pruebas de heterocedasticidad para evaluar las suposiciones.

Si las suposiciones no se cumplen, es importante considerar ajustes en el modelo o aplicar técnicas más avanzadas que puedan manejar mejor los datos y las relaciones subyacentes. Cumplir con estas suposiciones garantiza que el modelo de regresión lineal sea confiable y adecuado para la interpretación y las predicciones.

## **Contexto y datos**

El contexto y los datos son fundamentales al implementar un modelo de regresión lineal. El contexto se refiere al escenario o problema en el que se está aplicando la regresión lineal, mientras que los datos son la información que se utiliza para construir y evaluar el modelo. Aquí te proporciono detalles sobre el contexto y los datos al implementar un modelo de regresión lineal:

**Contexto:**

El contexto proporciona el marco en el que se aplicará el modelo de regresión lineal y cómo se interpretarán los resultados. Algunos aspectos clave del contexto incluyen:

1. **Definición del Problema:** ¿Qué problema estás tratando de resolver con la regresión lineal? Esto podría ser hacer predicciones, entender relaciones entre variables o realizar análisis de causa y efecto.

2. **Relevancia:** ¿Por qué es importante este análisis de regresión lineal en el contexto en el que se está aplicando? ¿Qué impacto tendrá en la toma de decisiones?

3. **Interpretación:** ¿Cómo se interpretarán los resultados del modelo en el contexto? ¿Cómo se comunicarán las conclusiones a las partes interesadas?

**Datos:**

Los datos son la base del modelo de regresión lineal. Deben ser recolectados y preparados de manera adecuada para que el modelo sea válido y útil. Algunos aspectos clave de los datos incluyen:

1. **Calidad de los Datos:** Los datos deben ser precisos, completos y confiables. Cualquier error o falta de calidad en los datos puede afectar negativamente la validez del modelo.

2. **Variables Relevantes:** Selecciona las variables independientes (predictoras) y la variable dependiente (objetivo) que sean relevantes para el problema y el contexto. Variables irrelevantes pueden introducir ruido en el modelo.

3. **Tamaño del Conjunto de Datos:** El tamaño del conjunto de datos debe ser suficientemente grande para obtener resultados significativos y generalizables. Demasiados datos pueden llevar a sobreajuste.

4. **Distribución de Datos:** La distribución de los datos y la relación entre las variables deben ser coherentes con las suposiciones de la regresión lineal.

5. **Datos Faltantes:** Si hay datos faltantes, es importante manejarlos adecuadamente, ya que pueden afectar la precisión y validez del modelo.

6. **Atipicidades:** Los valores atípicos pueden tener un impacto significativo en los resultados de la regresión lineal. Evalúa si es necesario tratarlos o considerar su influencia.

**Ejemplo de Contexto y Datos: Predicción de Ventas**

Supongamos que estás trabajando en una empresa minorista y deseas predecir las ventas mensuales en función del gasto en publicidad en diferentes canales (televisión, radio y redes sociales). El contexto sería mejorar la asignación de recursos publicitarios para maximizar las ventas. Los datos incluirían registros mensuales de ventas, gastos en publicidad y otras posibles variables relacionadas.

En este contexto, el objetivo sería construir un modelo de regresión lineal que relacione el gasto en publicidad con las ventas mensuales y permita realizar predicciones. Los datos tendrían que ser recopilados y preparados adecuadamente, asegurando que cumplan con las suposiciones de la regresión lineal y que sean representativos del problema en cuestión.

En resumen, el contexto y los datos son elementos cruciales al implementar un modelo de regresión lineal. La comprensión del problema y la selección adecuada de datos son esenciales para construir un modelo confiable y útil que pueda brindar información valiosa en el contexto adecuado.

## **Visualización de los datos**

La visualización de los datos es una parte esencial al implementar un modelo de regresión lineal, ya que te permite entender la relación entre las variables, identificar patrones, detectar posibles problemas y tomar decisiones informadas sobre cómo construir y ajustar el modelo. Aquí te proporciono algunas técnicas de visualización que son útiles al implementar un modelo de regresión lineal:

1. **Gráfico de Dispersión:** Un gráfico de dispersión es una herramienta fundamental para visualizar la relación entre la variable independiente $x$ y la variable dependiente $y$. Cada punto representa una observación en el conjunto de datos. Un patrón lineal en el gráfico puede indicar una relación adecuada para la regresión lineal.

2. **Línea de Regresión:** Dibuja la línea de regresión en el gráfico de dispersión para visualizar cómo se ajusta a los datos. Esto te ayudará a identificar si una relación lineal es apropiada y cómo la línea se adapta a los puntos.

3. **Gráficos de Residuos:** Los gráficos de residuos muestran los errores de predicción (residuos) en función de los valores predichos. Puedes usar gráficos de residuos para verificar las suposiciones de linealidad, homocedasticidad y normalidad de los residuos.

4. **Gráficos de Influencia:** Los gráficos de influencia pueden ayudarte a identificar observaciones que tienen un impacto significativo en el modelo. Los valores atípicos o influyentes pueden afectar la calidad del ajuste y la interpretación.

5. **Histogramas y Q-Q Plots de Residuos:** Estos gráficos te permiten evaluar la normalidad de los residuos. Un histograma de residuos normalmente distribuidos se asemejará a una curva de campana, y un Q-Q plot comparará la distribución de los residuos con una distribución teórica normal.

6. **Gráficos de Ajuste:** Una vez que hayas ajustado el modelo, puedes superponer la línea de regresión en el gráfico de dispersión para visualizar cómo se comparan las predicciones del modelo con los datos reales.

7. **Subgráficos por Categoría:** Si tienes variables categóricas en tu modelo, puedes crear subgráficos de dispersión para cada categoría para analizar si hay diferencias en las relaciones entre grupos.

8. **Matriz de Correlación:** Si tienes múltiples variables independientes, una matriz de correlación puede ayudarte a entender las relaciones entre todas las variables y su posible impacto en la variable dependiente.

9. **Gráficos de Tendencia Temporal:** Si los datos tienen una dimensión temporal, como series de tiempo, gráficos de línea o gráficos de dispersión de tiempo pueden mostrar patrones y tendencias.

La elección de la técnica de visualización depende del contexto y la naturaleza de los datos. Visualizar los datos te brinda una comprensión más profunda de las relaciones entre las variables, lo que te ayuda a tomar decisiones informadas sobre cómo construir y ajustar tu modelo de regresión lineal.

## **Creación del modelo con Scikit-Learn**

Crear un modelo de regresión lineal con Scikit-Learn en Python es un proceso sencillo y eficiente. Scikit-Learn es una biblioteca de aprendizaje automático que proporciona una interfaz fácil de usar para crear y entrenar modelos. A continuación, te proporcionaré una explicación detallada junto con un ejemplo de cómo crear un modelo de regresión lineal utilizando Scikit-Learn, y también te mostraré algunas gráficas relacionadas.

**Paso 1: Importar las bibliotecas necesarias**

Primero, importa las bibliotecas necesarias, que incluyen NumPy, Matplotlib y las clases de Scikit-Learn para el modelo de regresión lineal.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

**Paso 2: Preparar los datos**

Supongamos que tienes datos de horas de estudio $(x)$ y calificaciones $(y)$ como antes. Define los datos y conviértelos en matrices NumPy.

```python
hours = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
scores = np.array([65, 75, 80, 85, 90, 92, 95, 98, 100])
```

**Paso 3: Crear el modelo y entrenarlo**

Crea una instancia del modelo de regresión lineal y luego ajústalo a los datos utilizando el método `.fit()`.

```python
# Crear una instancia de Regresión Lineal
model = LinearRegression()

# Entrenar el modelo con los datos
model.fit(hours.reshape(-1, 1), scores)
```

**Paso 4: Hacer predicciones y graficar**

Una vez que el modelo está entrenado, puedes hacer predicciones y graficar los resultados.

```python
# Predicciones con el modelo
predicted_scores = model.predict(hours.reshape(-1, 1))

# Graficar los datos y la línea de regresión
plt.scatter(hours, scores, color='blue', label='Datos')
plt.plot(hours, predicted_scores, color='red', linewidth=2, label='Regresión Lineal')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal con Scikit-Learn')
plt.legend()
plt.show()
```

En esta gráfica, los puntos azules representan los datos reales, y la línea roja es la línea de regresión que se ajusta mejor a los datos.

**Resultado y Visualización:**

El gráfico te mostrará cómo se ajusta la línea de regresión a los datos, permitiéndote visualizar la relación entre las horas de estudio y las calificaciones. También puedes utilizar el modelo entrenado para hacer predicciones para nuevos valores de horas de estudio.

Recuerda que este es un ejemplo sencillo. En situaciones más complejas, es posible que tengas múltiples variables predictoras, necesites realizar pruebas de rendimiento y ajustar parámetros del modelo para obtener resultados óptimos.

La creación de un modelo de regresión lineal con Scikit-Learn es un proceso efectivo y Scikit-Learn proporciona una interfaz intuitiva y poderosa para ello.
