## **Distribuci贸n Normal (Gaussiana)**

La Distribuci贸n Normal, tambi茅n conocida como la Distribuci贸n Gaussiana o Campana de Gauss, es una de las distribuciones estad铆sticas m谩s importantes y comunes en la teor铆a de probabilidades y estad铆sticas. Es ampliamente utilizada debido a su aparici贸n natural en muchas situaciones del mundo real y su propiedad de describir numerosos fen贸menos observados. Aqu铆 tienes una explicaci贸n detallada, ejemplos y gr谩ficos que ilustran la Distribuci贸n Normal.

**Explicaci贸n:**

La Distribuci贸n Normal tiene las siguientes caracter铆sticas:

1. **Forma de Campana:** La curva de la distribuci贸n es sim茅trica alrededor de su valor medio (media). Esto significa que la mitad de los datos est谩n a la izquierda de la media y la otra mitad a la derecha.

2. **Media y Mediana Igual:** La media y la mediana de la distribuci贸n normal son iguales.

3. **Desviaci贸n Est谩ndar:** La dispersi贸n de los datos est谩 determinada por la desviaci贸n est谩ndar. Los datos m谩s cercanos a la media son m谩s frecuentes, mientras que los valores m谩s alejados son menos probables.

4. **Regla Emp铆rica:** Alrededor del 68% de los datos caen dentro de una desviaci贸n est谩ndar de la media, aproximadamente el 95% caen dentro de dos desviaciones est谩ndar y cerca del 99.7% caen dentro de tres desviaciones est谩ndar.

**Ejemplo: Distribuci贸n Normal y Generaci贸n de Datos**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par谩metros de la distribuci贸n normal
mu = 0         # Media
sigma = 1      # Desviaci贸n est谩ndar

# Generar datos de una distribuci贸n normal
np.random.seed(0)
data = np.random.normal(mu, sigma, 1000)

# Crear un histograma para visualizar los datos
plt.hist(data, bins=30, density=True, edgecolor='black')

# Calcular la funci贸n de densidad de probabilidad
x = np.linspace(-5, 5, 100)
pdf = norm.pdf(x, mu, sigma)

# Graficar la funci贸n de densidad de probabilidad
plt.plot(x, pdf, color='red', linewidth=2)
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribuci贸n Normal (Gaussiana)')
plt.show()
```

En este ejemplo, estamos generando datos aleatorios de una distribuci贸n normal con una media (mu) de 0 y una desviaci贸n est谩ndar (sigma) de 1. Luego, creamos un histograma para visualizar la distribuci贸n de los datos y superponemos la funci贸n de densidad de probabilidad (PDF) de la distribuci贸n normal utilizando `scipy.stats.norm.pdf()`.

**Interpretaci贸n del Gr谩fico:**

- La curva roja es la funci贸n de densidad de probabilidad de una distribuci贸n normal con media 0 y desviaci贸n est谩ndar 1.
- La mayor铆a de los datos caen cerca de la media (0) y se dispersan menos a medida que nos alejamos de la media.

**Usos y Conclusiones:**

- La Distribuci贸n Normal es fundamental en la estad铆stica y se utiliza en muchos modelos y m茅todos estad铆sticos.
- Describe muchos fen贸menos naturales y sociales, como la altura de las personas, puntajes de ex谩menes, mediciones cient铆ficas, etc.
- Ayuda a comprender y modelar la variabilidad en los datos y proporciona una base para muchas t茅cnicas estad铆sticas.

En resumen, la Distribuci贸n Normal es una distribuci贸n clave en estad铆sticas y ciencia de datos debido a su aparici贸n en numerosos contextos del mundo real y sus propiedades estad铆sticas bien definidas.

## **Distribuci贸n Normal, Uniforme y Exponencial**

**Distribuci贸n Normal:**

La Distribuci贸n Normal, tambi茅n conocida como la Distribuci贸n Gaussiana o Campana de Gauss, es una distribuci贸n continua que es fundamental en estad铆sticas debido a su aparici贸n en una variedad de fen贸menos naturales y su importancia en muchos m茅todos estad铆sticos.

**Caracter铆sticas Clave:**

- Forma de campana sim茅trica alrededor de la media.
- La media y la mediana son iguales.
- Aproximadamente el 68% de los datos caen dentro de una desviaci贸n est谩ndar de la media (Regla Emp铆rica).
- Se caracteriza por dos par谩metros: la media (渭) y la desviaci贸n est谩ndar ().

**Ejemplo: Distribuci贸n Normal**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par谩metros de la distribuci贸n normal
mu = 0         # Media
sigma = 1      # Desviaci贸n est谩ndar

# Generar datos de una distribuci贸n normal
data_normal = np.random.normal(mu, sigma, 1000)

# Crear un histograma para visualizar los datos
plt.hist(data_normal, bins=30, density=True, edgecolor='black')

# Calcular la funci贸n de densidad de probabilidad
x = np.linspace(-5, 5, 100)
pdf_normal = norm.pdf(x, mu, sigma)

# Graficar la funci贸n de densidad de probabilidad
plt.plot(x, pdf_normal, color='red', linewidth=2)
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribuci贸n Normal (Gaussiana)')
plt.show()
```

**Distribuci贸n Uniforme:**

La Distribuci贸n Uniforme describe una situaci贸n en la que todos los valores dentro de un intervalo tienen la misma probabilidad de ocurrencia. Es 煤til para modelar situaciones donde todos los resultados son igualmente probables.

**Caracter铆sticas Clave:**

- Los valores tienen la misma probabilidad de ocurrencia en un intervalo dado.
- La funci贸n de densidad es constante dentro del intervalo y cero fuera de 茅l.
- Se caracteriza por dos par谩metros: el valor m铆nimo (a) y el valor m谩ximo (b).

**Ejemplo: Distribuci贸n Uniforme**

```python
# Par谩metros de la distribuci贸n uniforme
a = 0       # Valor m铆nimo
b = 10      # Valor m谩ximo

# Generar datos de una distribuci贸n uniforme
data_uniform = np.random.uniform(a, b, 1000)

# Crear un histograma para visualizar los datos
plt.hist(data_uniform, bins=30, density=True, edgecolor='black')

# Graficar la funci贸n de densidad constante
x_uniform = np.linspace(-1, 11, 100)
pdf_uniform = np.full_like(x_uniform, 1 / (b - a))
plt.plot(x_uniform, pdf_uniform, color='red', linewidth=2)
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribuci贸n Uniforme')
plt.show()
```

**Distribuci贸n Exponencial:**

La Distribuci贸n Exponencial describe el tiempo que debe transcurrir antes de que ocurra un evento en un proceso de Poisson. Es com煤nmente utilizada para modelar tiempos de espera.

**Caracter铆sticas Clave:**
- Modela la probabilidad de que ocurra un evento despu茅s de un tiempo t desde el 煤ltimo evento.
- Tiene un 煤nico par谩metro: la tasa (位), que representa el n煤mero promedio de eventos en un intervalo de unidad.

**Ejemplo: Distribuci贸n Exponencial**

```python
from scipy.stats import expon

# Par谩metro de la distribuci贸n exponencial (tasa)
lambda_param = 0.5

# Generar datos de una distribuci贸n exponencial
data_exponential = np.random.exponential(scale=1/lambda_param, size=1000)

# Crear un histograma para visualizar los datos
plt.hist(data_exponential, bins=30, density=True, edgecolor='black')

# Calcular la funci贸n de densidad de probabilidad exponencial
x_exponential = np.linspace(0, 10, 100)
pdf_exponential = expon.pdf(x_exponential, scale=1/lambda_param)
plt.plot(x_exponential, pdf_exponential, color='red', linewidth=2)
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribuci贸n Exponencial')
plt.show()
```

**Conclusiones:**

- La Distribuci贸n Normal es sim茅trica y ampliamente utilizada en an谩lisis estad铆sticos.
- La Distribuci贸n Uniforme es 煤til para situaciones donde todos los valores son igualmente probables.
- La Distribuci贸n Exponencial modela tiempos de espera en procesos de Poisson.
