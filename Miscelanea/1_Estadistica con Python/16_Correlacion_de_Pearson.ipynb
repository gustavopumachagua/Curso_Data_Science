# **16. Correlación de Pearson y cómo crear Mapas de Calor de la Matriz de Correlaciones con Python**

## **Motivación para usar la correlación**

La correlación de Pearson es una medida estadística que evalúa la relación lineal entre dos variables numéricas. Es una herramienta útil para comprender la dirección y la fuerza de la relación entre las variables, lo que puede ser especialmente valioso al explorar datos y construir modelos, como la regresión lineal. A continuación, te proporcionaré una explicación detallada, ejemplos y gráficos para ilustrar la motivación detrás del uso de la correlación de Pearson.

**Explicación:**

La correlación de Pearson calcula un valor entre -1 y 1 que describe la relación lineal entre dos variables. Los valores extremos son los siguientes:
- 1: Indica una correlación positiva perfecta, lo que significa que a medida que una variable aumenta, la otra también lo hace en la misma proporción.
- -1: Indica una correlación negativa perfecta, lo que significa que a medida que una variable aumenta, la otra disminuye en la misma proporción.
- 0: Indica que no hay correlación lineal entre las variables.

**Ejemplo: Correlación entre Horas de Estudio y Calificaciones**

Supongamos que tienes datos de horas de estudio $(x)$ y calificaciones $(y)$ de varios estudiantes. Quieres determinar si hay una relación lineal entre estas dos variables y, en caso afirmativo, si es positiva o negativa.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Datos de ejemplo
hours = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
scores = np.array([65, 75, 80, 85, 90, 92, 95, 98, 100])

# Calcular la correlación de Pearson
correlation, _ = pearsonr(hours, scores)

# Graficar los datos
plt.scatter(hours, scores, color='blue', label='Datos')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title(f'Correlación de Pearson: {correlation:.2f}')
plt.legend()
plt.show()
```

En esta gráfica, puedes visualizar cómo se relacionan las horas de estudio y las calificaciones. Además, la correlación de Pearson $(correlation)$ se muestra en el título del gráfico. Si la correlación es cercana a 1, indica una correlación positiva, y si es cercana a -1, indica una correlación negativa.

**Motivación para Usar la Correlación de Pearson:**

- **Identificar Relaciones Lineales:** La correlación de Pearson es útil para identificar si existe una relación lineal entre dos variables. Esto puede ser esencial al elegir qué variables incluir en un modelo o para entender la relación entre características.

- **Información en Análisis Exploratorio:** En la etapa de análisis exploratorio de datos, la correlación de Pearson te permite descubrir patrones y relaciones que podrían guiar tu análisis y enfoque.

- **Validación de Suposiciones:** Algunos modelos, como la regresión lineal, asumen una relación lineal entre variables. La correlación de Pearson puede ayudar a validar si esta suposición es razonable.

**Nota:** La correlación de Pearson solo mide relaciones lineales. Si las relaciones son no lineales, la correlación de Pearson puede subestimar o no capturar la verdadera relación entre las variables. En tales casos, otras medidas de correlación, como la correlación de Spearman o la correlación de Kendall, pueden ser más apropiadas.

Luego, creamos un gráfico de dispersión para visualizar cómo se distribuyen los puntos de datos y si se observa alguna tendencia lineal. ## **Interpretación visual de la correlación**

La interpretación visual de la correlación de Pearson implica utilizar gráficos de dispersión para visualizar la relación entre dos variables numéricas y comprender la dirección y fuerza de la correlación. Los gráficos de dispersión son especialmente útiles para entender cómo las variables se distribuyen y si existe alguna tendencia lineal. A continuación, te proporcionaré una explicación detallada junto con un ejemplo y gráficos para ilustrar la interpretación visual de la correlación de Pearson.

**Explicación:**

La interpretación visual de la correlación de Pearson implica crear un gráfico de dispersión en el que se coloquen los puntos de datos en un plano cartesiano, con una variable en el eje horizontal y la otra en el eje vertical. Si los puntos de datos forman una tendencia lineal, es una indicación de que existe una correlación entre las dos variables.

**Ejemplo: Correlación entre Horas de Estudio y Calificaciones**

Continuaremos con el ejemplo anterior de las horas de estudio y las calificaciones de los estudiantes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
hours = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
scores = np.array([65, 75, 80, 85, 90, 92, 95, 98, 100])

# Graficar los datos
plt.scatter(hours, scores, color='blue', label='Datos')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Gráfico de Dispersión: Horas de Estudio vs. Calificaciones')
plt.legend()
plt.show()
```

En esta gráfica de dispersión, cada punto representa un estudiante. El eje horizontal representa las horas de estudio, y el eje vertical representa las calificaciones. Si observas que los puntos tienden a agruparse en una dirección, es una indicación visual de la dirección de la correlación.

**Interpretación Visual:**

1. **Correlación Positiva:** Si los puntos en el gráfico de dispersión tienden a formar una línea ascendente de izquierda a derecha, indica una correlación positiva. Esto significa que a medida que las horas de estudio aumentan, las calificaciones también tienden a aumentar.

2. **Correlación Negativa:** Si los puntos tienden a formar una línea descendente de izquierda a derecha, indica una correlación negativa. Esto significa que a medida que las horas de estudio aumentan, las calificaciones tienden a disminuir.

3. **Correlación Cercana a Cero:** Si los puntos están dispersos de manera aleatoria y no siguen ninguna tendencia clara, la correlación es cercana a cero, lo que sugiere que no hay una relación lineal significativa entre las variables.

**Resultado Visual:**

En el contexto de las horas de estudio y las calificaciones, si observamos un gráfico de dispersión en el que los puntos tienden a formar una línea ascendente, podríamos inferir que existe una correlación positiva entre las horas de estudio y las calificaciones.

La interpretación visual de la correlación de Pearson a través de gráficos de dispersión proporciona una forma intuitiva y rápida de comprender la relación entre dos variables. Sin embargo, recuerda que la interpretación visual debe respaldarse con valores numéricos de correlación para obtener una comprensión completa y precisa.

## **Obtener la matriz de correlaciones**

Puedes obtener la matriz de correlaciones utilizando bibliotecas como NumPy o pandas en Python. La matriz de correlaciones muestra las correlaciones entre todas las combinaciones posibles de variables en un conjunto de datos. A continuación, te mostraré cómo obtener la matriz de correlaciones utilizando ambas bibliotecas.

**Usando NumPy:**

```python
import numpy as np

# Datos de ejemplo
data = np.array([[2, 65],
                 [3, 75],
                 [4, 80],
                 [5, 85],
                 [6, 90],
                 [7, 92],
                 [8, 95],
                 [9, 98],
                 [10, 100]])

# Calcular la matriz de correlaciones
correlation_matrix = np.corrcoef(data, rowvar=False)

print("Matriz de Correlaciones:")
print(correlation_matrix)
```

**Usando pandas:**

```python
import pandas as pd

# Datos de ejemplo en forma de DataFrame
data = pd.DataFrame({'Horas de Estudio': [2, 3, 4, 5, 6, 7, 8, 9, 10],
                     'Calificaciones': [65, 75, 80, 85, 90, 92, 95, 98, 100]})

# Calcular la matriz de correlaciones
correlation_matrix = data.corr()

print("Matriz de Correlaciones:")
print(correlation_matrix)
```

En la matriz de correlaciones, los valores en la diagonal principal son siempre 1, ya que una variable se correlaciona perfectamente consigo misma. Los valores fuera de la diagonal principal representan las correlaciones entre pares de variables. En este ejemplo, puedes ver que la correlación entre "Horas de Estudio" y "Calificaciones" es aproximadamente 0.9618, lo que indica una correlación positiva fuerte.

La matriz de correlaciones es útil para entender las relaciones entre todas las variables en un conjunto de datos y puede ser especialmente útil en análisis exploratorios y toma de decisiones relacionadas con modelos.

## **Graficar la matriz de correlaciones**

Para graficar la matriz de correlaciones de manera visualmente atractiva, puedes utilizar la biblioteca Seaborn en combinación con matplotlib en Python. Seaborn ofrece una función llamada `heatmap` que permite crear una representación gráfica de la matriz de correlaciones utilizando colores para resaltar los valores.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Datos de ejemplo en forma de DataFrame
data = pd.DataFrame({'Horas de Estudio': [2, 3, 4, 5, 6, 7, 8, 9, 10],
                     'Calificaciones': [65, 75, 80, 85, 90, 92, 95, 98, 100]})

# Calcular la matriz de correlaciones
correlation_matrix = data.corr()

# Crear un mapa de calor (heatmap) utilizando Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(data=correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlaciones')
plt.show()
```

En este código, estamos utilizando la función `sns.heatmap()` de Seaborn para crear un mapa de calor de la matriz de correlaciones. Los argumentos clave utilizados son:

- `data`: La matriz de correlaciones que queremos visualizar.

- `annot`: Indica si deseamos mostrar los valores de correlación en los cuadros del mapa de calor.

- `cmap`: La paleta de colores a utilizar para representar los valores de correlación.

- `center`: El valor central de la paleta de colores (0 en este caso).

La matriz de correlaciones se visualizará como un mapa de calor con colores que representan los valores de correlación. Los colores más oscuros indican correlaciones más negativas, mientras que los colores más claros indican correlaciones más positivas.

Esta visualización te proporcionará una representación gráfica rápida y comprensible de cómo las variables se correlacionan entre sí en el conjunto de datos.
