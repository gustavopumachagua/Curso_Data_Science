# **3. OPERACIONES BAYESIANAS**

## **Posibilidades y sumas**

En el análisis bayesiano, las operaciones de posibilidades y sumas son dos enfoques utilizados para combinar distribuciones de probabilidad y obtener una nueva distribución resultante. Estas operaciones son útiles cuando se tienen múltiples fuentes de información o evidencia y se desea combinarlas para obtener una visión más completa y precisa de la incertidumbre.

* **Operaciones de Posibilidades:**

Las operaciones de posibilidades se utilizan para combinar distribuciones de probabilidad independientes y obtener una distribución resultante. La operación principal es la multiplicación de distribuciones. Si tenemos dos distribuciones independientes `A` y `B`, la distribución resultante `C` se obtiene multiplicando las distribuciones `A` y `B`.

En Python, podemos realizar operaciones de posibilidades utilizando bibliotecas como NumPy. A continuación se muestra un ejemplo de cómo multiplicar dos distribuciones normales independientes:

```python
import numpy as np
import matplotlib.pyplot as plt

# Distribución A (Normal media 0, desviación estándar 1)
A = np.random.normal(0, 1, 1000)

# Distribución B (Normal media 2, desviación estándar 0.5)
B = np.random.normal(2, 0.5, 1000)

# Distribución resultante C
C = A * B

# Gráfica de la distribución resultante C
plt.hist(C, bins=30, density=True, color='skyblue', edgecolor='black', linewidth=1.2, alpha=0.7)
plt.xlabel('Valores', fontsize=12)
plt.ylabel('Densidad de probabilidad', fontsize=12)
plt.title('Operación de Posibilidades: Multiplicación de Distribuciones', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.axvline(np.mean(C), color='red', linestyle='--', linewidth=2, label='Media de C')
plt.legend(fontsize=10)
plt.show()
```

En este ejemplo, generamos dos distribuciones normales independientes (`A` y `B`) y las multiplicamos para obtener la distribución resultante `C`. La distribución resultante reflejará la combinación de incertidumbre de ambas distribuciones originales.

* **Operaciones de Sumas:**

Las operaciones de sumas se utilizan para combinar distribuciones de probabilidad dependientes y obtener una distribución resultante. La operación principal es la suma de distribuciones. Si tenemos dos distribuciones dependientes `A` y `B`, la distribución resultante `C` se obtiene sumando las distribuciones `A` y `B`.

En Python, podemos realizar operaciones de sumas utilizando bibliotecas como SciPy. A continuación se muestra un ejemplo de cómo sumar dos distribuciones normales dependientes:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Distribución A (Normal media 0, desviación estándar 1)
A = norm(loc=0, scale=1)

# Distribución B (Normal media 2, desviación estándar 1)
B = norm(loc=2, scale=1)

# Valores para evaluar la distribución resultante
x = np.linspace(-5, 7, 1000)

# Función de densidad de probabilidad (PDF) de la distribución resultante C
pdf_C = np.convolve(A.pdf(x), B.pdf(x), mode='same')

# Normalización de la PDF resultante
pdf_C /= np.sum(pdf_C)

# Gráfica de la distribución resultante C
plt.plot(x, pdf_C, color='skyblue', linewidth=2, label='Suma de A y B')
plt.xlabel('Valores', fontsize=12)
plt.ylabel('Densidad de probabilidad', fontsize=12)
plt.title('Operación de Sumas: Suma de Distribuciones', fontsize=14)
plt.legend(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.axvline(np.mean(x), color='red', linestyle='--', linewidth=2, label='Media de C')
plt.legend(fontsize=10)
plt.show()
```

En este ejemplo, creamos dos distribuciones normales dependientes (`A` y `B`) utilizando la biblioteca SciPy. Luego, sumamos las distribuciones `A` y `B` para obtener la distribución resultante `C`. La distribución resultante representa la combinación de las distribuciones originales y refleja la incertidumbre resultante de la suma.

En resumen, las operaciones bayesianas de posibilidades y sumas permiten combinar distribuciones de probabilidad independientes o dependientes para obtener una distribución resultante que refleje la combinación de incertidumbres. Estas operaciones son fundamentales en el análisis bayesiano y son útiles para integrar múltiples fuentes de información y obtener estimaciones más precisas.

## **Mínimos, máximos y mezclas**

Las operaciones bayesianas de mínimos, máximos y mezclas son técnicas que permiten combinar distribuciones de probabilidad para obtener nuevas distribuciones resultantes. Estas operaciones son útiles en el contexto bayesiano para modelar y analizar situaciones en las que se requiere combinar información proveniente de diferentes fuentes o distribuciones.

* **Operación de Mínimos:**

La operación de mínimos se utiliza para obtener la distribución resultante que representa el mínimo de dos distribuciones de probabilidad. Esta operación es útil cuando se desea modelar una situación en la que se considera el peor escenario posible o la menor cantidad de alguna variable de interés.

A continuación, se presenta un ejemplo de cómo realizar la operación de mínimos utilizando Python y la biblioteca SciPy:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Distribución A (Normal media 0, desviación estándar 1)
A = norm(loc=0, scale=1)

# Distribución B (Normal media 2, desviación estándar 1)
B = norm(loc=2, scale=1)

# Valores para evaluar la distribución resultante
x = np.linspace(-5, 5, 1000)

# Obtener la PDF resultante de la operación de mínimos
pdf_min = np.minimum(A.pdf(x), B.pdf(x))

# Gráfica de la distribución resultante
plt.plot(x, pdf_min, color='green', linestyle='--', label='PDF Mínima')
plt.xlabel('Valores', fontsize=12)
plt.ylabel('Densidad de probabilidad', fontsize=12)
plt.title('Operación de Mínimos', fontsize=14)
plt.legend(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.axvline(np.argmin(pdf_min), color='red', linestyle='--', linewidth=2, label='Mínimo')
plt.legend(fontsize=10)
plt.show()
```

En este ejemplo, se crean dos distribuciones `A` y `B` utilizando la función `norm` de SciPy, que representa distribuciones normales. Luego, se evalúa la función de densidad de probabilidad (`PDF`) de cada distribución en una serie de valores `x`. Finalmente, se utiliza la función `np.minimum` para calcular la `PDF` resultante de la operación de mínimos y se grafica.

* **Operación de Máximos:**

La operación de máximos se utiliza para obtener la distribución resultante que representa el máximo de dos distribuciones de probabilidad. Esta operación es útil cuando se desea modelar una situación en la que se considera el mejor escenario posible o la mayor cantidad de alguna variable de interés.

A continuación, se presenta un ejemplo de cómo realizar la operación de máximos utilizando Python y la biblioteca SciPy:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Distribución A (Normal media 0, desviación estándar 1)
A = norm(loc=0, scale=1)

# Distribución B (Normal media 2, desviación estándar 1)
B = norm(loc=2, scale=1)

# Valores para evaluar la distribución resultante
x = np.linspace(-5, 5, 1000)

# Obtener la PDF resultante de la operación de máximos
pdf_max = np.maximum(A.pdf(x), B.pdf(x))

# Gráfica de la distribución resultante
plt.plot(x, pdf_max, label='Distribución Máxima', color='orange', linestyle='-.', linewidth=2)

# Añadir gráficas individuales para cada distribución
plt.plot(x, A.pdf(x), label='Distribución A', color='blue', alpha=0.7)
plt.plot(x, B.pdf(x), label='Distribución B', color='green', alpha=0.7)

plt.xlabel('Valores', fontsize=12)
plt.ylabel('Densidad de probabilidad', fontsize=12)
plt.title('Operación de Máximos', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

En este ejemplo, se crean dos distribuciones `A` y `B` utilizando la función `norm` de SciPy, que representa distribuciones normales. Luego, se evalúa la función de densidad de probabilidad (`PDF`) de cada distribución en una serie de valores `x`. Finalmente, se utiliza la función `np.maximum` para calcular la `PDF` resultante de la operación de máximos y se grafica.

* **Operación de Mezclas:**

La operación de mezclas se utiliza para combinar dos o más distribuciones de probabilidad mediante una combinación ponderada de las mismas. Esto es útil cuando se desea modelar una situación en la que se tienen múltiples fuentes de información y se desea obtener una distribución resultante que tome en cuenta todas las fuentes.

A continuación, se presenta un ejemplo de cómo realizar la operación de mezclas utilizando Python y la biblioteca SciPy:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Distribución A (Normal media 0, desviación estándar 1)
A = norm(loc=0, scale=1)

# Distribución B (Normal media 2, desviación estándar 1)
B = norm(loc=2, scale=1)

# Pesos de las distribuciones
weight_A = 0.6
weight_B = 0.4

# Valores para evaluar la distribución resultante
x = np.linspace(-5, 5, 1000)

# Obtener la PDF resultante de la operación de mezclas
pdf_mix = weight_A * A.pdf(x) + weight_B * B.pdf(x)

# Gráfica de la distribución resultante
plt.plot(x, pdf_mix, label='Distribución Mezclada', color='purple', linestyle='--', linewidth=2)

# Añadir gráficas individuales para cada distribución
plt.plot(x, weight_A * A.pdf(x), label='Distribución A', color='blue', alpha=0.7)
plt.plot(x, weight_B * B.pdf(x), label='Distribución B', color='green', alpha=0.7)

plt.xlabel('Valores', fontsize=12)
plt.ylabel('Densidad de probabilidad', fontsize=12)
plt.title('Operación de Mezclas', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

En este ejemplo, se crean dos distribuciones `A` y `B` utilizando la función `norm` de SciPy, que representa distribuciones normales. Se definen los pesos `weight_A` y `weight_B` para las distribuciones `A` y `B`, respectivamente. Luego, se evalúa la función de densidad de probabilidad (`PDF`) de cada distribución en una serie de valores `x` y se realiza una combinación ponderada de las mismas utilizando los pesos. Finalmente, se grafica la distribución resultante de la operación de mezclas.

Es importante tener en cuenta que estos ejemplos son simplificados y se basan en la suposición de distribuciones normales para ilustrar las operaciones bayesianas. Sin embargo, estas operaciones también se pueden aplicar a otras distribuciones de probabilidad y se pueden adaptar según las necesidades específicas del problema.

## **Procesos de Poisson**

Un proceso de Poisson es un modelo estocástico utilizado para describir la ocurrencia de eventos o sucesos aleatorios en un intervalo continuo de tiempo o espacio. Este proceso se basa en la suposición de que los eventos ocurren de forma independiente y con una tasa constante promedio.

El proceso de Poisson se caracteriza por el parámetro lambda (`λ`), que representa la tasa media de eventos por unidad de tiempo o espacio. La distribución de Poisson se utiliza para modelar el número de eventos que ocurren en un intervalo determinado, y la función de densidad de probabilidad de la distribución de Poisson se define como:

$P(X=k) = (λ^k * e^{-λ}) / k!$

donde $X$ es una variable aleatoria que representa el número de eventos, $k$ es un número entero no negativo y $e$ es la base del logaritmo natural.

A continuación, se presenta un ejemplo de cómo generar y graficar un proceso de Poisson utilizando Python y la biblioteca NumPy:

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro lambda
lmbda = 2.5

# Generar proceso de Poisson con lambda y tamaño de muestra
size = 1000
poisson_process = np.random.poisson(lmbda, size)

# Histograma de frecuencias
plt.hist(poisson_process, bins=range(12), rwidth=0.8, density=True, color='skyblue', edgecolor='black', alpha=0.7)

plt.xlabel('Número de eventos', fontsize=12)
plt.ylabel('Probabilidad', fontsize=12)
plt.title('Proceso de Poisson', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

En este ejemplo, se establece el parámetro lambda (`lmbda`) con un valor de 2.5, lo que indica que se espera un promedio de 2.5 eventos por intervalo. Luego, se utiliza la función `np.random.poisson` de NumPy para generar un proceso de Poisson con el parámetro lambda y un tamaño de muestra de 1000.

Finalmente, se muestra un histograma de frecuencias que representa la distribución de los eventos generados. Cada barra del histograma representa el número de ocurrencias de eventos y la altura de la barra indica la probabilidad relativa de que ocurran esos eventos.

Es importante destacar que la distribución de Poisson es discreta y que los eventos ocurren en números enteros no negativos. La variabilidad del proceso de Poisson está determinada por el parámetro lambda, siendo mayor para valores de lambda más altos.

## **La distribución exponencial**

La distribución exponencial es una distribución de probabilidad continua que se utiliza para modelar el tiempo entre eventos que ocurren de manera independiente y a una tasa constante. Es especialmente útil para analizar procesos de decaimiento, tiempos de espera y duraciones.

La distribución exponencial se caracteriza por un parámetro lambda $(λ)$, que es la tasa media de ocurrencia de los eventos. El parámetro lambda determina la forma de la distribución y controla la tasa de decaimiento exponencial.

La función de densidad de probabilidad $(PDF)$ de la distribución exponencial se define como:

$f(x|λ) = λ * e^{-λx}$

donde $x$ es la variable aleatoria que representa el tiempo entre eventos.

A continuación se presenta un ejemplo de cómo generar y graficar una distribución exponencial utilizando Python y la biblioteca NumPy:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parámetro lambda
lmbda = 0.5

# Generar datos de una distribución exponencial
size = 1000
exponential_data = expon.rvs(scale=1/lmbda, size=size)

# Histograma de frecuencias
plt.hist(exponential_data, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7)

# Añadir línea de la distribución teórica
x = np.linspace(0, 8, 1000)
pdf_expon = expon.pdf(x, scale=1/lmbda)
plt.plot(x, pdf_expon, 'r-', label='PDF Teórica')

plt.xlabel('Valor', fontsize=12)
plt.ylabel('Densidad de probabilidad', fontsize=12)
plt.title('Distribución Exponencial', fontsize=14)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

En este ejemplo, se establece el parámetro lambda `(lmbda)` con un valor de 0.5, lo que indica que se espera un evento en promedio cada 2 unidades de tiempo (ya que $λ = 1 / 0.5 = 2$). Luego, se utiliza la función `expon.rvs` de la biblioteca SciPy para generar datos aleatorios siguiendo una distribución exponencial, utilizando el parámetro scale que se calcula como `1/lambda`.

Finalmente, se muestra un histograma de frecuencias que representa la distribución de los datos generados. La forma del histograma sigue una curva exponencial decreciente, lo que indica la probabilidad de que ocurran eventos en diferentes intervalos de tiempo.

Es importante destacar que la distribución exponencial tiene una cola larga hacia la derecha, lo que significa que hay una pequeña pero no nula probabilidad de que ocurran eventos con tiempos muy largos entre ellos.
