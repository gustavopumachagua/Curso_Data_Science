# **3. ANÁLISIS DISCRIMINANTE**

## **Funciones discriminantes lineales**

Las funciones discriminantes lineales (LDF, por sus siglas en inglés) son un enfoque utilizado en el análisis discriminante para clasificar observaciones en diferentes categorías utilizando una función lineal de las variables predictoras. El objetivo es encontrar un hiperplano que maximice la separación entre las categorías.

A continuación, se proporciona un ejemplo de cómo implementar una función discriminante lineal en Python utilizando la biblioteca scikit-learn:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Generar datos de ejemplo
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Ajustar la función discriminante lineal
model = LinearDiscriminantAnalysis()
model.fit(X, y)

# Generar una malla de puntos en el espacio de características
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Predecir las categorías para cada punto en la malla
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Graficar los datos de entrada y las regiones de decisión
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Función discriminante lineal')
plt.show()
```

En este ejemplo, se generan datos de clasificación binaria utilizando `make_classification` de scikit-learn. Luego, se ajusta una función discriminante lineal utilizando `LinearDiscriminantAnalysis`. Se genera una malla de puntos en el espacio de características y se utilizan los coeficientes aprendidos por el modelo para predecir las categorías correspondientes. Se grafican los datos de entrada y las regiones de decisión resultantes.

La función discriminante lineal busca encontrar una función lineal que permita separar las categorías en el espacio de características. En el ejemplo, el área sombreada representa las regiones de decisión, donde cada punto se clasifica en función de la categoría predicha. Un punto se asignará a la categoría que esté asociada con la región en la que se encuentre.

Las funciones discriminantes lineales son útiles para problemas de clasificación en los que las categorías se pueden separar mediante un hiperplano en el espacio de características. Sin embargo, cuando las categorías no son linealmente separables, se pueden utilizar otros enfoques más avanzados, como las máquinas de vectores de soporte (SVM) o métodos de clasificación no lineales.

## **Interpretación**

La interpretación de las funciones discriminantes lineales (LDF) consiste en comprender cómo las variables predictoras contribuyen a la separación y clasificación de las categorías en el modelo. Esto implica analizar los coeficientes de las variables predictoras y su relación con la variable objetivo.

A continuación, se proporciona un ejemplo de cómo interpretar las funciones discriminantes lineales en Python utilizando la biblioteca scikit-learn:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Generar datos de ejemplo
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Ajustar la función discriminante lineal
model = LinearDiscriminantAnalysis()
model.fit(X, y)

# Obtener los coeficientes y las intersecciones de las funciones discriminantes lineales
coef = model.coef_
intercept = model.intercept_

# Graficar los datos de entrada y las líneas de separación
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')

# Graficar las líneas de separación
x = np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100)
y = -(coef[0, 0] * x + intercept[0]) / coef[0, 1]
plt.plot(x, y, color='r', label='LDF')

plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Función discriminante lineal')
plt.legend()
plt.show()
```

En este ejemplo, después de ajustar el modelo de función discriminante lineal, se obtienen los coeficientes (`coef`) y las intersecciones (`intercept`) de las funciones discriminantes lineales. Estos coeficientes representan la importancia relativa de cada variable predictora en la clasificación de las categorías.

En la gráfica, se muestran los datos de entrada y las líneas de separación que corresponden a las funciones discriminantes lineales. Las líneas de separación indican la región en la que los puntos se clasificarán en una u otra categoría.

La interpretación de los coeficientes implica analizar su magnitud y signo. Un coeficiente positivo indica que un aumento en el valor de la variable predictora correspondiente se asocia con una mayor probabilidad de pertenecer a una categoría específica, mientras que un coeficiente negativo indica lo contrario. La magnitud del coeficiente indica la importancia relativa de la variable predictora en la clasificación.

Es importante tener en cuenta que la interpretación de los coeficientes de las funciones discriminantes lineales puede ser más compleja en casos con más variables predictoras, ya que se deben considerar las interacciones y las relaciones entre las variables.

En resumen, la interpretación de las funciones discriminantes lineales implica analizar los coeficientes y las intersecciones de las funciones para comprender cómo las variables predictoras contribuyen a la separación y clasificación de las categorías. Esto permite identificar qué variables tienen un mayor impacto en la clasificación y cómo se relacionan con la variable objetivo.
