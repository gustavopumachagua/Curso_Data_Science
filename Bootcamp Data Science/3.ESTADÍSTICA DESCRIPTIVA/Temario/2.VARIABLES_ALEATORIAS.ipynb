# **2. VARIABLES ALEATORIAS**

## **Variables aleatorias discretas y continuas**

En la teoría de la probabilidad, existen dos tipos de variables aleatorias: discretas y continuas. Estos tipos de variables se diferencian en cómo toman valores y en la forma en que se pueden representar gráficamente. A continuación, te proporcionaré una explicación detallada de las variables aleatorias discretas y continuas, junto con ejemplos en código Python y gráficas para ilustrar cada caso.

* **Variables aleatorias discretas:**

Una variable aleatoria discreta es aquella que puede tomar solo un conjunto numerable de valores distintos, es decir, valores aislados. Por lo general, estos valores son números enteros o un conjunto finito de números. Las variables aleatorias discretas se caracterizan por tener una función de probabilidad de masa (`PMF`) que asigna probabilidades a cada uno de los posibles valores.

Aquí hay un ejemplo en código Python que ilustra una variable aleatoria discreta:

```python
import numpy as np
import matplotlib.pyplot as plt

# Variable aleatoria discreta: lanzamiento de un dado
valores = np.arange(1, 7)  # Valores posibles: del 1 al 6
probabilidades = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])  # Probabilidades correspondientes

# Graficar la función de probabilidad de masa (PMF)
markerline, stemlines, baseline = plt.stem(valores, probabilidades, basefmt=" ")  # Utilizar basefmt para mejorar la visualización

# Personalizar las líneas y las marcas
plt.setp(stemlines, linewidth=2, color='blue')  # Personalizar las líneas verticales
plt.setp(markerline, markersize=8, markerfacecolor='red', markeredgecolor='red', marker='o')  # Personalizar las marcas

plt.xlabel('Valor')
plt.ylabel('Probabilidad')
plt.title('Variable aleatoria discreta: Lanzamiento de un dado')

# Personalizar el gráfico
plt.xticks(valores)
plt.yticks(np.arange(0, 0.25, 0.05))
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir etiquetas a las barras
for i, prob in enumerate(probabilidades):
    plt.text(valores[i], prob + 0.01, f'{prob:.2f}', ha='center')

# Mostrar el gráfico
plt.show()
```

En este ejemplo, consideramos una variable aleatoria discreta que representa el lanzamiento de un dado. Los posibles valores son del 1 al 6, y asignamos una probabilidad de 1/6 a cada uno de estos valores. Graficamos la función de probabilidad de masa (PMF) utilizando el método `stem()` de Matplotlib, donde los valores se representan en el eje `x` y las probabilidades en el eje `y`.

* **Variables aleatorias continuas:**

Una variable aleatoria continua es aquella que puede tomar un rango infinito de valores dentro de un intervalo específico. Estas variables se caracterizan por tener una función de densidad de probabilidad (`PDF`) que describe la probabilidad relativa de que la variable tome un valor dentro de un intervalo dado.

Aquí hay un ejemplo en código Python que ilustra una variable aleatoria continua:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Variable aleatoria continua: distribución normal
media = 0
desviacion_estandar = 1

# Generar una muestra de la variable aleatoria
muestra = np.random.normal(loc=media, scale=desviacion_estandar, size=1000)

# Graficar el histograma y la función de densidad de probabilidad (PDF)
plt.hist(muestra, bins=30, density=True, alpha=0.5, color='blue', edgecolor='black', label='Histograma')
x = np.linspace(-4, 4, 100)
pdf = norm.pdf(x, loc=media, scale=desviacion_estandar)
plt.plot(x, pdf, 'r-', linewidth=2, label='PDF')

# Personalizar ejes y título
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Variable aleatoria continua: Distribución normal')

# Personalizar leyenda y cuadrícula
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.show()
```

En este ejemplo, consideramos una variable aleatoria continua que sigue una distribución normal con media `0` y desviación estándar `1`. Generamos una muestra de la variable aleatoria utilizando la función `np.random.normal()`. Luego, graficamos el histograma de la muestra y la función de densidad de probabilidad (`PDF`) utilizando el método `hist()` y la función `norm.pdf()` de la biblioteca SciPy respectivamente. El histograma representa la distribución empírica de los valores, mientras que la curva de la `PDF` representa la distribución teórica de la variable aleatoria.

En resumen, las variables aleatorias discretas toman valores aislados y se representan mediante una función de probabilidad de masa (`PMF`), mientras que las variables aleatorias continuas toman valores dentro de un intervalo y se representan mediante una función de densidad de probabilidad (`PDF`). Python, junto con las bibliotecas como NumPy y Matplotlib, proporciona herramientas para trabajar con variables aleatorias y visualizar sus distribuciones.

## **Distribución de probabilidades**

La distribución de probabilidades se refiere a la forma en que se distribuyen los posibles valores de una variable aleatoria y las probabilidades asociadas a cada uno de ellos. Es una función matemática que describe las probabilidades de ocurrencia de los distintos valores de una variable aleatoria.

Existen diferentes tipos de distribuciones de probabilidades, cada una con características y propiedades específicas. Algunas de las distribuciones más comunes incluyen la distribución uniforme, la distribución normal (o gaussiana), la distribución binomial y la distribución de Poisson, entre otras.

Aquí tienes ejemplos de cómo trabajar con algunas de estas distribuciones en Python utilizando las bibliotecas NumPy y Matplotlib:

* **Distribución uniforme:**

La distribución uniforme asigna la misma probabilidad a todos los valores posibles dentro de un intervalo dado.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de la distribución uniforme
a = 0  # Valor mínimo del intervalo
b = 10  # Valor máximo del intervalo
size = 1000  # Tamaño de la muestra

muestra_uniforme = np.random.uniform(a, b, size)

# Graficar el histograma de la muestra
plt.hist(muestra_uniforme, bins=20, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución uniforme')

# Personalizar ejes y cuadrícula
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.show()
```

En este ejemplo, generamos una muestra de la distribución uniforme en el intervalo `[0, 10]` utilizando la función `np.random.uniform()`. Luego, graficamos el histograma de la muestra utilizando el método `hist()` de Matplotlib. La forma del histograma será aproximadamente plana, ya que todos los valores dentro del intervalo tienen la misma probabilidad.

* **Distribución normal:**

La distribución normal, también conocida como distribución gaussiana, es una de las distribuciones más importantes y ampliamente utilizadas. Tiene una forma de campana y está completamente determinada por su media (`μ`) y desviación estándar (`σ`).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generar una muestra de la distribución normal
media = 0
desviacion_estandar = 1
size = 1000

muestra_normal = np.random.normal(media, desviacion_estandar, size)

# Graficar el histograma y la curva de la distribución normal
plt.hist(muestra_normal, bins=30, density=True, alpha=0.5, color='skyblue', edgecolor='black', label='Histograma')
x = np.linspace(-4, 4, 100)
pdf = norm.pdf(x, media, desviacion_estandar)
plt.plot(x, pdf, 'r', label='Distribución normal')

# Personalizar ejes y cuadrícula
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución normal')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.show()
```

En este ejemplo, generamos una muestra de la distribución normal con media `0` y desviación estándar `1` utilizando la función `np.random.normal()`. Luego, graficamos el histograma de la muestra y la curva de la distribución normal utilizando el método `hist()` y la función `norm.pdf()` de la biblioteca SciPy respectivamente. La curva de la distribución normal muestra la forma de campana característica de esta distribución.

Estos son solo dos ejemplos de distribuciones de probabilidades. Python, junto con las bibliotecas NumPy y Matplotlib, proporciona numerosas herramientas para trabajar con diferentes distribuciones de probabilidades, generar muestras aleatorias y visualizar sus características.

## **Esperanza y varianza de una variable aleatoria**

La esperanza y la varianza son dos medidas estadísticas importantes que describen las características de una variable aleatoria.

* **Esperanza:**

La esperanza, también conocida como valor esperado o media, es una medida de tendencia central que representa el promedio ponderado de los posibles valores de una variable aleatoria. Matemáticamente, se calcula como la suma de los productos de cada valor posible por su respectiva probabilidad.

En código Python, podemos calcular la esperanza de una variable aleatoria utilizando la función `np.sum()` para realizar la suma ponderada de los valores y las probabilidades correspondientes.

```python
import numpy as np

# Variable aleatoria discreta: lanzamiento de un dado
valores = np.array([1, 2, 3, 4, 5, 6])
probabilidades = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

# Calcular la esperanza
esperanza = np.sum(valores * probabilidades)

print("Esperanza:", esperanza)
```

En este ejemplo, consideramos una variable aleatoria discreta que representa el lanzamiento de un dado. Los posibles valores son del 1 al 6, y asignamos una probabilidad de 1/6 a cada uno de estos valores. Calculamos la esperanza multiplicando cada valor por su probabilidad correspondiente y luego sumando los resultados. El resultado es el promedio ponderado de los valores, que en este caso es 3.5.

* **Varianza:**

La varianza es una medida de dispersión que indica qué tan dispersos están los valores de una variable aleatoria alrededor de su media. Matemáticamente, se calcula como la suma de los cuadrados de las diferencias entre cada valor posible y la media, ponderadas por sus respectivas probabilidades.

En código Python, podemos calcular la varianza utilizando la función `np.sum()` para realizar la suma ponderada de los cuadrados de las diferencias.

```python
import numpy as np

# Variable aleatoria discreta: lanzamiento de un dado
valores = np.array([1, 2, 3, 4, 5, 6])
probabilidades = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

# Calcular la media
media = np.sum(valores * probabilidades)

# Calcular la varianza
varianza = np.sum((valores - media) ** 2 * probabilidades)

print("Varianza:", varianza)
```

En este ejemplo, utilizamos los mismos valores y probabilidades que en el ejemplo anterior. Calculamos primero la media (esperanza) de la variable aleatoria. Luego, calculamos la varianza sumando los cuadrados de las diferencias entre cada valor y la media, ponderadas por sus respectivas probabilidades. El resultado es una medida de dispersión que indica qué tan dispersos están los valores alrededor de la media.

La esperanza y la varianza son medidas importantes para comprender la distribución y las propiedades de una variable aleatoria. En Python, podemos utilizar NumPy u otras bibliotecas estadísticas para realizar estos cálculos de manera eficiente y conveniente.
