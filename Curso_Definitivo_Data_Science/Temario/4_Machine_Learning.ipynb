# **31. KNN Imputer | Multivariate Imputation | Handling Missing**

La imputaci√≥n de valores faltantes utilizando m√©todos m√°s avanzados como el KNN Imputer y la Imputaci√≥n Multivariante son estrategias m√°s sofisticadas que pueden ser √∫tiles cuando tienes varios atributos correlacionados. A continuaci√≥n, te proporciono una explicaci√≥n detallada con ejemplos en c√≥digo para ambos m√©todos.

1. **KNN Imputer:**
   - KNN Imputer es un m√©todo que utiliza los k vecinos m√°s cercanos para imputar valores faltantes. Calcula la distancia entre observaciones para encontrar vecinos cercanos y luego imputa los valores faltantes bas√°ndose en los valores de los vecinos.

````python
import pandas as pd
from sklearn.impute import KNNImputer
import numpy as np

# Crear un DataFrame de ejemplo con datos num√©ricos faltantes
data = pd.DataFrame({
       'A': [1, 2, np.nan, 4],
       'B': [5, np.nan, np.nan, 8],
       'C': [9, 10, 11, np.nan]})

# Imputar valores num√©ricos faltantes con KNN
imputer_knn = KNNImputer(n_neighbors=2)
data_knn_imputed = pd.DataFrame(imputer_knn.fit_transform(data), columns=data.columns)

print("Datos originales:")
print(data)

print("\nDatos con valores num√©ricos faltantes imputados con KNN:")
print(data_knn_imputed)
````

2. **Imputaci√≥n Multivariante:**
   - La imputaci√≥n multivariante se basa en modelos estad√≠sticos y utiliza la informaci√≥n de m√∫ltiples variables para imputar valores faltantes.

````python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Crear un DataFrame de ejemplo con datos num√©ricos faltantes
data = pd.DataFrame({
       'A': [1, 2, np.nan, 4],
       'B': [5, np.nan, np.nan, 8],
       'C': [9, 10, 11, np.nan]})

# Imputar valores num√©ricos faltantes con imputaci√≥n multivariante
imputer_multivariate = IterativeImputer(max_iter=10, random_state=42)
data_multivariate_imputed = pd.DataFrame(imputer_multivariate.fit_transform(data), columns=data.columns)

print("Datos originales:")
print(data)

print("\nDatos con valores num√©ricos faltantes imputados con imputaci√≥n multivariante:")
print(data_multivariate_imputed)
````

Ambos m√©todos son √∫tiles en situaciones donde los datos faltantes pueden tener una estructura m√°s compleja y pueden capturar patrones no lineales o interacciones entre variables. Es importante ajustar los par√°metros, como el n√∫mero de vecinos en el KNN Imputer o el n√∫mero m√°ximo de iteraciones en la imputaci√≥n multivariante, seg√∫n las caracter√≠sticas espec√≠ficas de tus datos.

# **32. Imputaci√≥n multivariada mediante ecuaciones encadenadas para el valor faltante | Algoritmo MICE | Computadora iterativa**

La imputaci√≥n multivariada mediante ecuaciones encadenadas (MICE, por sus siglas en ingl√©s) es una t√©cnica que utiliza modelos predictivos para imputar valores faltantes en un conjunto de datos. Este m√©todo se basa en la idea de que las variables est√°n condicionalmente relacionadas entre s√≠, y se pueden imputar utilizando modelos predictivos.

A continuaci√≥n, te proporciono una explicaci√≥n detallada con ejemplos en c√≥digo para la imputaci√≥n MICE utilizando el paquete `fancyimpute`.

**Imputaci√≥n Multivariante mediante Ecuaciones Encadenadas (MICE):**

1. **Instalaci√≥n del paquete:**
   ```bash
   pip install fancyimpute
   ```

2. **Ejemplo en C√≥digo:**

````python
import numpy as np
import pandas as pd
from fancyimpute import IterativeImputer

# Crear un DataFrame de ejemplo con datos num√©ricos faltantes
data = pd.DataFrame({
       'A': [1, 2, np.nan, 4],
       'B': [5, np.nan, np.nan, 8],
       'C': [9, 10, 11, np.nan]})

# Imputar valores num√©ricos faltantes con MICE
imputer_mice = IterativeImputer(max_iter=10, random_state=42)
data_mice_imputed = pd.DataFrame(imputer_mice.fit_transform(data), columns=data.columns)

print("Datos originales:")
print(data)

print("\nDatos con valores num√©ricos faltantes imputados con MICE:")
print(data_mice_imputed)
````

Este c√≥digo utiliza el algoritmo MICE para imputar valores faltantes en el conjunto de datos `data`. El `max_iter` especifica el n√∫mero m√°ximo de iteraciones, y `random_state` asegura la reproducibilidad de los resultados.

La imputaci√≥n MICE es √∫til cuando las variables est√°n correlacionadas y puede capturar patrones m√°s complejos que otras t√©cnicas de imputaci√≥n. Sin embargo, ten en cuenta que la elecci√≥n de la estrategia de imputaci√≥n depender√° de la naturaleza espec√≠fica de tus datos y tus objetivos anal√≠ticos.

# **33. ¬øQu√© son los valores at√≠picos? Valores at√≠picos en el aprendizaje autom√°tico**

**¬øQu√© son los valores at√≠picos?**

Los valores at√≠picos, tambi√©n conocidos como outliers, son puntos de datos que difieren significativamente del resto de los datos en un conjunto. Estos valores pueden ser m√°s grandes o m√°s peque√±os que la mayor√≠a de los datos y pueden afectar negativamente el rendimiento de los modelos de aprendizaje autom√°tico, ya que pueden introducir ruido y sesgar los resultados.

**Ejemplos en C√≥digo:**

**Identificaci√≥n de Valores At√≠picos:**

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un conjunto de datos con valores at√≠picos
np.random.seed(42)
data = np.concatenate([np.random.normal(0, 1, 1000), np.random.normal(8, 1, 50)])

# Visualizar el conjunto de datos
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=50, kde=True)
plt.title('Distribuci√≥n de Datos con Valores At√≠picos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
````

En este ejemplo, hemos creado un conjunto de datos con dos distribuciones normales. La segunda distribuci√≥n (con media 8) representa valores at√≠picos.

**Tratamiento de Valores At√≠picos:**

````python
from sklearn.ensemble import IsolationForest

# Identificar valores at√≠picos con Isolation Forest
data = data.reshape(-1, 1)  # Ajustar la forma para el modelo
iso_forest = IsolationForest(contamination=0.02, random_state=42)
outlier_preds = iso_forest.fit_predict(data)

# Filtrar los datos para excluir los valores at√≠picos
filtered_data = data[~(outlier_preds == -1)]

# Visualizar datos sin valores at√≠picos
plt.figure(figsize=(10, 6))
sns.histplot(filtered_data, bins=50, kde=True)
plt.title('Distribuci√≥n de Datos sin Valores At√≠picos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
````

En este caso, hemos utilizado el modelo Isolation Forest para identificar valores at√≠picos y luego filtramos el conjunto de datos original para excluir esos valores.

La identificaci√≥n y manejo adecuado de valores at√≠picos son esenciales para garantizar que los modelos de aprendizaje autom√°tico sean robustos y generalicen bien a nuevos datos. Sin embargo, el enfoque espec√≠fico puede depender del dominio y la naturaleza de los datos.

# **34. Detecci√≥n y eliminaci√≥n de valores at√≠picos mediante el m√©todo de puntuaci√≥n Z | Manejo de valores at√≠picos**

**Detecci√≥n y Eliminaci√≥n de Valores At√≠picos mediante el M√©todo de Puntuaci√≥n Z**

El m√©todo de puntuaci√≥n Z es una t√©cnica com√∫nmente utilizada para identificar valores at√≠picos en un conjunto de datos. Consiste en calcular la puntuaci√≥n Z para cada punto de datos, que mide cu√°ntas desviaciones est√°ndar un punto de datos est√° lejos de la media. Los puntos con puntuaciones Z muy altas o muy bajas se consideran valores at√≠picos.

**Ejemplo en C√≥digo:**

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un conjunto de datos con valores at√≠picos
np.random.seed(42)
data = np.concatenate([np.random.normal(0, 1, 1000), np.random.normal(8, 1, 50)])

# Visualizar el conjunto de datos
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=50, kde=True)
plt.title('Distribuci√≥n de Datos con Valores At√≠picos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
````

**Identificaci√≥n y Eliminaci√≥n de Valores At√≠picos:**

````python
from scipy import stats

# Calcular la puntuaci√≥n Z para cada punto de datos
z_scores = stats.zscore(data)

# Establecer un umbral para identificar valores at√≠picos (por ejemplo, umbral de 3)
threshold = 3
outliers = np.where(np.abs(z_scores) > threshold)[0]

# Filtrar los datos para excluir los valores at√≠picos
filtered_data = data[~np.isin(np.arange(len(data)), outliers)]

# Visualizar datos sin valores at√≠picos
plt.figure(figsize=(10, 6))
sns.histplot(filtered_data, bins=50, kde=True)
plt.title('Distribuci√≥n de Datos sin Valores At√≠picos (M√©todo de Puntuaci√≥n Z)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
````

En este ejemplo, hemos creado un conjunto de datos con dos distribuciones normales, una de las cuales representa valores at√≠picos. Luego, hemos utilizado el m√©todo de puntuaci√≥n Z para identificar y eliminar los valores at√≠picos del conjunto de datos.

Es importante ajustar el umbral seg√∫n los requisitos espec√≠ficos del problema y la distribuci√≥n de los datos. El m√©todo de puntuaci√≥n Z es una t√©cnica simple y efectiva, pero tambi√©n hay otros m√©todos disponibles, como el m√©todo de Tukey o el uso de modelos estad√≠sticos m√°s avanzados. La elecci√≥n del m√©todo depender√° de la naturaleza de los datos y del problema que est√©s abordando.

# **35. Detecci√≥n y eliminaci√≥n de valores at√≠picos mediante el m√©todo IQR | Entregando valores at√≠picos**

**Detecci√≥n y Eliminaci√≥n de Valores At√≠picos mediante el M√©todo IQR (Rango Intercuart√≠lico)**

El m√©todo del Rango Intercuart√≠lico (IQR) es una t√©cnica com√∫nmente utilizada para identificar y eliminar valores at√≠picos en un conjunto de datos. El IQR es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1) de un conjunto de datos. Los puntos de datos que caen fuera del rango definido por Q1 - 1.5 * IQR o Q3 + 1.5 * IQR se consideran valores at√≠picos y pueden ser eliminados.

**Ejemplo en C√≥digo:**

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un conjunto de datos con valores at√≠picos
np.random.seed(42)
data = np.concatenate([np.random.normal(0, 1, 1000), np.random.normal(8, 1, 50)])

# Visualizar el conjunto de datos
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=50, kde=True)
plt.title('Distribuci√≥n de Datos con Valores At√≠picos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
````

**Identificaci√≥n y Eliminaci√≥n de Valores At√≠picos mediante IQR:**

````python
# Calcular el primer y tercer cuartil (Q1 y Q3)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)

# Calcular el IQR
iqr = q3 - q1

# Establecer los l√≠mites para identificar valores at√≠picos
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identificar valores at√≠picos
outliers = np.where((data < lower_bound) | (data > upper_bound))[0]

# Filtrar los datos para excluir los valores at√≠picos
filtered_data = data[~np.isin(np.arange(len(data)), outliers)]

# Visualizar datos sin valores at√≠picos
plt.figure(figsize=(10, 6))
sns.histplot(filtered_data, bins=50, kde=True)
plt.title('Distribuci√≥n de Datos sin Valores At√≠picos (M√©todo IQR)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
````

En este ejemplo, hemos creado un conjunto de datos con dos distribuciones normales, una de las cuales representa valores at√≠picos. Luego, hemos utilizado el m√©todo IQR para identificar y eliminar los valores at√≠picos del conjunto de datos.

Es importante tener en cuenta que el m√©todo IQR asume que los datos siguen una distribuci√≥n aproximadamente normal. Si los datos tienen una distribuci√≥n diferente, otros m√©todos de detecci√≥n de valores at√≠picos pueden ser m√°s apropiados.

# **36. Detecci√≥n de valores at√≠picos mediante el m√©todo percentil | T√©cnica de winsorizaci√≥n**

**Detecci√≥n de Valores At√≠picos mediante el M√©todo de Percentiles y T√©cnica de Winsorizaci√≥n**

La t√©cnica de Winsorizaci√≥n es un m√©todo para manejar valores at√≠picos mediante la modificaci√≥n de los valores extremos en lugar de eliminarlos. Consiste en cambiar los valores at√≠picos por los valores m√°s cercanos que no son at√≠picos, a menudo utilizando percentiles. Esta t√©cnica ayuda a suavizar los efectos de los valores at√≠picos sin eliminarlos por completo.

**Ejemplo en C√≥digo:**

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un conjunto de datos con valores at√≠picos
np.random.seed(42)
data = np.concatenate([np.random.normal(0, 1, 1000), np.random.normal(8, 1, 50)])

# Visualizar el conjunto de datos
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=50, kde=True)
plt.title('Distribuci√≥n de Datos con Valores At√≠picos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
````

**Identificaci√≥n y Winsorizaci√≥n de Valores At√≠picos:**

````python
# Calcular los percentiles 5 y 95
percentile_5 = np.percentile(data, 5)
percentile_95 = np.percentile(data, 95)

# Winsorizar los valores at√≠picos
winsorized_data = np.where(data < percentile_5, percentile_5, np.where(data > percentile_95, percentile_95, data))

# Visualizar datos con t√©cnica de Winsorizaci√≥n
plt.figure(figsize=(10, 6))
sns.histplot(winsorized_data, bins=50, kde=True)
plt.title('Distribuci√≥n de Datos con Winsorizaci√≥n de Valores At√≠picos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
````

En este ejemplo, hemos creado un conjunto de datos con dos distribuciones normales, una de las cuales representa valores at√≠picos. Luego, hemos utilizado la t√©cnica de Winsorizaci√≥n para modificar los valores at√≠picos y suavizar la distribuci√≥n.

Es importante ajustar los percentiles de acuerdo con la naturaleza de los datos y la distribuci√≥n de los valores at√≠picos. La t√©cnica de Winsorizaci√≥n es √∫til cuando se desea mitigar el impacto de los valores at√≠picos sin eliminarlos por completo.

# **37. Construcci√≥n de caracter√≠sticas | Divisi√≥n de funciones**

La construcci√≥n de caracter√≠sticas implica la creaci√≥n de nuevas variables (caracter√≠sticas) a partir de las existentes para mejorar el rendimiento del modelo de aprendizaje autom√°tico. La divisi√≥n de funciones es una t√©cnica com√∫n en la construcci√≥n de caracter√≠sticas que implica dividir una variable existente en partes m√°s peque√±as o segmentos.

**Ejemplo en C√≥digo:**

Supongamos que tenemos una columna que representa la duraci√≥n de una pel√≠cula en minutos. Podemos dividir esta variable en segmentos para crear una nueva caracter√≠stica que indique la categor√≠a de duraci√≥n de la pel√≠cula.

````python
import pandas as pd

# Crear un DataFrame de ejemplo
data = pd.DataFrame({
    'titulo': ['Pelicula1', 'Pelicula2', 'Pelicula3', 'Pelicula4'],
    'duracion': [120, 95, 150, 110]
})

# Visualizar el DataFrame original
print("DataFrame Original:")
print(data)

# Dividir la columna 'duracion' en segmentos
bins = [0, 90, 120, 180]  # Definir los l√≠mites de los segmentos
labels = ['Corta', 'Normal', 'Larga']  # Definir etiquetas para los segmentos
data['duracion_categoria'] = pd.cut(data['duracion'], bins=bins, labels=labels, right=False)

# Visualizar el DataFrame con la nueva caracter√≠stica
print("\nDataFrame con Nueva Caracter√≠stica:")
print(data)
````

En este ejemplo, hemos creado un DataFrame con una columna de duraci√≥n de pel√≠culas. Luego, utilizamos la funci√≥n `pd.cut()` de pandas para dividir la columna de duraci√≥n en segmentos (corta, normal, larga) seg√∫n los l√≠mites definidos en `bins`. La nueva caracter√≠stica, `duracion_categoria`, representa la categor√≠a de duraci√≥n a la que pertenece cada pel√≠cula.

Este proceso de dividir funciones puede aplicarse a diferentes contextos y variables, adapt√°ndolo seg√∫n las necesidades espec√≠ficas del conjunto de datos y del problema que se est√© abordando.

# **38. Maldici√≥n de dimensionalidad**

La "maldici√≥n de la dimensionalidad" es un fen√≥meno en el aprendizaje autom√°tico y la estad√≠stica que se refiere a los desaf√≠os y limitaciones asociados con el aumento de la dimensionalidad de los datos. A medida que el n√∫mero de caracter√≠sticas o dimensiones de los datos aumenta, surgen varios problemas que pueden afectar negativamente el rendimiento de los modelos de aprendizaje autom√°tico.

**Problemas Asociados con la Maldici√≥n de la Dimensionalidad:**

1. **Dispersi√≥n de los Datos:** A medida que la dimensionalidad aumenta, la densidad de los datos en el espacio tambi√©n disminuye. Esto puede llevar a que los datos est√©n dispersos y a que haya menos ejemplos cercanos entre s√≠.

2. **Sobreajuste:** En conjuntos de datos de alta dimensionalidad, los modelos pueden volverse m√°s propensos al sobreajuste, es decir, aprender patrones que son espec√≠ficos del conjunto de entrenamiento pero no se generalizan bien a nuevos datos.

3. **Requerimientos Computacionales:** El aumento en la dimensionalidad conlleva un aumento en los requerimientos computacionales. La complejidad de los c√°lculos y el almacenamiento de datos puede hacer que sea m√°s costoso trabajar con conjuntos de datos de alta dimensionalidad.

**Ejemplo en C√≥digo:**

A continuaci√≥n, se presenta un ejemplo que ilustra la maldici√≥n de la dimensionalidad al generar un conjunto de datos en un espacio de alta dimensionalidad y calcular la distancia media entre los puntos.

````python
import numpy as np

# Generar un conjunto de datos en un espacio de alta dimensionalidad
num_dimensiones = 100
num_ejemplos = 1000
datos_alta_dimension = np.random.random((num_ejemplos, num_dimensiones))

# Calcular la distancia media entre los puntos
distancia_media = np.mean(np.linalg.norm(datos_alta_dimension - datos_alta_dimension[0], axis=1))

print(f"Dimensionalidad: {num_dimensiones}")
print(f"N√∫mero de Ejemplos: {num_ejemplos}")
print(f"Distancia Media entre Puntos: {distancia_media}")
````

En este ejemplo, se genera un conjunto de datos con 100 dimensiones y 1000 ejemplos. Luego, se calcula la distancia media entre los puntos en el espacio de alta dimensionalidad. Observar√°s que la distancia media puede ser significativamente grande, lo que refleja la dispersi√≥n de los datos en un espacio de alta dimensionalidad. Este fen√≥meno puede hacer que la interpretaci√≥n y el rendimiento de los modelos sean m√°s desafiantes.

# **39. An√°lisis de componentes principales (PCA) | Intuici√≥n geom√©trica**

El An√°lisis de Componentes Principales (PCA) es una t√©cnica utilizada en an√°lisis de datos y aprendizaje autom√°tico para reducir la dimensionalidad de un conjunto de datos, preservando al mismo tiempo la mayor cantidad posible de su variabilidad. La intuici√≥n detr√°s de PCA est√° relacionada con la identificaci√≥n y preservaci√≥n de las direcciones principales o "componentes principales" a lo largo de las cuales los datos tienen la mayor varianza.

**Pasos B√°sicos de PCA:**
1. **Centrar los Datos:** Restar la media de cada caracter√≠stica para centrar los datos.
2. **Calcular la Matriz de Covarianza:** Calcular la matriz de covarianza de los datos centrados.
3. **Calcular Autovectores y Autovalores:** Calcular los autovectores y autovalores de la matriz de covarianza.
4. **Ordenar Autovectores:** Ordenar los autovectores seg√∫n sus autovalores en orden descendente.
5. **Proyectar los Datos:** Proyectar los datos originales sobre los autovectores seleccionados para reducir la dimensionalidad.

**Intuici√≥n Geom√©trica:**

Imaginemos que tenemos un conjunto de datos en un espacio bidimensional. Los autovectores representar√°n las direcciones de m√°xima variaci√≥n en los datos. El primer autovector estar√° en la direcci√≥n de m√°xima varianza, y el segundo autovector ser√° ortogonal al primero y estar√° en la direcci√≥n de la segunda m√°xima varianza.

**Ejemplo en C√≥digo:**

Vamos a utilizar PCA con el conjunto de datos Iris para ilustrar el proceso. En este caso, el conjunto de datos tiene 4 dimensiones (longitud del s√©palo, ancho del s√©palo, longitud del p√©talo y ancho del p√©talo).

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data
y = iris.target

# Aplicar PCA para reducir la dimensionalidad a 2
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualizar los resultados
plt.figure(figsize=(8, 6))
for i in range(len(np.unique(y))):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=f'Clase {i}')

plt.title('PCA en el Conjunto de Datos Iris')
plt.xlabel('Primer Componente Principal')
plt.ylabel('Segundo Componente Principal')
plt.legend()
plt.show()
````

En este ejemplo, aplicamos PCA al conjunto de datos Iris y reducimos la dimensionalidad a 2 componentes principales. Luego, visualizamos los datos en el espacio reducido de dos dimensiones. Los puntos en el gr√°fico representan las muestras del conjunto de datos, y la direcci√≥n de m√°xima varianza est√° capturada por los dos primeros componentes principales. Este es un ejemplo simplificado, pero la idea general es aplicable a conjuntos de datos de mayor dimensionalidad.

# **40. An√°lisis de componentes principales (PCA) | Formulaci√≥n del problema y soluci√≥n paso a paso**

El An√°lisis de Componentes Principales (PCA) es una t√©cnica fundamental en estad√≠sticas y aprendizaje autom√°tico para reducir la dimensionalidad de los datos mientras se mantiene la mayor cantidad posible de su variabilidad. El objetivo es encontrar una proyecci√≥n de los datos en un nuevo espacio, donde las dimensiones (llamadas componentes principales) est√°n ordenadas seg√∫n la cantidad de variabilidad que capturan.

**Formulaci√≥n del Problema:**
Dado un conjunto de datos $(X)$ con $(n)$ muestras y $(m)$ caracter√≠sticas, el objetivo es encontrar una transformaci√≥n lineal $(W)$ que proyecta los datos originales en un nuevo espacio $(Z)$, donde $(Z)$ tiene $(m')$ componentes principales ordenadas seg√∫n la varianza.

**Pasos B√°sicos:**

1. **Centrar los Datos:**
   - Restar la media de cada caracter√≠stica para centrar los datos.

2. **Calcular la Matriz de Covarianza:**
   - Calcular la matriz de covarianza de los datos centrados.

3. **Calcular Autovectores y Autovalores:**
   - Calcular los autovectores y autovalores de la matriz de covarianza.

4. **Ordenar Autovectores:**
   - Ordenar los autovectores seg√∫n sus autovalores en orden descendente.

5. **Seleccionar Componentes Principales:**
   - Elegir los primeros $(m')$ autovectores como componentes principales.

6. **Proyectar los Datos:**
   - Proyectar los datos originales en el nuevo espacio definido por los autovectores seleccionados.

**Ejemplo en C√≥digo:**

Vamos a utilizar el conjunto de datos Iris y aplicar PCA paso a paso.

````python
import numpy as np
from sklearn.datasets import load_iris

# Paso 1: Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data

# Paso 2: Centrar los Datos
X_centered = X - np.mean(X, axis=0)

# Paso 3: Calcular la Matriz de Covarianza
cov_matrix = np.cov(X_centered, rowvar=False)

# Paso 4: Calcular Autovectores y Autovalores
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Paso 5: Ordenar Autovectores
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvectors_sorted = eigenvectors[:, sorted_indices]

# Paso 6: Seleccionar Componentes Principales (por ejemplo, los dos primeros)
num_components = 2
principal_components = eigenvectors_sorted[:, :num_components]

# Paso 7: Proyectar los Datos
X_pca = X_centered.dot(principal_components)

# Visualizar los resultados
print("Matriz de Covarianza:")
print(cov_matrix)
print("\nComponentes Principales:")
print(principal_components)
print("\nDatos Proyectados:")
print(X_pca)
````

Este ejemplo muestra c√≥mo llevar a cabo el An√°lisis de Componentes Principales paso a paso utilizando el conjunto de datos Iris. La matriz de covarianza se calcula, los autovectores se ordenan y se seleccionan los componentes principales. Luego, los datos se proyectan en el nuevo espacio definido por los autovectores seleccionados. Estos pasos ilustran el proceso fundamental de PCA.
