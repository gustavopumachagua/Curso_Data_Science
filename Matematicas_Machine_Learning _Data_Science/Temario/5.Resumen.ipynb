## **Funciones Matematicas**

Las funciones matemáticas son reglas que establecen una relación entre dos conjuntos de datos, de tal manera que cada elemento del primer conjunto se asocia con uno y solo uno del segundo conjunto. En otras palabras, una función es una forma de describir cómo una variable (o conjunto de variables) depende de otra.

Aquí hay algunos ejemplos de funciones matemáticas comunes:

* **La función lineal:**

 $y = mx + b$, donde $(m)$ es la pendiente de la línea y $(b)$ es la intersección con el eje $(y)$.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2*x + 3

x = np.linspace(-5, 5, 100)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, color='red', linestyle='--', label='$y = 2x + 3$')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfica de la Función Lineal')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.text(0, 6, 'Ejemplo de Anotación', fontsize=10, ha='center')
plt.show()
```

* **La función cuadrática:**

 $y = ax^2 + bx + c$, donde $(a)$, $(b)$ y $(c)$ son constantes y $(x)$ es la variable independiente.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 4*x + 5

x = np.linspace(-5, 5, 100)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, color='green', linestyle='-', linewidth=2, label='$y = x^2 - 4x + 5$')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfica de la Función Cuadrática')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.text(-3, 15, 'Mínimo', fontsize=10, ha='center')
plt.show()
```

* **La función exponencial:**

 $y = a^x$, donde $(a)$ es una constante y $(x)$ es la variable independiente.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2**x

x = np.linspace(-5, 5, 100)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, color='blue', linestyle='-', linewidth=2, label='$y = 2^x$')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfica de la Función Exponencial')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.text(1, 5, 'Ejemplo de Anotación', fontsize=10, ha='center')
plt.show()
```

* **La función logarítmica:**

 $y = log_a(x)$, donde $(a)$ es la base del logaritmo y $(x)$ es la variable independiente.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.log(x)

x = np.linspace(0.1, 5, 100)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, color='purple', linestyle='-', linewidth=2, label='$y = log(x)$')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfica de la Función Logarítmica')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.text(2, 0, 'Ejemplo de Anotación', fontsize=10, ha='center')
plt.show()
```

* **La función trigonométrica:**

 por ejemplo, la función seno: $y = sin(x)$, donde $(x)$ es la variable independiente y $(\sin)$ es la función trigonométrica.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)

x = np.linspace(-np.pi, np.pi, 100)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, color='green', linestyle='-', linewidth=2, label='$y = \sin(x)$')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfica de la Función Trigonométrica (Seno)')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.text(1, 0.8, 'Máximo', fontsize=10, ha='center')
plt.show()
```

## **Derivadas**

Las derivadas son una herramienta fundamental en el cálculo diferencial, que se utilizan para calcular la tasa de cambio instantánea de una función en un punto dado. Una derivada mide la rapidez con la que cambia una función en un punto específico, y se define como el límite de la tasa de cambio media de la función en un intervalo cada vez más pequeño alrededor del punto.

La derivada de una función se denota por la notación $f'(x)$ o $\frac{dy}{dx}$, y se puede calcular utilizando diferentes reglas y técnicas, como la regla de la cadena, la regla del producto y la regla de la potencia.

Aquí hay algunos ejemplos de cómo se pueden calcular las derivadas:

* **Derivada de una función lineal:**

 la función $f(x) = 2x + 1$ es una función lineal, y su derivada es simplemente el coeficiente de $(x)$, es decir, $f'(x) = 2$.

* **Derivada de una función cuadrática:**

 la función $f(x) = x^2$ tiene una derivada que se puede calcular utilizando la regla de la potencia: $f'(x) = 2x$.

* **Derivada de una función trigonométrica:**

 la función $f(x) = \sin(x)$ tiene una derivada que se puede calcular utilizando la regla del coseno: $f'(x) = \cos(x)$.

* **Derivada de una función exponencial:**

 la función $f(x) = e^x$ tiene una derivada que es igual a sí misma: $f'(x) = e^x$.

Las derivadas tienen muchas aplicaciones en la física, la ingeniería, la economía y otras áreas donde es necesario medir la tasa de cambio instantánea de una función en un punto dado.

* **Derivada de una función lineal:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return 2*x + 1

# Calcular la derivada
def f_deriv(x):
    return 2*np.ones_like(x)

# Graficar la función y su derivada
x = np.linspace(-5, 5, 100)
y = f(x)
y_deriv = f_deriv(x)

plt.plot(x, y, label='f(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y_deriv, label="f'(x)", color='green', linestyle='--', linewidth=2)
plt.legend(['f(x) = 2x + 1', "f'(x) = 2"])
plt.title("Función Lineal y su Derivada")
plt.xlabel("x")
plt.ylabel("y")

# Añadir línea de referencia horizontal en y=0
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')

plt.show()
```

* **Derivada de una función cuadrática:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return x**2

# Calcular la derivada
def f_deriv(x):
    return 2*x

# Graficar la función y su derivada
x = np.linspace(-5, 5, 100)
y = f(x)
y_deriv = f_deriv(x)

plt.plot(x, y, label='f(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y_deriv, label="f'(x)", color='green', linestyle='--', linewidth=2)
plt.legend(['$f(x) = x^2$', "$f'(x) = 2x$"])
plt.title("Función Cuadrática y su Derivada")
plt.xlabel("x")
plt.ylabel("y")

# Añadir línea de referencia horizontal en y=0
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return np.sin(x)

# Calcular la derivada
def f_deriv(x):
    return np.cos(x)

# Graficar la función y su derivada
x = np.linspace(-5, 5, 100)
y = f(x)
y_deriv = f_deriv(x)

plt.plot(x, y, label='f(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y_deriv, label="f'(x)", color='green', linestyle='--', linewidth=2)
plt.legend(['f(x) = sin(x)', "f'(x) = cos(x)"])
plt.title("Función Seno y su Derivada")
plt.xlabel("x")
plt.ylabel("y")

# Añadir línea de referencia horizontal en y=0
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')

plt.show()
```

Las derivadas en `3D` son un concepto matemático que se utiliza en cálculo vectorial y que permite calcular la tasa de cambio de una función en tres dimensiones. Para visualizar la derivada en `3D`, se pueden utilizar mapas de colores para representar la variación de la función en diferentes direcciones.

A continuación se muestra un ejemplo de cómo calcular y graficar la derivada parcial de una función en `3D` utilizando Python y la librería `Matplotlib`:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# Calcular la derivada parcial respecto a x
def f_deriv_x(x, y):
    return x*np.cos(np.sqrt(x**2 + y**2))/(np.sqrt(x**2 + y**2)*np.sqrt(x**2 + y**2 + 1))

# Calcular la derivada parcial respecto a y
def f_deriv_y(x, y):
    return y*np.cos(np.sqrt(x**2 + y**2))/(np.sqrt(x**2 + y**2)*np.sqrt(x**2 + y**2 + 1))

# Crear una figura en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Crear una malla de puntos en x y y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calcular la función y su derivada en la malla de puntos
Z = f(X, Y)
dZ_dx = f_deriv_x(X, Y)
dZ_dy = f_deriv_y(X, Y)

# Graficar la superficie de la función
surface = ax.plot_surface(X, Y, Z, cmap='viridis', rstride=1, cstride=1, alpha=0.8, linewidth=0.5, edgecolors='k')

# Graficar las derivadas parciales como mapas de colores
ax.contourf(X, Y, dZ_dx, zdir='x', offset=-5, cmap='plasma')
ax.contourf(X, Y, dZ_dy, zdir='y', offset=5, cmap='plasma')

# Ajuste de elevación y azimut
ax.view_init(elev=20, azim=30)

# Barra de colores
fig.colorbar(surface, shrink=0.5, aspect=10)

# Etiquetas y título
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Superficie y derivadas parciales')

# Anotación
ax.text(1, 1, 0.8, 'Máximo', color='red', fontsize=10, ha='center')

# Mostrar la gráfica
plt.show()
```

Este código grafica una superficie tridimensional de la función $f(x,y) = \sin(\sqrt{x^2 + y^2})$ y dos mapas de colores que representan la variación de la función en las direcciones $(x)$ e $(y)$. Para calcular las derivadas parciales se utilizan las funciones $f_{deriv_x}$ y $f_{deriv_y}$, las cuales se calculan utilizando la regla de la cadena y la regla del cociente.

La gráfica resultante muestra cómo la función varía en diferentes direcciones y permite visualizar el concepto de derivada parcial en `3D`.

## **Derivadas Parciales**

Las derivadas parciales son un concepto fundamental en el cálculo multivariado, que se refiere a la tasa de cambio de una función con respecto a cada una de sus variables independientes.

En otras palabras, la derivada parcial de una función en un punto dado es la tasa de cambio de la función en ese punto cuando se varía solo una de sus variables independientes, manteniendo las demás constantes.

Por ejemplo, si se tiene una función $f(x,y)$ que representa la temperatura en un punto $(x,y)$ en una placa de metal, entonces la derivada parcial de $(f)$ con respecto a $(x)$ mide cómo cambia la temperatura en ese punto cuando se varía solo la coordenada $(x)$, manteniendo constante la coordenada $(y)$. La derivada parcial de $(f)$ con respecto a $(y)$, por otro lado, mide cómo cambia la temperatura cuando se varía solo la coordenada $(y)$, manteniendo constante la coordenada $(x)$.

Las derivadas parciales son importantes en muchos campos de las matemáticas y la ciencia, incluyendo la física, la economía, la ingeniería, las estadísticas y la informática.

**Ejemplo:**

Una función $f(x,y) = 2x^2 + 3xy + 4y^2$ tiene dos derivadas parciales:

* La derivada parcial de $(f)$ con respecto a $(x)$ es $f_x = 4x + 3y$.
* La derivada parcial de $(f)$ con respecto a $(y)$ es $f_y = 3x + 8y$.

Estas derivadas parciales indican cómo cambia $(f)$ cuando se varía $(x)$ o $(y)$, respectivamente, manteniendo la otra variable constante.

Para calcular y graficar las derivadas parciales de una función en Python, se puede utilizar la librería `NumPy` y `Matplotlib`. A continuación se muestra un ejemplo para la función $f(x,y) = x^2 + y^2$:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Patch

# Definir la función f(x, y)
def f(x, y):
    return x**2 + y**2

# Definir las funciones de las derivadas parciales de f(x, y)
def f_x(x, y):
    return 2*x

def f_y(x, y):
    return 2*y

# Generar los datos para la superficie de f(x, y)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Generar los datos para la superficie de f_x(x, y)
Z_x = f_x(X, Y)

# Generar los datos para la superficie de f_y(x, y)
Z_y = f_y(X, Y)

# Graficar las superficies de f(x, y), f_x(x, y), y f_y(x, y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surface_f = ax.plot_surface(X, Y, Z, color='c', alpha=1.0, linewidth=0.5, edgecolors='k')
surface_fx = ax.plot_surface(X, Y, Z_x, color='b', alpha=0.7)
surface_fy = ax.plot_surface(X, Y, Z_y, color='r', alpha=0.7)

# Añadir leyenda directamente con Patch
legend_elements = [
    Patch(facecolor='c', edgecolor='k', alpha=1.0, label='f(x, y)'),
    Patch(facecolor='b', edgecolor='k', alpha=0.7, label='f_x(x, y)'),
    Patch(facecolor='r', edgecolor='k', alpha=0.7, label='f_y(x, y)'),
]

ax.legend(handles=legend_elements)

# Ajustes adicionales
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Función y sus derivadas parciales')
ax.view_init(elev=20, azim=30)

plt.show()
```

El resultado es una gráfica `3D` que muestra la superficie de la función $f(x,y)$, así como las superficies de sus derivadas parciales con respecto a $(x)$ y a $(y)$. La superficie de $f(x,y)$ es la más grande y se encuentra en color amarillo. Las superficies de $f_x(x,y)$ y $f_y(x,y)$ se encuentran por debajo de la superficie de $f(x,y)$ y en colores azul y rojo respectivamente.

Cada una de las superficies de las derivadas parciales muestra cómo cambia la función cuando se varía solo una de las variables, manteniendo la otra constante.

Las derivadas parciales en `3D` se utilizan para analizar cómo cambia una función de varias variables en diferentes direcciones. En esencia, se trata de la tasa de cambio de la función en una determinada dirección. Aquí hay un ejemplo de cómo calcular y graficar las derivadas parciales de una función en Python:

Supongamos que tenemos la siguiente función de dos variables:

$f(x, y) = x^2 + y^2$

Para calcular la derivada parcial con respecto a $(x)$, simplemente diferenciamos la función con respecto a $(x)$ mientras mantenemos y constante:

$\frac{∂f}{∂x} = 2x$

Para calcular la derivada parcial con respecto a $(y)$, simplemente diferenciamos la función con respecto a $(y)$ mientras mantenemos $(x)$ constante:

$\frac{∂f}{∂y} = 2y$

En Python, podemos usar la biblioteca `Matplotlib` para graficar estas derivadas parciales en `3D`. Aquí está el código:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función
def f(x, y):
    return x**2 + y**2

# Calcular las derivadas parciales
def f_x(x, y):
    return 2*x

def f_y(x, y):
    return 2*y

# Crear la malla de puntos para la superficie
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Evaluar la función en la malla
Z = f(X, Y)

# Evaluar las derivadas parciales en la malla
Z_x = f_x(X, Y)
Z_y = f_y(X, Y)

# Crear la figura
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
surface = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7, rstride=100, cstride=100, edgecolor='k', linewidth=0.5)

# Graficar la derivada parcial con respecto a x
ax.plot_surface(X, Y, Z_x, color='r', alpha=0.5)

# Graficar la derivada parcial con respecto a y
ax.plot_surface(X, Y, Z_y, color='g', alpha=0.5)

# Configurar los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Superficie y Derivadas Parciales')

# Añadir una barra de color
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Ajustar la vista
ax.view_init(elev=20, azim=30)

# Mostrar la figura
plt.show()
```

Este código generará una gráfica `3D` de la función y sus derivadas parciales con respecto a $(x)$ e $(y)$. La superficie de la función se mostrará en azul, mientras que la derivada parcial con respecto a $(x)$ se mostrará en rojo y la derivada parcial con respecto a $(y)$ se mostrará en verde.

## **Gradiente**

El gradiente es un vector que apunta en la dirección de la máxima tasa de crecimiento de una función en un punto dado. Matemáticamente, el gradiente es un vector compuesto por las derivadas parciales de una función en un punto dado.

Por ejemplo, si tenemos una función de dos variables $f(x, y)$, el gradiente de $(f)$ en un punto $(x_0, y_0)$ está dado por el vector:

$grad(f) = [\frac{∂f}{∂x}, \frac{∂f}{∂y}]$ en $(x_0, y_0)$

Este vector indica la dirección y la magnitud del cambio máximo en $(f)$ en el punto $(x_0, y_0)$.

En Python, podemos calcular el gradiente de una función utilizando la biblioteca `NumPy` y la función `gradient`:

```python
import numpy as np

# Definir la función
def f(x, y):
    return x**2 + y**2

# Definir el punto en el que se calculará el gradiente
x0, y0 = 1, 2

# Calcular el gradiente de f en (x0, y0)
df_dx = lambda x: 2*x
df_dy = lambda y: 2*y
grad_f = np.array([df_dx(x0), df_dy(y0)])

print("Gradiente de f en ({}, {}): {}".format(x0, y0, grad_f))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x, y):
    return x**2 + y**2

# Definir el punto en el que se calculará el gradiente
x0, y0 = 1, 2

# Calcular el gradiente de f en (x0, y0)
df_dx = lambda x: 2*x
df_dy = lambda y: 2*y
grad_f = np.array([df_dx(x0), df_dy(y0)])

# Crear una malla de puntos para la visualización de flechas
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)

# Calcular los valores de la función en la malla
Z = f(X, Y)

# Crear la figura
fig, ax = plt.subplots()

# Graficar las flechas que representan el gradiente en (x0, y0)
ax.quiver(x0, y0, grad_f[0], grad_f[1], angles='xy', scale_units='xy', scale=1, color='r', label='Gradiente en (1, 2)')

# Graficar la función de superficie
surface = ax.contour(X, Y, Z, cmap='viridis', levels=15)

# Configurar los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Gradiente de la función en (1, 2)')

# Añadir leyenda
ax.legend()

# Mostrar la figura
plt.show()
```

Este código calculará el gradiente de la función $f(x, y) = x^2 + y^2$ en el punto $(1, 2)$, y mostrará el resultado:


Gradiente de `f` en `(1, 2)`: `[2 4]`


Esto significa que en el punto `(1, 2)`, la función $f(x, y)$ crece más rápido en la dirección del eje $(x)$ (con una tasa de cambio de `2`), y más rápido en la dirección del eje $(y)$ (con una tasa de cambio de `4`).

Para mostrar un ejemplo del gradiente con gráfica, podemos utilizar la función $f(x,y) = x^2 + 2y^2$, que tiene como gradiente $\nabla f(x,y) = [2x, 4y]$.

Podemos usar el paquete `Matplotlib` para graficar la función y su gradiente.

Primero, creamos un arreglo `2D` de valores de $(x)$ e $(y)$ utilizando la función `meshgrid` de `NumPy`. Luego, calculamos los valores de la función y del gradiente para cada par $(x, y)$. Finalmente, graficamos la función y el gradiente utilizando la función `quiver` de `Matplotlib`.

Aquí está el código:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x, y):
    return x**2 + 2*y**2

# Definir el gradiente
def grad_f(x, y):
    return np.array([2*x, 4*y])

# Crear un arreglo 2D de valores de x e y
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)

# Calcular los valores de la función y del gradiente para cada par (x, y)
Z = f(X, Y)
U, V = grad_f(X, Y)

# Graficar la función y el gradiente
fig, ax = plt.subplots()

# Contornos de la función
contour = ax.contour(X, Y, Z, levels=10, cmap='viridis')

# Flechas del gradiente
quiver = ax.quiver(X, Y, U, V, angles='xy', scale=30, color='r')

# Configuración adicional
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_aspect('equal')
ax.set_title('Función y Gradiente')

# Personalizar el color de fondo
ax.set_facecolor('#f0f0f0')

# Añadir cuadrícula
ax.grid(True, linestyle='--', alpha=0.7)

# Mostrar la figura
plt.show()
```

Este código grafica la función de superficie y el gradiente como flechas en el plano $(x, y)$. La función de superficie se muestra como líneas de contorno, que son curvas en el plano que conectan puntos de igual valor de la función. Las flechas del gradiente apuntan en la dirección de la mayor pendiente ascendente y su longitud es proporcional a la magnitud del gradiente en ese punto.

Aquí te presento un ejemplo de cómo calcular y graficar el gradiente de una función de dos variables en `3D` utilizando Python y la librería `matplotlib`:

Supongamos que tenemos la función:

$f(x, y) = x^2 + y^2$

Para calcular el gradiente, necesitamos calcular las derivadas parciales de la función con respecto a cada una de las variables:

$\frac{∂f}{∂x} = 2x$

$\frac{∂f}{∂y} = 2y$

Entonces, el gradiente de la función es:

$grad(f) = (2x, 2y)$

Ahora, vamos a graficar la función y el gradiente usando la librería `matplotlib`:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función
def f(x, y):
    return x**2 + y**2

# Definir el rango de valores de x e y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Crear una malla de puntos (x, y)
X, Y = np.meshgrid(x, y)

# Calcular los valores de la función en cada punto de la malla
Z = f(X, Y)

# Calcular el gradiente de la función en cada punto de la malla
Gx, Gy = np.gradient(Z)
G = np.sqrt(Gx**2 + Gy**2)

# Graficar la función y el gradiente en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Superficie
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.8, antialiased=True)

# Flechas del gradiente
ax.quiver(X, Y, Z, Gx, Gy, 0, length=0.1, color='black')

# Configuraciones adicionales
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title('Función y Gradiente')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

Este código generará una figura en `3D` que muestra la función y el gradiente en cada punto de la malla. La función se representa como una superficie en rojo y azul, mientras que el gradiente se representa como una serie de flechas negras en la dirección del gradiente. La longitud de las flechas indica la magnitud del gradiente en ese punto.

## **Calculo Diferencial**

El cálculo diferencial es una rama de las matemáticas que se enfoca en el estudio de las funciones y sus propiedades, incluyendo la tasa de cambio de una función en un punto dado. Esta tasa de cambio se calcula mediante el concepto de derivada, que representa la pendiente de la recta tangente a la curva en un punto determinado.

A continuación se muestra un ejemplo de cómo utilizar Python para graficar la función $f(x) = x^2$ y su derivada $f'(x) = 2x$:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función y su derivada
def f(x):
    return x**2

def df(x):
    return 2*x

# Creamos un array con los valores de x
x = np.linspace(-5, 5, 100)

# Graficamos la función y su derivada en la misma figura
fig, ax = plt.subplots()

# Línea para la función f(x)
ax.plot(x, f(x), label='$f(x)=x^2$', color='blue', linewidth=2)

# Línea para la derivada df(x)
ax.plot(x, df(x), label='$f\'(x)=2x$', color='green', linestyle='--', linewidth=2)

# Añadimos etiquetas y leyenda
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.legend(fontsize=12)

# Añadimos una cuadrícula
ax.grid(True, linestyle='--', alpha=0.7)

# Añadimos un título
ax.set_title('Función y Derivada', fontsize=16)

# Mostramos la figura
plt.show()
```

En la gráfica resultante, se puede observar que la función $f(x) = x^2$ tiene un punto mínimo en $x=0$, donde su derivada $f'(x) = 2x$ es igual a cero. A partir de ese punto, la derivada es positiva, lo que indica que la función está aumentando, y a medida que $x$ se aleja de cero, la tasa de cambio se hace cada vez mayor.

## **Estadistica Inferencial**

La estadística inferencial es la rama de la estadística que se encarga de realizar inferencias o conclusiones sobre una población a partir de la información obtenida de una muestra representativa de esa población. Es decir, se utiliza para hacer generalizaciones o estimaciones sobre una población a partir de la información obtenida de una muestra.

Un ejemplo de estadística inferencial podría ser el estudio de la estatura media de los estudiantes de una universidad a partir de una muestra de `100` estudiantes seleccionados al azar.

A continuación, se presenta un ejemplo de código y gráfica para ilustrar cómo se puede utilizar la estadística inferencial en la práctica:

Supongamos que tenemos un conjunto de datos que representa la edad de un grupo de personas. Queremos estimar la edad media de la población a partir de una muestra aleatoria de `50` personas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generamos una muestra aleatoria de 50 personas
np.random.seed(123)
muestra = np.random.randint(low=18, high=60, size=50)

# Calculamos la media y la desviación estándar de la muestra
media_muestra = np.mean(muestra)
std_muestra = np.std(muestra)

# Estimamos el error estándar de la media
sem = std_muestra / np.sqrt(len(muestra))

# Calculamos el intervalo de confianza del 95% para la media poblacional
intervalo_confianza = stats.t.interval(0.95, df=len(muestra)-1, loc=media_muestra, scale=sem)

# Mostramos los resultados
print("Media muestral:", media_muestra)
print("Error estándar de la media:", sem)
print("Intervalo de confianza del 95%:", intervalo_confianza)

# Graficamos un histograma de la muestra
plt.hist(muestra, bins=10, alpha=0.5, color='skyblue', edgecolor='black')

# Línea vertical para la media muestral
plt.axvline(x=media_muestra, color='r', linestyle='--', label='Media muestral', linewidth=2)

# Líneas verticales para el intervalo de confianza
plt.axvline(x=intervalo_confianza[0], color='g', linestyle='--', label='Intervalo de confianza', linewidth=2)
plt.axvline(x=intervalo_confianza[1], color='g', linestyle='--', linewidth=2)

# Añadimos etiquetas y leyenda
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()

# Añadimos un título
plt.title('Histograma y Intervalo de Confianza')

# Mostramos la figura
plt.show()
```

En este ejemplo, generamos una muestra aleatoria de `50` personas y calculamos la media y la desviación estándar de la muestra. A continuación, utilizamos estas medidas para estimar el error estándar de la media y calcular el intervalo de confianza del `95%` para la media poblacional. Finalmente, graficamos un histograma de la muestra y agregamos la media muestral y el intervalo de confianza al gráfico.

## **Algebral Lineal**

El Álgebra Lineal es una rama de las matemáticas que se enfoca en el estudio de vectores, matrices y sistemas de ecuaciones lineales. Es una herramienta fundamental en áreas como la física, la ingeniería, la informática y la estadística, entre otras. Algunos de los conceptos más importantes del Álgebra Lineal son la multiplicación de matrices, el cálculo de determinantes, la resolución de sistemas de ecuaciones lineales y la diagonalización de matrices.

A continuación se presenta un ejemplo de cómo se pueden utilizar algunos de estos conceptos para realizar una operación de rotación en el plano.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos un vector en el plano
v = np.array([1, 1])

# Definimos el ángulo de rotación
theta = np.pi/4

# Definimos la matriz de rotación
R = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

# Rotamos el vector utilizando la matriz de rotación
v_rot = R @ v

# Graficamos los vectores antes y después de la rotación
fig, ax = plt.subplots()

# Vector original
ax.arrow(0, 0, v[0], v[1], head_width=0.1, head_length=0.1, fc='blue', ec='blue', label='Original Vector')

# Vector rotado
ax.arrow(0, 0, v_rot[0], v_rot[1], head_width=0.1, head_length=0.1, fc='red', ec='red', label='Rotated Vector')

# Establecemos límites y cuadrícula
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.grid(True)

# Añadimos etiquetas
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Añadimos una leyenda
ax.legend()

# Añadimos un título
plt.title('Vector Rotation')

# Mostramos la figura
plt.show()
```

En este ejemplo, primero definimos un vector $(v)$ en el plano con coordenadas `(1, 1)`. Luego, definimos un ángulo $(\theta)$ de `45` grados (o $(\frac{\pi}{4})$ radianes) y utilizamos esta información para construir una matriz de rotación $(R)$. Finalmente, rotamos el vector $(v)$ utilizando la matriz $(R)$ y obtenemos un nuevo vector $(v_{rot})$ con coordenadas $(0, \sqrt2)$. Para visualizar el efecto de la rotación, graficamos los vectores $(v)$ y $(v_{rot})$ en el plano, junto con un sistema de ejes cartesianos y un grid para facilitar la visualización.

La figura resultante muestra el vector original en azul y el vector rotado en rojo. Se puede observar que el vector rotado ha sido desplazado hacia la izquierda y hacia arriba, lo cual es consistente con la rotación de `45` grados que definimos anteriormente.
