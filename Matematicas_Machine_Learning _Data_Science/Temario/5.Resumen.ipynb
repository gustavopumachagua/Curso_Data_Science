## **Derivadas**

Las derivadas son una herramienta fundamental en el c谩lculo diferencial, que se utilizan para calcular la tasa de cambio instant谩nea de una funci贸n en un punto dado. Una derivada mide la rapidez con la que cambia una funci贸n en un punto espec铆fico, y se define como el l铆mite de la tasa de cambio media de la funci贸n en un intervalo cada vez m谩s peque帽o alrededor del punto.

La derivada de una funci贸n se denota por la notaci贸n $f'(x)$ o $\frac{dy}{dx}$, y se puede calcular utilizando diferentes reglas y t茅cnicas, como la regla de la cadena, la regla del producto y la regla de la potencia.

Aqu铆 hay algunos ejemplos de c贸mo se pueden calcular las derivadas:

* **Derivada de una funci贸n lineal:**

 la funci贸n $f(x) = 2x + 1$ es una funci贸n lineal, y su derivada es simplemente el coeficiente de $(x)$, es decir, $f'(x) = 2$.

* **Derivada de una funci贸n cuadr谩tica:**

 la funci贸n $f(x) = x^2$ tiene una derivada que se puede calcular utilizando la regla de la potencia: $f'(x) = 2x$.

* **Derivada de una funci贸n trigonom茅trica:**

 la funci贸n $f(x) = \sin(x)$ tiene una derivada que se puede calcular utilizando la regla del coseno: $f'(x) = \cos(x)$.

* **Derivada de una funci贸n exponencial:**

 la funci贸n $f(x) = e^x$ tiene una derivada que es igual a s铆 misma: $f'(x) = e^x$.

Las derivadas tienen muchas aplicaciones en la f铆sica, la ingenier铆a, la econom铆a y otras 谩reas donde es necesario medir la tasa de cambio instant谩nea de una funci贸n en un punto dado.

* **Derivada de una funci贸n lineal:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n
def f(x):
    return 2*x + 1

# Calcular la derivada
def f_deriv(x):
    return 2*np.ones_like(x)

# Graficar la funci贸n y su derivada
x = np.linspace(-5, 5, 100)
y = f(x)
y_deriv = f_deriv(x)

plt.plot(x, y, label='f(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y_deriv, label="f'(x)", color='green', linestyle='--', linewidth=2)
plt.legend(['f(x) = 2x + 1', "f'(x) = 2"])
plt.title("Funci贸n Lineal y su Derivada")
plt.xlabel("x")
plt.ylabel("y")

# A帽adir l铆nea de referencia horizontal en y=0
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')

plt.show()
```

* **Derivada de una funci贸n cuadr谩tica:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci贸n
def f(x):
    return x**2

# Calcular la derivada
def f_deriv(x):
    return 2*x

# Graficar la funci贸n y su derivada
x = np.linspace(-5, 5, 100)
y = f(x)
y_deriv = f_deriv(x)

plt.plot(x, y, label='f(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y_deriv, label="f'(x)", color='green', linestyle='--', linewidth=2)
plt.legend(['$f(x) = x^2$', "$f'(x) = 2x$"])
plt.title("Funci贸n Cuadr谩tica y su Derivada")
plt.xlabel("x")
plt.ylabel("y")

# A帽adir l铆nea de referencia horizontal en y=0
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')

plt.show()
```

Las derivadas en `3D` son un concepto matem谩tico que se utiliza en c谩lculo vectorial y que permite calcular la tasa de cambio de una funci贸n en tres dimensiones. Para visualizar la derivada en `3D`, se pueden utilizar mapas de colores para representar la variaci贸n de la funci贸n en diferentes direcciones.

A continuaci贸n se muestra un ejemplo de c贸mo calcular y graficar la derivada parcial de una funci贸n en `3D` utilizando Python y la librer铆a `Matplotlib`:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la funci贸n
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# Calcular la derivada parcial respecto a x
def f_deriv_x(x, y):
    return x*np.cos(np.sqrt(x**2 + y**2))/(np.sqrt(x**2 + y**2)*np.sqrt(x**2 + y**2 + 1))

# Calcular la derivada parcial respecto a y
def f_deriv_y(x, y):
    return y*np.cos(np.sqrt(x**2 + y**2))/(np.sqrt(x**2 + y**2)*np.sqrt(x**2 + y**2 + 1))

# Crear una figura en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Crear una malla de puntos en x y y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calcular la funci贸n y su derivada en la malla de puntos
Z = f(X, Y)
dZ_dx = f_deriv_x(X, Y)
dZ_dy = f_deriv_y(X, Y)

# Graficar la superficie de la funci贸n
surface = ax.plot_surface(X, Y, Z, cmap='viridis', rstride=1, cstride=1, alpha=0.8, linewidth=0.5, edgecolors='k')

# Graficar las derivadas parciales como mapas de colores
ax.contourf(X, Y, dZ_dx, zdir='x', offset=-5, cmap='plasma')
ax.contourf(X, Y, dZ_dy, zdir='y', offset=5, cmap='plasma')

# Ajuste de elevaci贸n y azimut
ax.view_init(elev=20, azim=30)

# Barra de colores
fig.colorbar(surface, shrink=0.5, aspect=10)

# Etiquetas y t铆tulo
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Superficie y derivadas parciales')

# Anotaci贸n
ax.text(1, 1, 0.8, 'M谩ximo', color='red', fontsize=10, ha='center')

# Mostrar la gr谩fica
plt.show()
```

Este c贸digo grafica una superficie tridimensional de la funci贸n $f(x,y) = \sin(\sqrt{x^2 + y^2})$ y dos mapas de colores que representan la variaci贸n de la funci贸n en las direcciones $(x)$ e $(y)$. Para calcular las derivadas parciales se utilizan las funciones $f_{deriv_x}$ y $f_{deriv_y}$, las cuales se calculan utilizando la regla de la cadena y la regla del cociente.

La gr谩fica resultante muestra c贸mo la funci贸n var铆a en diferentes direcciones y permite visualizar el concepto de derivada parcial en `3D`.

## **Derivadas Parciales**

Las derivadas parciales son un concepto fundamental en el c谩lculo multivariado, que se refiere a la tasa de cambio de una funci贸n con respecto a cada una de sus variables independientes.

En otras palabras, la derivada parcial de una funci贸n en un punto dado es la tasa de cambio de la funci贸n en ese punto cuando se var铆a solo una de sus variables independientes, manteniendo las dem谩s constantes.

Por ejemplo, si se tiene una funci贸n $f(x,y)$ que representa la temperatura en un punto $(x,y)$ en una placa de metal, entonces la derivada parcial de $(f)$ con respecto a $(x)$ mide c贸mo cambia la temperatura en ese punto cuando se var铆a solo la coordenada $(x)$, manteniendo constante la coordenada $(y)$. La derivada parcial de $(f)$ con respecto a $(y)$, por otro lado, mide c贸mo cambia la temperatura cuando se var铆a solo la coordenada $(y)$, manteniendo constante la coordenada $(x)$.

Las derivadas parciales son importantes en muchos campos de las matem谩ticas y la ciencia, incluyendo la f铆sica, la econom铆a, la ingenier铆a, las estad铆sticas y la inform谩tica.

**Ejemplo:**

Una funci贸n $f(x,y) = 2x^2 + 3xy + 4y^2$ tiene dos derivadas parciales:

* La derivada parcial de $(f)$ con respecto a $(x)$ es $f_x = 4x + 3y$.
* La derivada parcial de $(f)$ con respecto a $(y)$ es $f_y = 3x + 8y$.

Estas derivadas parciales indican c贸mo cambia $(f)$ cuando se var铆a $(x)$ o $(y)$, respectivamente, manteniendo la otra variable constante.

Para calcular y graficar las derivadas parciales de una funci贸n en Python, se puede utilizar la librer铆a `NumPy` y `Matplotlib`. A continuaci贸n se muestra un ejemplo para la funci贸n $f(x,y) = x^2 + y^2$:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Patch

# Definir la funci贸n f(x, y)
def f(x, y):
    return x**2 + y**2

# Definir las funciones de las derivadas parciales de f(x, y)
def f_x(x, y):
    return 2*x

def f_y(x, y):
    return 2*y

# Generar los datos para la superficie de f(x, y)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Generar los datos para la superficie de f_x(x, y)
Z_x = f_x(X, Y)

# Generar los datos para la superficie de f_y(x, y)
Z_y = f_y(X, Y)

# Graficar las superficies de f(x, y), f_x(x, y), y f_y(x, y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surface_f = ax.plot_surface(X, Y, Z, color='c', alpha=1.0, linewidth=0.5, edgecolors='k')
surface_fx = ax.plot_surface(X, Y, Z_x, color='b', alpha=0.7)
surface_fy = ax.plot_surface(X, Y, Z_y, color='r', alpha=0.7)

# A帽adir leyenda directamente con Patch
legend_elements = [
    Patch(facecolor='c', edgecolor='k', alpha=1.0, label='f(x, y)'),
    Patch(facecolor='b', edgecolor='k', alpha=0.7, label='f_x(x, y)'),
    Patch(facecolor='r', edgecolor='k', alpha=0.7, label='f_y(x, y)'),
]

ax.legend(handles=legend_elements)

# Ajustes adicionales
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Funci贸n y sus derivadas parciales')
ax.view_init(elev=20, azim=30)

plt.show()
```

El resultado es una gr谩fica `3D` que muestra la superficie de la funci贸n $f(x,y)$, as铆 como las superficies de sus derivadas parciales con respecto a $(x)$ y a $(y)$. La superficie de $f(x,y)$ es la m谩s grande y se encuentra en color amarillo. Las superficies de $f_x(x,y)$ y $f_y(x,y)$ se encuentran por debajo de la superficie de $f(x,y)$ y en colores azul y rojo respectivamente.

Cada una de las superficies de las derivadas parciales muestra c贸mo cambia la funci贸n cuando se var铆a solo una de las variables, manteniendo la otra constante.

Las derivadas parciales en `3D` se utilizan para analizar c贸mo cambia una funci贸n de varias variables en diferentes direcciones. En esencia, se trata de la tasa de cambio de la funci贸n en una determinada direcci贸n. Aqu铆 hay un ejemplo de c贸mo calcular y graficar las derivadas parciales de una funci贸n en Python:

Supongamos que tenemos la siguiente funci贸n de dos variables:

$f(x, y) = x^2 + y^2$

Para calcular la derivada parcial con respecto a $(x)$, simplemente diferenciamos la funci贸n con respecto a $(x)$ mientras mantenemos y constante:

$\frac{f}{x} = 2x$

Para calcular la derivada parcial con respecto a $(y)$, simplemente diferenciamos la funci贸n con respecto a $(y)$ mientras mantenemos $(x)$ constante:

$\frac{f}{y} = 2y$

En Python, podemos usar la biblioteca `Matplotlib` para graficar estas derivadas parciales en `3D`. Aqu铆 est谩 el c贸digo:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la funci贸n
def f(x, y):
    return x**2 + y**2

# Calcular las derivadas parciales
def f_x(x, y):
    return 2*x

def f_y(x, y):
    return 2*y

# Crear la malla de puntos para la superficie
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Evaluar la funci贸n en la malla
Z = f(X, Y)

# Evaluar las derivadas parciales en la malla
Z_x = f_x(X, Y)
Z_y = f_y(X, Y)

# Crear la figura
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
surface = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7, rstride=100, cstride=100, edgecolor='k', linewidth=0.5)

# Graficar la derivada parcial con respecto a x
ax.plot_surface(X, Y, Z_x, color='r', alpha=0.5)

# Graficar la derivada parcial con respecto a y
ax.plot_surface(X, Y, Z_y, color='g', alpha=0.5)

# Configurar los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Superficie y Derivadas Parciales')

# A帽adir una barra de color
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Ajustar la vista
ax.view_init(elev=20, azim=30)

# Mostrar la figura
plt.show()
```

Este c贸digo generar谩 una gr谩fica `3D` de la funci贸n y sus derivadas parciales con respecto a $(x)$ e $(y)$. La superficie de la funci贸n se mostrar谩 en azul, mientras que la derivada parcial con respecto a $(x)$ se mostrar谩 en rojo y la derivada parcial con respecto a $(y)$ se mostrar谩 en verde.

## **Gradiente**

El gradiente es un vector que apunta en la direcci贸n de la m谩xima tasa de crecimiento de una funci贸n en un punto dado. Matem谩ticamente, el gradiente es un vector compuesto por las derivadas parciales de una funci贸n en un punto dado.

Por ejemplo, si tenemos una funci贸n de dos variables $f(x, y)$, el gradiente de $(f)$ en un punto $(x_0, y_0)$ est谩 dado por el vector:

## **Calculo Diferencial**

El c谩lculo diferencial es una rama de las matem谩ticas que se enfoca en el estudio de las funciones y sus propiedades, incluyendo la tasa de cambio de una funci贸n en un punto dado. Esta tasa de cambio se calcula mediante el concepto de derivada, que representa la pendiente de la recta tangente a la curva en un punto determinado.

A continuaci贸n se muestra un ejemplo de c贸mo utilizar Python para graficar la funci贸n $f(x) = x^2$ y su derivada $f'(x) = 2x$:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la funci贸n y su derivada
def f(x):
    return x**2

def df(x):
    return 2*x

# Creamos un array con los valores de x
x = np.linspace(-5, 5, 100)

# Graficamos la funci贸n y su derivada en la misma figura
fig, ax = plt.subplots()

# L铆nea para la funci贸n f(x)
ax.plot(x, f(x), label='$f(x)=x^2$', color='blue', linewidth=2)

# L铆nea para la derivada df(x)
ax.plot(x, df(x), label='$f\'(x)=2x$', color='green', linestyle='--', linewidth=2)

# A帽adimos etiquetas y leyenda
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.legend(fontsize=12)

# A帽adimos una cuadr铆cula
ax.grid(True, linestyle='--', alpha=0.7)

# A帽adimos un t铆tulo
ax.set_title('Funci贸n y Derivada', fontsize=16)

# Mostramos la figura
plt.show()
```

En la gr谩fica resultante, se puede observar que la funci贸n $f(x) = x^2$ tiene un punto m铆nimo en $x=0$, donde su derivada $f'(x) = 2x$ es igual a cero. A partir de ese punto, la derivada es positiva, lo que indica que la funci贸n est谩 aumentando, y a medida que $x$ se aleja de cero, la tasa de cambio se hace cada vez mayor.

## **Estadistica Inferencial**

La estad铆stica inferencial es la rama de la estad铆stica que se encarga de realizar inferencias o conclusiones sobre una poblaci贸n a partir de la informaci贸n obtenida de una muestra representativa de esa poblaci贸n. Es decir, se utiliza para hacer generalizaciones o estimaciones sobre una poblaci贸n a partir de la informaci贸n obtenida de una muestra.

Un ejemplo de estad铆stica inferencial podr铆a ser el estudio de la estatura media de los estudiantes de una universidad a partir de una muestra de `100` estudiantes seleccionados al azar.

A continuaci贸n, se presenta un ejemplo de c贸digo y gr谩fica para ilustrar c贸mo se puede utilizar la estad铆stica inferencial en la pr谩ctica:

Supongamos que tenemos un conjunto de datos que representa la edad de un grupo de personas. Queremos estimar la edad media de la poblaci贸n a partir de una muestra aleatoria de `50` personas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generamos una muestra aleatoria de 50 personas
np.random.seed(123)
muestra = np.random.randint(low=18, high=60, size=50)

# Calculamos la media y la desviaci贸n est谩ndar de la muestra
media_muestra = np.mean(muestra)
std_muestra = np.std(muestra)

# Estimamos el error est谩ndar de la media
sem = std_muestra / np.sqrt(len(muestra))

# Calculamos el intervalo de confianza del 95% para la media poblacional
intervalo_confianza = stats.t.interval(0.95, df=len(muestra)-1, loc=media_muestra, scale=sem)

# Mostramos los resultados
print("Media muestral:", media_muestra)
print("Error est谩ndar de la media:", sem)
print("Intervalo de confianza del 95%:", intervalo_confianza)

# Graficamos un histograma de la muestra
plt.hist(muestra, bins=10, alpha=0.5, color='skyblue', edgecolor='black')

# L铆nea vertical para la media muestral
plt.axvline(x=media_muestra, color='r', linestyle='--', label='Media muestral', linewidth=2)

# L铆neas verticales para el intervalo de confianza
plt.axvline(x=intervalo_confianza[0], color='g', linestyle='--', label='Intervalo de confianza', linewidth=2)
plt.axvline(x=intervalo_confianza[1], color='g', linestyle='--', linewidth=2)

# A帽adimos etiquetas y leyenda
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()

# A帽adimos un t铆tulo
plt.title('Histograma y Intervalo de Confianza')

# Mostramos la figura
plt.show()
```

En este ejemplo, generamos una muestra aleatoria de `50` personas y calculamos la media y la desviaci贸n est谩ndar de la muestra. A continuaci贸n, utilizamos estas medidas para estimar el error est谩ndar de la media y calcular el intervalo de confianza del `95%` para la media poblacional. Finalmente, graficamos un histograma de la muestra y agregamos la media muestral y el intervalo de confianza al gr谩fico.

A continuaci贸n se presenta un ejemplo de c贸mo se pueden utilizar algunos de estos conceptos para realizar una operaci贸n de rotaci贸n en el plano.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos un vector en el plano
v = np.array([1, 1])

# Definimos el 谩ngulo de rotaci贸n
theta = np.pi/4

# Definimos la matriz de rotaci贸n
R = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

# Rotamos el vector utilizando la matriz de rotaci贸n
v_rot = R @ v

# Graficamos los vectores antes y despu茅s de la rotaci贸n
fig, ax = plt.subplots()

# Vector original
ax.arrow(0, 0, v[0], v[1], head_width=0.1, head_length=0.1, fc='blue', ec='blue', label='Original Vector')

# Vector rotado
ax.arrow(0, 0, v_rot[0], v_rot[1], head_width=0.1, head_length=0.1, fc='red', ec='red', label='Rotated Vector')

# Establecemos l铆mites y cuadr铆cula
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.grid(True)

# A帽adimos etiquetas
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# A帽adimos una leyenda
ax.legend()

# A帽adimos un t铆tulo
plt.title('Vector Rotation')

# Mostramos la figura
plt.show()
```

En este ejemplo, primero definimos un vector $(v)$ en el plano con coordenadas `(1, 1)`. Luego, definimos un 谩ngulo $(\theta)$ de `45` grados (o $(\frac{\pi}{4})$ radianes) y utilizamos esta informaci贸n para construir una matriz de rotaci贸n $(R)$. Finalmente, rotamos el vector $(v)$ utilizando la matriz $(R)$ y obtenemos un nuevo vector $(v_{rot})$ con coordenadas $(0, \sqrt2)$. Para visualizar el efecto de la rotaci贸n, graficamos los vectores $(v)$ y $(v_{rot})$ en el plano, junto con un sistema de ejes cartesianos y un grid para facilitar la visualizaci贸n.

La figura resultante muestra el vector original en azul y el vector rotado en rojo. Se puede observar que el vector rotado ha sido desplazado hacia la izquierda y hacia arriba, lo cual es consistente con la rotaci贸n de `45` grados que definimos anteriormente.
